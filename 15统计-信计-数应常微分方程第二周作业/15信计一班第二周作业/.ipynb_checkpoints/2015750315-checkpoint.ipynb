$$ \frac{\mathrm dy}{\mathrm dx} = \frac{y^6 - 2x^2}{2xy^5+x^2y^2} = \frac {(y^3)-2x^2}{y^2(2xy^3+x^2)} = \frac {\mathrm dy^3}{\mathrm dx}= \frac{3[(y^3)^2-2x^2]}{2xy^3+x^2}$$ ,令 y^3 = u , 则原方程为$$ \frac{\mathrm d u}{\mathrm dx}= \frac{3u^2-6x^2}{2xu+x^2}= \frac{\frac{3u^2}{x^2}-6}{\frac{2u}{x}+1}$$ 则此方程为齐次方程，令$$ \frac{u}{x}=z$$ ,则$$ \frac {\mathrm du}{\mathrm dx}= z + x\frac{\mathrm dz}{\mathrm dx}$$ ,所以 $$ \frac{\mathrm du}{\mathrm dx}=z + x\frac{\mathrm dz}{\mathrm dx}$$ , 所以 $$\frac {3z^2-6}{2z+1}= z+x\frac{\mathrm dz}{\mathrm dx}$$, $$ x\frac{\mathrm dz}{\mathrm dx} = \frac{z^2-z-6}{2z+1}$$,  当$$z^2-z-2 = 0$$, 得 $$z = 3 $$ 或 $$ z = -2 $$ 是上式方程的解。 即$$ y^3 = 3x $$ 或 $$ y^3 = -2x $$是方程的解。 当$$ z^2 - z -6 \neq 0 $$时， 变量分离 $$\frac {2z+1}{z^2-z-d}dz=\frac{1}{x}dx$$,  两边积分得 $$(z-3)^7(z+2)^3 = x^5c$$, 即 $$ (y^3-3x)^7(y^3+2x)^3 = x^5c$$, 又因为$$ y^3 = 3x$$ 或 $$y^3 = -2x$$ 包含在通解中当 $$c = 0$$时。 故原方程的解为$$(y^3-3x)^7(y^3+2x)^3 = x^15c$$

(2)

原方程化为$$ \frac{\mathrm dy}{\mathrm dx} = \frac{x(2x^2+3y^2+1)}{y(3x^2+2y^2-1)}$$,$$ \frac{\mathrm dy^2}{\mathrm dx^2}= \frac {2x^2+3y^2+1}{3x^2+2y^2-1}$$ 令 $$y^2= u$$ , $$x^2 = v$$ , 则 $$\frac {\mathrm du}{\mathrm dv} = \frac{2v + 3u + 1}{3v + 2u - 1}（1）$$ 方程组   $$\begin{cases}
2v+3u+1 = 0 \\ 
3v +2u -1 = 0 
\end{cases}$$,从而方程 （1）化为$$ \frac {\mathrm dy}{\mathrm dz} = \frac {2+3\frac{y}{z}}{3+2\frac{y}{z}}$$
令 $$ t = \frac {y}{z}$$,则有$$ \frac {\mathrm dy}{\mathrm dz} = t + z\frac{\mathrm dt}{\mathrm dz}$$,所以 $$ t +z\frac{\mathrm dt}{\mathrm dz} = \frac {2+3t}{3+2t}$$, $$z\frac{2 - 2t^2}{3+2t}(2)$$, 当  $$ 2 - 2t^2 = 0$$时， 即 $$t = \pm 1$$,是方程（2）的解。得 $$ y^2 = x^2-2 $$或$$ y^2 = -x^2 $$是原方程的解。 当$$2-2t^2 \neq 0$$时， 分离变量得$$ \frac{3+2t}{2-2t^2}dt = \frac{1}{z}dz$$两边积分得 $$y^2+x^2 = (y^2-x^2+x)^5c$$ 另外$$ y^2 = x^2-2$$ ,或$$ y^2 = -x^2$$，包含在其通解中，故原方程的解为 $$y^2+x^2 = (y^2 - x^2 +x)^5c$$
