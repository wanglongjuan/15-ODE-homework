1.求下列方程的解：

（6）$x\frac{\mathrm d y}{\mathrm d x}-y+\sqrt{x^2-y^2}=0$;

$解：令\frac{y}{x}=u,y=ux,\frac{\mathrm d y}{\mathrm d x}=u+x\frac{\mathrm d u}{\mathrm d x},$则原方程化为：
$$\frac{\mathrm d u}{\mathrm d x}=\frac{\sqrt{x^2(1-u^2)}}{x},$$
分离变量得：$\frac{1}{\sqrt{1-u^2}}\mathrm d u=sgn x*\frac{1}{x}\mathrm d x$
两边积分得：$$arcsin u=sgn x*ln|x|+c$$
带回原来变量得：$$\arcsin\frac{y}{x}=sgn x*ln|x|+c$$
另外,$y^2=x^2也是方程的解$

7)$\tan y\mathrm d x=\cot x\mathrm d y$

解：变量分离得：$\frac{\tan y}{\mathrm d y}=\frac{\cot x}{\mathrm d x};$
两边积分得：$$ln|\sin y|=-ln|cos x|+c$$

（7）$\frac{\mathrm d y}{\mathrm d x}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$

$\frac{\mathrm d y}{\mathrm d x}=\frac{(y^3)^2-2x^2}{y^2(2xy^3+x^2)}=\frac{\mathrm d y^3}{\mathrm d x}=\frac{3((y^3)^2-2x^2)}{2xy^3+x^2}$
,令y^3=u,则化为：$\frac{\mathrm d u}{\mathrm d x}=\frac{\frac{3u^2}{x^2}-6}{2\frac{u}{x}+1}$,令$\frac{u}{x}=z$,则$\frac{\mathrm d u}{\mathrm d x}=z+x\frac{\mathrm d z}{\mathrm d x}$,所以当$z^2-z-6=0$,得z=3或z=-2是方程解。即$y^3=3x$或$y^3=-2x$是方程的解。
当$z^2-z-6 \neq 0$时，
$$\frac{2z+1}{z^2-z-d}\mathrm d z=\frac{1}{x}\mathrm d x$$,解得：$$(y^3-3x)^7(y^3+2x)^3=x^5c$$
因为$y^3=3x$或$y^3=-2x$包含在通解中当c=0时。故原方程解为：$$(y^3-3x)^7(y^3+2x)^3=x^(15)c$$
