一.验证下列各函数是相应微分方程的解：

(1)   $y=x^2+1$,$y'=y^2-(x^2+1)y+2x$

   解:       $y'=2x$
             $$y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)(x^2+1)+2x=2x=y'$$
              $y=x^2+1$是方程解             

(2)$y=-\frac{g(x)}{f(x)}$,$y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$    

   解:$$y'=-\frac{g'(x)f(x)-g(x)f'(x)}{(f(x))^2}=-\frac{g'(x)}{f'(x)}+\frac{f'(x)y^2}{g(x)}$$
       所以$y=-\frac{g（x）}{f（x)}$是方程解

二.给定一阶微分方程$\frac{\mathrm d y}{\mathrm d x}=2x$

(1)求出它的通解；

解：    等式两边积分得：
$$y=x^2+c$$

（2）求通过点（1,4）的特解

解： 把（1，4）代入$y=x^2+c$中得：
      $$c=3$$
      再把c=3代入$y=x^2+c$得：
      $$y=x^2+3$$

(3)求出与直线$y=2x+3$相切的解

解：立方程，方程解唯一求c，
$$\begin{cases}   y=x^2+c\\y =2x+3\\
\end{cases}$$
$$c=4$$
与直线$y=2x+3$相切的解:$y=x^2+4$

(4)求出满足条件$\int\limits_0^1 y\mathrm d x=2x$的解

解：$$\int\limits_0^1 y\mathrm d x =\frac{1}{3}+c=2$$
    则$c=\frac{5}{3}$
    所以满足题意的解为$y=x^2+\frac{5}{3}$

$x\frac{\mathrm d y}{\mathrm d x}-y+\sqrt{x^2-y^2}=0$

$$\frac{\mathrm d u}{\mathrm d x}=\frac{\sqrt{x^2(1-u^2)}}{x},$$
分离变量得：$\frac{1}{\sqrt{1-u^2}}\mathrm d u=sgn x\frac{1}{x}\mathrm d x$
两边积分得：$$arcsin u=sgn xln|x|+c$$
带回原来变量得：$$\arcsin\frac{y}{x}=sgn xln|x|+c$$
另外,$y^2=x^2也是方程的解$

四.作适当的变量变换求解下列方程：

$\frac{\mathrm d y}{\mathrm d x}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$

解：$\frac{\mathrm d y}{\mathrm d x}=\frac{(y^3)^2-2x^2}{y^2(2xy^3+x^2)}=\frac{\mathrm d y^3}{\mathrm d x}=\frac{3((y^3)^2-2x^2)}{2xy^3+x^2}$
,令y^3=u,则化为：$\frac{\mathrm d u}{\mathrm d x}=\frac{\frac{3u^2}{x^2}-6}{2\frac{u}{x}+1}$,令$\frac{u}{x}=z$,则$\frac{\mathrm d u}{\mathrm d x}=z+x\frac{\mathrm d z}{\mathrm d x}$,所以当$z^2-z-6=0$,得z=3或z=-2是方程解。  即$y^3=3x$或$y^3=-2x$是方程的解。
当$z^2-z-6 \neq 0$时，
$$\frac{2z+1}{z^2-z-d}\mathrm d z=\frac{1}{x}\mathrm d x$$,解得：$$(y^3-3x)^7(y^3+2x)^3=x^5c$$
因为$y^3=3x$或$y^3=-2x$包含在通解中当c=0时。故原方程解为：$$(y^3-3x)^7(y^3+2x)^3=x^(15)c$$
