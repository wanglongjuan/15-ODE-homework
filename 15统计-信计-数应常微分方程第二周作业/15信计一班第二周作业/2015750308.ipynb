$1(6)$

原方程化为$xdy-ydx+\sqrt{x^2-y^2}dx=0$.

当$y^\neq x^2$时,两边同时除以$\sqrt{x^2-y^2}$,

得$\frac{d(\frac{x}{y})}{\sqrt{1-(\frac{y}{x})^2}}+\frac{1}{x}dx=0$.

两边积分得到$arcsin{\frac{y}{x}}+In |x|=c$,c为任意常数。

当$y^2=x^2$时，可以知道这也是一个解。

所以方程的解为$y^2=x^2$及$arcsin{\frac{y}{x}}+In |x|=c$。

$2(6)$

令$u=y^3$

方程可化为$\frac{du}{dx}=\frac{3u^2-6x^2}{2xu+x^2}$,

令$v=\frac{u}{x}$,方程可化为$x\frac{dv}{dx}+v=\frac{3v^2-6}{2v+1}$

可以化为$7/5\frac{dv}{v-3}+3/5\frac{dv}{v+2}=\frac{dx}{x}$

积分得到$(v-3)^7(v+2)^3=cx^5$,c为正常数

$2(7)$

方程可化为$\frac{ydy}{xdx}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$

令$u=x^2-1$,$v=y^2+1$,方程可化为$\frac{dv}{du}=\frac{2u+3v}{3u}{3u+2v}$

令$w=\frac{v}{u}$,得$\frac{dv}{du}=w+u\frac{dw}{du}$,$u\frac{dw}{du}=\frac{2-2w^2}{3+2w}$,

化简得$\frac{4}{u}du=(\frac{5}{1-w}+\frac{1}{1+w})dw$

积分得$u^4(1-w)^5=c(1+w)$,

方程的解为$(x^2-y^2-2)^5=c(x^2+y^2)$,c为正常数.
