3.验证下列各函数是相应微分方程的解

（7）$ y=x^2+1,y'=y^2-(x^2+1)y+2x $<br>

解：$因为y=x^2+1$<br>
   $所以 y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)(x^2+1)+2x=2x=y'$<br>
   $所以 y=x^2+1是方程y'=y^2-(x^2+1)y+2x的解$<br>

（8）$y=-\frac{g(x)}{f(x)},y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$<br>

 解：$因为y=-\frac{g(x)}{f(x)}$<br>
    $所以\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}=\frac{f'(x)}{g(x)}(-\frac{g(x)}{f(x)})^2-\frac{g'(x)}{f(x)}=\frac{f'(x)g(x)-g'(x)f(x)}{f(x)^2}=y'$<br>
    $所以y=-\frac{g(x)}{f(x)}是y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}的解$<br>

    $所以y=x^2+\frac{5}{3}$<br>
