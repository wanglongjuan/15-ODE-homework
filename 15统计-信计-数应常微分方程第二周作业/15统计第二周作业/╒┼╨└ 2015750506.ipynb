### 3.请验证下列各函数是相应微分方程的解：

**(7)** $ y = x^2 + 1 $ , $ y' = y^2 - ( x^2 + 1 ) y + 2x $

解：$y'= 2x $  ,  $y^2 - ( x^2 + 1 ) y + 2x = (x^2 + 1 )^2 + 2x$

∴微分方程左右两端相等，即函数是微分方程的解。

**(8)** $y = - \frac{g(x)}{f(x)}$  ,  $y' = \frac{f'(x)}{g(x)}y^2 - \frac{g'(x)}{f(x)}$

解：$ y'= - \frac{g'(x)f(x)-g(x)f'(x)}{f^2(x)} $  ,  $ y^2= \frac{g^2(x)}{f^2(x)} $

∴$\frac{f'(x)}{g(x)}y^2 - \frac{g'(x)}{f(x)} = \frac{f'(x)}{g(x)}\frac{g^2(x)}{f^2(x)}- \frac{g'(x)}{f(x)}=\frac{f'(x)g(x)-f(x)g'(x)}{f^2(x)}= - \frac{g'(x)f(x)-g(x)f'(x)}{f^2(x)}= y' $

∴微分方程左右两端相等，即函数是微分方程的解。

### 1.求下列方程的解：

**(6)** $x\frac{dy}{dx}- y + \sqrt{x^2-y^2}=0 $；

解：令$ \frac{y}{x}=u $，$ y=ux $，$\frac{dy}{dx}= u+ \frac{du}{dx} $，

则原方程化为：$ \frac{du}{dx} = \frac{\sqrt {x^2(1-u^2)}}{x} $，分离变量得：$ \frac{1}{\sqrt{1-u^2}} du=\frac{1}{x}dx$

两边积分得：$ \arcsin (\frac{y}{x}) + \ln\mid x\mid = c$，c为任意常数，

另外，可验证当$ y^ =x^2 $也为方程的解，

所以原方程的解为$ \arcsin (\frac{y}{x}) + \ln\mid x\mid = c$ 和 $ y^ =x^2 $。

**(7)** $ \tan ydx -\cot xdy =0$

解：分离变量得：$ \frac{dx}{\cot x}=\frac{dy}{\tan y} $ ，

两边积分得：$ \ln \mid \sin y\mid= -\ln \mid \cos x\mid +c$，

所以原方程的解为$ \ln \mid \sin y\mid= -\ln \mid \cos x\mid +c$.

### 2.做适当的变换求解下列方程：

**(6)** $ \frac{dy}{dx}= \frac{y^6-2x^2}{2xy^5+x^2y^2} $；

解：$ \frac{dy}{dx}= \frac{y^6-2x^2}{2xy^5+x^2y^2} = \frac{dy^3}{dx} = \frac{3[(y^3)^2-2x^2]}{2xy^3+x^2} $

令$ y^3=u $，则原方程化为$ \frac{du}{dx}= \frac{3u^2-6x^2}{2xu+x^2}=\frac{\frac{3u^2}{x^2}-6}{\frac{2u}{x}+1} $，为齐次方程，

令$ \frac{u}{x}=z $，则$ \frac{du}{dx}=z+\frac{dz}{dx} $，所以$ \frac{3z^2-6}{2z+1}=z+x\frac{dz}{dx} $，$x\frac{dz}{dx}=\frac{z^2-z-6}{2z+1}  \cdots①$

当$z^2-z-6= 0$时，得$ z=3 $或$ z=-2 $是方程①的解，即$ y^3=3x $或$ y^3=-2x $是方程的解；

当$z^2-z-6\neq 0$时，分离变量并积分得：$(z-3)^7(z+2)^5= x^5c $，即$(y^3-3x)^7(y^3+2x)^5= x^5c $，

又因为$ y^3=3x $和$ y^3=-2x $包括在通解中，所以原方程的解为$(y^3-3x)^7(y^3+2x)^5= x^5c $。

**(7)** $ \frac{dy}{dx}= \frac{2x^3+ 3xy^2+x}{3x^2y+2y^3-y} $；

解：原方程可化为：$ \frac{dy}{dx} = \frac{x(2x^2+3y^2+1)}{y(3x^2+2y^2-1)}$，$ \frac{dy^2}{dx^2}= \frac{2x^2+3y^2+1}{3x^2+2y^2-1}$，

令$ y^2=u ,x^2=v$,则$ \frac{du}{dv}= \frac{2v+3u+1}{3v+2u-1}$,  方程组$\begin{cases}
2v+3u+1=0 \\3v+2u-1=0
\end{cases}$的解为(1,-1)

令$Z=v-1,Y=u+1$,则有$\begin{cases}
2Z+3Y=0 \\3Z+2Y=0
\end{cases}$,所以①可化为$ \frac{dy}{dx} =\frac{2+3\frac{y}{z}}{3+2\frac{y}{z}}   $

令$ t=\frac{y}{z} $，则有$ \frac{dy}{dz}=y+z\frac{dt}{dz} $，所以$ t+z\frac{dt}{dz}=\frac{2+3t}{3+2t} $，$ z\frac{dt}{dz}=\frac{2-2t^2}{3+2t^2}\cdots② $

当$ 2-2t^2=0 $时，即$ t=\pm 1 $，是方程②的解，所以$ y^2=x^2-2 $或$ y^2=-x $是原方程的解；

当$ 2-2t^2\neq0 $时，分离变量并积分得$y^2+x^2=(y^2-x^2+2)^5c $；

因为$ y^2=x^2-2 $和$ y^2=-x $包括在通解中，所以原方程的解是$y^2+x^2=(y^2-x^2+2)^5c $。
