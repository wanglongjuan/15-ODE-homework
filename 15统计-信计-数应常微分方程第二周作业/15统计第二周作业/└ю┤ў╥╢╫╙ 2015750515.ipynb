## 3.验证下列各函数是相应微分方程的解 ##
（7）$ y=x^2+1,y'=y^2-(x^2+1)y+2x $<br>
解：$因为y=x^2+1$<br>
   $所以 y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)(x^2+1)+2x=2x=y'$<br>
   $所以 y=x^2+1是方程y'=y^2-(x^2+1)y+2x的解$<br>
（8）$y=-\frac{g(x)}{f(x)},y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$<br>
 解：$因为y=-\frac{g(x)}{f(x)}$<br>
    $所以\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}=\frac{f'(x)}{g(x)}(-\frac{g(x)}{f(x)})^2-\frac{g'(x)}{f(x)}=\frac{f'(x)g(x)-g'(x)f(x)}{f(x)^2}=y'$<br>
    $所以y=-\frac{g(x)}{f(x)}是y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}的解$<br>

## 1.求下列方程的解  ##
（6）$x\frac{dy}{dx}-y+\sqrt{x^2-y^2}$<br>
解：$令\frac{y}{x}=u,y=ux,\frac{dy}{dx}=u+x\frac{du}{dx},则原方程化为：$<br>
   $\frac{du}{dx}=\frac{\sqrt{x^2(1-u^2)}}{x}，分离变量得：\frac{1}{1-u^2}du=sgn x \frac{1}{x}dx$<br>
   $两边积分得：\arcsin u =sgn x \ln |x|+c$
   $代回原来变量，得\arcsin \frac{y}{x}=sgn x \ln |x|+c $<br>
   $另外，y^2=x^2也是方程的解$<br>
（7）$\tan y dx-\cot x dy=0$<br>
解：$变量分离。得\frac{dy}{\tan y}=\tan x dx$<br>
   $两边积分得：\ln |\sin y|=-\ln |\cos x|+c$<br>
   $即\sin y\cos x=c$<br>
   $另外，当y=k\pi,k=0,\pm1,\pm2,\cdots时，也容易验证是方程的解$<br>
## 2.做适当的变量变换求解下列方程 ##
（6）$\frac{dy}{dx}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$<br>
解：$\frac{dy}{dx}=\frac{(y^3)^2-2x^2}{y^2(2xy^3)+x^2}=\frac{dy^3}{x}=\frac{3[(y^3)^2-2x^2]}{2xy^3+x^2},令y^3=u,则原方程化为$<b2xur>
   $\frac{du}{dx}=\frac{3u^2-6x^2}{2xu+x^2}=\frac{\frac{3u^2}{x^2}-6}{2\frac{u}{x}+1},这是齐次方程，令$<br>
   $\frac{u}{x}=z,则\frac{du}{dx}=z+x\frac{du}{dx},所以\frac{3z^2-6}{2z+1}=z+x\frac{dz}{dx},x\frac{dz}{dx}=\frac{z^2-z-6}{2z+1}$<br>
   $当z^2-z-6=0,得z=3或z=-2,所以y^3=3x或y^3=-2是方程的解$<br>
   $当z^2-z-6\neq 0,变量分离\frac{2z+1}{z^2-z-6}dz=\frac{1}{x}dx，两边积分$<br>
   $(y^3-3x)^7(y^3+2x)^3=x^5c,又因为y^3=3x或y^3=-2x包含在通解中当c=0时，故原方程的解为(y^3-3x)^7(y^3+2x)^3=cx^{15} $<br>
（7）$\frac{dy}{dx}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$<br>
解：$原方程化为\frac{dy}{dx}=\frac{x(2x^2+3y^2+1)}{y(3x^2+2y^2-1)},\frac{dy^2}{dx^2}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$<br>
   $令y^2=u,x^2=v,则\frac{du}{dv}=\frac{2v+3u+1}{3v+2u-1}$<br>
   $方程组\left\{\begin{array}\\2v+3u+1\\{3v+2u-1} \end{array}\right.的解为（1,-1），令q=v-1,p=u+1$<br>
   $则有\left\{\begin{array}\\2q+3p=0\\{3q+2p=0} \end{array}\right.可得\frac{dp}{dq}=\frac{2+3\frac{p}{q}}{3+2\frac{p}{q}}$<br>
   $令t=\frac{p}{q},则有\frac{dp}{dq}=t+z\frac{dp}{dq},所以t+q\frac{dt}{dq}=\frac{2+3t}{3+2t},q\frac{dt}{dq}=\frac{2-2t^2}{3+2t}$<br>
   $当2-2t^2=0,即t=\pm ,得y^2=x^2-2或y^2=-x^2是原方程的解$<br>
   $当2-2t^2\neq 0时，分离变量得\frac{3+2t}{2-2t^2}dt=\frac{1}{q}dq两边积分得y^2+x^2=(y^2-x^2+2)^5c$<br>
   $另外，y^2=x^2-2或y^2=-x^2,包含在其通解中，故原方程的解为(y^2+x^2)c=y^2-x^2$<br>
