$1.求下列方程的解:$  
$(6)xdy/dx-y+(x^2+y^2)½=0$  
$解：两边同除以x,$  
$原式化为dy/dx-y/x+sgn(x)[1-(y/x)^2]½=0$  
$作变换y=ux,$  
$化为xdu/dx+sgn(x)(1-u^2)½=0,$  
$变量分离du/（1-u^2）½=-sgn(x)dx/x,$  
$积分得\arcsin u+sgn (x)ln|x|=c$ 
$∴原方程的解为\arcsin y/x+sgn(x)ln|x|=c$  
$∵u^2=1亦为方程的解,$  
$∴原方程还有解y=±x$  

$(7)\tan ydx-\cot xdy=0$  
$解：\tan y≠0时,\cot ydy=\tan xdx$  
$此时解为\sin y\cos x=c$  
$\tan y=0时，有解y=nπ，n=0,±1,±2,...$  

$2.作适当的变量变换求解下列方程$  
$(6)dy/dx=(y^6-2x^2)/(2xy^5+x^2y^2)$  
$解：令u=y^2,原式化为齐次方程du/dx=(3u^2-6x^2)/(2xu+x^2)$  
$再由变换u=tx化为变量分离方程dx/x=(7/5(t-3)+3/5(t+2))dt$  
$解为（y^3-3x)^7(y^3+2x)^3=cx^15$  

$(7)dy/dx=(2x^3+3xy^2+x)/(3x^2y+2y^3-y)$  
$解：方程可改写为d(y^2+1)/d(x^2+1)=[2(x^2-1)+3(y^2+1)]/[3(x^2-1)+2(y^2+1)]$  
$取变换[t=x^2-1,s=y^2+1],化为齐次方程ds/dt=(2t+3s)/(3t+2s)$  
$再令s=wt,得tdw/dt=(2-2w^2)/(3+2w),dt/t=(3+2w)dw/2(1-w^2)=1/4[5/(1-w)+1/(1+w)]dw$  
$积分得-5ln|1-w|/4+ln|1+w|/4=ln|t|+c,(1+w)=ct^4(1-w)^5$  
$即t+s=c(t-s)^5$  
$通解为x^2+y^2=c(x^2-y^2-2)^5,其中c为任意常数$ 

$验证题：设y_1（t）是dy/dt=f(y),y(0)=y_0的解，证明y_2(t)=y1(t-t_0)是dy/dt=f(y),y(t_0)=y_0的解$
