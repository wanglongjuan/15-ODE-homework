## (7)$y=x^2+1$, $y^\prime = y^2 - (x^2+1)y +2x$
解：

 ∵$ y = x^2+1 $,∴$ y^\prime = 2x $

将$ y = x^2+1 $代入$y^\prime = y^2 - (x^2+1)y +2x$

  得：$ y^\prime = 2x $

  ∴$ y = x^2+1 $ 是 $ y\prime = y^2 - (x^2+1)y +2x $ 的解

## (8)$y = - \frac{g(x)}{f(x)}$,$ y = -\frac{f(x)^\prime}{g(x)}y^2-\frac{g(x)^\prime}{f(x)}$
解：

∵$y = - \frac{g(x)}{f(x)}$，∴$ y^\prime = - \frac{f(x)g(x)^\prime-f(x)^\prime g(x)}{f(x)^2} $

将$y = -\frac{g(x)}{f(x)}$代入$ y = -\frac{f(x)^\prime}{g(x)}y^2-\frac{g(x)^\prime}{f(x)} $

得：$ y^\prime = - \frac{f(x)g(x)^\prime-f(x)^\prime g(x)}{f(x)^2} $

∴$ y = -\frac{g(x)}{f(x)} $ 是 $ y = -\frac{f(x)^\prime}{g(x)}y^2-\frac{g(x)^\prime}{f(x)} $ 的解

# 4.给定一阶微分方程$\frac{dy}{dx} = 2x $，
## （1）求出它的通解
解：

将方程改写为$dy= 2xdx $

对方程两边同时积分：$ \int{ydx} =\int 2x{dx}+c $

得：$y=x^2+c$即为含一个任意常数c的通解

## （2）求出通过点（1,4）的特解
解 ：

∵过点（1，4）

∴$ y = 1 $$ x = 4 $ 

则 $  c = - 3 $ 

∴$y = x^2-3 $

## （3）求出与直线$y=2x+3$相切的解
解：

若与直线 $ y = 2x+3 $相切 

则$ 2x+3 = x^2+c $ 

∴$ 4 - 4(c-3) = 0 $即$ c = 3 $ 

∴$ y = x^2+3 $

## （4）求出满足条件$ \int_{0}^{1}{ydx} = 2 $的解
解：

∵$ \int_{0}^{1}{ydx} = 2 $ 

则$  \frac{1}{3} + c = 2 $


∴$ c = \frac{5}{3} $


即$ y = x^2 + \frac{5}{3} $

得$\frac{7}{5}\frac{\mathrm dv}{v-3}+\frac{3}{5}\frac{\mathrm dv}{v+2}=\frac{\mathrm dx}{x}$

两边积分$(v-3)^7(v+2)^3=cx^5$

代回得$(y^3-3x)^7(y^3+2x)^3=cx^{15}$为解
