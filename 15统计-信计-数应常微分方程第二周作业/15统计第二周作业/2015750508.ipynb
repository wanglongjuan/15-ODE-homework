3.验证下列各函数是相应微分方程的解：

(7)$y=x^2+1,y′=y^2-(x^2+1)y+2x$;

解：因为$y=x^2+1$,$y′=x(-2)$

故$y^2-(x^2+1)y+2x=(x^2+1)(x^2+1)-(x^2+1)(x^2+1)+2x=2x=y′$

因此$y=x^2+1$是微分方程的解。

故该函数是该微分方程的解。
