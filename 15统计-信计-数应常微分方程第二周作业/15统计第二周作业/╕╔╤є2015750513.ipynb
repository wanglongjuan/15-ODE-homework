## 验证下列各函数是相应微分方程的解：##
(7) $ y=x^2+1, y’=y^2-(x^2+1)y+2x $
(8)$ y=-\frac{g(x)}{f(x)},y’=\frac{f’(x)}{g(x)}y^2-\frac{g’(x)}{f(x)}$
解：(7)## 因为$y=x^2+1,y’=2x$, ##
## 所以$y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)(x^2+1)+2x=2x=y’$,##
## 因此,$y=x^2+1$是方程$y’=y^2-(x^2+1)y+2x$的解.##
(8)因为$y=-\frac{g(x)}{f(x)}$,所以$y’=-\frac{g’(x)f(x)-g(x)f’(x)}{f(x)^2}=\frac{g(x)f’(x)}{f(x^2)}-\frac{g’(x)}{f(x)}$,
又$\frac{f’(x)}{g(x)}y^2-\frac{g’(x)}{f(x)}=\frac{f’(x)}{g(x)}*(-\frac{g(x)}{f(x)})^2-\frac{g’(x)}{f(x)}$,
$=\frac{g(x)f’(x)}{f(x)^2}-\frac{g’(x)}{f(x)}=y’$,
所以，$y=-\frac{g(x)}{f(x)}$是方程$y'=\frac{f'(x)}{g(x)} y^2-\frac{g’(x)}{f(x)}$的解.

## 求下列方程的解：##
(6)$x*\frac{dy}{dx}-y+\sqrt{x^2+y^2}=0$
(7)$\tan ydx-\cot xdy=0$
解：(6)变形原方程为$xdx-ydy+\sqrt{x^2+y^2}dx=0$.
当$y^2\neq x^2$,容易验证$y^2=x^2$也是方程的解.
$ \frac{xdy-ydx}{\sqrt{x^2+y^2}}+dx=0 $,
即$\frac{d(\frac{y}{x})}{\sqrt{1-(\frac{y}{x})^2}}+\frac{1}{x}dx=0$,两边积分的方程的解为$\arcsin \frac{y}{x}+\ln|x|=c$,
这里c为任意常数.
另外当$y^2=X^2$时，容易验证$y^2=x^2$也是方程的解.
所以原方程的解为$y^2=x^2$或$\arcsin \frac{y}{x}+\ln|x|=c$.
(7)当$\tan y\neq 0$时,即$y \neq k\pi ,k=0,\pm1,\pm2, \cdots$时分离变量得$ \cot ydy=\tan xdx $.
两边同时积分,既得方程的解为$ \ln|\sin y|=-\ln|\cos x|+c1 $,这里c_1为任意常数.
即$\sin y\cos c=c$，
这里$c=e^c_1$为任意正常数.
另外，当 $ y=k\pi,k=0,\pm 1,\pm 2, \cdots $时，也容易验证是方程的解.所以,原方程的解为$\sin y\cos x=c$$y=k\pi, k=0,\pm1,\pm2, \cdots $.

(7)$\frac{dy}{dx}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$
解：令$u=y^3$,则方程可变为$ \frac{dv}{dx}=\frac{3u^2-6x^2}{2xu+x^2} $,
令$v=\frac{u}{x}$,即$u=vx$,则方程可化为$ x\frac{dv}{dx}+v=\frac{3v^2-6}{2v+1} $
即$\frac{7}{5}(\frac{dv}{v-3})\frac{3}{5}(\frac{dv}{v+2}=\frac{dx}{x}$
两边同时积分，得方程的解$(v-3)^7(v+2)^3=cx^5$,
这里c为任意正常数.
因此原方程的解为$(y^3-3x)^7(v+2)^3=cx^5$.
(7)原方程变形为$\frac{ydy}{xdx}=\frac{2x^2+3y^2+1}{3x^2-2y^2-1}$,
即$\frac{dy^2}{dx^2}=2x^2+3y$，
令$u=x^2-1,v=y^2+1$,则上面的方程为$ \frac{dv}{du}+\omega=\frac{2+3\omega }{3+2\omega } $
分离变量并且两边积分，得方程的解$c(1+\omega)=u^4(1- \omega )^5 $

## 第27页第4题##
## 给定一阶微分方程$\frac{dy}{dx}=2x$,##
## (1)求出它的通解;##
## (2)求通过点(1,4)的特解; ##
## (3)求出与直线$y=2x+3$相切的解;##
## (4)求出满足条件$\int_0^1 y {\rm d}x=2$的解;##
## (5)绘出（2）（3）(4)中的解得图形.##
解：(1)由$\frac{dy}{dx}=2x$,得$$dy=2xdx$$
   方程两边积分，即得$y=x^2+c$.其中c为任意常数. 
所以，方程$\frac{dy}{dx}=2x$的通解为 $ y=x^2+c$，c为任意常数。
(2)把$x=1,y=4$代入 $ y=x^2+c$得 $c=3$
则过点（1，4）的特解为$y=x^2+3$
(3)因为与直线相切，所以方程组
$$
\begin{cases}
 y=x^2+c
 y=2x+3
\end{cases}
$$
有且只有唯一解，即$x^2+c=2x+3有唯一解，故 $c=4$.因此，与直线 $y=2x+3$相切的解是$y=x^2+4$ 
(4)因为$$\int_0^1 y {\rm d}x= \int_0^1 (x^2+c) {\rm d}x=\frac{1}{3}+c=2$$, 所以，$c=\frac{5}{3}$因此满足条件$$ \int_0^1 y {\rm d}x=2 $$的解为 $y=x^2+\frac{5}{3}$
