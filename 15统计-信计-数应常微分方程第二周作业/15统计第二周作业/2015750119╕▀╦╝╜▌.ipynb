## 1、验证下列各函数是相应微分方程的解

## （1）$y=x^2+1，y^\prime=y^2-(x^2+1)y+2x$;

## （2）$y=-\frac{g(x)}{f(x)},y^\prime=\frac{f^\prime(x)}{g(x)}y^2-\frac{g^\prime(x)}{f(x)}$.

$y^\prime=2x$

$y^\prime=(x^2+1)^2-(x^2+1)(x^2+1)+2x=2x$

$y^\prime=-\frac{g^\prime(x)f(x)-g(x)f^\prime(x）}{f(x)^2}$

$y^\prime=\frac{f^\prime(x)}{g(x)}(-\frac{g(x)}{f(x)})^2-\frac{g^\prime（x）}{f(x)}=-\frac{g^\prime(x)f(x)-g(x)f^\prime(x）}{f(x)^2}$

证毕.

## 3、求下列方程的解
## （1）$x\frac{dy}{dx}-y+\sqrt(x^2-y^2)=0$;
## （2）$\tan ydx-\cot xdy=0$.

解：

（1）将方程$x\frac{dy}{dx}-y+\sqrt(x^2-y^2)=0$改写为$\frac{dy}{dx}=\frac{y}{x}-\frac{\sqrt(x^2-y^2)}{x}$

以$\frac{y}{x}=u$以及$\frac{dy}{dx}=x\frac{du}{dx}+u$代入，得

$x\frac{du}{dx}=-\sqrt(1-u^2)$

分离变量，得

$-\frac{du}{\sqrt(1-u^2)}=\frac{dx}{x}$

两边积分，得

$-arcsinu=ln|x|+c$，代入$u=\frac{y}{x}$得$-arcsin\frac{y}{x}=ln|x|+c$(c是任意常数)

此外，方程还有解$-\sqrt(1-(u)^2)=0$即$y^2=x^2$

综上，方程的解为$-arcsin\frac{y}{x}=ln|x|+c$ ,$y^2=x^2$


(2)将变量分离得

$\frac{\mathrm d y}{dx}=tanytanx$,$\frac{dy}{tany}=tanxdx$

两边积分得

$lnsiny=-lncosx+c$(c是任意正常数)

此外，方程还有解$tany=0$即$y=kpi$,k是任意整数

即方程的解为$sinycosx=c$，$y=kpi$,k是任意整数

## 5、作适当的变量变换解下列方程

## （1）$\frac{\mathrm d y}{\mathrm d x}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$、

## （2）$\frac{d y}{d x}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$


解：

（1）$\frac{dy}{dx}=\frac{(y^3)^2-2x^2}{y^2(2xy^3+x^2)}=\frac{dy^3}{dx}=\frac{3[(y^3)^2]}{2xy^3+x^2}$

令u=y^3,原方程变成$\frac{du}{dx}=\frac{3u^2-6x^2}{2ux-+x^2}=\frac{3u^2/x^2-6}{2u/x+1}$

令$z=\frac{u}{x}$则$\frac{du}{dx}=z+x\frac{dz}{dx}$

所以$\frac{3z^2-6}{2z+1}=z+x\frac{dz}{dx},x\frac{dz}{dx}=\frac{z^2-z-6}{2z+1}$

当$\frac{z^2-z-6}{2z+1}$等于0，得z=3或z=-2，$y^3=3x$或$y^3=-2x$是方程的解

当$\frac{z^2-z-6}{2z+1}$不等于0，

将变量分离得

$\frac{2z+1}{z^2-z-d}dz=\frac{1}{x}dx$

两边积分得

$(z-3)^7(z+2)^3=x^5c,代入y为（y^3-3x）(y^3+2x)^3=x^15c$
                      
又因为$y^3=3x$或$y^3=-2x$包含在通解中

所以，解为$(y^3-3x)(y^3+2x)^3=x^{15}c$

（2）$\frac{dy}{dx}=\frac{x(2x^2+3y^2+1}{y3x^2+2y^2-1}$,$\frac{dy^2}{dx^2}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$

令$y^2=u,x^2=v$

则$\frac{du}{dx}=\frac{2v+3u+1}{3v+2u-1}$

由方程$2v+3u+1=0$和$3v+2u-1=0$组成的方程组的解为（1，-1）

令$z=v-1,w=u+1,由2z+3w=0和3z+2w=0的方程组，化简为\frac{dw}{dz}=\frac{2+3w/z}{3+2w/z}$

令$t=\frac{w}{z}$,则有$\frac{dw}{dz}=t+z\frac{dt}{dz}$,所以$t+z\frac{dt}{dz}=\frac{2+3t}{3+2t}$,$z\frac{dt}{dz}=\frac{2-2t^2}{3+2t^2}$

当$2-2t^2=0$，即t=-1，得$y^2=x^2-2$或$y^2=-x^2$是方程的解$

所以原方程的解为$(y^2+x^2)c=(y^2-x^2+2)^5$
