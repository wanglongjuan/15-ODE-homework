## 4 作适当的变量变换解下列方程

1.$\frac{\mathrm d y}{\mathrm d x}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$

解：
$\frac{dy}{dx}=\frac{(y^3)^2-2x^2}{y^2(2xy^3+x^2)}=\frac{dy^3}{dx}=\frac{3[(y^3)^2]}{2xy^3+x^2}$

$令u=y^3,原方程变成\frac{du}{dx}=\frac{3u^2-6x^2}{2ux-+x^2}=\frac{3u^2/x^2-6}{2u/x+1}$

$令z=\frac{u}{x}则\frac{du}{dx}=z+x*\frac{dz}{dx}$

$所以：\frac{3z^2-6}{2z+1}=z+x*\frac{dz}{dx},x*\frac{dz}{dx}=\frac{z^2-z-6}{2z+1}$

$当\frac{z^2-z-6}{2z+1}等于0，得z=3或z=-2，y^3=3x或y^3=-2x是方程的解$

$当\frac{z^2-z-6}{2z+1}不等于0，将变量分离得：\frac{2z+1}{z^2-z-d}*dz=\frac{1}{x}*dx$

$两边积分得：(z-3)^7*(z+2)^3=x^5*c,代入y为（y^3-3x）*(y^3+2x)^3=x^5*c$
                      
$又因为y^3=3x或y^3=-2x包含在通解中$

$综上，原方程的解为（y^3-3x）*(y^3+2x)^3=x^5*c$

2.$\frac{d y}{d x}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$
