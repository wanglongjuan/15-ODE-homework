# 1.求下列方程的解

### (6)$x\frac{dy}{dx}-y+\sqrt{x^2-y^2}=0$

**解：**由题可知，$xdy-ydx+\sqrt{x^2-y^2}dx=0$

* 易知$y^2=x^2$时，是方程的解

* 当$y^2\not=x^2$时

方程变形为$$\frac{xdy-ydx}{\sqrt{x^2-y^2}}+dx=0$$

即$$\frac{d(\frac{y}{x})}{\sqrt{1-(\frac{y}{x})^2}}+\frac{1}{x}dx=0$$

两边积分的解为$\arcsin \frac{y}{x}+ln|x|=c$,c为任意常数

综上所述，原方程的解为：$y^2=x^2$，$\arcsin \frac{y}{x}+ln|x|=c$

### (6)$\frac{dy}{dx}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$

**解：**令$u=y^3$,得

$$\frac{du}{dx}=\frac{3u^2-6x^2}{2xu+x^2}$$

令$v=\frac{u}{x}$,得

$$x\frac{dv}{dx}+v=\frac{3v^2-6}{2v+1}$$

的方程的解为$$(v-3)^7(v+2)^3=cx^5$$

取关于x,y的式子，原方程解为：$$(y^3-3x)^7(y^3+2x)=cx^{15}$$  c为任意常数

### (7)$\frac{dy}{dx}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$

**解：**上式可变形为$$\frac{ydy}{xdx}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$$

即$$\frac{dy^2}{dx^2}=2x^2+3y$$

令$u=x^2-1,v=y^2+1$,则有

$$\frac{dv-1}{du+1}=\frac{2u+3v}{3u+2v}$$

令m=$\frac{v}{u}$带入上式并进行积分 得到$$\frac{1}{4}ln(4+4m)-\frac{5}{4}ln(1-m)=ln|u|+c$$

对上式两边取e得指数，并取为关于x,y的式子，得到原方程的解为：$$c(x^2+y^2)=y^2-x^2+2$$
