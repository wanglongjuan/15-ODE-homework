3、验证下列各函数是相应微分方程的解：
(7)$y=x^2+1,y'=y^2-(x^2+1）y+2x$
(8)$y=-\frac{g(x)}{f (x)},y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$

解：(7)对$y=x^2+1$求导得$$ y=2x$$
再将其平方得$$y^2=x^4+2x^2+1$$
将平方代入到微分方程中可得$$y^2-(x^2+1)y+2x=2x=y'$$
证毕

（8）对y进行求导得$$y'=-\frac{g'(x)f(x)-f'(x)g(x)}{f^2(x)}$$
对方程平方得$$y^2=\frac{g^2(x)}{f^2(x)}$$
将平方后的方程代入微分方程得$$\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}=-\frac{g'(x)f(x)-f'(x)g(x)}{f^2(x)}=y'$$
证毕
