(8)$y=-\frac{g(x)}{f(x)},y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$

解:（7）$$y'=2x$$

$$y^2=x^4+2x^2+1$$

代入得$$y^2-(x^2+1)y+2x=2x=y'$$


(8)$$y'=-\frac{g'(x)f(x)-f'(x)g(x)}{f^2(x)}$$

$$y^2=\frac{g^2(x)}{f^2(x)}$$

代入得$$\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}=-\frac{g'(x)f(x)-f'(x)g(x)}{f^2(x)}=y'$$

（6）$x\frac{dy}{dx}-y+\sqrt{x^2-y^2}=0$；

（7）$tanydx-cotxdy=0$；


解：（6）原式变形得$$\frac{dy}{dx}=\frac{y}{x}-\sqrt{1-(\frac{y}{x})^2}$$
当$\frac{y}{x}≠1$时，令$u=\frac{y}{x}$
则上式可化为
$$\frac{du}{\sqrt{1-u^2}}=-\frac{dx}{x}$$
两边分别积分得$$arcsin\frac{y}{x}=ln|x|+c,其中c为任意常数$$

当$\frac{y}{x}=1$时，$y^2=x^2$也是方程的解

则其解为$$y^2=x^2或arcsin\frac{y}{x}=ln|x|+c$$

解：(7)当$y≠kπ$时$$cotydy=tanxdx$$

两边分别积分得$$ln|siny|=-ln|cosx|+c_1$$

其解为$$sinycosx=c,其中c=e^{c_1}为任意常数$$

当$y=kπ$时，其为方程的解

所以方程的解为$$sinycosx=c或y=kπ$$

3、设$y_1(t)$是$\frac{dy}{dx}=f(y),y(0)=y_0的解,$证明$y_2(t)=y_1(t-t_0)是\frac{dy}{dt}=f(y),y(t_0)=y_0的解$。

t为自变量可以为任意值，所以当t取$t-t_0时$

即$y_2=y_1(t-t_0)是方程\frac{dy}{dt}=f(y)的解$

∵$y(0)=y_0,即y中自变量t=0时,y_1(0)=y_0$

∴当$t=t_0$,$y_1(t-t_0)=y_0$，即$y_2(t)=y_0$
