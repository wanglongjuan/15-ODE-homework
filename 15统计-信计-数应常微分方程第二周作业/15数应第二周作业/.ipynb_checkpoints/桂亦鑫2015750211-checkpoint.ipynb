## P27. 3.验证下列各函数是相应微分方程的解

### (7) $ y=x^2+1 , y'=y^2-(x^2+1)y+2x $

### (8) $ y= \frac{-g(x)}{f(x)} , y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)} $

(7)解: 先对$$ y=x^2+1 $$进行求导,可得$$ y'=2x $$ 
   
   再将$$ y=x^2+1 , y'=2x $$两式带入方程$$ y'=y^2-(x^2+1)y+2x $$
   
   有 $$ (x^2+1)^2-(x^2+1)(x^2+1)+2x=2x=y' $$
   
   所以$$ y=x^2+1 $$是$$ y'=y^2-(x^2+1)y+2x $$方程的解 
   


(8)解: 先对$$ y= \frac{-g(x)}{f(x)} $$进行求导,可得$$ y'=\frac{f'(x)g(x)-f(x)g'(x)}{f^2(x)} $$


   再将$$ y= \frac{-g(x)}{f(x)} , y'=\frac{f'(x)g(x)-f(x)g'(x)}{f^2(x)} $$两式带入方程$$ y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$$
   
   有 $$ \frac{f'(x)g(x)}{f^2(x)}-\frac{g'(x)}{f(x)}=\frac{f'(x)g(x)-f(x)g'(x)}{f^2(x)}=y'$$
   
   所以$$ y= \frac{-g(x)}{f(x)} $$是$$ y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$$方程的解 
   


##  P27. 4.给定一阶微分方程 $\frac{dy}{dx} = 2x$
### (1)求出它的通解;
### (2)求出通过(1,4)点的特解;
### (3)求出与直线$y=2x+3$相切的解;
### (4)求出满足条件$\int ^1_0 y=2$的解;
### (5)绘出(2),(3),(4)中的解的图形;

(1)解:对微分方程$$ \frac{dy}{dx} = 2x $$移项得$$ dy=2xdx $$

   对方程$$ dy=2xdx $$两边进行求积分$$\int dy=\int 2xdx $$
   
   可得$$ y=x^2+c （c为任意常数）$$
   

(2)解:由(1)可知$$ y=x^2+c $$为方程通解。将(1,4)代入$$ y=x^2+c $$

   解得$$c=3$$,所以通过(1,4)点的特解为$$ y=x^2+3 $$
 

(3)解:联立方程组$$ \begin{cases}  y = x^2+c\\ y=2x+3\end{cases}$$

   可得$$ x^2-2x+c-3 $$,令$$\delta = 0$$即$$4-4(c-3)=0$$
   
   所以与直线$$y=2x+3$$相切的解为$$ y=x^2+4 $$
   

(4)解:由(1)可知$$ y=x^2+c $$为方程通解,对方程$$ y=x^2+c $$求定积分$$\int ^1_0 x^2+c=\frac{1}{3} x^3+cx\mid^1_0$$

   又知$$\int ^1_0 y=2$$,所以解得$$c=\frac{5}{3}$$
   
   所以满足条件$$\int ^1_0 y=2$$的解为$$y=x^2+\frac{5}{3}$$

##  P42.1. 求下列方程的解

### (6)  $x\frac{dy}{dx}-y+\sqrt{x^2-y^2}=0$

### (7)  $\tan y dx-\cot xdy=0$

(6)解:方程两边同除以x，则方程化为 $$  \frac{dy}{dx}=\frac{y}{x}-\frac{|x|}{x}\sqrt{1-\frac{y^2}{x^2}} （其中$x\neq0$）$$

   令$$y=ux$$,于是$$\frac{dy}{dx}=x\frac{du}{dx}+u$$
   
   将其带入原方程,所以有$$\frac{du}{dx}=\frac{\sqrt{1-u^2}}{x}$$
   
   分离变量即有$$\frac{du}{\sqrt{1-u^2}}=\frac{dx}{x}  （u\neq1或-1）$$
   
   两边同时积分可得$$arcsinu=\ln|x|+c （u\neq1或-1）$$
   
   另外，当$u=1或-1$时，有$$y^2=x^2$$
   
   所以原方程的通解为$$\begin{cases} arcsinu=\ln|x|+c （u\neq1或-1）\\ y^2=x^2\end{cases}$$
   

（7）解：对方程分离变量可得$$\frac{dy}{\tan y}=\frac{dx}{\cot x}$$
  
   对方程两边同时积分可得$$\int \cot ydy=\int \tan ydy$$
   
   即$$\ln|\sin y|=-\ln|\cos x|+c$$
   
   可改写为$$\cos x \sin y=c$$
   
   另外，当$tany=0$时亦满足方程，所以$$y=k\pi (k\in Z)$$也是方程的解。
   
   所以原方程的通解为$$\begin{cases} \cos x \sin y=c\\ y=k\pi (k\in Z)\end{cases}$$
   
   
   
   

## P43.2.作适当的变量变换求解下列方程


### (6)$\frac{dy}{dx}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$

### (7)$ \frac{dy}{dx}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$

（6）解：将原方程的分母因式分解可得$$\frac{dy}{dx}=\frac{y^6-2x^2}{y^2(2xy^3+x^2)}$$
   将分母中提取的$y^2$移项可得$$\frac{y^2dy}{dx}=\frac{y^6-2x^2}{2xy^3+x^2}$$
   将方程右边分子分母同时除以$x^2$可得$$\frac{y^2dy}{dx}=\frac{\frac{y^6}{x^2}-2}{2\frac{y^3}{x}+1}$$
   令$u=\frac{y^3}{x}$可得$$\frac{dy^3}{dx}=\frac{3u^2-6}{2u+1}$$
   又可知 $$\frac{dy}{dx}=x\frac{du}{dx}+u$$
   将其带入$$\frac{dy^3}{dx}=\frac{3u^2-6}{2u+1}$$可得$$x\frac{du}{dx}=\frac{u^2-u-6}{2u+1}$$
   对变化后的方程分离变量有$$\frac{2u+1}{u^2-u-6}du=\frac{dx}{x}$$
   对两边同时积分可得$$(u+2)^3(u-3)^7=cx^5$$
   将$u=\frac{y^3}{x}$代入有$$(\frac{y^3}{x}+2)^3(\frac{y^3}{x}-3)^7=cx^5$$
   化简后得到方程的通解为$$(y^3-3x)^7(y^3+2x)^3=cx^{15}$$

（7）解：提取公因式可得$$\frac{dy}{dx}=\frac{x(2x^2+3y^2+1)}{y(3x^2+2y^2-1)}$$
   亦即$$\frac{dy^2}{dx^2}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$$
   令$a=x^2,b=y^2$即有$$\frac{db}{da}=\frac{2a+3b+1}{3a+2b-1}$$
   联立方程组$$\begin{cases} 2a+3b+1=0 \\ 3a+2b-1=0\end{cases}$$解得其交点为$$(1,-1)$$
   所以令$\begin{cases} X=a-1 \\ Y=b+1\end{cases}$
   则方程变为$$\frac{dY}{dX}=\frac{2X+3Y}{3X+2Y}$$
   令$Y=uX$可得$$\frac{du}{dX}=\frac{\frac{3u+2}{2u+3}-u}{X}$$
   分离变量两边同时积分可得$$ (u-1)^5X^4=c(u+1)$$
   所以得到方程的解为$$(y^2-x^2+2)^5=c(x^2+y^2)$$
   

##  3.设$ y_1(t)$是$\frac{dy}{dt}=f(y),y(0)=y_0$的解.证明$y_2(t)=y_1(t-t_0)$是$\frac{dy}{dt}=f(y),y(t_0)=y_0$的解.


证明:因为$$ y_1(t)$$是$$\frac{dy}{dt}=f(y),y(0)=y_0$$的解.

所以我们有$$ y_1(0)=y_0,y_1'(t)=f(y_1(t))$$

自然的，我们令$t=t'+c$.得$$y_1'(t'+c)=f(y_1(t'+c))$$

明显的$$y_1(t+c)$$也是方程$$\frac{dy}{dt}=f(y)$$的解.

又在题设条件中有$$y(t_0)=y_0$$,所以$$y_2(t_0)=y_1(t_0+c)=y_0$$

当$t_0+c=0$等式成立,得$$c=-t_0$$

所以$$y_2(t)=y_1(t-t_0)$$是$$\frac{dy}{dt}=f(y),y(t_0)=y_0$$的解.

当然方程是否有其他解我们并不确定.
