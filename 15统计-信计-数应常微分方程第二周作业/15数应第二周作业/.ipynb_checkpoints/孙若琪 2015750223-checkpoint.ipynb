$1.求下列方程的解：$

$(6)x\frac{dy}{dx}-y+\sqrt{x^2-y^2}=0$

$两边同时除以x可以得到：\frac{dy}{dx}=\frac{y}{x}-\sqrt{1-(\frac{y}{x})^2}①$

$以\frac{y}{x}=u及\frac{dy}{dx}=x\frac{du}{dx}+u代入，得到x\frac{du}{dx}+u=u-\sqrt{1-u^2}$

$整理得:-\frac{du}{\sqrt{1-u^2}}=\frac{dx}{x}+c$

$两边求积分，得到\arcsin \frac{y}{x}+\ln |x|=c$

$由①可得u^2=1也是方程的解，所以原方程的解还有y=±x$

$(7)\tan ydx-\cot xdy=0即\tan ydx=\cot xdy$

$①\tan y≠0，\tan xdx=\cot ydy,两边积分得到\cos x+\sin y=0$

$②\tan y=0，y=nπ，n=…，-2，-1，0，1，2，…$

$2.(6)令u=y^3,则du=3y^2dy$

$所以\frac{du}{dx}=\frac{3u^2-6x^2}{2xu+x^2}$

$令u=tx,化为变量分离方程：\frac{dx}{x}=(\frac{7}{5}\frac{1}{t-3}+\frac{3}{5}\frac{1}{t+2})dt$

$解得：（y^3-3x）^7(y^3+2x)^3=cx^{15}.$

$(7)方程整理得\frac{d(y^2+1)}{d(x^2-1)}=\frac{2(x^2-1)+3(y^2+1)}{3(x^2-1)+2(y^2+1)}$

$令t=x^2-1,s=y^2+1,可得:\frac{ds}{dt}=\frac{2t+3s}{3t+2s}$

$再令s=wt,得到\frac{dt}{t}=\frac{3+2w}{2(1-w^2)}dw=\frac{1}{4}(\frac{5}{1-w}+\frac{1}{1+w})dw,$

$积分得到-\frac{5}{4}\ln |1-w|+\frac{1}{4}\ln |1+w|=\ln |t|+c,即t+s=c(t-s)^5$

$通解为x^2+y^2=c(x^2-y^2-2)^5$

# 3.设$ y_1(t)$是$\frac{dy}{dt}=f(y),y(0)=y_0$的解.证明$y_2(t)=y_1(t-t_0)$是$\frac{dy}{dt}=f(y),y(t_0)=y_0$的解.
$ 证明：因为y_1(t)是\frac{dy}{dt}=f(y),y(0)=y_0的解.$

$所以 y_1(0)=y_0,y_1'(t)=f(y_1(t))$

$令t=t'+c.得y_1'(t'+c)=f(y_1(t'+c))$

$y_1(t+c)也是方程\frac{dy}{dt}=f(y)的解$

$由条件有y(t_0)=y_0,所以y_2(t_0)=y_1(t_0+c)=y_0$

$当t_0+c=0等式成立,得c=-t_0$

$所以y_2(t)=y_1(t-t_0)是\frac{dy}{dt}=f(y),y(t_0)=y_0的解.$
