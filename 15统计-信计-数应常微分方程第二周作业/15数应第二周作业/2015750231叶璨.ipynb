$$y=x^2+1$$
对$$y=x^2=1$$求导可知
$$y'=2x$$
且$$y^2=x^4+2x^2+1$$
将$$y'=y^2-(x^2+1)y+2x$$中的y用x代换得
$$y'=(x^2+1)^2-(x^2+1)^2+2x$$
$$y'=2x$$
即证



2、$$y=-\frac{g(x)}{f(x)}，y'=\frac{f'(x)y^2}{g(x)}$$
对$$y=-\frac{g(x)}{f(x)}$$求导可得
 $$y'=-\frac{g'(x)f(x)-g(x)f'(x)}{f(x)^2}$$
 又有
  $$\frac{f'(x)y^2}{g(x)}-\frac{g'(x)}{f(x)}=\frac{f'(x)(\frac{-g(x)}{f(x)})^2}{g(x)}-\frac{g'(x)}{f(x)}$$
  $$=\frac{f'(x)[g(x)]^2}{g(x)[f(x)]^2}-\frac{g'(x)}{[f(x)]^2}$$
  $$=-\frac{g'(x)f(x)-g(x)f'(x)}{f(x)^2}$$
  即可验证
   $$所以y=\frac{-g(x)}{f(x)}是y'=\frac{f'(x)y^2}{g(x)}-\frac{g'(x)}{f(x)}微分方程的解$$

二、给定一阶微分方程$$\frac{dy}{dx}=2x$$
（1）求出它的通解
由于$$\frac{dy}{dx}=2x$$
则dx移项可知$$dy=2xdx$$
即$$y=x^2+c$$
（2）求通过点（1,4）的特解
由（1）可知$$y=x^2+c$$
将（1,4）代入可得$$c=3$$
(3)求出与直接$$y=2x+3$$相切的解
由题可知$$y'=2x$$
二次求导可得$$y''=2$$
建立方程组\begin{cases}
x^2+c=2x+3\\
2x=2\\
\end{cases}
可得$$x=1,c=4$$
即$$y=x^2+4$$
(4)求出满足条件$\int_0^1\mathrm y\,\mathrm{d}x=2$的解
由题知$$y=x^2+c$$
代入已知条件得$$\int_0^1\mathrm x^2+c\,\mathrm{d}x=2$$
求积分可知$$\frac{x^3}{3}+cx=2$$
解得$$c=\frac{5}{3}
$$
即$$y=x^2+\frac{5}{3}$$

$$\frac{dy}{dx}=t-\sqrt{1-t^2}$$

3、设$y_1(t)$是$\frac{dy}{dx}=f(y),y(0)=y_0的解,$证明$y_2(t)=y_1(t-t_0)是\frac{dy}{dt}=f(y),y(t_0)=y_0的解$。

解：由题意， $\frac{dy_1(x)}{dx}=f(y_1(x))$

令$x=t-t_0$有

$$\frac{dy_1(t-t_0)}{dx}=f(y_1(t-t_0))$$

即$$\frac{dy_2(t)}{dx}=f(y_2(t))$$

则$y_2(t)=y_1(t-t_0)也是方程\frac{dy}{dt}=f(y)的解$

由题意$y_1(0)=y_0$

当$t=t_0$,$$y_2(t_0)=y_0$$

$y_2(t)=y_1(t-t_0)是y(t_0)=y_0的解$。
