1.验证下列各函数是相应微分方程的解。

(1)$y=-\frac{1}{x}，x^2y'=x^2y^2+xy+1$

解：由于$$y'=2x，y=-\frac{1}{x}$$,将其代入第二式得$$x^2y'=x^2y^2+xy+1$$

所以验证得此函数是该微分方程的解

(2)$y=-\frac{g(x)}{f(x)},y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$


解：由于$$y'=\frac{f'(x)g(x)}{f^2(x)}-\frac{g'(x)}{f(x)},y^2=\frac{g^2(x)}{f^2(x)}$$,将其代入第二式得$$y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$$

所以验证得此函数是该微分方程的解

3.求下列方程的解

(1)$x\frac{dy}{dx}-y+\sqrt{x^2-y^2}=0$


解：原方程可化为$$xdy-ydx+\sqrt{x^2-y^2}dx=0$$

当${y^2}\neq{x^2}$时，两边同时除以$$\sqrt{x^2-y^2}$$得$$\frac{xdy-ydx}{\sqrt{x^2-y^2}}+dx=0$$

两边积分得$$arcsin\,\frac{y}{x}+ln\,x=c$$,这里c为任意常数。

另外，当$y^2=x^2$时，容易验证$y^2=x^2$也是方程的解。

所以，原方程的解为$$y^2=x^2$$或$$arcsin\,\frac{y}{x}+ln\,x=c$$



(2)$tanydx-cotxdy=0$


解：当$${tany}\neq{0}$$时即$${y}\neq{k\pi},k=0,\pm:1,2,......$$时分离变量得$$cotydy=tanxdx$$两边同时积分得$$ln|siny|=-ln|cosx|+c_1$$这里c1为任意常数.即$$sinycosx=c$$这里$c=e^c1$为任意常数.

另外，当$y=k\pi,k=0,\pm:1,2,......$时，容易验证也是方程的解

所以，原方程的解为$$sinycosx=c$$或$$y=k\pi,k=0,\pm:1,2,......$$
