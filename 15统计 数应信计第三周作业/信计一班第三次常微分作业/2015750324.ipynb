（4）$\frac{dy}{dx}-\frac{n}{x}y=e^xx^n$

（15）$\frac{dy}{dx}=\frac{1}{xy+x^3y^3}$

（16）$y=e^x+\int_0^x y(t)dt$

（4）

解：

因为$P(x)=\frac{n}{x},Q(x)={e^x}\cdot{x^n}$

所以
$$
\begin{aligned}
y& =e^{\int\frac{n}{x}dx}(\int{e^x}\cdot{x^n}\cdot{e^{-\int{\frac{n}{x}dx}}\cdot{dx}+c})\\
& =x^n(\int{e^x}\cdot{x^n}\cdot{x^{-n}}\cdot{dx}+c)\\
& ={(e^x+c)}\cdot{x^n}
\end{aligned}
$$

（15）

解：

令$u=y^2$,则原方程为
$$
\frac{du}{dx}=\frac{2}{x+x^3u}
$$

将u看做自变量，x看做函数变量，则方程可变为
$$
\frac{dx}{du}=\frac{x}{2}+\frac{u}{2}x^3
$$

即为$n=3$时的伯努利方程。

令$z=x^{-2}$

得$$
\frac{dz}{du}={-2}\cdot{x^{-3}}\cdot{\frac{dx}{du}}
$$

代入$\frac{dx}{du}$的表达式，得到
$$
\frac{dz}{du}=-z-u
$$

可得到通解
$$
z=e^{\int{-1du}}(\int{(-u){e^{\int{{1}\cdot{du}}}}}\cdot{du}+c)
$$

即
$$
z={c}\cdot{e^{-u}}-u+1
$$

把$z=x^{-2},u=y^2$代入得
$$
\frac{z}{x^2}={c}\cdot{e^{-y^2}}-y^2+1
$$

即为原方程的通解。

（16）

解：

对方程$y=e^x+\int_0^xy(t)dt$两边关于x求导

得
$$
\frac{dy}{dx}=e^x+y
$$

可得通解
$$
y=e^x(\int{e^x}\cdot{e^{-x}}\cdot{dx}+c)
$$

即
$$
y=e^x(x+c)
$$

当$x=0$时，由原方程得
$$
y=e^0+\int_0^0y(t)dt=1
$$

可求出
$$
c=1
$$

因此原方程的解为
$$
y=e^x(x+1)
$$

7.求解下列方程：

（3）${y'\sin{x}}\cdot{\cos{x}}-y-\sin^3{x}=0$

解：

此方程可变形为
$$
\frac{dy}{dx}=\frac{1}{\sin{x}\cos{x}}y+\frac{\sin^2{x}}{\cos{x}}\ \ （\sin{x}\neq0,\cos{x}\neq0）
$$

可得通解
$$
y=e^{\int\frac{1}{\sin{x}\cos{x}}dx}({\int\frac{\sin^2{x}}{\cos{x}}\cdot{e^{-\int\frac{1}{\sin{x}\cos{x}}}dx}}\cdot{dx}+c)
$$

即
$$
y=\tan{x}(\int{\frac{\sin^2{x}}{\cos{x}}}\cdot{\cot{x}dx}+c)
$$

化简得
$$
y=\tan{x}(-\cos{x}+c)
$$

当$\sin{x}=0$时，原方程化为$y=0$.

当$\cos{x}=0$时，原方程化为$y=-\sin^3{x}=\pm1$

综上，原方程的解为
$$
y={c}\cdot{\tan{x}}-\sin{x}\ \ (x\neq{kπ}\pm\frac{π}{2})
$$

1.验证下列方程是恰当微分方程，并求出方程的解：

（1）$(x^2=y)dx=(x-2y)dy=0$

（2）$(y-3x^2)dx-(4y-x)dy=0$

（1）

解：

由$M=X^2+y,N=x-2y$

得到
$$
\frac{\partial M}{\partial y}=1,\ \ \frac{\partial N}{\partial x}=1
$$

此方程为恰当微分方程，令u满足：
$$
\frac{\partial u}{\partial x}=x^2+y,\ \ \frac{\partial u}{\partial y}=x-2y
$$

第一个方程关于x积分，得到
$$
u=\frac1 3x^3+xy+\phi(y)
$$

两边关于y求导，得
$$
\frac{\partial u}{\partial y}=x+\phi'(y)
$$

与$\frac{\partial u}{\partial y}=x-2y$比较得到
$$
\phi{y}=-2y
$$

积分后得到
$$
\phi(y)=-y^2
$$

所以
$$
u=\frac1 3x^3+xy-y^2
$$

即得到通解
$$
\frac1 3x^3+xy-y^2=c
$$

其中，c为任意常数

（2）

解：

由$M=y-3x^2,N=x-4y$

得到
$$
\frac{\partial M}{\partial x}=1,\frac{\partial N}{\partial x}=1
$$

其方程为恰当微分方程，令u满足：
$$
\frac{\partial u}{\partial x}=y-3x^2,\ \ \frac{\partial u}{\partial y}=x-4y
$$

对第一式关于x求积分，得到
$$
u=xy-x^3+\phi(y)
$$

并关于y求导数，代入第二式，得到
$$
\frac{\partial u}{\partial y}=x+\phi'(y)=x-4y
$$

得到
$$
\phi(y)=-4y
$$

积分后得到
$$
\phi(y)=-2y^2
$$

代入上式，得到
$$
u=xy-x^3-2y^2
$$

因此，得到通解
$$
xy-x^3-2y^2=c
$$

其中，c是任意常数
