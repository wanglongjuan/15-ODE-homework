## 1、求下列方程的解。 ##

# （4）$\frac{dy}{dx}-\frac{n}{x}y=e^xx^n$ #

# （15）$\frac{dy}{dx}=\frac{1}{xy+x^3y^3}$ #

# （16）$y=e^x+\int_0^x\mathrm y(t)\,\mathrm{d}t$ #

解：（4）先求齐次方程$\frac{dy}{dx}-\frac{n}{x}y=0$的通解

因为$\frac{dy}{dx}=\frac{n}{x}y$,所以$\frac{dy}{y}=n\frac{dx}{x}$

两边积分，得$y=c^xn$

设所求方程的通解为$y=c(x)x^n$

求导后代入原方程，得$\frac{dy}{dx}=\frac{d(c(x))}{dx}x^n=e^xx^n$

$c(x)=e^x+c_1$

原方程的通解为
$$y=e^xx^n+c_1x^n(其中c_1为任意常数)$$

（15）将原方程改写为$y\frac{dy}{dx}=\frac{1}{x+x^3y^2}$

令$u=y^2,则\frac{du}{dx}=\frac{2}{x+x^3u}$

视u为自变量，x为因变量，则$\frac{dx}{du}=\frac{x}{2}+\frac{u}{2}x^3$

两边同时乘以${x}^{-3},得{x}^{-3}\frac{dx}{du}=\frac{{x}^{-2}}{2}+\frac{u}{2}$

令$z={x}^{-3},则\frac{dz}{du}=-z-u$

应用一阶线性方程的求解，代回原变量，则原方程的解为
$$(1-x^2+x^2y^2){e}^{y^2}=cx^2(其中c为任意常数)$$

（16）对方程$y=e^x+\int_0^x\mathrm y(t)\,\mathrm{d}t两边关于x求导，得\frac{dy}{dx}=e^x+y$

先求齐次方程$\frac{dy}{dx}-y=0$的通解

$\frac{dy}{y}=dx$

$y=e^xc$

设所求方程的通解为$y=c(x)e^x$

求导后代入原方程得 $\frac{dy}{dx}=\frac{d(c(x))}{dx}=e^x$

$c(x)=x+c_1$

所以原方程的解为
$$y=e^x(x+c_1)(其中c_1为任意常数)$$

## 7、求下列方程的解##

# （3）$y'\sin x\cos x-Y-\sin^3x=0$#

解：（3）化简移项，得：

$\frac{dy}{dx}=\frac{1}{\sin x\cos x}y+\frac{\sin^2 x}{\cos x}$

先求其次线性微分方程$\frac{dy}{dx}-\frac{1}{\sin x\cos x}y=0$的通解：

$y=c\tan x.$

$应用常数变易法求非齐次线性微分方程的通解：$

$y=c(x)\tan x$

两边求导，有：

$\frac{dy}{dx}=\frac{dc(x)}{dx}\tan x+c(x)\sec^2 xc(x)$

联立以上方程有$\frac{1}{\sin x\cos x}c(x)\tan x+\frac{\sin^2 x}{\cos x}=\frac{dc(x)}{dx}\tan x+c(x)\sec^2 xc(x)$

化简，得：

$\frac{dc(x)}{dx}=\sin x$

两边求积分，有：

$c(x)=-\cos x$

所以原方程的解为：

$y=c\tan x-\sin x$

当$\sin x=0$时，原方程为y=0.

当$\cos x=0$时，原方程化为$y=±1$，曲线退化成$（2kπ+π/2,-1)和（2kπ-π/2,1)，k=0，±1，±2.....$

综上所述，原方程的解为:

$$y=c\tan x-\sin x,(x≠kπ±π/2),c为任意常数，或者点（2kπ+π/2,-1)，（2kπ-π/2,1)，k=0，±1，±2.....$$

## 1、验证下列方程是恰当微分方程，并求出方程的解:##

# (1)、$(x^2+y)dx+（x-2y)dy=0$ #

# (2)、$(y-3x^2)dx-(4y-x)dy=0$ #

解：

$M=x^2+y, N=x-2y$

因为$\frac{∂M}{∂y}=1,\frac{∂N}{∂x}=1$,

$\frac{∂M}{∂y}=\frac{∂N}{∂x}$

所以方程是恰当微分方程.

把方程重新“分项组合”，得到：

$x^2dx-2ydy+(ydx+xdy)=0$

$d(\frac{x^3}{3}+xy-y^2)=0$

所以方程的通解为：

$$\frac{x^3}{3}+xy-y^2=c (其中c为任意常数）$$

(2)$M=y-3x^2 N=x-4y$

因为$\frac{∂M}{∂y}=1,\frac{∂N}{∂x}=1$

$\frac{∂M}{∂y}=\frac{∂N}{∂x}$

所以方程是恰当微分方程.

把方程重新“分项组合”，得到：

$(xdy+ydx)-3x^2dx-4ydy=0$

$d(xy-x^3-2y^2)=0$

所以方程的通解为：

$$xy-x^3-2y^2=c (其中c为任意常数）$$

对$P(t)=\int_{x_0}^x \mathrm p(t)\,\mathrm{d}t$求导,

解：

由区间的可加性，有：

$P(x+△x)-P(x)=\int_{x_0}^{x+△x} \mathrm p(t)\,\mathrm{d}t-\int_{x_0}^x \mathrm p(t)\,\mathrm{d}t=\int_x^{x+△x} \mathrm p(t)\,\mathrm{d}t$

由积分第一中值定理，有：

$P(x+△x)-P(x)=η△x或P(ξ)△x$

不管在哪种情况下，当$△x→0时都有P(x+△x)-P(x)→0，即P(x)在（x_0,x)$上连续。

若$P(x)在（x_0,x)上连续，当△x→0时有ξ→x,因而P(ξ)-P(x)$，于是;

$$P'(x)=lim\frac{P(x+△x)-P(x)}{△x}=limP(ξ)=P(x).$$
