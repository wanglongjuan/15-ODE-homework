**2.solve following equation**  
**(8)$(x+2y)\mathrm d x+x\mathrm d y=0$**  
**answer:**suppose $\frac{y}{x}=t$  
so $\mathrm d y=x\mathrm d t+t\mathrm d x$  
both sides of the equation multiply $\frac{1}{x}$ and substitute $\frac{y}{x}$ with $t$  
\begin{align}
(1+2\frac{y}{x})\mathrm d x+\mathrm d y&=0\\
(1+2t)\mathrm d x+x\mathrm d t+t\mathrm d x&=0\\
(1+3t)\mathrm d x+x\mathrm d t&=0\\
\frac{\mathrm d x}{x}&=-\frac{\mathrm d t}{1+3t}\\
\mathrm{ln} x&=-\frac{1}{3}\mathrm{ln}(1+3t)+c\\
x^3&=c\frac{1}{1+3t}
\end{align}
substitute $t$ with $\frac{y}{x}$,we got  
\begin{align}
x^3&=c\frac{1}{1+\frac{3y}{x}}\\
y&=\frac{c}{3x^2}-\frac{x}{3}\\
\end{align}
that is the solution    
**(9)$[xcos(x+y)+sin(x+y)]\mathrm d x+xcos(x+y)\mathrm d y=0$**  
**answer:**
\begin{align}
[xcos(x+y)+sin(x+y)]\mathrm d x+xcos(x+y)\mathrm d y&=0\\
xcos(x+y)(\mathrm d x+\mathrm d y)+sin(x+y)\mathrm d x&=0\\
xcos(x+y)\mathrm d (x+y)+sin(x+y)\mathrm d x&=0\\
x\mathrm d (sin(x+y))+sin(x+y)\mathrm d x&=0\\
\mathrm d(xsin(x+y))&=0\\
xsin(x+y)&=c
\end{align}
so $xsin(x+y)=c$ is the solution  
**(10)$(ycosx-xsinx)\mathrm d x+(ysinx+xcosx)\mathrm d y=0$**  
**answer:**$M=ycosx-xsinx$,$N=ysinx+xcosx$  
$\frac{\partial M}{\partial y}\neq\frac{\partial N}{\partial x}$  
because $$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{-M}=1$$  
the integral factor is $\mu=e^{\int\mathrm d y}=e^y$  
both sides of the equation multiply $\mu$  
$e^y(ycosx-xsinx)\mathrm d x+e^y(ysinx+xcosx)\mathrm d y=0$  
this is an exact ODE,the solution is  
$\int {e^y(ycosx-xsinx)\mathrm d x}+\int{(N-\frac{\partial}{\partial y}\int{e^y(ysinx+xcosx)\mathrm d y)}}=0$  
that is $e^ysinx(y-1)+e^ycosx=c$

**4.suppose $f(x,y)$ and $\frac{\partial f}{\partial y}$ are continuous,certifacate that the necessary and suffcient condition of that the equation $\mathrm d y-f(x,y)\mathrm d x=0$ is a liner differential equation is that the euqation has an integral factor which only depends on $x$.**  
**answer:**  
necessary:if the equation is a liner differential equation,it will be
$\frac{\mathrm d y}{\mathrm d x}=P(x)y+Q(x)$  
this equation has the integral factor:$\mu(x)=e^{-\int{P(x)\mathrm d x}}$,and $\mu(x)$ only depends on x  
sufficent:if the euqation has the integral factor $\mu(x)$ which only depends on $x$,  
the equation $\mu(x)\mathrm d y-\mu(x)f(x,y)\mathrm d x=0$ is an exact ODE.  
so $\frac{\partial(-\mu(x)f(x,y))}{\partial y}=\frac{\mathrm d (\mu(x))}{\mathrm d x}$,
$\frac{\partial f}{\partial y}=-\frac{\mu^\prime(x)}{\mu(x)}$  
$f=-\int{\frac{\mu^\prime(x)}{\mu(x)}\mathrm d y}+Q(x)=-\frac{\mu^\prime(x)}{\mu(x)}y+Q(x)=P(x)y+Q(x)$  
so the equation can be change into a first order linear equation.  
**8.solve the integral factor of the bernoulli equation **  
**answer:**the form of the bernoulli equation is $\frac{\mathrm d y}{\mathrm d x}=P(x)y+Q(x)y^n,y\neq0$  
both sides of the equation multiply $\frac{1}{y^n}$,suppose $z=\frac{1}{y^{n-1}}$    
$\frac{\mathrm d z}{\mathrm d x}=(1-n)P(x)z+(1-n)Q(x)$  
this equation has the integral factor:$\mu=e^{\int{-(1-n)P(x)\mathrm d x}}=e^{(n-1)\int{P(x)\mathrm d x}}$  
so the integral factor of the original equation is $\mu=e^{(n-1)\int{P(x)\mathrm d x}}$

**1.solve following equation**  
**(1)$x(\frac{\mathrm d y}{\mathrm d x})^3=1+\frac{\mathrm d y}{\mathrm d x}$**  
**answer:**suppose $\frac{\mathrm d y}{\mathrm d x}=p=\frac{1}{t}$  
so $x(\frac{1}{t})^3=1+\frac{1}{t}$,$x=(1+\frac{1}{t})t^3=t^3+t^2$ then
\begin{align}
y&=\int{p\mathrm d x}+c\\
&=\int{\frac{1}{t}\mathrm d (t^3+t^2)}+c\\
&=\int{(3t+2)\mathrm d t}+c\\
&=\frac{3}{2}t^2+2t+c\\
\end{align}
so the solution of parameter form is
\begin{cases}
x=t^3+t^2\\
y=\frac{3}{2}t^2+2t+c
\end{cases}
**(2)$(\frac{\mathrm d y}{\mathrm d x})^3-x^3(1-\frac{\mathrm d y}{\mathrm d x})=0$**  
**answer:**suppose $\frac{\mathrm d y}{\mathrm d x}=p=tx$  
so $(tx)^3-x^3(1-tx)=0$,$x=\frac{t^3-1}{t}=t^2-\frac{1}{t}$ then
\begin{align}
y&=\int{p\mathrm d x}+c\\
&=\int{t(t^2-\frac{1}{t})\mathrm d (t^2-\frac{1}{t})}+c\\
&=\int{(t^3-1)(2t+\frac{1}{t^2})\mathrm d t}+c\\
&=\int{(2t^4-t-\frac{1}{t^2})\mathrm d t}+c\\
&=\frac{2}{5}t^5-\frac{1}{2}t^2+\frac{1}{t}+c\\
\end{align}
so the solution of parameter form is
\begin{cases}
x=t^2-\frac{1}{t}\\
y=\frac{2}{5}t^5-\frac{1}{2}t^2+\frac{1}{t}+c
\end{cases}
**(3)$y=(\frac{\mathrm d y}{\mathrm d x})^2e^{\frac{\mathrm d y}{\mathrm d x}}$**  
**answer:**suppose $\frac{\mathrm d y}{\mathrm d x}=p$ then $y=p^2e^p$  
so 
\begin{align}
x&=\int{\frac{1}{p}\mathrm d (p^2e^p)}+c\\
&=\int{\frac{1}{p}(2pe^p+p^2e^p)\mathrm d p}+c\\
&=\int{(2e^p+pe^p)\mathrm d p}+c\\
&=(1+p)e^p+c
\end{align}
so the solution of parameter form is
\begin{cases}
x=(1+p)e^p+c\\
y=p^2e^p
\end{cases}
in addition,$y=0$ is also the solution.  
**(4)$y(1+\frac{\mathrm d y}{\mathrm d x})=2a$("a" is a constant.)**  
**answer:**suppose $\frac{\mathrm d y}{\mathrm d x}=tan\phi$ 
then $y=\frac{2a}{1+tan^2\phi}=\frac{2a}{sec^2\phi}=2acos^2\phi$  
so
\begin{align}
x&=\int{\frac{1}{tan\phi}\mathrm d (2acos^2\phi)}+c\\

\end{align}

\begin{cases}
x=-a(2\phi+sin2\phi)+c\\
y=2acos^2\phi
\end{cases}
**(5)$x^2+(\frac{\mathrm d y}{\mathrm d x})^2=1$**  
**answer:**suppose $\frac{\mathrm d y}{\mathrm d x}=cos\phi$
then $x^2+(cos\phi)^2=1$,$x=\sqrt{1-cos^2\phi}=sin\phi$  
so 
\begin{align}
y&=\int{cos\phi\mathrm d (sin\phi)}+c\\
&=\int{cos^2\phi\mathrm d \phi}+c\\
&=\int{\frac{1+cos2\phi}{2}\mathrm d \phi}+c\\
&=\frac{1}{2}\phi+\frac{1}{4}sin2\phi+c
\end{align}
so the solution of parameter form is
\begin{cases}
x=sin\phi\\
y=\frac{1}{2}\phi+\frac{1}{4}sin2\phi+c
\end{cases}
**(6)$y^2(\frac{\mathrm d y}{\mathrm d x}-1)=(2-\frac{\mathrm d y}{\mathrm d x})^2$**  
**answer:**suppose $2-\frac{\mathrm d y}{\mathrm d x}=yt$,
then $1-\frac{\mathrm d y}{\mathrm d x}=yt-1$,
$y=t+\frac{1}{t}$  
so 
\begin{align}
\mathrm d x&=\frac{\mathrm d y}{y^\prime}\\
&=\frac{t+\frac{1}{t}}{2-t(t+\frac{1}{t})}\\
&=\frac{(1-t^{-2})\mathrm d t}{1-t^2}\\
&=\frac{t^2-1}{t^2(1-t^2)}\mathrm d t\\
&=-\frac{1}{t^2}\mathrm d t
\end{align}
so $x=\int{(-\frac{1}{t^2})\mathrm d t}+c=\frac{1}{t}+c$  
so the solution of parameter form is
\begin{cases}
x=\frac{1}{t}+c\\
y=y=t+\frac{1}{t}
\end{cases}
the general solution is $y=\frac{1}{x-c}+x-c$
