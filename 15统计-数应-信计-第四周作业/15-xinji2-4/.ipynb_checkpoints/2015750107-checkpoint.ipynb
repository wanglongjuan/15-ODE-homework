# 习题2.3

2.求下列方程的解： 

（8）(x+2y)dx+xdy=0;  

（9）$[x\cos (x+y)+\sin (x+y)]dx+\cos (x+y)dy=0$;  

（10）$（y\cos x-x\sin x）dx+(y\sin x+x\cos x)dy$;  

（11）$x(4ydx+2xdy)+y^3(3ydx+5xdy)=0$.  

解：

（8）

$M=x+2y,N=x$

$\frac{\partial{M}}{\partial{y}}=2$,$\frac{\partial{N}}{\partial{x}}=1$

$\frac{\partial{M}}{\partial{y}}\neq\frac{\partial{N}}{\partial{x}}$，方程是不恰当的

$\frac{\frac{\partial{M}}{\partial{y}}-\frac{\partial{N}}{\partial{x}}}{N}=\frac{1}{x}$

有积分因子：$μ=e^{\int\frac{1}{x}dx}=x$

则：$2xydx+x^2dy+x^2dx=0$

即：$d(x^2y+\frac{1}{3}x^3)=0$

方程的通解为：$x^3+3x^2y=c$

（9）

$M=x\cos (x+y)+\sin (x+y),N=x\cos (x+y)$

$\frac{\partial{M}}{\partial{y}}=\frac{\partial{N}}{\partial{x}}=\cos (x+y)-x\sin (x+y)$

又方程可写为：$d(x\sin (x+y))=0$

方程的通解为：$x\sin (x+y)=c$

（10）

$M=y\cos x-x\sin x$,$N=y\sin x+x\cos x$

$\frac{\partial{M}}{\partial{y}}=\cos x,\frac{\partial{N}}{\partial{x}}=y\cos x+\cos x-x\sin x$，方程不是恰当的

$\frac{\frac{\partial{M}}{\partial{y}}-\frac{\partial{N}}{\partial{x}}}{M}=1$

有积分因子：$μ=e^{\int1dy}=e^y$

则：$(ye^y\cos xdx+ye^y\sin xdy)+(e^y\cos xdy-e^yx\sin xdx)=0$

即：$d[(y-1)e^y\sin x+e^yx\cos x]=0$

方程的通解为：$x\cos x+(y-1)\sin x=ce^{-y}$

（11）

$x(4ydx+2xdy)=d(2x^2y)$

对于方程:$y^3(3ydx+5xdy)=0$

有:$M=3y^4,N=5xy^3,\frac{\partial{M}}{\partial{y}}=12y^3,\frac{\partial{N}}{\partial{x}}=5y^3$

$\frac{\frac{\partial{M}}{\partial{y}}-\frac{\partial{N}}{\partial{x}}}{5xy^3}=\frac{7}{5x}$

有积分因子：$μ=e^{\int\frac{7}{5x}dx}=x^{\frac{7}{5}}$

则：$3y^4x^{\frac{7}{5}}dx+5x^{\frac{12}{5}}y^3dy=0$

即：$d(\frac{5}{4}x^{\frac{12}{5}}y^4)=0$

综上，原方程的通解为：$8x^2y+5x^{\frac{12}{5}}y^4=c$

4.设f(x,y)及$\frac{\partial{f}}{\partial{y}}$连续，试证方程dy-f(x,y)dx=0为线性微分方程的充要条件是它有仅依赖于x的积分因子。

证明：

充分性：设dy-f(x,y)dx=0有仅依赖于x的积分因子，

因为积分因子依赖于x，所以存在某个$F(x)$满足：

$\frac{\frac{\partial{M}}{\partial{y}}-\frac{\partial{N}}{\partial{x}}}{N}=\frac{-\frac{\partial{f}}{\partial{y}}-0}{1}=F(x)$

$f(x,y)=-F(x)y+h(x)$  ($h(x)$是关于x的任意可微函数)

原方程可写为：$\frac{dy}{dx}=-F(x)y+h(x)$

则$dy-f(x,y)dx=0$是线性的  

必要性：设$dy-f(x,y)dx=0$是线性微分方程

即,存在g(x),h(x)

使f(x,y)=yg(x)+h(x)

$M=-f(x,y)=-yg(x)-h(x),N=1$

$\frac{\frac{\partial{M}}{\partial{y}}-\frac{\partial{N}}{\partial{x}}}{N}=\frac{-g(x)}{1}=-g(x)$

所以，方程有积分因子：$μ=e^{\int-g(x)dx}$  

则证明了方程有仅依赖于x的积分因子

8.求出伯努利微分方程的积分因子。

解:

伯努利方程为：$\frac{dy}{dx}=p(x)y+q(x)y^n(n≠0,1)$

化为：$dy-p(x)ydx-q(x)y^ndx=0$

两边乘以$y^{-n}$

得，$y^{-n}dy-p(x)y^{1-n}dx-q(x)dx=0$

即：$d(y^{1-n})-(1-n)p(x)y^{1-n}dx-(1-n)q(x)dx=0$

再乘以$e^{-(1-n)\int p(x)dx}$

得，$e^{-(1-n)\int p(x)dx}[d(y^{1-n})-(1-n)p(x)y^{1-n}dx]-e^{-(1-n)\int p(x)dx}(1-n)q(x)dx=0$

即：$d[y^{1-n}e^{-(1-n)\int p(x)dx}]-d[\int(1-n)q(x)e^{-(1-n)\int p(x)dx}(1-n)q(x)dx]=0$

这是全微分方程，

因此，所求积分因子为：$μ=y^{-n}e^{(n-1)\int p(x)dx}$

# 习题2.4

1.求解下列方程：

（1）$xy'^3=1+y'$

（2）$y'^3-x^3(1-y')=0$

（3）$y=y'^2e^{y'}$

（4）$y(1+y'^2)=2a(a为常数）$

（5）$x^2+y'^2=1$

（6）$y^2(y'-1)=(2-y')^2$

解：

（1）

令$\frac{dy}{dx}=p$

则，$x=\frac{1}{p^3}+\frac{1}{p^2}\neq0$

对y求导得，$(3+2p)dp+p^3dy=0$

对其积分得，$y=\frac{3}{2p^2}+\frac{2}{p}+c$

设$t=\frac{1}{p}$

方程通解为

$
\begin{cases}
x=t^3+t^2\\
y=\frac{1}{2t^2}+2t+c
\end{cases}
$

（2）
令y'=tx

原方程化为：$x^3(t^3-1+tx)=0$

得，$x=\frac{1}{t}-t^2$

两边对y求导、变形得，$dy=[(1-t^3)(-\frac{1}{t^2}-2)]dt$

对上式积分得，$y=-\frac{1}{2}t^2+\frac{2}{5}t^5+\frac{1}{t}+c$

方程的通解为：
$
\begin{cases}
x=\frac{1}{t}-t^2\\
y=-\frac{1}{2}t^2+\frac{2}{5}t^5+\frac{1}{t}+c
\end{cases}
$  

（3）

令y'=p,

原方程化为：$y=p^2e^p$

对x求导、变形得，$dx=(p+2)e^pdp$

对上式积分即得，$x=(p+1)e^p+c$

方程的通解为：
$
\begin{cases}
x=(p+1)e^p+c\\
y=p^2e^p
\end{cases}
$  

（4）

令$y'=\tan t$

原方程化为：$y=2a\cos^2t$

两端对x求导、变形得，$dx=-4acos^2tdt$

对其积分得，$x=-a(2t+\sin 2t+c)$

方程的通解为：
$
\begin{cases}
x=-a(2t+\sin 2t+c)\\
y=2acos^2t
\end{cases}
$  

（5）

令$y'=\cos t$

原方程化为：$x^2+\cos^2t=1$

$x=\sin t$

对y求导、变形得，$dy=\cos^2tdt$

对上式积分得，$y=\frac{t}{2}+\frac{1}{4}\sin 2t+c$

方程的通解为:
$
\begin{cases}
x=\sin t\\
y=\frac{t}{2}+\frac{\sin 2t}{4}+c
\end{cases}
$

（6）

令$2-y'=yt$

原方程化为:$y^2(1-yt)=y^2t^2$

$y=\frac{1}{t}-t$

又$y'=1+t^2$

这是原微分方程的参数形式，

因此，$dx=\frac{dy}{y'}=\frac{(-\frac{1}{t^2}-1)dt}{1+t^2}=-\frac{1}{t^2}dt$

对其积分得，$x=\frac{1}{t}+c$

方程的参数形式的通解为：$\begin{cases}x=\frac{1}{t}+c\\y=\frac{1}{t}-t.\end{cases}$

消参得，$y=x-\frac{1}{x-c}-c  (c为任意常数)$

此外，当y'=0时，

原方程变为$-y^2=4$

此时方程无解
