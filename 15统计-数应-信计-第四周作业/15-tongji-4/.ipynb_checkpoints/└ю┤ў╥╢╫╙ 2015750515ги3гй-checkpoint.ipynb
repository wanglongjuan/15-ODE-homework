## 二、求下列方程的解 
**（8）**$（x+2y）dx+xdy=0$;<br>
解：这里$M=x+2y,N=x$<br>
$ \frac{\partial M}{\partial y}=2,\frac{\partial N}{\partial x}=1 $,方程不是恰当的。<br>
$ \frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac{1}{x} $<br>
所以方程有积分因子$ \mu=e^\int 1/x{\rm d}x=x $.以$ \mu=x^-2 $乘以方程两边得到$ 2xydx+x^2dy+x^2dx=0 $<br>
即$ d(x^2y+\frac{1}{3}x^3)=0 $,于是，方程的通解为$ x^3+3x^2y=c $.<br>
**（9）**$[xcos(x+y)+sin(x+y)]dx+xcos(x+y)dy=0;$<br>
解：这里$M=xcos(x+y)+sin(x+y),N=xcos(x+y)$<br>
$ \frac{\partial M}{\partial y}=-xsin(x+y)+cos(x+y) $,$ \frac{\partial N}{\partial x}=cos(x+y)-xsin(x+y) $,方程是恰当的<br>
并且方程可以写成$ d(xsin(x+y))=0 $<br>
于是方程的通解为$ xsin(x+y)=c $.<br>
**（10）**$（ycosx-xsinx）dx+(ysinx+xcosx)dy=0$<br>
解：这里$M=ycosx-xsinx,N=ysinx+xcosx$<br>
$ \frac{\partial M}{\partial y}=cosx $,$ \frac{\partial N}{\partial x}=ycosx+cosx-xsinx $,方程不是恰当的。<br>
因为$ \frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial y}}{M}=1 $<br>
所以方程有积分因子$  \mu=e^\int 1{\rm d}y=e^y $<br>
以$ \mu=e^y $乘以方程两边得到$ (ye^ycosxdx+ye^ysinxdy)+(e^ycosxdy-e^ysinxdx)=0 $<br>
即$ d[(y-1)e^ysinx+e^y(y-1)sinx]=0 $<br>
于是方程的通解为$ (y-1)e^ysinx+e^y（y-1)sinx=c $.<br>
**（11）**$ x（4ydx+2xdy）+y^3(3ydx+5xdy)=0.$<br>
解：显然$x(4ydx+2xdy)=d(2x^2y)$.<br>
对于方程$y^3(3ydx+5xdy)=0$，有$M=3y^4.N=5xy^3$,$ \frac{\partial M}{\partial y}=12y^3,\frac{\partial N}{\partial x}=5y^3 $<br>
因为$ \frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{5xy^3}=\frac{7}{5x} $<br>
所以方程有积分因子$ \mu=e^\int\frac{7}{5}x{\rm d}x=x^\frac{7}{5} $<br>
用$ \mu=x^\frac{7}{5} $乘以方程$y^3(3ydx+5xdy)=0$两边<br>
得$ 3y^4x^\frac{7}{5}dx+5x^\frac{12}{5}y^3dy=0 $<br>
即$ d(5/4x^\frac{12}{5}y^4)=0 $<br>
综上所述，得到原方程的通解为$ 2x^2y+\frac{5}{4}x^\frac{12}{5}y^4=c $.
## 四、设$f(x,y)$及$ \frac{\partial f}{\partial y} $连续，试证方程$dy-f(x,y)dx=0$为线性微分方程的充要条件是它有仅依赖于$x$的积分因子。
 解：充分性：设方程$dy-f(x,y)dx=0$有仅依赖于x的积分因子，欲证该方程是线性的。<br>
 由于积分因子依赖于$x$<br>
 所以存在某个$\varphi(x)$满足$ \frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac{-\frac{\partial f}{\partial y}-0}{1}=\varphi(x)$<br>
 即$ \frac{\partial f}{\partial y}=-\varphi(x)$<br>
 对上式两边积分后得$ f(x,y)=-\varphi(x)y+h(x)$<br>
 这里$h(x)$是关于x的任意可微函数，所以，原方程可以成$ \frac{dx}{dy}=-\varphi(x)y+h(x)$，这就证明了$dy-f(x,y)dx=0$是线性的。<br>
必要性：设方程$dy-f(x,y)dx=0$是线性微分方程。<br>
即存在函数$g(x),h(x)$使得$f(x,y)=yg(x)+h(x)$.<br>
这样$ M=-f(x,y)=-(yg(x)+h(x)),N=1$.<br>
$ \frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac{-g(x)}{1}=-g(x)$<br>
所以方程具有积分因子$\mu=e^\int -g(x){\rm d}x$这就证明了方程有仅依赖于$x$的积分因子。
## 八、求出伯努利微分方程的积分因子。
解：伯努利方程为$ \frac{dx}{dy}=p(x)y+q(x)y^n $<br>
可化为$ dy-p(x)ydx-q(x)y^ndx=0 $<br>
两边同时乘以$y^-n$得$ y^{-n}dy-p(x)y^{1-n}dx-q(x)dx=0 $<br>
即$ d(y^{1-n})-(1-n)p(x)y^{1-n}x-(1-n)q(x)dx=0 $<br>
再乘以$ e^{-(1-n)\int p(x){\rm d}x} $<br>
得$ e^{-(1-n)\int p(x){\rm d}x}[ d(y^{1-n})-(1-n)p(x)y^{1-n}dx]-e^{-(1-n)\int p(x){\rm d}x}(1-n)q(x)dx=0 $这是全微分方程<br>
因此所求的积分因子是$ \mu= y^{-n}e^{(n-1)\int p(x){\rm d}x} $.
## 一、求解下列方程：
**（1）**$xy'^3=1+y'$;
解：解出$x$，并以$ \frac{dx}{dy}=p$代入<br>
得到$x=\frac{1}{p^3}\neq0$,上式两边同时对y求导<br>
得到$\frac{1}{p}=(-\frac{3}{p^4}-\frac{2}{p^3})\frac{dp}{dy}$,两边积分<br>
得到$y=\frac{3}{2p^2}+\frac{2}{p}+c$则方程的通解为
$ \begin{cases}
x=\frac{1}{p^3},\\y=\frac{1}{2p^2}+\frac{2}{p}+c
\end{cases} p\neq0 $<br>
**（2）**$y'^3-x^3(1-y')=0$;
解：令$y'=tx$,则原方程化为$x^3(t^3-1+tx)=0$<br>
解出x得$x=\frac{1}{t}-t^2$<br>
两边对y求导得$\frac{1}{1-t^3}=(-\frac{1}{t^2}-2t)\frac{dt}{dy} dy=[(1-t^3)(-\frac{1}{t^2}-2)]dt$<br>
两边积分得$y=-\frac{1}{2}t^2+\frac{2}{5}t^5+\frac{1}{t}+c$,故方程的通解为
$ \begin{cases}
x=\frac{1}{t}-t^2,\\
y=-\frac{1}{2}t^2+\frac{2}{5}t^5+\frac{1}{t}+c
\end{cases} $<br>
**（3）** $y=(y')^2e^{y'}$;
解：令$y'=p$则原方程化为$y=p^2e^p$<br>
对x求导，即得$p=(2p+p^2)e^p\frac{dp}{dx}$<br>
两边积分，即得$x=(p+1)e^p+c$.因此，方程的通解为
$ \begin{cases}
x=(p+1)e^p+c,\\y=p^2e^p
\end{cases} $<br>
**（4）**$y(1+y'^2)=2a$(a为常数）；
解：令y'=tant，则原方程化为$y=2acot^2t$<br>
两端对x求导数，可得$tant=-2asin2t\frac{dt}{dx}$<br>
两边积分，得$x=-a(2t+sin2t+c)$,因此，方程的通解为
$ \begin{cases}
x=-a(2t+sin2t+c),\\y=2acot^2t
\end{cases} $<br>
**（5）**$x^2+y'^2=1$;
解：令$y'=cost$,则方程化为$x^2+cos^2t=1$<br>
解出x，得$x=sint$.对y求导<br>
得$dy=cos^2tdt$,两边积分<br>
得$y=\frac{t}{2}+\frac{1}{4}sin2t+c$因此，方程的通解为
$ \begin{cases}
x=sint,\\y=\frac{t}{2}+\frac{1}{4}sin2t+c
\end{cases} $<br>
**（6）**$y^2(y'-1)=(2-y')^2$.
解：令$2-y'=yt$,则原方程化为$y^2(1-yt)=y^2t^2$<br>
由此，得$y=\frac{1}{t}-t$,并且$y'=1+t^2$.<br>
这是常微分方程的参数形式，因此$dx=-\frac{2}{t^2}dt$，两边积分<br>
得$x=\frac{1}{t}+c$于是求得方程的参数形式的通解为
$ \begin{cases}
x=\frac{1}{t}+c,\\y=\frac{1}{t}-t
\end{cases} $<br>
其中c为任意常数。此外，当y'=0时，原方程变为$-y^2=4$,此时方程无解。
