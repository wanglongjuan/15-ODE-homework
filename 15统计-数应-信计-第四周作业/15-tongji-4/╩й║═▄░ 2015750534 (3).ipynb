$1.求下列方程的解：$

$（8）(x+2y)dx+xdy=0$

$解：令M=x+2y;N=x;$

$\frac{\partial M}{\partial y}=2;\frac{\partial N}{\partial x}=1;$

$\frac{(\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x})}{N}=\frac{1}{x};$

$上式只与x有关，故方程有只与x有关的积分因子:$

$μ=e^{\int\mathrm (\frac{1}{x})\,\mathrm{d}x}=e^{ln|x|}=x$

$在原方程的两边同时乘以μ得：\\2xydx+x^2dy+x^2dx=0$

$解得通解为：\\x^3+3x^2y=c$

$(9)[xcos(x+y)+sin(x+y)]dx+xcos(x+y)dy=0;$

$解：令M=xcos(x+y)+sin(x+y);N=xcos(x+y);$

$\frac{\partial M}{\partial y}=-xsin(x+y)+cos(x+y)=\frac{\partial N}{\partial x}=cos(x+y)-xsin(x+y)$

$所以该方程是恰当微分方程$

$在原方程的两边同时求积分得通解为:xsin(x+y)=c$

$(10)(ycosx-xsinx)dx+(ysinx+xcosx)dy=0;$

$解：令M=ycosx-xsinx;N=ysinx+xcosx;$

$\frac{\partial M}{\partial y}=cosx;\\\frac{\partial N}{\partial x}=ycosx+cosx-ysinx;$

$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{-M}=1;$

$故方程有积分因子：μ=e^{\int\mathrm 1\,\mathrm{d}y}=e^y$

$在原方程的两边同时乘以μ得：(ye^ycosxdx+ye^ysinxdy)+(xe^ycosxdy-xe^ysinxdx)=0$

即;$$d[(y-1)e^ysinx+e^yxcosx)]=0$$

$所以方程的通解为：(y-1)e^ysinx+e^yxcosx=c$

$(11)x(4ydx+2xdy)+y^3(3ydx+5xdy)=0$

$解：整理原方程得：（4xy+3y^4）dx+(2x^2+5xy^3)dy=0；$

$令M=4xy+3y^4;N=2x^2+5xy^3;$

$\frac{\partial M}{\partial y}=4x+12y^3,\frac{\partial N}{\partial x}=4x+5y^3;$

$故此方程为非恰当微分方程，设积分因子为μ=x^my^n$

$在原方程的两边同时乘以μ得：(4x^{1+m}y^{1+n}+3x^my^{4+n})dx+(2x^{2+m}y^n+5x^{1+m}y^{3+n})dy=0;$

函数μ（x,y）为方程的积分因子的充要条件为;$$\frac{\partial (μM)}{\partial y}=\frac{\partial (μN)}{\partial x}$$

$即：4(1+n)x^{1+m}y^n+3(4+n)x^my^{3+n}=2(2+m)x^{1+m}y^n+5(1+m)x^my^{3+n}$

对比系数得$\begin{cases}4n-2m=0\\ 3n-5m+7=0\end{cases}$

解得：$\begin{cases}m=2\\n=1\end{cases}$

$所以积分因子μ=x^2y,原方程乘以积分因子后得：4x^3y^2dx+3x^2y^5dx+2x^4ydy+5x^3y^4dy=d(x^4y^2)+d(x^3y^5)=0$

$故方程的通解为：x^4y^2+x^3y^5=c;x=0,y=0也是方程的解.$

$4.设f(x,y)及\frac{\partial f}{\partial y}连续，试证方程dy-f(x,y)dx=0为线性微分方程的充要条件是它有仅依赖于x的积分因子.$

解：充分性：设方程dy-f(x,y)dx=0有仅依赖于x的积分因子，则存在某个ψ(x)满足$$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=-\frac{\frac{\partial f}{\partial y}-0}{1}=ψ(x)$$

$即;\frac{\partial f}{\partial y}=-ψ(x);$

$积分得：f(x,y)=ψ(x)y+h(x),h(x)是关于x的任意可微分函数.$

$即证明了方程dy-f(x,y)dx=0为线性微分方程.$

$必要性：$

$设方程dy-f(x,y)dx=0为线性微分方程.$

$即存在g(x),h(x)使得f(x,y)=yg(x)+h(x);$

$令M=-f(x,y)=-yg(x,y)-h(x），N=1，则\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=-\frac{g(x)}{1}=-g(x);$

$所以方程具有积分因子μ=e^{\int\mathrm -g(x)\,\mathrm{d}x},集方程有仅依赖于x得积分因子.$

$8.求出伯努利微分方程的积分因子.$

解：\frac{dy}{dx}=P(x)y+Q(x)y^n整理得;$$dy-P(x)ydx+Q（x）y^ndx=0$$

$两边同时乘以y^{-n}得y^{-n}dy-P(x)y^{1-n}dx-Q（x）dx=0$

$即：d(y^{1-n})-(1-n)P(x)y^{1-n}dx-(1-n)Q（x）dx=0$

$再乘e^{-(1-n)\int\mathrm P（x）\,\mathrm{d}x}得：$

$e^{-(1-n)\int\mathrm P（x）\,\mathrm{d}x}[d(y^{1-n})-(1-n)P(x)y^{1-n}dx-(1-n)Q（x）dx]=0$

$即d[y^{1-n}e^{-(1-n)\int\mathrm P(x)\,\mathrm{d}x}]-d[（1-n）Q（x）e^{-(1-n)\int\mathrm P（x）\,\mathrm{d}x}dx]=0$

$上式为全微分方程，因此所以的积分因子是y^{-n}e^{-(1-n)\int\mathrm P(x)\,\mathrm{d}x}.$


$1.求下列方程的解.$

$(1)xy'^3=1+y'$

$解;令y'=p,则x=\frac{1+p}{p^3}$

$两边再关于y求导得：\frac{1}{p}=(-\frac{3}{p^4}-\frac{2}{p^3})\frac{dp}{dy};$

$两边同时积分得：y=\frac{3}{2p^2}+\frac{2}{p}+c;$

所以方程的通解为:$$\begin{cases}x=\frac{1+p}{p^3}\\y=\frac{3}{2p^2}+\frac{2}{p}+c(p为参数且不等于0，c位任意常数）\end{cases}$$

$（2）y'^3-x^3(1-y')=0$

$解;令y'=tx则x=\frac{1}{t}-t^2;$

$两边关于y求导得：\frac{1}{1-t^3}=(-\frac{1}{t^2}-2t)\frac{dt}{dy}$;

$即dy=[(1-t^3)(-\frac{1}{t^2}-2t)\frac{dt}{dy})dt;$

$积分得：y=-\frac{1}{2}t^2+\frac{2}{5}t^5+\frac{1}{t}+c;$

所以方程的通解为$$\begin{cases}x=\frac{1}{t}-t^2\\y=-\frac{1}{2}t^2+\frac{2}{5}t^5+\frac{1}{t}+c(t为参数，c为任意常数）\end{cases}$$

$(3)y=y'^2e^{y'}$

$解：令y'=p则y=p^2e^p;$

$对x求导得：p=(2p+p^2）e^p\frac{dp}{dx},即：dx=(p+2)e^pdp;$

$积分得x=(p+1)e^p+c$

所以方程的通解为:$$\begin{cases}x=(p+1)e^p+c\\y=p^2e^p(p为参数，c位任意常数）\end{cases}$$

$y=0也是方程的解$

$(4)y(1+y'^2)=2a$

$解：令y'=tant则y=2acos^2t$

$两边关于x求导得：tant=-2asin2t\frac{dt}{dx};$

$化简得：dx=-4acos^2tdt$

$积分得x=-a(2t+sin2t)+c$

所以方程的通解为:$$\begin{cases}x=-a(2t+sin2t)+c\\y=2acos^2t(t为参数，c为任意常数）\end{cases}$$

$y=2a也是方程的解$

$(5)x^2+y'^2=1;$

$解：令y'=cost则x=sint$

$对y求导得：dy=dxcost=cos^2tdt=\frac{cos2t+1}{2}dt$

$两边积分得;y=\frac{t}{2}+\frac{1}{4}sin2t+c.$

所以方程的通解为$$\begin{cases}x=sint\\y=(\frac{t}{2})+(\frac{1}{4}sin2t)+c(t为参数，c为任意常数）\end{cases}$$

$(6)y^2(y'-1)=(2-y')^2.$

$解：令2-y'=ty则原方程为：y^2(1-ty)=y^2t^2$

$解得：y=\frac{1}{t}-t,y'=1+t^2.$

$两边关于x求导得：dx=\frac{dy}{y'}=\frac{-\frac{1}{t^2}-1}{1+t^2}dt=-\frac{1}{t^2}dt;$

$积分得x=\frac{1}{t}+c;$

所以原方程的通解为：$$\begin{cases}x=\frac{1}{t}+c\\y=\frac{1}{t}-t(t为参数，c为任意常数）\end{cases}$$
