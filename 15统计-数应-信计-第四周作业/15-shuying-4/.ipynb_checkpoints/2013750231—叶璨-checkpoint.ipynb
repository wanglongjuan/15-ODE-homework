一，求下列方程的解。
（1）$$(x+2y)dx+xdy=0$$
   解：这里$M=x+2y,N=x$
   
   $$\frac{\partial M}{\partial y}=2, \frac{\partial N}{\partial x}=1$$
   
   此方程是不恰当的。
   
   又因为$$ \frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac{1}{x}$$
   是只关于x的函数
   
   故方程有积分因子$m=e^{\int\mathrm{\frac{1}{x}}\,\mathrm dx}=x$
   在方程两边同时乘以$x$可得
   
  $$（x^2+2xy）dx+x^2dy=0$$
  $$ d(\frac{x^3}{3}+x^2y)=0$$
  故方程的通解为$$\frac{x^3}{3}+x^2y=c$$






（2）$$[xcos(x+y)+sin(x+y)]dx+xcos(x+y)dy=0$$
  解： 这里$M=xcos(x+y)+sin(x+y),N=xcos(x+y)$
       $$ \frac{\partial M}{\partial y}=cos(x+y)-xsin(x+y)$$, 
       $$ \frac{\partial N}{\partial x}=cos(x+y)-xsin(x+y)$$
        此方程是恰当的。
        所以$$[xcos(x+y)+sin(x+y)]dx+xcos(x+y)dy=0$$
        即$$d[xsin(x+y)]=0$$
        此方程的通解为$$[xsin(x+y)]=c$$


（3）$$（ycosx-xsinx）dx+（ysinx+xcosx）dy=0$$

解：这里$M=ycosx-xsinx,N=ysinx+xcosx$
       $$ \frac{\partial M}{\partial y}=cosx$$
       $$\frac{\partial N}{\partial x}=ycosx+cosx+xsinx$$
   此方程是不恰当的。
   又因为$$ \frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{M}=1$$
   是只关于y的函数。
       故方程有积分因子$m=e^{\int\mathrm{\frac{1}{-1}}\,\mathrm dx}=e^y$
       在方程两边同时乘以$e^y$可得
       $$(ye^ycosxdx+ye^ysinxdy)+(e^ycosxdy-e^yxsinxdx)=0$$
       $$d[(y-1)e^ysinx+e^yxcosx=0$$
       故方程的通解是$$e^yxcosx+e(y-1)sinx=c$$
       


（4）$$x(4ydx+2xdy)+y^3(3ydx+5xdy)=0$$
  
 解：首先对方程进行变形可得
 $$x(4ydx+2xdy)=-y^3(3ydx+5xdy)$$
  易观察知：$$x(4ydx+2xdy)=d(2x^2y)$$
  对于方程$y^3(3ydx+5xdy)=0$
  这里$$M=3y^4,N=5xy^3$$
  
  $$\frac{\partial M}{\partial y}=12y^3$$
  
  $$\frac{\partial N}{\partial x}=5y^3$$
  
  又因为$$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac{7}{5x}$$
  
  是只关于x的函数。
  $故方程有积分因子$
  
  $$m==e^{\int\mathrm{\frac{7}{5x}}\,\mathrm dx}$$
  
  在方程两边同时乘以$x^{\frac{7}{5}}$
  
  $$3y^4x^{\frac{7}{5}}dx+5x^{\frac{12}{5}}y^3dy=0$$
  
  即$$d(\frac{5}{4}x^{\frac{12}{5}}y^4)=0$$
  
  综上，原方程的通解为
  $$2x^2y+\frac{5}{4}x^{\frac{12}{5}}y^4=c$$

三，求出伯努利微分方程的积分因子。

解：伯努利微分方程的形式为：

$$\frac{dy}{dx}=P(x)y+Q(x)y^n$$

将伯努利方程变形为
$$dy-p(x)ydx-q(x)y^ndx=0$$
方程两边同时乘以y^-n得：
$$y^-ndy-p(x)y^{1-n}dx-q(x)dx$$
即：$$d(y^{1-n})-(1-n)p(x)y^{1-n}dx-(1-n)q(x)dx=0$$

令$M=y^{1-n}P(x)+Q(x),N=y^{-n}$
这时$$\frac{\partial M}{\partial y}=(1-n)P(x)y^{-n},\frac{\partial N}{\partial x}=0$$
此方程不是恰当方程
$$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=(1-n)P(x)$$
因此$u=y^{-n}e^{\int\mathrm {-(1-n)P(x)}\,\mathrm dx}$
伯努利微分方程的积分因子为$u=y^{-n}e^{\int\mathrm {-(1-n)P(x)}\,\mathrm dx}$
$=u=y^{-n}e^{\int\mathrm {(n-1)P(x)}\,\mathrm dx}$


四，求解下列方程。
（1）$xy'^3=1+y'$
解：解出x，并令$\frac{dy}{dx}=p$，得到
$$x=\frac{1}{p^3}+\frac{1}{p^2}$$
两边对y求导数，得到
$$\frac{1}{p}=(\frac{-3}{p^4}-\frac{2}{p^3}\frac{dp}{dy}$$
积分之，可得：
$$y=\frac{3}{2p^2}+\frac{2}{p}+c$$
即原方程的通解为：
$$x=\frac{1}{p^3}+\frac{1}{p^2}$$
$$y=\frac{3}{2p^2}+\frac{2}{p}+c$$

（2）$y'^3-x^3(1-y')=0$

解：解：令y'=p=tx,则由方程得
$$(tx)^3-x^3(1-tx)=0$$
$$x=\frac{1}{t}-t^2$$
从而$$p=1-t^3$$
于是$$dy=pdx=-(1-t^3)(\frac{1}{t^2}+2t)dt$$
$$dy=(-\frac{1}{t^2}-t+2t^4)dt$$
积分之，得到
$$y=-\frac{t^2}{2}+\frac{2t^5}{5}+\frac{1}{t}+c$$
故方程的通解为：
$$x=\frac{1}{t}-t^2$$
$$y=-\frac{t^2}{2}+\frac{2t^5}{5}+\frac{1}{t}+c$$

（3）$y=y'^2e^{y'}$

解：令y'=p则原方程可化为：
$$y=p^2e^p$$
对x求导得：
$$p=(2p+p^2)e^p\frac{dp}{dx}$$
两边积分可得
$$x=(p+1)e^p+c$$
故方程的通解为：
$$x=(p+1)e^p+c$$
$$y=p^2e^p$$
（4）$y(1+y'^2)=2a(a为常数)$
解：令y'=tant,则
因此令$y'=tant$
即$$y=2acos^2t$$
对x求导得$$\frac{dy}{dx}=-2asin2t\frac{dt}{dx}$$
即$$dx=-4acos^2tdt$$
两边同时积分得$$x=-a(2t+sin2t+c)$$
方程的通解为$$y=2acos^2t$$
$$x=-a(2t+sin2t+c)$$
（5）$x^2+y'^2=1$

解：令y'=cost,将方程化为
$$x^2+cos^2x=1$$
解出x，可得：
$$x=sint$$
对y求导可得：
$$\frac{1}{cost}=cos\frac{dt}{dy}$$
两边积分可得：
$$y=\frac{t}{2}+\frac{sin2t}{4}+c$$
故方程得解为
$$x=sint$$
$$y=\frac{t}{2}+\frac{sin2t}{4}+c$$

（6）$y^2(y'-1)=(2-y')^2$

解:令2-y'=ty,则与原微分方程消去y'后，有
$$y^2(1-yt)=y^2t^2$$
由此可得：
$$y=\frac{1}{t}-t$$
并且$$y'=t^2+1$$
这是原微分方程的参数形式，因此
$$dx=\frac{dy}{y'}=-\frac{1}{t^2}dt$$
积分可知：$$x=\frac{1}{t}+c$$
于是求得方程的参数形式的通解为
$$x=\frac{1}{t}+c$$
$$y=\frac{1}{t}-t$$
或者消去参数t，得到$$y=x-\frac{1}{x-c}-c$$
此外，当y'=0时，原方程变为$$-y^2=4$$,此时方程无解。
