### 1.有理整式的概念。
有理整式(rational integral expression)亦称整式.一种基本的有理式.对变数字母与数仅限于有限次加、减、乘、乘方运算的代数式，即除式不含变数字母的有理式称为有理整式.

### 2.通解中解对常数的独立性是指什么？
解对常数的独立性是指，对$\varphi$及其$n-1$阶偏导数关于$n$个常数$c_1,c_2,…,c_n$的雅可比行列式不为0.

### 3.设$y_1(t)$是$\frac{dy}{dt}=f(y)$,$y(0)=y_0$的解,证明$y_2(t)=y_1(t-t_0)$是$\frac{dy}{dt}=f(y)$,$y(t_0)=y_0$的解.
证明： 由$$y_1(0)=y_0$$得$$y_2(t_0)=y_1(t_0-t_0)=y_1(0)=y_0$$
     由$$\frac{dy_1(t)}{dt}=f(y_1(t))$$
     得$$\frac{dy_2(t)}{dt}=\frac{dy_1(t-t_0)}{d(t-t_0)}=f(y_1(t-t_0))=f(y_2(t))$$
     证毕.

1.求下列方程的解：

(6)$x\frac{dy}{dx}-y+\sqrt{x^2-y^2}=0;$

解：首先易知当$x=0$时$y=0$,

当$x>0$时可化为$$\frac{dy}{dx}=\frac{y}{x}-\sqrt{1-(\frac{y}{x})^2}$$当$y=\pm{x}$时成立

当$y\neq{x}且y\neq{-x}时$做$t=\frac{y}{x}$代换
$$\frac{dy}{dx}=x\frac{dt}{dx}+t=t-\sqrt{1-t^2}$$即$$\frac{dt}{\sqrt{1-t^2}}=-\frac{dx}{x}$$两边积分得$$\arcsin{t}=-\ln{x}+c$$
即$$\arcsin{\frac{y}{x}}=-\ln{x}+c$$
当$x<0$时可化为$$\frac{dy}{dx}=\frac{y}{x}+\sqrt{1-(\frac{y}{x})^2}$$当$y=\pm{x}$时成立

当$y\neq{x}且y\neq{-x}时$做$t=\frac{y}{x}$代换
$$\frac{dy}{dx}=x\frac{dt}{dx}+t=t+\sqrt{1-t^2}$$即$$\frac{dt}{\sqrt{1-t^2}}=\frac{dx}{x}$$两边积分得$$\arcsin{t}=\ln{(-x)}+c$$
即$$\arcsin{\frac{y}{x}}=\ln{(-x)}+c$$
综上所述：$$y=\begin{cases}\arcsin{\frac{y}{x}}=-\ln{x}+c,x>0\\\arcsin{\frac{y}{x}}=\ln{(-x)}+c,x<0\end{cases} ,及y=\pm{x}$$
(7)$\tan{y}dx-\cot{x}dy=0;$ 

解：易知当$y=n\pi,n\in{Z}$时，及$x=\frac{\pi}{2}+n\pi,n\in{Z}$时成立      （1）

否则可得$$\frac{dy}{\tan{y}}=\frac{dx}{\cot{x}}$$两边积分得
$$\ln{|\sin{y}|}=-\ln{|\cos{x}|}+c_1\iff|\sin{y}\cos{x}|=e^{c_1}\iff\sin{y}\cos{x}=\pm{e^{c_1}}$$
又由微分方程的定义域可得$$\sin{y}\neq\pm{1}且\cos{x}\neq\pm{1}$$所以$$e^{c_1}\in(0,1)$$
由（1）结果可得$$e^{c_1}=0$$综上$$\sin{y}\cos{x}=c,c\in[0,1)$$

 2.作适当的变量变换求解下列方程：
 
(6)$\frac{dy}{dx}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$;

解:做$y=（tx）^\frac{1}{3}$的代换得$$\frac{dy}{dx}=\frac{t^2-2}{y^2(2t+1)}$$
即$$\frac{dy^3}{dx}=\frac{3(t^2-2)}{2t+1}\iff\frac{dtx}{dx}=\frac{3(t^2-2)}{2t+1}\iff x\frac{dt}{dx}+t=\frac{3(t^2-2)}{2t+1}$$
即$$\frac{(2t+1)dt}{t^2-t-6}=\frac{dx}{x}$$两边积分得$$7\ln{|t-3|}+3\ln{|t+2|}=5\ln{|x|}+c$$
代入t整理可得$$|y^3-3x|^7|y^3+2x|^3=|x|^{15}e^c$$
即$$|y^3-3x|^7|y^3+2x|^3=|x|^{15}c_0,c_0>0$$经验证$c_0=0$时也成立，综上通解为$$(y^3-3x)^7(y^3+2x)^3=x^{15}c$$


(7)$\frac{dy}{dx}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y};$

解：
可变为$$\frac{dy^2}{dx^2}=\frac{(2x^2-2)+(3y^2+3)}{(2y^2+2)+(3x^2-3)}$$
令$X=x^2-1,Y=y^2+1$得$$\frac{dy^2}{dx^2}=\frac{dY}{dX}=\frac{2X+3Y}{3X+2Y}$$
可变为$$\frac{dY}{dX}=\frac{2+3\frac{Y}{X}}{3+2\frac{Y}{X}}$$
令$t=\frac{Y}{X}$得$$\frac{dY}{dX}=\frac{Xdt}{dX}+t=\frac{2+3t}{3+2t}$$
即$$\frac{Xdt}{dX}=\frac{2-2t^2}{3+2t}$$
当$t=\pm1$时即$Y=\pm{X}$时显然成立(1)

当$t\neq\pm1$时可化为$$\frac{(3+2t)dt}{2-2t^2}=\frac{dX}{X}$$
左右积分并整理得$$|t+1||t-1|^5=|X|^4e^{c_1}$$结合(1)可得$$|t+1||t-1|^5=|X|^4c_2,c_2>0或c_2=0$$
代入t并整理得$$|Y+X||Y-X|^5=|X|^{10}c_2$$
将$X,Y$代入得$$|x^2+y^2||y^2-x^2+2|^5=|x^2-1|^{10}c_2$$
即通解为$$(x^2+y^2)(y^2-x^2+2）^5=（x^2-1）^{10}c$$
