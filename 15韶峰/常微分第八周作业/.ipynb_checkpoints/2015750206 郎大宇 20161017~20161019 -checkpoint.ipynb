#### 第一题
证明： 利用反证法。假设x(t)和y(t)在区间$a \leq t\leq b $上线性相关。则$ \exists $不全为零的$l_1,l_2,$

$$s.t:l_1x(t)+l_2y(t)=0$$

此时，$\frac{x(t)}{y(t)}$ 与 $\frac{y(t)}{x(t)}$之中至少有一个为常数，与已知条件矛盾，故假设不成立，原命题得证。

#### 第二题
证明:由于$x_1(t)$是方程$（1）$的解，则有

$$\frac{d^{n}x_1(t)}{dt_n}+a_1(t)\frac{d^{n-1}x_1(t)}{dt_{n-1}}+ \cdots +a_n(t)x_1(t)=f_1(t)$$

由于$x_2(t)$是方程$（2）$的解，则有

$$\frac{d^{n}x_2(t)}{dt_n}+a_1(t)\frac{d^{n-1}x_2(t)}{dt_{n-1}}+ \cdots +a_n(t)x_2(t)=f_2(t)$$

将上面两式相加，根据求导的性质，可得

$$\frac{d^{n}[x_1(t)+x_2(t)]}{dt_n}+a_1(t)\frac{d^{n-1}[x_1(t)+x_2(t)]}{dt_{n-1}}+ \cdots +a_n(t)[x_1(t)+x_2(t)]=f_1(t)+f_2(t)$$

故$x_1(t)+x_2(t)$是方程$$\frac{d^{n}x}{dt_n}+a_1(t)\frac{d^{n-1}x}{dt_{n-1}}+ \cdots +a_n(t)x=f_1(t)+f_2(t)$$的解。

#### 第三题

(1)解：应用常数变易法，令
$x=c_1(t)e^t+c_2(t)e^{-t}$,将它代入方程，则可得决定$c_1'(t)$和$c_2'(t)$的两个方程：

$$e^tc_1'(t)+e^{-t}c_2'(t)=0$$

$$e^tc_1'(t)+e^{-t}c_2'(t)=\frac{1}{cost}$$

联立方程解得：

$$c_1'(t)=\frac{1}{2e^t cost}$$

$$c_1'(t)=-\frac{1}{2e^{-t} cost}$$

积分得：$$c_1(t)=\frac{sint-cost}{4}e^{-t}+r_1 $$

$$c_1(t)=-\frac{sint+cost}{4}e^t+r_2$$

代入即得方程通解：

$$
\begin{aligned}
x &=\frac{sint-cost}{4}e^{-t}-\frac{sint+cost}{4}+r_1e^{t}+r_2e^{-t}\\
&=-\frac{1}{2}cost+r_1e^{t}+r_2e^{-t}
\end{aligned}
$$

(2)解：应用常数变易法，令

$$x=c_1(t)cos2t+c_2(t)sin2t$$

将它代入方程，则可得决定$c_1'(t)$和$c_2'(t)$的两个方程：

$$cos2t\cdot c_1'(t)+sin2t\cdot c_2'(t)=0$$

$$-2sin2t\cdot c_1'(t)+2cos2t\cdot c_2'(t)=tsin2t$$

联立解得：

$$c_2'(t)=\frac{1}{4}tsin4t$$

$$c_1'(t)=-\frac{t}{2}sin^{2}2t$$

进行积分得：

$$c_2(t)=\frac{1}{4}t^{2}-\frac{1}{8}tsin4t-\frac{1}{32}cos4t+r_2$$

$$c_1(t)=-\frac{1}{8}t^{2}cos2t+\frac{1}{16}sin4tcos2t+\frac{1}{64}cos2tcos4t+r_1$$

代入解得：

$$x=r_1cos2t+r_2sin2t-\frac{1}{8}t^{2}cos2t+\frac{1}{16}sin2t$$

(3)应用常数变易法，令

$$x=c_1(t)t+c_2(t)tlnt$$

将它代入方程，则可得决定$c_1'(t)$和$c_2'(t)$的两个方程：

$$t \cdot c_1'(t)+tlnt \cdot c_2'(t)=0,$$

$$c_1'(t)+(lnt+1)c_2'(t)=\frac{6}{t}+34,$$

解得：$$c_1'(t)=-lnt(\frac{6}{t}+34)$$

$$c_2'(t)=\frac{6}{t}+34,$$

积分得：

$$c_1(t)=-34(tlnt-t)-3(lnt)^2+c_1,$$

$$c_2(t)=34t+6lnt+c_2.$$

代入解得：

$$x=c_1\cdot t+c_2tlnt +34t^{2}+3t(lnt)^{2}.$$

#### 第四题

证明：(1)①充分性.因为
$$W'[x_1,x_2]=
\begin{vmatrix}
x_1'&x_2'\\
x_1'&x_2'\\
\end{vmatrix}
+\begin{vmatrix}
x_1&x_2\\
x_1''&x_2''\\
\end{vmatrix}
=\begin{vmatrix}
x_1&x_2\\
x_1''&x_2''\\
\end{vmatrix},
$$

$$W'[x_1,x_2]+a_1(t)W[x_1,x_2]=
\begin{vmatrix}
x_1&x_2\\
x_1''&x_2''\\
\end{vmatrix}
+a_1(t)\begin{vmatrix}
x_1&x_2\\
x_1'&x_2'\\
\end{vmatrix}
=\begin{vmatrix}
x_1&x_2\\
x_1''+a_1(t)x_1'&x_2''+a_1(t)x_2'\\
\end{vmatrix}
=0,
$$
而$x_1(t)$是方程的解，并且不为0，故
$$
\begin{vmatrix}
x_1&x_2\\
-a_1(t)x_1&x_2''+a_1(t)x_2'\\
\end{vmatrix}
=x_1\begin{vmatrix}
1&x_2\\
-a_2(t)&x_2''+a_1(t)x_2'\\
\end{vmatrix}
=0
$$

故有

$$x_2''+a_1(t)x_2'+a_2(t)x_2=0,$$

综上，$x_2(t)$是方程的解。

②必要性.由于W[x_1,x_2]是方程的解$x_1(t),x_2(t)$的朗斯基行列式，故
$$W'[x_1,x_2]=
\begin{vmatrix}
x_1&x_2\\
x_1''&x_2''\\
\end{vmatrix}=
\begin{vmatrix}
x_1&x_2\\
x_1''+a_1(t)x_1&x_2''+a_1(t)x_2\\
\end{vmatrix}=
\begin{vmatrix}
1&x_2\\
-a_1(t)&x_1'+a_1(t)x_2'\\
\end{vmatrix}=
-a_1(t)\begin{vmatrix}
x_1&x_2\\
x_1'&x_2'\\
\end{vmatrix}=
-a_1(t)W[x_1,x_2],
$$

故$W[x_1,x_2]$是$W'[x_1,x_2]+a_1(t)W[x_1,x_2]=0$的解，必要性得证，

(2)由于$x_1,x_2$是方程的解，则由刘维尔公式得
$$
\begin{vmatrix}
x_1&x_2\\
x_1'&x_2'\\
\end{vmatrix}=
W(t_0)e^{-\int_{t_0}^t a_1(s)ds},
$$

即
$$x_1x_2'-x_1'x_2=W(t_0)e^{-\int_{t_0}^t a_1(s)ds}$$

方程两边同乘以$\frac{1}{{x_1}^2}$,则有 

$$\frac{d(\frac{x_2}{x_1})}{dt}=\frac{W(t_0)}{{x_1}^2}e^{-\int_{t_0}^t a_1(s)ds}$$ 

于是

$$\frac{x_2}{x_1}=c_1\int \frac{1}{{x_1}^2}e^{-\int_{t_0}^t a_1(s)ds}+c_2,$$

即

$$x_2=(c_1\int \frac{1}{{x_1}^2}e^{-\int_{t_0}^t a_1(s)ds}+c_2)x_1.$$

取 $c_1=1,c_2=0,$得

$$x_2=x_1\int \frac{1}{{x_1}^2}e^{-\int_{t_0}^t a_1(s)ds}dt,$$

又因为
$$
W[t]=
\begin{vmatrix}
x_1&x_2\\
x_1'&x_2'\\
\end{vmatrix}
=e^{-\int_{t_0}^t a_1(s)ds} \neq 0,
$$

因此方程的通解可以表示为
$$x=x_1[c_1\int \frac{1}{{x_1}^2}e^{-\int_{t_o}^t a_1(s)ds}+c_2],$$

命题得证。
