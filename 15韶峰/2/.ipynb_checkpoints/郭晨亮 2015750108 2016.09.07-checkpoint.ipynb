## 1.有理整式

有理整式亦称整式，是一种基本的有理式，对变数字母与数有限次加、减、乘、乘方运算的代数式，除式不含变数字母的有理式。

## 2.通解中，解对常数的独立性

我们把$n$个独立的任意常数$c_1,c_2,\cdots,c_n$的解

$$y = \varphi(x,c_1,c_2,\cdots,c_n)$$

称为$n$阶方程的通解。

关于解对常数的独立性是指，对$/varphi$及其$n-1$阶偏导数关于$n$个常数$c_1,c_2,\cdots,c_n$的雅可比行列式不为0。

## 3.验证常微分方程的解

设$\mathbf y_1(t)$是
$
\begin{cases}
\frac{\mathrm d\mathbf y}{\mathrm dt}=\mathbf f(\mathbf y) \\
\mathbf y(0) = \mathbf y_0
\end{cases}
$
的解，证明$\mathbf y_2(t)= \mathbf y_1(t - t_0)$是
$
\begin{cases}
\frac{\mathrm d\mathbf y}{\mathrm dt}=\mathbf f(\mathbf y) \\
\mathbf y(t_0) = \mathbf y_0
\end{cases}
$
的解

解：

$$\mathbf y_2(t_0)=\mathbf y_1(t_0-t_0)=\mathbf y_1(0)=\mathbf y_0$$

$$\frac{\mathrm d\mathbf y_2(t)}{dt}=\frac{\mathrm d\mathbf y_1(t-t_0)}{\mathrm dt}=\mathbf f(\mathbf y_1(t-t_0))=\mathbf f(\mathbf y_2(t))$$

所以$\mathbf y_2(t)= \mathbf y_1(t - t_0)$是
$
\begin{cases}
\frac{\mathrm d\mathbf y}{\mathrm dt}=\mathbf f(\mathbf y) \\
\mathbf y(t_0) = \mathbf y_0
\end{cases}
$
的解

## 4.验证下列各函数是相应微分方程的解

（1）$\mathbf y=x^2+1，\mathbf y\prime=\mathbf y^2-(x^2+1)\mathbf y+2x$

解：

$\mathbf y^2-(x^2+1)\mathbf y+2x=(x^2+1)^2-(x^2+1)^2+2x=2x=\mathbf y\prime$

所以$\mathbf y=x^2+1$是$\mathbf y\prime=\mathbf y^2-(x^2+1)\mathbf y+2x$的解

<font color="red">**正确**

（2）$\mathbf y=-\frac{\mathbf g(x)}{\mathbf f(x)},\mathbf y\prime=\frac{\mathbf f\prime(x)}{\mathbf g(x)}\mathbf y^2-\frac{\mathbf g\prime(x)}{\mathbf f(x)}$

解：

$\frac{\mathbf f\prime(x)}{\mathbf g(x)}\mathbf y^2-\frac{\mathbf g\prime(x)}{\mathbf f(x)}$

$=\frac{\mathbf f\prime(x)}{\mathbf g(x)}(-\frac{\mathbf g(x)}{\mathbf f(x)})^2-\frac{\mathbf g\prime(x)}{\mathbf f(x)}$

$=\frac{\mathbf f\prime(x)\mathbf g(x)-\mathbf g\prime(x)\mathbf f(x)}{\mathbf f^2(x)}$

$=\mathbf y\prime$


所以$\mathbf y=-\frac{\mathbf g(x)}{\mathbf f(x)}$是$\mathbf y\prime=\frac{\mathbf f\prime(x)}{\mathbf g(x)}\mathbf y^2-\frac{\mathbf g\prime(x)}{\mathbf f(x)}$的解

<font color="red">**正确**

## 5.给定一阶微分方程$\frac{\mathrm d\mathbf y}{\mathrm dx}=2x$

（1）求出它的通解

（2）求通过点 $(1,4)$ 的特解

（3）求出与直线$\mathbf y=2x+3$相切的解

（4）求出满足条件$\int_0^1\mathbf y\mathrm dx=2$的解

（5）绘出（2）,（3）,（4）中的解的图形

解：（1）$\mathrm d\mathbf y=2x\mathrm dx$

$\int \mathrm d\mathbf y=\int 2x\mathrm dx$

$\mathbf y=x^2+c$

（2）把 $y=4,x=1$ 代入得 $c=3$

通过点 $(1,4)$ 的特解是 $\mathbf y=x^2+3$

（3）$y=x^2+c$ 与 $y=2x+3$ 相切，即方程组
$
\begin{cases}
y=x^2+c \\
y=2x+3
\end{cases}
$
有唯一解

$x^2-2x-3+c=0$

$4-4(c-3)=0$

$c=4$

与直线 $\mathbf y=2x+3$ 相切的解是 $\mathbf y=x^2+4$

（4）$\int_0^1\mathbf y\mathrm dx=2$

$\int_0^1(x^2+c)\mathrm dx=2$

$\left(\frac{x^3}{3}+cx\middle)\right|_0^1=2$

$c={5 \over 3}$

满足条件$\int_0^1\mathbf y\mathrm dx=2$的解是 $\mathbf y=x^2+{5 \over 3}$

（5）$y=x^2+3,y=x^2+4,y=x^2+{5 \over 3}$的图像

```python
var('x')
y=x^2+3
show(plot(y,x))
y=x^2+4
show(plot(y,x))

y=x^2+5/3
show(plot(y,x))
```

<font color="red">**正确**

## 6.求下列方程的解

（1）$x\frac{\mathrm dy}{\mathrm dx}-y+\sqrt{x^2-y^2}=0$

解：两边同时乘${1 \over x}$

$${\mathrm dy \over \mathrm dx}-{y \over x}+\sqrt{1-({y \over x})^2}=0$$

设$u={y \over x}$

$${\mathrm dy \over \mathrm dx}=x{\mathrm du \over \mathrm dx}+u$$

$$x{\mathrm du \over \mathrm dx}+u=u-\sqrt{1-u^2}$$

$$-{\mathrm du \over \sqrt{1-u^2}}={\mathrm dx \over x}$$

$$\int -{\mathrm du \over \sqrt{1-u^2}}=\int {\mathrm dx \over x}$$

$$\arcsin u=\ln |x|+c$$

$$\arcsin {y \over x}=\ln |x|+c$$

<font color="red">**$y^2=x^2$也是解**

（2）$\tan y\mathrm ~dx-\cot x\mathrm ~dy=0$

解：

$${\mathrm dy \over \tan y}={\mathrm dx \over \cot x}$$

$$\int \cot y\mathrm ~dy=\int \tan x\mathrm ~dx$$

$$\ln |\sin y|=-\ln |\cos x|+c$$

$$\sin y\cos x=c$$

<font color="red">**$y=k\pi \quad k=0,\pm1,\cdots$也是解**

## 7.作适当的变量变换求解下列方程

（1）$${\mathrm dy \over dx}={y^6-2x^2 \over 2xy^5+x^2y^2}$$

解：

$${\mathrm dy \over \mathrm dx}={y \over x}{1-2({x \over y^3})^2 \over 2+{x \over y^3}}$$

设$u={x \over y^3}$

$${\mathrm dy \over \mathrm dx}={y \over x}{1-2u^2 \over 2+u}$$

$${y^2\mathrm dy \over \mathrm dx}={y^3 \over x}{1-2u^2 \over 2+u}$$

$${\mathrm dy^3 \over 3\mathrm dx}=u{1-2u^2 \over 2+u}$$

$${\mathrm dy^3 \over \mathrm dx}=3u{1-2u^2 \over 2+u}$$

$${\mathrm dy^3 \over \mathrm dx}=x{\mathrm du \over \mathrm dx}+u$$

$$x{\mathrm du \over \mathrm dx}+u=3u{1-2u^2 \over 2+u}$$

$$\frac{3(2u + 1)\mathrm d u}{(u - 3)(u + 2)} = \frac{3\mathrm d x}{x}$$

$$\int\frac{3(2u + 1)\mathrm d u}{(u - 3)(u + 2)} = \int\frac{3\mathrm d x}{x}$$

$$(y^3 - 3x)^7(y^3 + 2x)^3 = cx^{15}$$
