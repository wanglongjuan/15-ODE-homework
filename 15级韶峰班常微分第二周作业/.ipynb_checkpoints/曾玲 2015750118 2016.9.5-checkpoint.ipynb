# **1.求下列方程的解：**

（6）、$x\frac{\mathrm dy}{\mathrm dx}-y+\sqrt{x^2-y^2}=0$;

x>0时，微分方程可化为$\frac{\mathrm dy}{\mathrm dx}=\frac{y}{x}-\sqrt{1-(\frac{y}{x})^2} $.

令$u=\frac{y}{x} $,则$y=ux,\frac{\mathrm dy}{\mathrm dx}=u+x\frac{\mathrm du}{\mathrm dx}$ .

得到微分方程$u+x\frac{\mathrm du}{\mathrm dx}=u-\sqrt{1-u^2}$ 。

分离变量得$\frac{\mathrm du}{\sqrt{1-u^2}}=-\frac{1}{x}\mathrm dx$ ,

两边积分，$arcsin(\frac{y}{x})=-ln|x|+c(c为任意常数)$ 。

x<0时，可得到$arcsin(\frac{y}{x})=ln|x|+c(c为任意常数)$。

另一解为y^2=x^2。

<font color="red">**正确**

可化为$\sin y\cos x=c$(c为任意常数)。
