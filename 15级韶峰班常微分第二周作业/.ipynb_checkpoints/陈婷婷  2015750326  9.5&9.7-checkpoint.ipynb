三、设y₁（t）是dy/dt=f(y），y（0）=y。的解，证明y₂(t)=y₁(t-t。)是dy/dt=f(y），y（t。）=y。的解.
   证明：dy₂(t)/dt=dy₁(t-t。)/dt=dy₁(t-t。)/(dt-t。)=f(y₁(t-t。))=f(y₂)；
   故y₂=y₁(t-t。)是dy/dt=f(y)的解；
   由y₁（0）=y。得t=0时y₁（0）=y。；
   故t-t。=0时y₁(t-t。）=y。，所以y₂（t。）=y。

27页4、给定一阶微分方程dy/dx=2x,
   (1)求出它的通解；
   解：∵dy/dx=2x;∴dy=2xdx
   则y=∫dy=∫2xdx=x^2+c (c为任意常数）
   故dy/dx=2x的通解为y=x^2+c (c为任意常数）
   
   (2)求通过点（1,4）的特解；
   解：将x=1,y=4代入y=x^2+c得c=3
   故通过点（1,4）的特解是y=x^2+3
   
   (3)求出与直线y=2x+3相切的解；
   解：由题意：y=x^2+c与y=2x+3有且只有一组解，即x^2+c=2x+3有唯一解，解得c=4
   故与直线y=2x+3相切的解为y=x^2+4

   ∴满足条件∫。¹ydx=2的解为y=x^2+5/3
