$(8)y=-\frac{g(x)}{f(x)},y'=\frac{f'(x)}{g(x))}y^2-\frac{g'(x)}{f(x)}$

验证 ：1. 求y的导数：

```python
var('x')
g=function('g')(x)
f=function('f')(x)
y=-g/f
show(diff(y,x))
```

2.代入等式右边运算

```python
var('x s')
s=(diff(f,x)/g)*y^2-diff(g,x)/f
show(s)
```

3.综上，得到等式左边等于等式右边，验证完毕。

## 五. 求下列方程的解

$（6）x\frac{\mathrm dy}{\mathrm dx}-y+\sqrt{x^2-y^2}=0$

* 当x不等于0时，得到$\frac{\mathrm dy}{\mathrm dx}-\frac{y}{x}+\sqrt{1-\frac{y^2}{x^2}}=0$
* 令$u=\frac{y}{x}$,则$\frac{\mathrm dy}{\mathrm dx}=x\frac{\mathrm du}{\mathrm dx}+\frac{y}{x}$
* 代入计算得： $x\frac{\mathrm du}{\mathrm dx}=\sqrt{1-u^2}$
* 化简得： $\frac{\mathrm du}{\sqrt{1-u^2}=-\frac{\mathrm dx}{x}}$
* 得到： $arcsin\frac{y}{x}=ln|x|+c$
* 当x等于0时，有$\sqrt{1-u^2}=0$,即$y^2=x^2$
* 所以解为$arcsin\frac{y}{x}=ln|x|+c$或$y^2=x^2$

$(7)\tan y \mathrm dx-\cot x \mathrm dy=0$

* 由已知得$\tan x\mathrm dx=\cot y\mathrm dy$
* 两边求积分得：$ln|\cos x|=-ln|\sin y|+c$
* 化简得：$\cos x\sin y=e^c$；可以看做$\cos x\sin y=c$,其中c为任意常数
* 又或者有 $\frac{\mathrm dx}{\mathrm dy}=\tan x\tan y$
* 所以又有解为$y=k\pi,k=0,\pm 1,\dots$
* 所以解为：$\cos x\sin y=c,其中c为任意常数或y=k\pi,k=0,\pm 1,\dots$

<font color="red">**正确**

## 六. 作适当的变量变换求解下列方程

$(6)\frac{\mathrm dy}{\mathrm dx}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$

* 由已知：令$u=y^3,则原式=\frac{\mathrm dy}{\mathrm dx}=\frac{u^2-2x^2}{y^2(2xu+x^2)}$
* 从而有：$x\frac{\mathrm du}{\mathrm dx}+u=\frac{3z^2+6}{2z+1}$
* 化简得：$x\frac{\mathrm du}{\mathrm dx}=\frac{z^2-z+6}{2z+1}$
* 当x不等于0时，有$\frac{2z+1}{z^2-z+6}\mathrm dz=\frac{\mathrm dx}{x}$
* 整理得：$\frac{7}{5}\frac{\mathrm dz}{z-3}+\frac{3}{5}\frac{\mathrm dz}{z+2}=\frac{\mathrm dx}{x}$
* 两边同时积分得：$(z-3)^7(z+2)^3=cx^5$；这里c为常数
* 代入并化简得：解为：$（y^3-3x）^7(y^3+2x)^3=cx^{15}$

<font color="red">**正确**

$(7)\frac{dy}{dx}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$

* 原方程可变形为$\frac{y\mathrm dy}{x\mathrm dx}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$
* 即$\frac{\mathrm dy^2}{\mathrm dx^2}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$；
* 令$u=x^2-1,v=y^2+1$；则上面方程化为：$\frac{\mathrm dv}{\mathrm du}=\frac{2u+3v}{3u+2v}$
* 再令$w=\frac{v}{u},即v=uw$,则上面方程化为$u\frac{\mathrm dw}{\mathrm du}+w=\frac{2u+3v}{3u+2v}$
* 分离变量并积分得：$c(1+w)=u^4(1-w)^5$
* 代入并计算得：$c(x^2+y^2)=(x^2-y^2-2)^5$；这里c是任意常数.

<font color="red">**正确**
