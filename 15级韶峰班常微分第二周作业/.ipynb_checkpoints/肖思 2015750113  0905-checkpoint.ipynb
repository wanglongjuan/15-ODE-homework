**无理整式**:与有理整式相对应,即除式含有变数字母的有理整式。

**4.验证下列各函数是相应微分方程的解**

$(1)y=x^2+1,y^\prime=y^2-(x^2+1)y+2x;$

$(2)y=-\frac {g(x)}{f(x)},y^\prime=\frac {f^\prime(x)}{g(x)}y^2-\frac {g^\prime(x)}{f(x)}.$

**验证**:

$(1)$等式左边=$2x$

等式右边=$(x^2+1)^2-(x^2+1)(x^2+1)+2x=2x$

等式左边=等式右边

$(2)$等式左边=$- \frac {g^\prime(x)f(x)-g(x)f^\prime(x)}{f^2(x)}$

等式右边=$ \frac {f^\prime(x)}{g(x)}(-\frac {g(x)}{f(x)})^2-\frac {g^\prime(x)}{f(x)}=- \frac {g^\prime(x)f(x)-g(x)f^\prime(x)}{f^2(x)}$

等式左边=等式右边

**6.求下列方程的解:**

(1)$x \frac {\mathrm d y}{\mathrm d x}- y + \sqrt{x^2-y^2} = 0$;

(2)$\tan y \mathrm d x - \cot x \mathrm d y =0$.

**解:**(1)将方程改写为
$$\frac {\mathrm d y}{\mathrm d x}=\frac {y}{x}-\sqrt{1-(\frac {y}{x})^2}$$

令$\frac {y}{x}=u$及$\frac {\mathrm d y}{\mathrm d x}=x\frac{\mathrm d u}{\mathrm d x}+u$代入,则方程变为
$$x\frac{\mathrm d u}{\mathrm d x}=-\sqrt{1-u^2}$$

1.$\sqrt{1-u^2}>0$时

分离变量,得到
$$\frac {\mathrm d u}{\sqrt{1-u^2}}=-\frac {1}{x}\mathrm d x$$

两边积分,得到通解$$\arcsin u =\ln |x|+c$$

将$u=\frac {y}{x}$代入得$$\arcsin\frac {y}{x}  =\ln |x|+c$$

2.$\sqrt{1-u^2}=0$时

$$y^2=x^2$$

**所以解为$$\arcsin\frac {y}{x}  =\ln |x|+c , \\y^2=x^2$$**


(2)分离变量得$$\frac {\mathrm d x}{\cot x}=\frac{\mathrm d y}{\tan y}$$

1.$\tan y \neq 0$

求积分$$-\ln |\cos x| =\ln |\sin y|+c$$

化解得$$\sin y \cos x =c^\prime$$($c^\prime$为常数)

2.$\tan y = 0$

则$y=k \pi,k \in Z$

**所以解为$$\sin y \cos x =c^\prime,\\y=k \pi,k \in Z$$**

<font color="red">**正确**

**7.作适当的变量变换求解下列方程:**

(1)$\frac {\mathrm d y}{\mathrm d x}=\frac {y^6-2x^2}{2xy^5+x^2y^2}$;

(2)$\frac {\mathrm d y}{\mathrm d x}=\frac {2x^3+3xy^2+x}{3x^2y+2y^3-y}$.

**解:**

$(1)由\frac {\mathrm d y}{\mathrm d x}=\frac {(y^3)^2-2x^2}{y^2(2xy^3+x^2)}$得到

$$\frac {\mathrm d y^3}{\mathrm d x}=\frac {3[(y^3)^2-2x^2]}{2xy^3+x^2}$$

令$u=y^3$,即可得到$$\frac {\mathrm d u}{\mathrm d x}=\frac {3(u^2-2x^2)}{2xu+x^2}=\frac {3[(\frac {u}{x})^2-2]}{2(\frac {u}{x}+1)}$$

令$\frac {u}{x}=a$,则$\frac {\mathrm d u}{\mathrm d x}=a +x\frac {\mathrm d a}{\mathrm d x}$代入上式得
$$x\frac {\mathrm d a}{\mathrm d x}=\frac {a^2-a-6}{2a+1}$$
**1.**  $a^2-a-6=0$时,$a=-2,a=3$,则$$y^3=-2x,y^3=3x$$

**2.**  $a^2-a-6\neq 0$时,$\frac {2a+1}{(a+2)(a-3)}\mathrm d a =\frac {\mathrm d x}{x}$,两边求积分得:
$$\frac {3}{5}\ln |a+2|+\frac{7}{5}\ln |a-3|=\ln |x|+c$$
代入$$\frac {3}{5}\ln |\frac {y^3}{x}+2|+\frac{7}{5}\ln |\frac {y^3}{x}-3|=\ln |x|+c$$

**则解为**$$(y^3+2x)^3(y^3-3x)^7=c^\prime x^{15}$$


$(2)由\frac {\mathrm d y}{\mathrm d x}=\frac {x(2x^2+3y^2+1)}{y(3x^2+2y^2-1)}$可得

$$\frac {\mathrm d y^2}{\mathrm d x^2}=\frac {2x^2+3y^2+1}{3x^2+2y^2-1}$$

令$x^2=a,y^2=b$,则$$\frac {\mathrm d b}{\mathrm d a}=\frac {2a+3b+1}{3a+2b-1}$$
解方程组$\begin {cases}
2a+3b+1=0\\3a+2b-1=0
\end {cases}$得交点$\begin {cases}
a=1\\b=-1
\end {cases}$

令$\begin {cases}
X=a-1\\Y=b+1
\end {cases}$代入得
$$\frac {\mathrm d Y}{\mathrm d X}=\frac {2X+3Y}{3X+2Y}=\frac {2+3\frac{Y}{X}}{3+2\frac {Y}{X}}$$
令$\frac {Y}{X}=u$,将$\frac {\mathrm d Y}{\mathrm d X}=u+X\frac {\mathrm d u}{\mathrm d X}$代入得
$$\frac {(3+2u)\mathrm d u}{2(1-u^2)}=\frac {\mathrm d X}{X}$$

**1.**$1-u^2=0$时,解为$y^2=x^2-2$或$y^2=-x^2$

**2.**$1-u^2\neq 0$时,两边积分得
$$y^2+x^2=c(y^2-x^2+2)^5$$
$\because y^2=x^2-2$或$y^2=-x^2$包含在$y^2+x^2=c(y^2-x^2+2)^5$内

$\therefore $ **解为**$$y^2+x^2=c(y^2-x^2+2)^5$$



<font color="red">**正确**
