3.验证下列各函数是相应微积分方程的解:

（7）$ y=x^2+1,y^\prime=y^2-(x^2+1)y+2x $

$证明:  ∵ y=x^2+1,y^\prime=2x $

$ y^\prime=y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)(x^2+1)+2x=2x=y^\prime $

∴$ y=x^2+1是方程 y^\prime=y^2-(x^2+1)y+2x的解$.

（8）$ y=-\frac{g(x)}{f(x)},y^\prime=\frac{f^\prime(x)}{g(x)}y^2-\frac{g^\prime(x)}{f(x)} $

$证明：∵ y=-\frac{g(x)}{f(x)} $

∴$ y^\prime=-\frac{g^\prime(x)f(x)-f^\prime(x)g(x)}{f^2(x)}$

$y^\prime=\frac{f^\prime(x)}{g(x)}y^2-\frac{g^\prime(x)}{f(x)}=\frac{f^\prime(x)}{g(x)}(-\frac{g(x）}{f(x)})^2-\frac{g^\prime(x)}{f(x)}=\frac{f^\prime(x)g(x)}{f^2(x)}-\frac{g^\prime(x)}{f(x)}=-\frac{g^\prime(x)f(x)-f^\prime(x)g(x)}{f^2(x)}=y^\prime$

∴$y=-\frac{g(x)}{f(x)}是方程y^\prime=\frac{f^\prime(x)}{g(x)}y^2-\frac{g^\prime(x)}{f(x)}的解$.

4.$给定一阶微分方程\frac{\mathrm d y}{\mathrm d x}=2x,$

（1）求出它的通解；

$∵\frac{\mathrm d y}{\mathrm d x}=2x,∴\mathrm d y=2x\mathrm d x $

$在方程\mathrm d y=2x\mathrm d x两边求积分，得y=x^2+c,c是任意常数.$

$即y=x^2+c,(c是任意常数)是方程\frac{\mathrm d y}{\mathrm d x}=2x的通解.$

（2）求出过点$（1,4）$的特解；

$把（1,4）代入方程y=x^2+c得c=3.$

$即过点（1,4）的特解为y=x^2+3.$

（3）求出与直线$y=2x+3$相切的解；

$∵相切  ∴\begin{cases}y=x^2+c\\y=2x+3\end{cases}有唯一解$

$即方程x^2+c-2x-3=0有唯一解,所以c=4.$

$∴与直线y=2x+3相切的解是y=x^2+4.$

（4）求出满足条件$\int_0^1 \mathrm y\,\mathrm{d}x=2$的解；

$由（1）可知，y=x^2+c是通解，∴\int_0^1 \mathrm y\,\mathrm{d}x=\int_0^1\mathrm (x^2+c)\,\mathrm{d}x=\left.(\frac{x^3}{3}+cx)\right|_0^1=\frac{1}{3}+c=2$

$解得c=\frac{5}{3},即满足条件的解是y=x^2+\frac{5}{3}.$

1.求下列方程的解：

（6）$x\frac{\mathrm d y}{\mathrm d x}-y+\sqrt{x^2-y^2}=0$

$原方程为 \frac{\mathrm d y}{\mathrm d x}=\frac{y}{x}-\sqrt{1-(\frac{y}{x})^2}=0.$

$令\frac{y}{x}=u,可得\frac{\mathrm d y}{\mathrm d x}=x\frac{\mathrm d u}{\mathrm d x}+u,$

$方程可化为x\frac{\mathrm d u}{\mathrm d x}+u=u-\sqrt{1-u^2}.$

$再代入\frac{y}{x}=u,可得-\frac{\mathrm d u}{\sqrt{1-u^2}}=\frac{1}{x}\mathrm d x.$

$两边同时积分，即得方程的解为\arcsin\frac{y}{x}+\ln|x|+c=0,c为任意常数.$

（7）$\tan y\mathrm d x-\cot x\mathrm d y=0$

$原方程为\tan x\mathrm d x=\cot y\mathrm d y$

$即方程的解为\ln |\cos x|+\ln|\sin x|+c=0，c为任意常数.$

2.作适当的变量变换求解下列方程：

（6）$\frac{\mathrm d y}{\mathrm d x}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$

$令u=y^3,则\frac{\mathrm d y}{\mathrm d x}=\frac{\mathrm d u}{3y^2\mathrm d x}$,

$方程可化为\frac{\mathrm d u}{\mathrm d x}=\frac{3u^2-6x^2}{2xu+x^2}.$

$再令v=\frac{u}{x},即u=vx,$

$可得方程为x\frac{\mathrm d v}{\mathrm d x}+v=\frac{3v^2-6}{2v+1},$

$即\frac{7}{5}\frac{\mathrm d v}{v-3}+\frac{3}{5}\frac{\mathrm d v}{v+2}=\frac{\mathrm d x}{x}$

$两边同时积分，得(v-3)^7(v+2)^3=cx^5$

$所以原方程的解为(y^3-3x)^7(y^3+2x)^3=cx^{15},c为任意常数.$

（7）$\frac{\mathrm d y}{\mathrm d x}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$

$原方程可化为\frac{y\mathrm d y}{x\mathrm d x}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1},即\frac{\mathrm d y^2}{\mathrm d x^2}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}.$

$令u=x^2-1,v=y^2+1,$

$可得\frac{\mathrm d v}{\mathrm d u}=\frac{2u+3v}{3u+2v}$

$分离变量后两边积分得到c(1+w)=u^4(1-w)^5$

$所以原方程的解为c(x^2+y^2)=(x^2-y^2+2)^5,c为任意常数.$
