$$(8).y=-\frac{g(x)}{f(x)};$$
$$证明：对y=-\frac{g(x)}{f(x)}求导得：$$
$$y'=-\frac{g'(x)f(x)-g(x)f'(x)}{f^2(x)}；$$
$$等式左边=y'=-\frac{g'(x)f(x)-g(x)f'(x)}{f^2(x)};$$
$$将y=-\frac{g(x)}{f(x)}代入等式右边有：$$
$$右边=\frac{f'(x)}{g(x)}(-\frac{g(x)}{f(x)})^2-\frac{g'(x)}{f(x)};$$
$$推出右边=-\frac{g'(x)f(x)-g(x)f'(x)}{f^2(x)}=y'=左边；$$
$$等式得证$$

$$(7)\frac{dy}{dx}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3_y}$$
$$解：将方程化为$$
$$\frac{d(y^2+1)}{d(x^2-1)}=\frac{2(x^2_1)+3(y^2+1）}{3（x^2-1)+2(y^2+1)};$$
$$取变换\begin{cases}
t=x^2-1,\\
s=y^2+1,
\end{cases}$$
$$化为齐次方程\frac{ds}{dt}=\frac{2t+3s}{3t+2s}.$$
$$再令s=wt,得t\frac{dw}{dt}=\frac{2-2w^2}{3+2w},$$
$$\frac{dt}{t}=\frac{3+2w}{2(1-w^2)}dw$$
$$=\frac{1}{4}(\frac{5}{1-w}+\frac{1}{1+w})dw;$$
$$对两边同时积分得-\frac{5}{4}ln|1-w|+\frac{1}{4}ln|1+w|=ln|t|=c,$$
$$(1+w)=ct^4(1-w)^5,$$
$$即t+s=c(t-s)^5,$$
$$所以通解为x^2+y^2=c(x^2-y^2-2)^5,其中c为任意数$$
