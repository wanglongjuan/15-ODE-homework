1.（6）求下列方程的解$$x*\frac{\mathrm dy}{\mathrm dx}-y+\sqrt{x^2-y^2}=0$$
解：设$u=\frac{y}{x}$，则$y=u*x$,则有：
$$\frac{\mathrm dy}{\mathrm dx}=u+x*\frac{\mathrm du}{\mathrm dx}$$
原式可化简为：$$x*(u+x*\frac{\mathrm du}{\mathrm dx})=u*x-\sqrt{x^2-y^2}$$
$$\frac{\mathrm du}{\mathrm dx}=\frac{\sqrt{x^2-y^2}}{x^2}$$
$$\frac{1}{\sqrt{1-u^2}}*\mathrm du=\frac{1}{|x|}*\mathrm dx$$
   等式两边同时求积分得：$$arcsin\frac{y}{x}=ln|x|+c$$
   另外还有解：$$x^2=y^2$$

2.(6)作适当的变量变换求解下列方程：$$\frac{\mathrm dy}{\mathrm dx}=\frac{y^6-2*x^2}{2*x*y^5+x^2*y^2}$$
解：令$u=y^3$,则$\mathrm du=3*y^2*\mathrm dy$，则：
$$\frac{\mathrm du}{\mathrm dx}=\frac{3*(u^2-2*x^2)}{2*x*u+x^2}$$
令$z=\frac{u}{x}$，则$u=z*x$，则：$$\frac{\mathrm du}{\mathrm dx}=z+x*\frac{\mathrm dz}{\mathrm dx}$$
原式可化为：$$z+x*\frac{\mathrm dz}{\mathrm dx}=\frac{3*z^2-6}{2*z+1}$$
$$x*\frac{\mathrm dz}{\mathrm dx}=\frac{z^2-z-6}{2*z+1}$$
$$\frac{2*z+1}{z^2-z-6}*\mathrm dz=\frac{1}{x}*\mathrm dx$$
等式两边同时求积分得到$$(z-3)^7(z+2)^3=x^5*c$$
当$z^2-z-6=0$,即$z=3$或者$z=-2$时，这个解包含在$c=0$的情况中。因此，解为：$$(y^3-3*x)^7(y^3+2*x)^3=x^{15}*c$$

2.(7)作适当的变量变换求解下列方程：$$\frac{\mathrm dy}{\mathrm dx}=\frac{2*x^3+3*x*y^2+x}{3*x^2*y+2*y^3-y}$$
令$y^2=u$,$x^2=v$,则$$\mathrm du=2*y*\mathrm dy,\mathrm dv=2*x*\mathrm dx$$
则：$$\frac{\mathrm du}{\mathrm dv}=\frac{2*v+3*u+1}{3*v+2*u-1}$$
联立方程：$$2*v+3*u+1=0和3*v+2*u-1=0$$
解得：$$u=-1,v=1$$
令$v=X+1,u=Y-1$,得到$$\frac{\mathrm dY}{\mathrm dX}=\frac{2*X+3*Y}{3*X+2*Y}$$
设$m=\frac{Y}{X}$,则$Y=mX$,则：$$\frac{\mathrm dY}{\mathrm dX}=m+X*\frac{\mathrm dm}{\mathrm dX}$$
原式可化为：$$m+X\frac{\mathrm dm}{\mathrm dX}=\frac{2+3*m}{3+2*m}$$
$$X*\frac{\mathrm dm}{\mathrm dX}=\frac{2-2*m^2}{3+2*m}$$
$$\frac{3+2*m}{2-2*m^2}*\mathrm dm=\frac{1}{X}*\mathrm dX$$
等式两边同时求积分，得到：$$y^2+x^2=(y^2-x^2+2)^5*c$$
当$2-2*m^2=0即m=\pm1即y^2=x^2-2或y^2=-x^2$时，包涵在$c=0$的解中。
所以原方程的解为：
$$y^2+x^2=(y^2-x^2+2)^5*c$$
