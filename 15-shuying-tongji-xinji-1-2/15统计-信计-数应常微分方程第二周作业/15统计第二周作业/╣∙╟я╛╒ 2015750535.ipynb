3.验证下列各函数是相应微分方程的解。  

$(7)y=x^2+1,y'=y^2-(x^2+1)y+2x$  

$解：因为y=x^2+1,所以对其求导y'=2x$  
$所以y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)^2+2x=y'$  
$即可证得函数y=x^2+1是该方程的解.$  




$(8)y=-g(x)/f(x),y'=f'(x)y^2/g(x)-g'(x)/f(x)$  

$解：对y求导得y'=-g'(x)f(x)+g(x)f'(x)/f(x)^2.$  
$f'(x)y^2/g(x)-g'(x)/f(x)=f'(x)g(x)/f(x)^2-g'(x)/f(x)=y'$  
$即可证得函数y=-g(x)/f(x)是该方程的解。$

$1.求下列方程的解：$  

$(6)xdy/dx-y+$$\sqrt{x^2-y^2}=0$  

$解：当x^2不等于y^2时，将该式变形为dy/dx=y/x-$$\sqrt{1-(y/x)^2}$  
$令y/x=u  ,dy/dx=xdu/dx+u  ,该式化为dx/x=du/-$$\sqrt{1-u^2}$  
$然后分别对两边求积分得arcsin(u)+ln|x|=c,c为任意常数。$  
$当x^2=y^2时，即y^2=x^2也是其的解$  
$所以该方程的解为y^2=x^2和arcsin(u)+ln|x|=c$  




$(7)tanydx-cotxdy=0$  

$解：将该式变形为dx/cotx=dy/tany$  
$对两边分别求积分得：-ln|cosx|=ln|sinx|+c$  
$因而，其通解为：ln|sinx|+ln|cosx|=c$  


$2.作适当的变量变换求解下列方程：$  

$(6)dy/dx=(y^6-2x^2)/(2xy^5+x^2y^2)$  

$解：令u=y^3,化为齐次方程du/dx=(3u^2-6x^2)/(2xu+x^2)$  
$再由变换u=tx将其化为变量分离方程dx/x=[7/5*1/(t-3)+3/5*1/(t+2)]dt.$  
$求得其解为（y^3-3x)^7(y^3+2x)^3=cx^(15)$  




$(7)dy/dx=(2x^3+3xy^2+x)/(3x^2y+2y^3-y)$  

$解：先将方程改写为d(y^2+1)/d(x^2+1)=[2(x^2-1)+3(y^2+1)]/[3(x^2-1)+2(y^2+1)]$  
$取变换s=y^2+1  t=x^2-1,将其化为其次方程ds/dt=(2t+3s)/(3t+2s),再令s=wt得$  
$tdw/dt=(2-2w^2)/(3+2w),dt/t=(3+2w)dw/2(1-w^2)=1/4[5/(1-w)+1/(1+w)]dw,积分得$  
$-5/4ln|1-w|+1/4ln|1+w|=ln|t|+c$  
$即可求得其通解为（x^2+y^2)c=(y^2-x^2+2)^5,其中c为任意常数。$  








$证明题：设y_1(t)是dy/dt=f(y),y(0)=y_0的解，证明y_2(t)=y_1(t-t_0)是$  
$dy/dt=f(y),y(t_0)=y_0的解。$  

$解：当t_0=0时，显然成立y_2(t)是其的解$  
$dy_2(t)/dt=dy_1(t-t_0)/dt=dy_1(t-t_0)/d(t-t_0)=f[y_1(t-t_0)]=f(y_2)$  
$所以y_2=y_1(t-t_0)也为dy/dt=f(y)的解$  
$因为y_1(0)=y_0,得出t=0时，y_1(0)=y_0$  
$所以t-t_0=0时，y_1(t-t_0)=y_0,y_2(t_0）=y_0$
