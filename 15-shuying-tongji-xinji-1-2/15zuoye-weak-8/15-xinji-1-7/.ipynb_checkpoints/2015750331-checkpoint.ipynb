$一,设x(t)和y(t)是区间a\leq t\leq b上的连续函数，证明：如果在区间a \leq t\leq b上有\frac {x(t)}{y(t)}\neq常数或\frac{y(t)}{x(t)}\neq常数,则x(t)和y(t)在区间a\leq t\leq b上线性无关.(提示用反证法)$

$证明:$

$假设x(t)和y(t)在区间[a,b]上线性相关，则存在不全为零的常数C_1和C_2使得$

$$C_1 x(t)+C_2 y(t)\equiv0,t\in[a,b].$$

$若C_2\neq 0$

$$\frac {y(t)}{x(t)}=-\frac{C_1}{C_2}=常数$$

$若C_1\neq 0$

$$\frac{x(t)}{y(t)}=-\frac{C_2}{C_1}=常数$$

$在两种情况下，都与已知条件\frac{x(t)}{y(t)}\neq常数或\frac{y(t)}{x(t)}\neq常数矛盾$

$所以假设不成立.$

$二,证明非齐次线性微分方程的叠加原理：设x_1(t),x_2(t)分别是非齐次线性微分方程$

$$\frac{d^n x}{dt^n}+a_1(t)\frac{{d^{n-1}}x}{d{t^{n-1}}}+\cdots+a_n(t)x=f_1(t),$$

$$\frac{d^n x}{dt^n}+a_1(t)\frac{{d^{n-1}}x}{d{t^{n-1}}}+\cdots+a_n(t)x=f_2(t)$$

$的解，则x_1(t)+x_2(t)是方程\frac{d^n x}{dt^n}+a_1(t)\frac{{d^{n-1}}x}{d{t^{n-1}}}+\cdots+a_n(t)x=f_1(t)+f_2(t)的解.$

$证明:$

$由于x_1(t),x_2(t)分别是非齐次线性微分方程$

$$\frac{d^n x}{dt^n}+a_1(t)\frac{{d^{n-1}}x}{d{t^{n-1}}}+\cdots+a_n(t)x=f_1(t),$$

$$\frac{d^n x}{dt^n}+a_1(t)\frac{{d^{n-1}}x}{d{t^{n-1}}}+\cdots+a_n(t)x=f_2(t)$$

$的解$

$所以满足:$

$$\frac{d^n x_1(t)}{dt}+a_1(t)\frac{{d^{n-1}}x_1(t)}{dt}+\cdots+a_n(t)x_1(t)=f_1(t),$$

$$\frac{d^n x_2(t)}{dt}+a_1(t)\frac{{d^{n-1}}x_2(t)}{dt}+\cdots+a_n(t)x_2(t)=f_2(t)$$

$把以上两式相加，得$$

$$\frac{d^{n}(x_1(t)+x_2(t))}{\mathrm d t^n}+a_1(t)\frac{d^{n-1}(x_1(t)+x_2(t))}{\mathrm d t^{n-1}}+\cdots+a_n(t)(x_1(t)+x_2(t))=f_1(t)+f_2(t)$$

$$说明(x_1(t)+x_2(t))是上式的解。$$

$三已知齐次线性微分方程的基本解组x_1,x_2,求下列方程对应的非齐次线性微分方程的通解：$

$1,\quad x^{{\prime}{\prime}}-x=cost,x_1=e^t,x_2=e^{-t}$

$解:$

$令x=C_1(t)e^t+C_2(t)e^{-t}$

$则可得决定C_1^{\prime}(t)和C_2^{\prime}(t)的两个方程$

$$e^tC_1^{\prime}(t)+e^{-t}C_2^{\prime}(t)=0$$

$及$

$$e^tC_1^{\prime}(t)-e^{-t}C_2^{\prime}(t)=cost$$

$解得$

$$C_1^{\prime}(t)=\frac 1 2 coste^{-t}$$

$$C_2^{\prime}(t)=\frac {-1}{2}e^t cost$$

$由此$

$$C_1(t)=\frac {sint-cost}{4}*e^{-t}+\gamma_1$$

$$C_2(t)=-\frac{1}{4}(sint+cost)e^t+\gamma_2$$

$于是原方程的通解为$

$$x=\gamma_1 e^t+\gamma_2 e^{-t}-\frac{1}{2}cost$$

$2,\quad  x^{{\prime}{\prime}}+4x=tsin2t,x_1=cos2t,x_2=sin2t$

$解$

$令x=C_1(t)cos2t+C_2(t)sin2t$

$则可得决定C_1^{\prime}(t)和C_2^{\prime}(t)的两个方程$

$$cos2t*C_1^{\prime}(t)+sin2t*C_2^{\prime}(t)=0$$

$及$

$$-2sin2tC_1^{\prime}(t)+2cos2t*C_2^{\prime}(t)=tsint$$

$解得$

$$C_1^{\prime}(t)=-\frac{1}{2}tsintsin2t,C_2^{\prime}(t)=\frac{1}{2}tsintcos2t$$

$由此$

$$C_1(t)=\frac{1}{4}[\frac 1 3 tsin3t+\frac 1 9 cos3t-tsint-cost]+C_1$$

$$C_2(t)=\frac 1 4[-\frac 1 3 tcos3t+\frac 1 9 sin3t+tcost-sint]+C_2$$

$于是方程的解为$

$$x=C_1cos2t+C_2sin2t+\frac {t}{3} sint-\frac 2 9 cost$$

$3,\quad t^2x^{{\prime}{\prime}}-tx^{\prime}+x=6t+34t^2,x_1=t,x_2=t\ln t;$

$解$

$方程可以变形为$

$$x^{{\prime}{\prime}}-\frac {1}{t}x^{\prime}+\frac{1}{t^2}x=\frac t 6+34$$

$令方程的解为x=tC_1(t)+t\ln(t)C_2(t)$

$则可得决定C_1^{\prime}(t)和C_2^{\prime}(t)的两个方程$

$$C_1^{\prime}(t)+t\ln(t)C_2^{\prime}(t)=0$$

$及$

$$C_1^{\prime}(t)+(\ln t+1)C_2^{\prime}(t)=\frac 6 t +34$$

$解得$

$$C_1^{\prime}(t)=-\ln{t(\frac 6 t+34)},C_2^{\prime}(t)=\frac 6 t +34$$

$由此$

$$C_1(t)=-34(t\ln t-t)-3(\ln t)^2 +C_1,C_2(t)=34t+6\ln t+C_2$$

$于是方程的通解为$

$$x=C_1t+C_2t\ln t+34t^2+3t(ln t)^2$$

$四\quad假设x_1(t)\neq 0是二阶齐次线性微分方程x^{{\prime}{\prime}}+a_1(t)x^{\prime}+a_2(t)x=0的解，这里a_1(t)和a_2(t)于区间[a,b]上连续，试证：$

$(1),x_2(t)为方程的解的充要条件是：W^{\prime}[x_1,x_2]+a_1W[x_1,x_2]=0;$

$证明$

$充分性:$

$因为$

$$W^{\prime}[x_1,x_2]=\begin{bmatrix}x_1^{\prime} & x_2^{\prime}\\ x_1^{\prime} & x_2^{\prime}\\ \end{bmatrix}+
\begin{bmatrix}x_1 & x_2\\x_1^{{\prime}{\prime}} & x_2^{{\prime}{\prime}}\\ \end{bmatrix}=\begin{bmatrix}x_1 & x_2\\x_1^{{\prime}{\prime}} & x_2^{{\prime}{\prime}}\\ \end{bmatrix}$$

$$W^{\prime}[x_1,x_2]+a_1(t)W[x_1,x_2]=\begin{bmatrix}x_1 & x_2\\x_1^{{\prime}{\prime}} & x_2^{{\prime}{\prime}}\\ \end{bmatrix}+
a_1(t)\begin{bmatrix}x_1 & x_2\\ x_1^{\prime} & x_2^{\prime}\\ \end{bmatrix}=\begin{bmatrix}x_1 & x_2\\x_1^{{\prime}{\prime}}+a_1(t)x_1^{\prime}& x_2^{{\prime}{\prime}}+a_1(t)x_2^{\prime}\\ \end{bmatrix}$$

$而x_1(t)\neq 0是已知方程的解，所以$

$$\begin{bmatrix}x_1 & x_2\\-a_2(t)x_1& x_2^{{\prime}{\prime}}+a_1(t)x_2^{\prime}\\ \end{bmatrix}=x_1\begin{bmatrix} 1&x_2\\-a_2(t)& x_2^{{\prime}{\prime}}
+a_1(t)x_2^{\prime}\\ \end{bmatrix}=0$$

$故有$

$$x_2^{{\prime}{\prime}}+a_1(t)x_2^{\prime}+a_2(t)x_2=0$$

$即x_2(t)是已知方程的解$

$必要性$

$因为W^{\prime}[x_1,x_2]是方程的解x_1(t),x_2(t)的朗斯基行列式，所以$

$$W^{\prime}[x_1,x_2]=\begin{bmatrix}x_1 & x_2\\x_1^{{\prime}{\prime}}
&x_2^{{\prime}{\prime}}\\ \end{bmatrix}=-a_1(t)\begin{bmatrix}x_1 & x_2\\ x_1^{\prime} & x_2^{\prime}\\ \end{bmatrix}=-a_1(t)W[x_1,x_2]$$

$即W[x_1,x_2]满足W^{\prime}[x_1,x_2]+a_1(t)W[x_1,x_2]=0$

$(2),因为x_1,x_2为方程的解，则由公式$

$$\begin{bmatrix}x_1 & x_2\\ x_1^{\prime} & x_2^{\prime}\\ \end{bmatrix}=
W(t_o)e^{-\int_{t_o}^{t} a_1(s)ds},即x_1x_2^{\prime}-x^{\prime}x^2=W(t_o)e^{\int_{t_o}^{t} a_1(s)ds}$$

$解得$

$$x_2=(c_1{\int \frac {1}{x^2} e^{-\int_{t_o}^{t} a_1(s) ds }dt}+c_2)x_1$$

$取c_1=1,c_2=0,得$

$$x_2=x_1{\int \frac {1}{x^2}e^{\int_{t_o}^{t} a_1(s)ds }dt}$$

$又$

$$W(t)=\begin{bmatrix}x_1 & x_2\\ x_1^{\prime} & x_2^{\prime}\\
\end{bmatrix}=e^{-\int_{t_o}^{t} a_1(s)ds} \neq 0$$  

$从而原方程的通解可表示为$

$$x=x_1[c_1\int \frac {1}{x^2}exp(-\int_{t_o}^{t} a_1(s) ds )dt+c_2],其中c_1,c_2为常数，t_0,t \in[a,b]$$
