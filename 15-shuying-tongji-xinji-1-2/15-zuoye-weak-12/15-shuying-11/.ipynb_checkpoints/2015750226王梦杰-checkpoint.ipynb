求解下列方程

$(2)xx''-(x')^2+(x')^3=0$

$(4)x''+\sqrt{1-(x')^2}=0$

$(6)x''-\frac{1}{t}x'+(x')^2=0(提示：方程两端除以x')$

解
（4）由题可知
令$x'=y,则方程化为$

$$
xy\frac{dy}{dx}-y^2+y^3=0
$$

解得

$$
y=0或者\frac{dy}{dx}=\frac{y-y^2}{x}
$$

$\frac{dy}{dx}=\frac{y-y^2}{x}积分后可得$

$$
\frac{y}{1-y}=c_1x(c_1\neq 0)
$$

即

$$
\frac{x'}{1-x'}=c_1x
$$

再次积分后可得

$$
x+\frac{1}{c_1}ln|x|=t+c_2
$$

当$y=0时，易得x=c$

综上所述

$$
x+\frac{1}{c_1}ln|x|=t+c_2或x=c
$$

(4)解  由题可知
$令x'=y,则方程可化为$

$$
y\frac{dy}{dx}+\sqrt{1-y^2}=0
$$

当$y=±1时，即x=±t+c_1$

当y≠1时，方程可以化为

$$
y\frac{dy}{dx}+\sqrt{1-y^2}=0
$$

积分后可得

$$
\sqrt{1-y^2}=x+c
$$

即

$$
\sqrt{1-(x')^2}=x+c
$$

积分后可得

$$
x=sin(t+c_1)-c或x=cos(t+c_1)-c
$$

综上所述

$$
x=sin(t+c_1)-c或x=cos(t+c_1)-c或x=±t+c_1
$$

其中$c_1,c是任意常数$

（6）解 由题可知
$当x'=0时$

易得$$x=c$$

$当x'≠0时，两边同时除以x'得$

$$
\frac{x''}{x'}-\frac{1}{t}+x'=0
$$

即$$d(ln|x'|-ln|t|+x)=0$$

积分后可得

$$
ln|\frac{x'}{t}|+x=c_1
$$

解得

$$
x'=t(e^{-x+c_1})
$$

再次积分后可得

$$
e^x=e^{c_1}(\frac{t^2}{2})+c_2
$$

综上所述

$$
e^x=e^{c_1}(\frac{t^2}{2})+c_2或x=c
$$

其中$c_1,c_2,c为任意常数$

用幂级指数法求解下列方程

$（2）(1-t)x''+x=0$

 由题可知

$设x=a_0+a_1t+a_2t^2+a_3t^3+……+a_nt^n是方程的解$

$$
x'=a_1+2a_2t+3a_3t^2+……+na_nt^{n-1}
$$

$$
x''=2a_2+6a_3t+……+n(n-1)a_nt^{n-2}
$$

将$x'',x',x代入原方程得

$$
\begin{cases}
a_0+2a_2=0\\
a_1-2a_2+6a_3=0\\
n(n-1)a_n-(n-1)(n-2)a_{n-1}+a_{n-2}=0
\end{cases}
$$

设$x_1(t),x_2(t)满足初值条件x_1(0)=0,x'_1(0)=1或者x_1(0)=1,x'(0)=0的 两个解$

易得

$$
a_0=0,a_1=1,a_2=0,a_3=\frac{-1}{3*2},a_4=\frac{-1}{4*3}……
$$

$$
a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)}
$$

即

$$
x_1(t)=t-\frac{t^3}{3!}-\frac{2t^4}{4!}-……
$$

同理得

$$
x_2(t)=1-\frac{t^2}{2!}-\frac{t^3}{3!}-\frac{t^4}{4!}-……
$$

方程通解为

$$
x=c_1x_1(t)+c_2x_2(t)
$$

$$
=c_1(t-\frac{t^3}{3!}-\frac{2t^4}{4!}-……)+c_2(1-\frac{t^2}{2!}-\frac{t^3}{3!}-\frac{t^4}{4!}-……)
$$
