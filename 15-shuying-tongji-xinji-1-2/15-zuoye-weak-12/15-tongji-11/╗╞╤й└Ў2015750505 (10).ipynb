$1.求解下列方程：$

$(2)xx"-(x')^2+(x')^3=0;$

$解：$

$该方程属于不显含t的类型，令x'=y则有：$

$$x"=\frac{dy}{dx}x'$$

$即$

$$x"=y\frac{dy}{dx}$$

$原方程变为：$

$$xy\frac{dy}{dx}-y^2+y^3=0.$$

$解得：$

$$y=0或\frac{dy}{dx}=\frac{y-y^2}{x}=0.$$

$积分，得：$

$$\frac{y}{1-y}=\tilde{c}x.$$

$即$

$$y=\frac{\tilde{c}x}{1+\tilde{c}x},$$

$从而：$

$$x'=\frac{\tilde{c}x}{1+\tilde{c}x},$$

$对x'积分，得：$

$$x+\frac{1}{\tilde{c}}ln|x|=t+c_2.$$

$当y=0时，即$

$$x'=0,所以x=c.$$

$综合以上，方程的通解为：$

$$x+c_1ln|x|=t+c_2,或x=c.(其中c_1,c_2,c为任意常数。）$$




$(4)x"+\sqrt{1-(x')^2}=0;$

$解：$

$该方程不显含t、x，可令x'=y则有：$

$$x"=\frac{dy}{dx}x'$$

$即$

$$x"=y\frac{dy}{dx}$$

$原方程变为：$

$$y\frac{dy}{dx}+\sqrt{1-y^2}=0.$$

$当y\neq\pm1时，有$

$\frac{ydy}{\sqrt{1-y^2}}=-dy.$

$积分，得：$

$$\sqrt{1-y^2}=x+c,$$

$所以$

$$x'=\pm\sqrt{1-(x+c)^2}.$$

$再次积分，得：$

$$x=\sin (c_1+t)+c_2或者x=\cos (c_1+t)+c_2.$$

$当y=\pm 1时，x'=\pm1,$

$从而，$

$$x=\pm t+c.$$

$综合以上，方程的通解为：$

$$x=x=\sin (c_1+t)+c_2或者x=\cos (c_1+t)+c_2、x=\pm t+c.（其中c_1,c_2,c_3为任意常数。）$$




$(6)x"-\frac{1}{t}x'+(x')^2=0(提示：方程两端除以x').$

$解：$

$当x'\neq0时，方程两边同除以x',则有：$

$$\frac{x"}{x'}-\frac{1}{t}+x'=0.$$

$即$

$$d(ln|x'|-ln|t|+x)=0$$

$积分，得：$

$$ln|\frac{x'}{t}|+x=c_1.$$

$解出x':$

$$x'=t(e^{-x+c_1}.$$

$再次积分，有：$

$$e^x=e^{c_1}(\frac{t^2}{2})+c_2.$$

$即$

$$e^x=c_1'(\frac{t^2}{2})+c_2.$$

$当x'=0时，有$

$$x=c.$$

$综合以上，方程的通解为：$

$$e^x=c_1'(\frac{t^2}{2})+c_2，或者x=c.（其中c_1,c_2,c为任意常数。）$$



$2.用幂级数解法求解下列方程：$

$(2)(1-t)x"+x=0;$

$解:$

$化简方程：$

$$x"+\frac{x}{1-t}=0.$$

$此时,tp(t)=0,t^2q(t)=\frac{t^2}{1-t},故存在幂级数解.$

$设x=a_0+a_1t+a_2t^2+…a_nt^n…是方程的解，则：$

$$x'=a_1+2a_2t+3a_3t^3+…+na_nt^{n-1}+….$$

$$x"=2a-2+3*2a_3t+…+n(n-1)a_nt^{n-2}+…$$

$将x、x'、x"的表达式代入原方程，比较同此次幂函数，得：$

$$2a_2+a_0=0,3*2a_3-2a_2+a_1=0,$$

$$k(k-1)a_k-(k-1)(k-2)a_{k-1}+a_{k-2}=0,k>3$$

$进而找两个线性无关的解：$

$设x_1(t)和x_2(t)是满足初值条件x_1(0)=0,x'_1(0)=1,x_2(0)=1,x'_2(0)=0的两个解$

$对于x_1(t)，有a_0=0,a_1=1，将其代入k(k-1)a_k-(k-1)(k-2)a_{k-1}+a_{k-2}=0，可得：$

$a_2=0,a_3=-\frac{1}{1*2*3},a_4=-\frac{1}{3*4}=-\frac{2}{1*2*3*4},a_5=-\frac{5}{1*2*3*4*5},...a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)}$

$即：$

$$x_1(t)=t-\frac{t^3}{3!}-\frac{2}{4!}t^4-\frac{5}{5!}t^5-\frac{18}{6!}t^6-...$$

$对于x_2(t)有a_0=1,a_1=0，代入k(k-1)a_k-(k-1)(k-2)a_{k-1}+a_{k-2}=0，可得：$

$a_2=-\frac{1}{2},a_3=-\frac{1}{1*2*3},a_4=-\frac{1}{1*2*3*4},...a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)}$

$即：$

$$x_2(t)=1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-...$$

$综合以上，方程的通解为$

$$x=c_1x_1(t)+c_2x_2(t)=c_1(t-\frac{t^3}{3!}-\frac{2}{4!}t^4-\frac{5}{5!}t^5-\frac{18}{6!}t^6-...)+c_2(1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-...)$$
