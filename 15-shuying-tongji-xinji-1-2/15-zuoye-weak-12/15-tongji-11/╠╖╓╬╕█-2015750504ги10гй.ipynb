$x'=y,可化为xy\frac{dy}{dx}-y^2+y^3=0.\quad 得到y=0,或者\frac{dy}{dx}=\frac{y-y^2}{x}$

$积分后得\frac{y}{1-y}=c_1x，\quad (c_1\neq 0) \quad即有\frac{x'}{1-x’}=c_1x$

$再次积分得x+\frac{1}{c_1} \ln |x|=t+c_2，\quad 当y=0时得x'=0或x=c$

$则方程的解为x+\frac{1}{c_1} \ln |x|=t+c_2或x=c  \quad c_1,c_2,c是任意常数$

$(4)y=x',则可化为y\frac{dy}{dx}+\sqrt{1-y^2}=0，\quad 当y \neq \pm 1时，即x'=\pm 1,x=\pm t+c$

$当y=\pm 1,方程可变形为\frac{ydy}{\sqrt{1-y^2}}=-dx,积分得\sqrt{1-y^2}=x+c,即\sqrt{1-x'^2}=x+c$

$再次积分得1.当x'=\sqrt{1-(x+c)^2}时,x=\sin (t+c_1)+c_2$

$2.当x'=-\sqrt{1-(x+c)^2}时,x=\cos (t+c_1)+c_2;$



$则方程的解为x=\sin (t+c_1)+c_2或x=\cos (t+c_1)+c_2或x=\pm t+c  \quad其中c_1,c_2，c为任意常数$

$(6)当x' \neq 0时，方程两边同时除以x'，则得到\frac{x"}{x'}-\frac{1}{t}+x'=0，$

$即d(\ln |x'|- \ln |t|+x)=0,积分得\ln |\frac{x'}{t}|+x=c_1,得x'=t(e^{-x+c}),再次积分得e^x=e^{c_1}(\frac{t^2}{2})+c_2$

$当x'=0时，即得x=c$

$则方程的解为e^x=c_1(t^2+c_2),或x=c  \quad c,c_1,c_2为任意常数$
