# 常微分作业十

**P182页**

$1.求解下列方程。$

$(2)xx''-(x'^2)+(x'^3)=0$

$解：令x'=y,x''=\frac{\mathrm dy}{\mathrm dx}\frac{\mathrm dx}{\mathrm dt}=yy'$

$所以原式就可以化为：xyy'-y^2+y^3=0$

$可解得：y=0或x\frac{\mathrm dy}{\mathrm dx}-y+y^2=0$

$\frac{\mathrm dy}{y-y^2}=\frac{\mathrm dx}{x},分别对两边积分得：$

$y=(1-y)xc\to y=\frac{xc}{xc-1}$

$即是\frac{\mathrm dx}{\mathrm dt}=\frac{xc}{xc-1}$

$x-\frac{1}{c}ln\mid x\mid=t+c_2,即方程有解为：x+c_1ln\mid x\mid=t+c_2(c_1=-\frac{1}{c}以及x=c.$

$~~~~~~~~$

$~~~~~~~~~$

$(4)x''+\sqrt 1-(x')^2=0$

$解：令x'=y,因为方程不含x和t,所以x''=y',即方程可化为：y'+\sqrt 1-y^2=0$

$\frac{\mathrm dy}{\sqrt 1-y^2}=-{\mathrm dt}\to y=\sin(c_1-t)\to x=\cos(t-c_1)+c_2$

$又因为y^2=1,y=\pm 1,x'=y,所以x=\pm t+c也是方程的解。即方程的解为：x=\cos(t-c_1)+c_2，x=\pm t+c$

$~~~~~~~~~$

$~~~~~~~~~$

$(6)x''-\frac{1}{t}x'+(x')^2=0 (提示两端除以x')$

$解：将两端同时除以x'得：\frac{x''}{x'}-\frac{1}{t}+x'=0$

$将两边分别积分得：x+ln\mid x'\mid=ln\mid t\mid \to x'=\frac{ct}{e^x}$

$e^x{\mathrm dx}=ct{\mathrm dt},即得方程的解为：e^x=c_1t^2+c_2$

$2.用幂级数解法求解下列方程。$

$(2)(1-t)x''+x=0$

$解：先将方程化为：x''+\frac{x}{1-t}=0$

$因为tp(t)=0,t^2q(t)=\frac{t^2}{1-t},故方程存在幂级数解。$

$设x=a_0+a_1t+a_2t^2+a_3t^3+\ldots+a_nt^n+\ldots$

$即得x''=2a_2+6a_3t+\ldots+n(n-1)a_nt^{n-2}+\ldots$

$将x,x''带入方程得：2a_2+a_0=0  3*2a_3-2a_2+a_1=0$

$k(k-1)a_k-(k-1)(k-2)a_{k-1}+a_{k-2}=0,k>3$

$设想x(0)=c_1,x'(0)=c_2,则有a_0=c_1,a_1=c_2,a_2=-\frac{c_1}{2},a_3=-\frac{c_1}{3!}-\frac{c_2}{3!}$

$a_k=\frac{k-2}{k}a_{k-1}-\frac{1}{k(k-1)}a_{k-2}(k>3)$

$即可求得幂级数解为：x=c_1(1-\frac{1}{2}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-\frac{2}{5!}t^5-\frac{7}{6!}t^6_\ldots)+c_2(t-\frac{1}{3!}t^3-\frac{2}{4!}t^4-\frac{5}{5!}t^5-\frac{18}{6!}t^6-\ldots)$

$(2)将下面的初值问题化为与之等价的一阶方程组的初值问题：$

$(3)\begin {cases}x''+5y'-7x+6y=e^t\\y''-2y+13y'-15x=\cos t\end{cases}\\x(0)=1,x'(0)=0,y(0)=0,y'(0)=1\\(提示：令w_1=x,w_2=x',w_3=y,w_4=y')$

3.试用逐步逼近法求方程组$${\mathrm x'}=\begin {bmatrix} 0&1\\-1&0\end {bmatrix}$$

$满足初值条件{\mathrm x(0)}=\begin {bmatrix} 0\\ 1 \end {bmatrix}的第三次近似解。$

$解：令方程组的解为\varphi(t),取\varphi_0(t)=x(0)=\begin {bmatrix} 0\\ 1 \end {bmatrix},于是第一次近似解为：$

$\varphi_1(t)=\varphi_0(t)+\int_0^tA(s)\varphi_0(s){\mathrm ds}=\begin {bmatrix} 0\\ 1 \end {bmatrix}+\int_0^t\begin {bmatrix} 0&1\\ -1&0 \end {bmatrix}\begin {bmatrix} 0\\ 1 \end {bmatrix}{\mathrm ds}=\begin {bmatrix} 0\\ 1 \end {bmatrix}+\int_0^t\begin {bmatrix} 0\\ 1 \end {bmatrix}{\mathrm ds}=\begin {bmatrix} 0\\ 1 \end {bmatrix}+\begin {bmatrix} 0\\ t \end {bmatrix}=\begin {bmatrix} t\\ 1 \end {bmatrix}$

$第二次近似解为：$

$\varphi_2(t)=\varphi_0(t)+\int_0^tA(s)\varphi_1(s){\mathrm ds}=\begin {bmatrix} 0\\ 1 \end {bmatrix}+\int_0^t\begin {bmatrix} 0&1\\ -1&0 \end {bmatrix}\begin {bmatrix} s\\ 1 \end {bmatrix}{\mathrm ds}=\begin {bmatrix} 0\\ 1 \end {bmatrix}+\int_0^t\begin {bmatrix}1\\ -s \end {bmatrix}{\mathrm ds}=\begin {bmatrix} 0\\ 1 \end {bmatrix}+\begin {bmatrix}t\\ -\frac{t^2}{2} \end {bmatrix}=\begin {bmatrix} t\\ 1-\frac{t^2}{2} \end {bmatrix}$

$第三次近似解为：$

$\varphi_3(t)=\varphi_0(t)+\int_0^tA(s)\varphi_2(s){\mathrm ds}=\begin {bmatrix} 0\\ 1 \end {bmatrix}+\int_0^t\begin {bmatrix} 0&1\\ -1&0 \end {bmatrix}\begin {bmatrix} s\\ 1-\frac{s^2}{2} \end {bmatrix}{\mathrm ds}=\begin {bmatrix} 0\\ 1 \end {bmatrix}+\int_0^t\begin {bmatrix}1-\frac{s^2}{2}\\ -s \end {bmatrix}{\mathrm ds}=\begin {bmatrix} 0\\ 1 \end {bmatrix}+\begin {bmatrix}t-\frac{t^3}{3!}\\ -\frac{t^2}{2!} \end {bmatrix}=\begin {bmatrix} t-\frac{t^3}{3!}\\ 1-\frac{t^2}{2!} \end {bmatrix}$
