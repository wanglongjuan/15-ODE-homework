$\bf P183.$

$1.求解下列方程$


$\bf(2)$

$xx"-(x')^2+(x')^3=0$

$\bf解：$

$令y=x'$

方程可变形为
$$xy\frac{dy}{dx}-y^2+y^3=0$$

$y(x\frac{dy}{dx}-y-y^2)=0$
$$y=0或\frac{dy}{dx}=\frac{y-y^2}{x}$$
$当\frac{dy}{dx}=\frac{y-y^2}{x}时$

$积分得$

$\frac{y}{1-y}=cx$

$y=\frac{cx}{1+cx}=x'$

$积分得$
$$x+\frac{1}{c}ln|x|=t+c_2$$
$x+c_1ln|x|=t+c_2$

$当y=0,x'=0$

$x=c_3$

$\therefore 方程的解为x+c_1ln|x|=t+c_2，或x=c_3$
$(c_1,c_2,c_3为任意常数)$

$\bf(4)$

$x"+\sqrt{1-(x')^2}=0$


$\bf解：$

$令y=x'$

$则y\frac{dy}{dx}+\sqrt{1-y^2}=0$

$当y\neq \pm1$

$$\frac{ydy}{\sqrt{1-y^2}}=-dx$$
$积分得$

$\sqrt{1-y^2}=x+c$

$\therefore x'=\pm \sqrt{1-(x+c)^2}$

积分

$$①   x'= \sqrt{1-(x+c)^2},x=\sin (t+c_1)+c_2$$
$$②   x'= -\sqrt{1-(x+c)^2},x=\cos (t+c_1)+c_3$$
$当y=\pm1$

$x'=\pm 1,x=\pm t+c$

$方程的解为x=\sin (t+c_1)+c_2,x=\cos (t+c_1)+c_3,x=\pm t+c$


$\bf(6)$

$x"-\frac{1}{t}x'+(x')^2=0$

$\bf解：$

$当x'\neq 0$

$\frac{x"}{x'}-\frac{1}{t}+x'=0$
$$d(ln|x'|-ln|t|+x)=0$$
$积分可得$

$ln|\frac{x'}{t}|+x=c_1$

$x'=t(e^{-x+c_1})$

$积分可得$
$$e^x=e^{c_1}(\frac{t^2}{2})+c_2$$

$即e^x=c_1'(t^2+c_2')（c_1,c_2为任意常数）$

$当x'= 0$

$x=c$

$方程的解为$

$e^x=c_1'(t^2+c_2')或者x=c$


$\bf P183.$

$2.用幂级数解法求解下列方程$


$\bf（2）$

$(1-t)x"+x=0$

$\bf解：$

$方程可化为$

$x"+\frac{x}{1-t}=0$

$tp(t)=0,t^2q(t)=\frac{t^2}{1-t}$
$$设x=a_0+a_1t+a_2t^2+a_3t^3+...+a_nt^n+...是方程的解$$
$$x'=a_1+2a_2t+3a_3t^2+...+na_nt^{n-1}+...,$$
$$x"=2a_2+6a_3t+...+n(n-1)a_n^{n-2}+...$$
$代入原方程,比较同次幂系数$

$2a_2+a_0=0,6a_3-2a_2+a_1=0,k(k-1)a_k-(k-1)(k-2)a_{k-1}+a_{k-2}=0,k>3$

$设x_1(t)和x_2(t)是满足初值条件x_1(0)=0,x'_1(0)=1,x_2(0)=1,x'_2(0)=0的两个解$

$对x_1(t)$

$a_0=0,a_1=1$

$代入k(k-1)a_k-(k-1)(k-2)a_{k-1}+a_{k-2}=0$

$a_2=0,a_3=-\frac{1}{1*2*3},a_4=-\frac{1}{3*4}=-\frac{2}{1*2*3*4},a_5=-\frac{5}{1*2*3*4*5},...a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)}$

$x_1(t)=t-\frac{t^3}{3!}-\frac{2}{4!}t^4-\frac{5}{5!}t^5-\frac{18}{6!}t^6-...$

$对x_2(t)$

$a_0=1,a_1=0$

$代入k(k-1)a_k-(k-1)(k-2)a_{k-1}+a_{k-2}=0$

$a_2=-\frac{1}{2},a_3=-\frac{1}{1*2*3},a_4=-\frac{1}{1*2*3*4},...a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)}$

$x_2(t)=1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-...$

$\therefore 方程的通解为$

$x=c_1x_1(t)+c_2x_2(t)=c_1(t-\frac{t^3}{3!}-\frac{2}{4!}t^4-\frac{5}{5!}t^5-\frac{18}{6!}t^6-...)+c_2(1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-...)$
