$1.求解下列方程：$  
$(2)xx''-(x')^2+(x')^3=0;$  
$解：方程不显含t，令y=x',则方程变为xy\frac{dy}{dx}-y^2+y^3=0,得到y=0或\frac{dy}{dx}=\frac{y-y^2}{x},$  
$积分后得到\frac{y}{1-y}=\overline{c}x(\overline{c}\neq0),$  
$即y=\frac{\overline{c}x}{1+\overline{c}x},从而x'=\frac{\overline{c}x}{1+\overline{c}x},$  
$积分得x+\frac{1}{\overline{c}}\ln|x|=t+c_2.$  
$即方程有解x+c_1\ln|x|=t+c_2.$  
$当y=0时得x'=0,故x=c.$  
$综上，方程的解为x+c_1\ln|x|=t+c_2或x=c,其中c_1,c_2,c是任意常数.$  

$(4)x''+\sqrt{1-(x')^2}=0;$  
$解：方程不显含t,x,令y=x',则方程化为y\frac{dy}{dx}+\sqrt{1-y^2}=0.$  
$当y\neq\pm1时，方程又可以变形为\frac{ydy}{\sqrt{1-y^2}}=-dx,$  
$积分后得到\sqrt{1-y^2}=x+c,$  
$即\sqrt{1-x'^2}=x+c,故x'=\pm\sqrt{1-(x+c)^2},再次积分得$  
$1'\quad当x'=\sqrt{1-(x+c)^2}时,x=\sin{(t+c_1)}+c_2;$  
$2'\quad当x'=-\sqrt{1-(x+c)^2}时,x=\cos(t+c_1)+c_2.$  
$当y=\pm1时x'=\pm1,从而x=\pm t+c.$  
$综上所述，原方程的解为x=\sin(t+c_1)+c_2或x=\cos(t+c_1)+c_2,以及x=\pm t+c,其中c_1,c_2,c为任意常数.$  

$(6)x''-\frac{1}{t}x'+(x')^2=0$  
$解：当x'\neq0时，方程两边同除以x',则得到\frac{x''}{x'}-\frac{1}{t}+x'=0,即$  
$d(\ln|x'|-\ln|t|+x)=0,$  
$积分可得\ln|\frac{x'}{t}|+x=c_1,$  
$解出x',则有x'=t(e^{-x+c_1}),再次积分可得e^x=e^{c_1}(\frac{t^2}{2})+c_2.$  
$由于c_1,c_2为任意常数，所以可以写为e^x=c'_1(t^2+c'^2).$  
$当x'=0时，即得x=c,显然它也是原方程的解.$  
$综上，方程的解为e^x=c'_1(t^2+c'_2)或x=c.$  

$2.用幂级数解法求解下列方程：$  
$(2)(1-t)x''+x=0;$  
$解：方程可化为x''+\frac{x}{1-t}=0,因tp(t)=0,t^2q(t)=\frac{t^2}{1-t},故存在幂级数解.$  
$设x=a_0+a_1t+a_2t^2+a_3t^3+\cdots+a_nt^n+\cdots是方程的解，则$  
$x'=a_1+2a_2t+3a_3t^2+\cdots+na_nt^{n-1}+\cdots,$  
$x''=2a_2+3·2a_3t+\cdots+n(n-1)a_nt^{n-2}+\cdots.$  
$将x、x'、x''的表达式代入原方程，比较同次幂级数，可得$  
$$2a_2+a_0=0,3·2a_3-2a_2+a_1=0,$$  
$$k(k-1)a_k-(k-1)(k-2)a_{k-1}+a_{k-2}=0,k>3.\quad[1]$$  
$本方程的解应该为通解，为此，需要找到两个线性无关的解.$  
$设x_1(t)和x_2(t)是满足初值条件x_1(0)=0,x'_1(0)=1,x_2'(0)=0的两个解$  
$对于x_1(t),显然有a_0=0,a_1=1,代入[1]式，可得$  
$a_2=0,a_3=\frac{-1}{1·2·3},a_4=\frac{-1}{3·4}=\frac{-2}{1·2·3·4}，a_5=\frac{-5}{1·2·3·4·5},\cdots,a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)},$  
$也就是x_1(t)=t-\frac{t^3}{3!}-\frac{2}{4!}t^4-\frac{5}{5!}t^5-\frac{18}{6!}t^6-\cdots,$  
$对于x_2(t),显然有a_0=1,a_1=0,代入[1]式可得$  
$a_2=-\frac{1}{2},a_3=\frac{-1}{1·2·3},a_4=\frac{-1}{1·2·3·4},\cdots,$  
$a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)},$  
$也就是x_2(t)=1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-\cdots,$  
$所以方程的通解为$  
$x=c_1x_1(t)+c_2x_2(t)=c_1(t-\frac{1}{3!}t^3-\frac{2}{4!}t^4-\frac{5}{5!}t^5-\frac{18}{6!}t^6-\cdots)+c_2(1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-\cdots).$
