$1.求解下列方程$  

$(2).xx''-(x')^2+(x')^3=0$  
$        \quad 解：令x'=y，则方程化为xy\frac{dy}{dx}-y^2+y^3=0$  
$            \quad 得到y=0或\frac{dy}{dx}=\frac{y-y^2}{x}$  
$            \quad 积分后得\frac{y}{1-y}=c_1x,(c_1 \neq 0)$  
$            \quad 即\frac{x'}{1-x'}=c_1x$  
$            \quad 再次积分得到x+\frac{1}{c_1}ln|x|=t+c_2$  
$            \quad 当y=0时，得x'=0,故x=c$  
$            \quad 综上所述，方程的解为x+\frac{1}{c_1}ln|x|=t+c_2或x=c，这里c_1,c_2,c是任意常数$  

$(4).x''+\sqrt{1-(x')^2}=0$  
$       \quad 解：令x'=y，则方程化为有y\frac{dy}{dx}+\sqrt{1-y^2}=0$  
$            \quad 当y=±1时，即x'=±t+c_1$  
$            \quad 当y\neq ±1时，方程又可化为\frac{ydy}{\sqrt{1-y^2}}=-dx$  
$            \quad 积分后得到\sqrt{1-y^2}=x+c$  
$            \quad 也就是\sqrt{1-x'^2}=x+c,或者x'=±\sqrt{1-(x+c)^2}$  
$            \quad 再次积分则得到$  
$            \quad ①当x'=+\sqrt{1-(x+c)^2}时，x=\sin (t+c_1)-c$  
$            \quad ②当x'=-\sqrt{1-(x+c)^2}时，x=\cos (t+c_1)-c$  
$            \quad 综合以上几种情况，即得到原方程的解为$  
$            \quad x=\sin (t+c_1)-c或x=\cos (t+c_1)-c，以及x=±t+c_1，这里c_1,c是任意常数$  

$(6).x''-\frac{1}{t}x'+(x')^2=0,(提示：方程两端除以x')$  
$       \quad 解：当x'\neq 0时，方程两边同除以x',则得到\frac{x''}{x'}-\frac{1}{t}+x'=0$  
$            \quad 即d(ln|x'|-ln|t|+x=0$  
$            \quad 两边积分可得ln|\frac{x'}{t}|+x=c_1$  
$            \quad 解出x'，则有x'=t（e^{-x+c_1}）$  
$            \quad 再次积分可得e^x=e^{c_1}(\frac{t^2}{2}+c_2$  
$            \quad 由于这里的c_1,c_2是任意常数，所以可以写成e^x=c_1(t^2+c_2)$  
$            \quad 当x'=0时，即得x=c,显然它也是原方程的解$  
$            \quad 综上所述，方程的解是e^x=c_1(t^2+c_2)或x=c$  

$2.用幂级数解法求解下列方程$  

$(2).(1-t)x''+x=0$  
$        \quad 解：设x=a_0+a_1t+a_2t^2+a_3t^3+\cdots+a_nt^n+\cdots是方程的解$  
$            \quad 则x'=a_1+2a_2t+3a_3t^2+\cdots+na_nt^{n-1}+\cdots$  
$            \quad  x''=2a_2+3 \bullet 2a_3t+\cdots+n(n-1)a_nt^{n-2}+\cdots$  
$            \quad 把x_1,x',x''的表达式代入原方程，合并t的各同次幂的项，并令各项系数等于零，得到$  
$$
\begin{cases}
a_0+2a_2=0\\
a_1-2a_2+3 \bullet 2a_3=0\\
n(n-1)a_n-(n-1)(n-2)a_{n-1}+a_{n-2}=0\\
\end{cases}
$$
$            \quad 由于本题设有初值条件，所以求出的应该是方程的通解，为此，需要找到两个线性无关的解，$  
$            \quad 我们设x_1(t)和x_2(t)是满足初值条件x_1(0)=0,x_1'(0)=1,x_2(0)=1,x_2'(0)=0的两个解$  
$            \quad 对于x_1(t),显然有a_0=0,a_1=1,代入方程组，可得$  
$            \quad a_2=0,a_3=\frac{-1}{3 \bullet 2},a_4=\frac{-1}{4 \bullet 3}=\frac{-2}{4 \bullet 3 \bullet 2},a_5=\frac{-5}{5 \bullet 4 \bullet 3 \bullet 2}，\cdots$  
$            \quad a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)},$  
$            \quad x_1(t)=t-\frac{t^3}{3!}-\frac{2}{4!}t^4-\frac{5}{5!}t^5 \cdots$  
$            \quad 对于x_2(t),显然有a_0=1,a_1=0,代入上边的方程组，可得$  
$            \quad a_2=-\frac{1}{2},a_3=\frac{-1}{3 \bullet 2},a_4=\frac{-1}{4 \bullet 3 \bullet 2},\cdots$  
$            \quad a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)},$  
$            \quad 即x_2(t)=1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-\cdots$  
$            \quad 所以方程的通解为x=c_1x_1(t)+c_2x_2(t)$ 
$                             \quad =c_1(t-\frac{1}{3!}t^3-\frac{2}{4!}t^4-\frac{5}{5!}t^5 \cdots)+c_2(1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4 \cdots)$
