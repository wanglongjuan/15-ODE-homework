$1.求解下列方程：$

$(2)xx''-(x')^2+(x')^3=0$

$(4)x''+\sqrt{1-(x')^2}=0$

$(6)x''-\frac{1}{t}x'+(x')^2=0$

$解：（2）令x'=y，则xy\frac{\mathrm d y}{\mathrm d x}-y^2+y^3=0$

$∴y=0或者\frac{\mathrm d y}{\mathrm d x}=\frac{y-y^2}{x}$

$积分得\frac{y}{1-y}=c_1x,即\frac{x'}{1-x'}=c_1x$

$再积分得x+\frac{1}{c_1}\ln \mid x\mid=t+c_2$

$当y=0时，x'=0,∴x=c$

$∴方程的解为x+\frac{1}{c_1}\ln \mid x\mid=t+c_2或x=c,(c_1,c_2,c_3为任意常数$

$（4）令x'=y，则y\frac{\mathrm d y}{\mathrm d x}+\sqrt{1-y^2}=0$

$当y=\pm 1时，即x'=\pm 1，则x=\pm t+c_1$

$当y\neq \pm 1时，\frac{y\mathrm d y}{\sqrt{1-y^2}}=-\mathrm d x$

$积分得\sqrt{1-y^2}=x+c,即\sqrt{1-{x'}^2}=x+c或x'=\pm \sqrt{1-(x+c)^2}$

$再积分得，当x'=+\sqrt{1-(x+c)^2}时，x=\sin(t+c_1)-c$

$当x'=-\sqrt{1-(x+c)^2}时，x=\cos(t+c_1)-c$

$∴原方程的解为x=\sin(t+c_1)-c,x=\cos(t+c_1)-c或x=\pm t+c_1(c_1,c_2为任意常数）$

$（6）当x'\neq =0时，方程可得\frac{x''}{x'}-\frac{1}{t}+x'=0$

$即\mathrm d (\ln\mid x'\mid -\ln \mid t\mid+x)=0$

$积分得\ln \mid \frac{x'}{t} \mid+x=c_1$

$解得x'=t(e^{=x+c_1})$

$再积分得e^x=e^{c_1}(\frac{t^2}{2})+c_2$

$当x'=0时，x=c$

$∴原方程的解为e^x=c_1(t^2+c_2)或x=c(因为c_1,c_2为任意常数$

$2.用幂级数解法求解下列方程：$

$(2)(1-t)x''+x=0$

$解：设x=a_0+a_1t+a_2t^2+\dots+a_nt^n+\dots是方程的解$

$则x'=a_1+2a_2t+3a_3t^2+\dots+na_nt^{n-1}+\dots$

$x''=2a_2+3\bullet 2a_3t+\dots+n(n-1)a_nt^{n-2}+\dots$

$将x,x',x''代入方程，可得$

$\begin{cases}a_0+2a_2=0\\a_1-2a_2+3\bullet 2a_3=0\\n(n-1)a_n-(n-1)(n-2)a_{n-1}+a_{n-2}=0\\ \end{cases}$

$设x_1(t)和x_2(t)是满足初值条件x_1(0)=0,x'_1(0)=1,x_2(0)=1,x'_2(0)=0的两个解$

$对于x_1(t),有a_0=0,a_1=1,带入方程组$

$得a_2=0,a_3=\frac{-1}{3\bullet 2},a_4=\frac{-1}{4\bullet 3}=\frac{-2}{4\bullet 3 \bullet 2},a_5=\frac{-5}{5\bullet 4\bullet 3\bullet 2},a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)}$

$即x_1(t)=t-\frac{t^3}{3!}-\frac{2}{4!}t^4-\frac{5}{5!}t^5\dots$

$对于x_2(t),有a_0=1,a_1=0,带入方程组$

$得a_2=-\frac{1}{2},a_3=\frac{-1}{3\bullet 2},a_4=\frac{-1}{4\bullet 3 \bullet 2},\dots,a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)}$

$即x_2(t)=1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-\dots$

$∴方程的通解为x=c_1x_1(t)+c_2x_2(t)=c_1(t-\frac{t^3}{3!}-\frac{2}{4!}t^4-\frac{5}{5!}t^5\dots)+c_2(1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4\dots)$

$2.将下面的初值问题化为与之等价的一阶方程组的初值问题：$

$(3)\begin{cases}x''+5y'-7x+6y=e^t\\y''-2y+13y'-15x=\cos t\\ \end{cases}$

$x(0)=1,x'(0)=0,y(0)=0,y'(0)=1$

$解：令{\omega}_1=x,{\omega}_2=x',{\omega}_3=y,{\omega}_4=y',则化方程组为$

$\begin{pmatrix}{\omega}_1\\{\omega}_2\\{\omega}_3\\{\omega}_4\\ \end{pmatrix}'=\begin{pmatrix}0&1&0&0\\7&0&-6&-5\\0&0&0&1\\15&0&-2&-13 \end{pmatrix}\begin{pmatrix}{\omega}_1\\{\omega}_2\\{\omega}_3\\{\omega}_4\\ \end{pmatrix}+\begin{pmatrix}0\\e^t\\0\\ \cos t \end{pmatrix}$

$初值条件为{\omega}_1(0)=1,{\omega}_2(0)=0,{\omega}_3(0)=0,{\omega}_4(0)=1$
        
$3.试用逐步逼近法求方程组x'=\begin{bmatrix}0&1\\-1&0\\ \end{bmatrix}x,满足初值条件x(0)=\begin{bmatrix}0\\1\\ \end{bmatrix}的第三次近似解.$

解：$${\varphi}_0=\begin{bmatrix}0\\1 \end{bmatrix}$$

$${\varphi}_1(t)=\begin{bmatrix}0\\1 \end{bmatrix}+\int^t_0 \begin{bmatrix}0&1\\-1&0 \end{bmatrix}\begin{bmatrix}0\\1 \end{bmatrix}\mathrm d s=\begin{bmatrix}0\\1 \end{bmatrix}+\int^t_0\begin{bmatrix}1\\0 \end{bmatrix}\mathrm d s=\begin{bmatrix}0\\1 \end{bmatrix}+\begin{bmatrix}t\\0 \end{bmatrix}=\begin{bmatrix}t\\1 \end{bmatrix}$$

$${\varphi}_2(t)=\begin{bmatrix}0\\1 \end{bmatrix}+\int^t_0 \begin{bmatrix}0&1\\-1&0 \end{bmatrix}\begin{bmatrix}t\\1 \end{bmatrix}\mathrm d t=\begin{bmatrix}0\\1 \end{bmatrix}+\int^t_0\begin{bmatrix}1\\-t \end{bmatrix}\mathrm d t=\begin{bmatrix}0\\1 \end{bmatrix}+\begin{bmatrix}t\\-\frac{1}{2}t^2 \end{bmatrix}=\begin{bmatrix}t\\1-\frac{1}{2}t^2 \end{bmatrix}$$

$${\varphi}_3(t)=\begin{bmatrix}0\\1 \end{bmatrix}+\int^t_0 \begin{bmatrix}0&1\\-1&0 \end{bmatrix}\begin{bmatrix}s\\1-\frac{1}{2}s^2\end{bmatrix}\mathrm d s=\begin{bmatrix}0\\1 \end{bmatrix}+\int^t_0\begin{bmatrix}1-\frac{1}{2}s^2\\-s \end{bmatrix}\mathrm d s=\begin{bmatrix}0\\1 \end{bmatrix}+\begin{bmatrix}t-\frac{1}{6}t^3\\-\frac{1}{2}t^2 \end{bmatrix}=\begin{bmatrix}t-\frac{1}{6}t^3\\1-\frac{1}{2}t^2 \end{bmatrix}$$

$∴满足初值条件x(0)=\begin{bmatrix}0\\1 \end{bmatrix}的第三次近似解为{\varphi}_3(t)=\begin{bmatrix}t-\frac{1}{6}t^3\\1-\frac{1}{2}t^2 \end{bmatrix}$
