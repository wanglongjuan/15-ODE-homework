$1.$求解下列方程：

$(1).xx''-(x')^2+(x')^3=0$

解：令$x'=y,$方程化为$$xy\frac{dy}{dx}-y^2+y^3=0$$  于是得$$y=0或\frac{dy}{dx}=\frac{y-y^2}{x}$$  积分后可得$$\frac{y}{1-y}=c_1x（c_1\neq 0）$$将$x'=y,$代入得$$\frac{x'}{1-x'}=c_1x$$  再次积分得到$$x+\frac{1}{c_1}ln|x|=t+c_2$$  当$y=0$时得$x'=0,所以x=c$

综合得，方程的解为$$x+\frac{1}{c_1}ln|x|=t+c_2或x=c$$  这里$c,c_1,c_2$为任意常数

$(2).x''+\sqrt{1-(x')^2}=0$

解：令$x'=y$,方程化为$$y\frac{dy}{dx}+\sqrt{1-y^2}=0$$  当$y=\pm1时，即x'=\pm1,那么x=\pm t+c_1$ 

当$y\neq\pm1$时，方程变形为$$\frac{ydy}{\sqrt{1-y^2}}=-dx$$  积分后得$$\sqrt{1-y^2}=x+c$$  即$$x'=\pm\sqrt{1-(x+c)^2}$$
再次积分得$$①当x'=+\sqrt{1-(x+c)^2}时，x=sin(t+c_1)-c$$   $$②当x'=-\sqrt{1-(x+c)^2}时，x=cos(t+c_1)-c$$
综上得，原方程的解为$$x=\pm t+c_1，x=sin(t+c_1)-c,或x=cos(t+c_1)-c$$  这里$c,c_1$为任意常数

$(3).x''-\frac{1}{t}x'+(x')^2=0$(提示：方程两端除以$x'$)

解：当$x'\neq 0$时，方程两边同时除以$x'$得$$\frac{x''}{x'}-\frac{1}{t}+x'=0$$   即$$d(ln|x'|-ln|t|+x)=0$$  积分得$$ln|\frac{x'}{t}|+x=c_1$$   解得$$x'=t(e^{-x+c_1})$$  再次积分得$$e^x=e^{c_1}(\frac{t^2}{2})+c_2$$  因为这里$c_1,c_2$是任意常数，所以可以写成$$e^x=c_1(t^2+c_2)$$  当$x'=0$时，得$x=c$,显然它是方程的解

综合得，方程的解是$$e^x=c_1(t^2+c_2)或x=c$$

$2.$用幂级数解法求解下列方程：

$(1)(1-t)x''+x=0$

解：设$x=a_0+a_1t+a_2t^2+...+a_nt^n+...$是方程的解，则$$x'=a_1+2a_2t+...+na_nt^{n-1}+...$$  

$$x''=2a_2+3.2a_3t...+n(n-1)a_nt^{n-2}+...$$  

把$x_1,x',x''$的表达式代入原方程，同时合并t的各同次幂的项，令各项系数等于零，得①式如下


$$
\begin{cases}
a_0+2a_2=0,\\
a_1-2a_2+3.2a_3=0,\\
n(n-1)a_n-(n-1)(n-2)a_{n-1}+a_{n-2}=0
\end{cases}
$$


设$x_1(t),x_2(t)$是满足初始条件$x_1(0)=0,x_1'(0)=1,x_2(0)=1,x_2'(0)=0$的两个解。那么对于$x_1(t)$,显然有$a_0=0,a_1=1,$代入①式得

$$a_2=0,a_3=\frac{-1}{3.2},a_4=\frac{-2}{4.3.2},a_5=\frac{-5}{5.4.3.2},...$$  

$$a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)}$$  则$$x_1(t)=t-\frac{1}{3!}t^3-\frac{2}{4!}t^4-\frac{5}{5!}t^5...$$


对于$x_2(t)$,显然有$a_0=1,a_1=0,$代入①式得$$a_2=-\frac{1}{2},a_3=\frac{-1}{3.2},a_4=\frac{-1}{4.3.2},...$$  

$$a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)}$$  则$$x_1(t)=1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-...$$

所以方程的通解为$$x=c_1(t-\frac{1}{3!}t^3-\frac{2}{4!}t^4-\frac{5}{5!}t^5...)+c_2(1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-...)$$  这里$c_1,c_2$为任意常数

$3.$将下面的初值问题化为与之等价的一阶方程组的初值问题：

$(1)$
$
\begin{cases}
x''+5y'-7x+6y=e^t\\
y''-2y+13y'-15x=cost
\end{cases}
$

$x(0)=1,x'(0)=0,y(0)=0,y'(0)=1$

（提示：令$\omega_1=x,\omega_2=x',\omega_3=y,\omega_4=y'$）

解：令$\omega_1=x,\omega_2=x',\omega_3=y,\omega_4=y'$，则化为方程组
$$
\begin{bmatrix}
\omega_1\\
\omega_2\\
\omega_3\\
\omega_4\\
\end{bmatrix}'=
\begin{bmatrix}
0&1&0&0\\
7&0&-6&-5\\
0&0&0&1\\
15&0&2&-13\\
\end{bmatrix}\begin{bmatrix}
\omega_1\\
\omega_2\\
\omega_3\\
\omega_4\\
\end{bmatrix}+\begin{bmatrix}
0\\
e^t\\
0\\
cost\\
\end{bmatrix}
$$
初值条件为$\omega_1(0)=1,\omega_2(0)=0,\omega_3(0)=0,\omega_4(0)=1$

$4.$试用逐步逼近法求方程组$$x'=
\begin{bmatrix}
0&1\\
-1&0
\end{bmatrix}
$$
满足初始条件$$x(0)=
\begin{bmatrix}
0\\
1
\end{bmatrix}$$的第三次近似解

解：$$
\phi_0(t)=
\begin{bmatrix}
0\\
1
\end{bmatrix}$$

$$\begin{aligned}
\phi_1(t)&=
\begin{bmatrix}
0\\
1
\end{bmatrix}+\int_0^t \begin{bmatrix}
0&1\\
-1&0
\end{bmatrix}\begin{bmatrix}
0\\
1
\end{bmatrix}ds\\
&=\begin{bmatrix}
0\\
1
\end{bmatrix}+\int_0^t \begin{bmatrix}
1\\
0
\end{bmatrix}ds\\
&=\begin{bmatrix}
0\\
1
\end{bmatrix}+\begin{bmatrix}
t\\
0
\end{bmatrix}=\begin{bmatrix}
t\\
1
\end{bmatrix}
\end{aligned}
$$


$$\begin{aligned}
\phi_2(t)&=
\begin{bmatrix}
0\\
1
\end{bmatrix}+\int_0^t \begin{bmatrix}
0&1\\
-1&0
\end{bmatrix}\begin{bmatrix}
0\\
1
\end{bmatrix}dt\\
&=\begin{bmatrix}
t\\
1
\end{bmatrix}+\int_0^t \begin{bmatrix}
1\\
-t
\end{bmatrix}dt\\
&=\begin{bmatrix}
0\\
1
\end{bmatrix}+\begin{bmatrix}
t\\
-\frac{1}{2}t^2
\end{bmatrix}=\begin{bmatrix}
t\\
1-\frac{1}{2}t^2
\end{bmatrix}
\end{aligned}
$$


$$\begin{aligned}
\phi_3(t)&=
\begin{bmatrix}
0\\
1
\end{bmatrix}+\int_0^t \begin{bmatrix}
0&1\\
-1&0
\end{bmatrix}\begin{bmatrix}
s\\
1-\frac{1}{2}s^2
\end{bmatrix}ds\\
&=\begin{bmatrix}
0\\
1
\end{bmatrix}+\int_0^t \begin{bmatrix}
1-\frac{1}{2}s^2\\
-s
\end{bmatrix}ds\\
&=\begin{bmatrix}
0\\
1
\end{bmatrix}+\begin{bmatrix}
t-\frac{1}{6}t^3\\
-\frac{1}{2}t^2
\end{bmatrix}=\begin{bmatrix}
t-\frac{1}{6}t^3\\
1-\frac{1}{2}t^2
\end{bmatrix}
\end{aligned}
$$

所以满足初始条件$$x(0)=
\begin{bmatrix}
0\\
1
\end{bmatrix}$$的第三次近似解为$$\phi_3(t)=\begin{bmatrix}
t-\frac{1}{6}t^3\\
1-\frac{1}{2}t^2
\end{bmatrix}
$$
