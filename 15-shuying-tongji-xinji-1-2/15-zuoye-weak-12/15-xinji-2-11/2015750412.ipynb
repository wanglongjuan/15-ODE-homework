1.（2）

令$y=x'$,则原式变为:

$$\quad xy\frac{\mathrm dy}{\mathrm dx}-y^2+y^3=0$$

解得$y=0\quad或\quad\frac{\mathrm dy}{\mathrm dx}=\frac{\mathrm y-y^2}{\mathrm x}$

积分得：

$$y=\frac{\mathrm cx}{\mathrm cx-1}$$

所以$x'=\frac{\mathrm cx}{\mathrm cx-1}$,积分得:

$$x-\frac{1}{\mathrm c}\ln{x}=t+c_2$$

即得方程有解

$x+c_1\ln{x}=t+c_2$,当y=0时，$x=c$

所以方程有解$x+\frac{\mathrm 1}{\mathrm c_1}ln|x|=t+c_2或 x=c$，$c_1,c_2,c为任意常数$

1.(4)

令：$y=x'$，则原式变为：

$$y\frac{\mathrm dy}{\mathrm dx}+\sqrt{1-y^2}=0$$

当$y=\pm 1时：x'=\pm 1,则x=\pm t+c_1$

当$y\ne \pm1$时,方程变为：

$$\frac{y dy}{\sqrt{1-y^2}}=-dx$$

积分得；

$$\sqrt{1-x'^2}=x+c$$

①当$x’=\sqrt{1-(x+c)^2}时：x=\sin(t+c_1)-c$

②当$x’=-\sqrt{1-(x+c)^2}，x=\cos(t+c_1)-c$

所以原方程的解为:

$x=\pm t+c_1；x=\cos(t+c_1)-c；x=\sin(t+c_1)-c$

1.(6)

当$x'\ne 0$时，方程两边同时除以$x'$，得：

$$\frac{\mathrm x''}{\mathrm x'}-\frac{\mathrm 1}{\mathrm t}+x'=0$$

即$d(ln|x'|-ln|t|+x)=0$

积分得：

$$ln|\frac{\mathrm x'}{\mathrm t}|+x=c_1$$

解得$x'=t(e^{-x+c_1})$

再次积分:

$$e^x=e^{c_1}(\frac{\mathrm t^2}{\mathrm 2})+c_2$$

化简:

$$e^x=c_1(t^2+c_2)$$

当$x'=0$时，可知$x=c$，为方程的解。

所以方程的解为：$e^x=c_1(t^2+c_2)或x=c$

2.（2）

设$x=a_0+a_1t+a_2t^2+a_3t^3+ \dots +a_nt^n+ \dots$是方程的解。

$则有：
\begin{aligned}
x'&=a_1+2a_2t+3a_3t^2+\dots +nat^{n-1}+\dots\\
x''&=2a_2+3·2a_3t+\dots+n(n-1)a_nt^{n-2}+\dots\\
\end{aligned}
$

代入方程得：

$$
\begin{cases}
a_0+2a_2=0\\
a_1-2a_2+3·2a_3=0\\
n(n-1)a_n-(n-1)(n-2)a_{n-1}+a_{n-2}=0\\
\end{cases}
$$

设$x_1(t)$和$x_2(t)$是方程的两个线性无关的解，且$x_1(0)=0,x_1'(0)=1,x_2(0)=1,x_2'(0)=0$

①$x_1(t)，a_0=0,a_1=1,$将

$$a_2=0,a_3=-\frac{\mathrm 1}{\mathrm 3!},a_4=-\frac{\mathrm 1}{\mathrm 4*3}=-\frac{\mathrm 2}{\mathrm 4!},a_5=-\frac{\mathrm 5}{\mathrm 5!},\dots ,a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)}$$

代入

可得$x_1(t)=t-\frac{\mathrm 1}{\mathrm 3!}t^3-\frac{\mathrm 2}{\mathrm 4!}t^4-\frac{\mathrm 5}{\mathrm 5!}t^5 \dots $

②$x_2(t),a_0=1,a_1=0$，将

$$a_2=-\frac{\mathrm 1}{\mathrm 2},a_3=-\frac{\mathrm 1}{\mathrm 3!},a_4=-\frac{\mathrm 1}{\mathrm 4!},\dots,a_n=\frac{\mathrm (n-1)(n-2)a_{n-1}-a_{n-2}}{\mathrm n(n-1)}$$

得

可得$x_1(t)=1-\frac{\mathrm 1}{\mathrm 2!}t^2-\frac{\mathrm 1}{\mathrm 3!}t^3-\frac{\mathrm 1}{\mathrm 4!}t^4\dots $

所以通解为:$$x_1(t)=c_1(t-\frac{\mathrm 1}{\mathrm 3!}t^3-\frac{\mathrm 2}{\mathrm 4!}t^4-\frac{\mathrm 5}{\mathrm 5!}t^5 \dots)+c_2(1-\frac{\mathrm 1}{\mathrm 2!}t^2-\frac{\mathrm 1}{\mathrm 3!}t^3-\frac{\mathrm 1}{\mathrm 4!}t^4\dots)$$
