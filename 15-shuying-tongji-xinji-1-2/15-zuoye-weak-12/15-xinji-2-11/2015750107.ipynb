# 习题4.3

1.求解下列方程：

（2）$xx''-(x')^2+(x')^3=0$；

（4）$x''+\sqrt{1-(x')^2}=0$；

（6）$x''-\frac1tx'+(x')^2=0$（提示：方程两端除以$x'$）.

解： 

（2）

令$x'=y$，则方程化为，$xy\frac{dy}{dx}-y^2+y^3=0$

得，$y=0$

或，$\frac{dy}{dx}=\frac{y-y^2}{x}$

积分后得，$\frac y{1-y}=c_1x,(c_1\neq0)$

即，$\frac{x'}{1-x'}=c_1x$

再次积分得，$x+\frac1{c_1}\ln|x|=t+c_2$

当$y=0$时，$x'=0$

故，$x=c$

综上，方程的解为，$x+\frac1{c_1}\ln|x|=t+c_2或x=c$，其中$c_1,c_2,c$是任意常数

（4）

令$x'=y$,则方程化为，$y\frac{dy}{dx}+\sqrt{1-y^2}=0$

当y=±1时，即$x'=±t+c_1$

当$y\neq±1$时，方程又可以变形为

$\frac{ydy}{\sqrt{1-y^2}}=-dx$

积分后得，$\sqrt{1-y^2}=x+c$

即，$\sqrt{1-x'^2}=x+c

或，x'=±\sqrt{1-(x+c)^2}$

再次积分得，

①当$x'=+\sqrt{1-(x+c)^2}时，x=\sin (t+c_1)-c$

②当$x'=-\sqrt{1-(x+c)^2}时，x=\cos (t+c_1)-c$

综上，可得原方程的解为，$x=\sin (t+c_1)-c$,或$x=\cos (t+c_1)-c$,以及$x=±t+c_1，c_1,c$是任意常数。

（6）

当$x'\neq0$时，

方程两边同时除以$x'$

得，$\frac{x''}{x'}-\frac1t+x'=0$

即，$d(\ln|x'|-\ln|t|+x)=0$

两边积分得，$\ln|\frac{x'}t|+x=c_1$

即，$x'=t(e^{-x+c_1})$

再次积分得，$e^x=e^{c_1}(\frac{t^2}2)+c_2$

由于这里的$c_1,c_2$时任意常数，所以可以写成

$e^x=c_1(t^2+c_2)$

当$x'=0$时，

$x=c$

显然它也是原方程的解

综上，方程的解为，$e^x=c_1(t^2+c_2)$或$x=c$

2.用幂级数解法求解下列方程:

（2）(1-t)x''+x=0

解：

设$x=a_0+a_1t+a_2t^2+a_3t^3+\cdots+a_nt^n+\cdots$是方程的解，则

$x'=a_1+2a_2t+3a_3t^2+\cdots+na_nt^{n-1}+\cdots$

$x''=2a_2+6a_3t+\cdots+n(n-1)a_nt^{n-2}+\cdots$

把$x_1,x',x''$的表达式代入原方程，合并t的各同次幂的项，并令各项系数等于0，得，

$
\begin{cases}
a_0+2a_2=0\\
a_1-2a_2+6a_3=0\\
n(n-1)a_n-(n-1)(n-2)a_{n-1}+a_{a-2}=0
\end{cases}
$

对$x_1(t)$，有$a_0=0,a_1=1$，

代入上面的方程组，可得，

$a_2=0,a_3=\frac{-1}{6},a_4=\frac{-1}{4*3}=\frac{-2}{4*3*2}$

$a_5=\frac{-5}{5*4*3*2},\cdots$

$a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)}$

即，$x_1(t)=t-\frac1{3!}t^3-\frac2{4!}t^4-\frac{5}{5!}t^5\cdots$

对$x_2(t)$，有$a_0=1,a_1=0$，

代入上面的方程组，可得，

$a_2=-\frac12,a_3=\frac{-1}{3*2},a_4=\frac{-1}{4*3*2},\cdots$

$a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)}$

即，$x_2(t)=1-\frac1{2!}t^2-\frac1{3!}t^3-\frac1{4!}t^4-\cdots$

所以方程的通解为，

$
\begin{aligned}
x&=c_1x_1t+c_2x_2(t)\\\
&=c_1(t-\frac1{3!}t^3-\frac2{4!}t^4-\frac5{5!}t^5\cdots)+c_2(1-\frac1{2!}t^2-\frac1{3!}t^3-\frac1{4!}t^4\cdots）
\end{aligned}
$

# 习题5.1

2.将下面的初值问题化为与之等价的一阶方程组的初值问题：

（1）$\begin{cases}x''+5y'-7x+6y=e^t\\y''-2y+13y'-15x=\cos t\end{cases}，x(0)=1,x'(0)=0,y(0)=0,y'(0)=1$

提示：$\omega_1=x,\omega_2=x',\omega_3=y,\omega_4=y'$

解：

令$\omega_1=x,\omega_2=x',\omega_3=y,\omega_4=y'$,可化为方程组

$$
\begin{bmatrix}
\omega_1\\
\omega_2\\
\omega_3\\
\omega_4
\end{bmatrix}
=
\begin{bmatrix}
0&1&0&0\\
7&0&-6&-5\\
0&0&0&1\\
15&0&2&-13
\end{bmatrix}
\begin{bmatrix}
\omega_1\\
\omega_2\\
\omega_3\\
\omega_4
\end{bmatrix}
+
\begin{bmatrix}
0\\
e^t\\
0\\
\cos t
\end{bmatrix}
$$

初值条件为$\omega_1(0)=1,\omega_2(0)=0,\omega_3(0)=0,\omega_4(0)=1$

3.试用逐步逼近法求方程组
$$
x'=
\begin{bmatrix}
0&1\\
-1&0
\end{bmatrix}
x
$$
满足初值条件
$$
x(0)=
\begin{bmatrix}
0\\
1
\end{bmatrix}
$$
的第三次近似解。

解：

$
\phi_0(t)=
\begin{bmatrix}
0\\
1
\end{bmatrix}
$

$
\begin{aligned}
\phi_1(t)&=
\begin{bmatrix}0\\1\end{bmatrix}
+
\int^0_1\begin{bmatrix}0&1\\-1&0\end{bmatrix}
\begin{bmatrix}0\\1\end{bmatrix}
ds\\\
&=
\begin{bmatrix}0\\1\end{bmatrix}
+
\int^t_0\begin{bmatrix}1\\0\end{bmatrix}
ds\\\
&=
\begin{bmatrix}0\\1\end{bmatrix}
+
\begin{bmatrix}t\\0\end{bmatrix}\\\
&=
\begin{bmatrix}t\\1\end{bmatrix}
\end{aligned}
$

$
\begin{aligned}
\phi_2(t)&=
\begin{bmatrix}0\\1\end{bmatrix}
+
\int^t_0\begin{bmatrix}0&1\\-1&0\end{bmatrix}
\begin{bmatrix}t\\1\end{bmatrix}
dt\\\
&=
\begin{bmatrix}0\\1\end{bmatrix}
+
\int^t_0\begin{bmatrix}1\\-t\end{bmatrix}
dt\\\
&=
\begin{bmatrix}0\\1\end{bmatrix}
+
\begin{bmatrix}t\\-\frac12 t^2\end{bmatrix}\\\ 
&=
\begin{bmatrix}t\\1-\frac12 t^2\end{bmatrix}
\end{aligned}
$

$
\begin{aligned}
\phi_3(t)&=
\begin{bmatrix}0\\1\end{bmatrix}
+
\int^t_0\begin{bmatrix}0&1\\-1&0\end{bmatrix}
\begin{bmatrix}s\\1-\frac12s^2\end{bmatrix}
ds\\\
&=
\begin{bmatrix}0\\1\end{bmatrix}
+
\int^t_0\begin{bmatrix}1-\frac12s^2\\-s\end{bmatrix}
ds\\\
&=
\begin{bmatrix}0\\1\end{bmatrix}
+
\begin{bmatrix}t-\frac16t^3\\-\frac12t^2\end{bmatrix}\\\
&=
\begin{bmatrix}t-\frac16t^3\\1-\frac12t^2\end{bmatrix}
\end{aligned}
$

所以满足初值条件$x(0)=\begin{bmatrix}0\\1\end{bmatrix}$的第三次近似解为，

$
\phi_3(t)=
\begin{bmatrix}t-\frac16t^3\\1-\frac12t^2\end{bmatrix}
$
