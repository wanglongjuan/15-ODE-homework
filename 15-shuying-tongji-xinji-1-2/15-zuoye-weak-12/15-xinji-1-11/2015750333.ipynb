# 1、$xx''-(x')^2+(x')^3=0$

解：令$ x’= y ,$得 

$ x''= y \frac{\mathrm dy}{\mathrm dx} $ 

所以原方程化为
$$
xy \frac{\mathrm dy}{\mathrm dx} - y^2 +y^3 = 0
$$
得到 $ y=0,$或
$\frac{1}{y - y^2}\mathrm dy = \frac{1}{x}\mathrm dx$

积分得 

$\frac{y}{1-y} = c_1x$

即$\quad \frac{x^{\prime}}{1 -x^{\prime}} = c_1x $ 

所以通解为

$ x + c_1ln|x| = t +c_2 $或$ x=c$


# 2、$x''+\sqrt{1-(x')^2}=0$

解：令 $ x’= y $,得 

$ x’’ = y \frac{\mathrm dy}{\mathrm dx}$ 

于是
$$
y \frac{\mathrm dy}{\mathrm dx} + \sqrt {1 -y^2} = 0
$$
当 $ y = \pm 1 $ 时

$ x^{\prime} = 0 $  

即 $ \quad x = \pm t +c $

当$ y \neq \pm1$时，积分得

$ \sqrt {1 -y^2} = x +c $

即
$\quad x’= \pm \sqrt{1-(x +c)}$

(1) $ x’ = \sqrt{1-(x +c)} $ ，积分得
 
$ x = \sin(t +c_1) - c_2 $

(2) $ x’= -\sqrt{1-(x +c)} $ ，积分得
$ x = \cos(t +c_1) - c_2 $

所以方程的解为 $ x = \pm t +c ,  x = \sin(t +c_1) - c_2 $ , $ x = \cos(t +c_1) - c_2 $


# 3、$x''-\frac{1}{t}x'+(x')^2=0$

当$ x’ = 0 $时， $ x = c $

当$ x’ \neq0 $ 时,两边同除$ x’$ ,得

$$
\frac{x’’} {x’} -\frac{1}{t} +x’= 0
$$
积分得

$
\ln|x'| - \ln|t| +x =0
$

所以 $ x’= te^{-x+c_1} $

积分得

$
e^x = \frac{e^{c_1}}{t^2} +c_2
$

所以解为$e^x = \frac{1}{t^2}e^{c_1} +c_2 $ , $ x = c $ 


# 4、$(1-t)x''+x=0$

原方程可化为 $ {x^{\prime}}^{\prime} + \frac{x}{1-t} = 0$ 

设 $ x = a_0+a_1t+a_2t^2 + a_3t^3+ \ldots +a_nt^n + \cdots $ 是方程的解,则

$ x’ = a_1+2a_2t+3a_3t^2+ \ldots +na_nt^{n-1} +\dots $

$x’’= 2a_2 + 3*2a_3 t+\cdots +n(n-1)a_nt^{n-2} +\cdots $

将上式表达式代入原方程，比较同次幂系数，得

$$
2a_2 +a_0 =0,3*2a_3 -2a_2+a_1=0\\
k(k-1)a_k-(k-1)(k-2)a_{k-1} + a_{k-2} = 0 , k>3 \quad \quad \quad \quad \quad \quad  (1)
$$

设 $x_1(t)$ 和$x_2(t)$ 是满足初值条件 $x_1(0) =0 ,x’(0)=1,x_2(0) = 1,x’_2(0)=0$ 的两个解。

对于 $ x_1(t)$ ,显然有 

$a_0=0 ,a_1 =1$

代入(1)得

$
a_2=0\\
a_3=\frac{-1}{1*2}\\
a_4=\frac{-1}{1*2*3*4}\\
a_5=\frac{-5}{1*2*3*4*5}\\
\cdots\\
a_n\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)},
$

所以

$ x_1(t) = t-\frac{t^3}{3!}-\frac{2}{4!}t^4-\frac{5}{5!}t^5-\frac{18}{6!}t^6-\cdots,$ 

对于 $x_2(t)$,显然有 

$ a_0=1,a_1=0$,

代入(1)得

$
a_2=-\frac{1}{2}\\
a_3=\frac{-1}{1*2*3}\\
a_4=\frac{-1}{1*2*3*4}\\
\cdots\\
a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)}$

所以$x_2(t) = 1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-\cdots $

所以通解为

$ x=c_1(t-\frac{t^3}{3!}-\frac{2}{4!}t^4-\frac{5}{5!}t^5-\frac{18}{6!}t^6-\cdots)+c_2(1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-\cdots)$
