**1.解：（2）**方程不显含 $ t $ ,令 $ y=x' $ ,则方程变形为 $$ xy\frac{dy}{dx}-y^2+y^3=0 $$

得 $$ y=0 或 \frac{dy}{dx}=\frac{y-y^2}{x} $$

积分后，得 $$ \frac{y}{1-y}=c'x(c'\neq0) $$

即 $ y=\frac{c'x}{1+c'x} $,从而 $ x'=\frac{c'x}{1+c'x} $

积分，得 $$ x+\frac{1}{c'}ln|x|=t+c_2 $$

即方程有解 $$ x+c_1ln|x|=t+c_2 $$

当 $ y=0 $ 时，得 $ x'=0 $

∴ $ x=c $

综上，方程的解为 $$ x+c_1ln|x|=t+c_2 或 x=c $$ ,其中 $ c_1,c_2,c_3 $ 是任意常数

**(4)**方程不显含 $ t $ ,令 $ y=x' $ ,则方程变形为 $$ y\frac{dy}{dx}-\sqrt{1-y^2}=0 $$

当 $ y\neq±1 $ 时，方程又可以变形为 $$ \frac{ydy}{\sqrt{1-y^2}}=-dx $$

积分后，得 $$ \sqrt{1-y^2}=x+c $$

即 $ \sqrt{1-x'^2}=x+c $

∴ $ x'=±\sqrt{1-(x+c)^2} $

再次积分，得

①当 $ x'=\sqrt{1-(x+c)^2} $ 时，$ x=sin(t+c_1)+c_2 $

②当 $ x'=-\sqrt{1-(x+c)^2} $ 时，$ x=cos(t+c_1)+c_2 $

当 $ y=±1 $ 时，$ x'=±1 $ ,从而 $ x=±t+c $

综上，方程的解为 $$ x=sin(t+c_1)+c_2 ,或 x=cos(t+c_1)+c_2 ，和 x=±t+c $$ ,其中 $ c_1,c_2,c $ 是任意常数

**(6)**当 $ x'\neq0 $ 时，方程两边同除以 $ x' $ ,则，得 $$ \frac{x''}{x'}-\frac{1}{t}+x'=0 $$

即 $$ d(ln|x'|-ln|t|+x)=0 $$

积分后，得 $$ ln|\frac{x'}{t}|+x=c_1 $$

解出 $ x' $ ,有 $ x'=t(e^{-x+c_1}) $ ,再次积分，得 $$ e^x=c'_1(t^2+c'_2) $$

当 $ x'=0 $ 时，即得 $ x=c $ ,显然它也是原方程的解

综上，方程的解 $$ e^x=c_1(t^2+c'_2) 或 x=c $$

**2.解：(2)**方程可化为 $$ x''+\frac{x}{1-t}=0 $$

∵ $ tp(t)=0,t^2q(t)=\frac{t^2}{1-t} $

∴存在幂级数解

设 $$ x=a_0+a_1t+a_2t^2+a_3t^3+…+a_nt^n+… $$ 是方程的解

∴ $$ x'=a_1+2a_2t+3a_3t^2+…+na_nt^{n-1}+… $$

$$ x''=2a_2+3·2a_3t+n(n-1)a_nt^{n-2}+… $$

将 $ x,x',x'' $的表达式代入原方程，比较同次幂系数，得

$$ 2a_2+a_0=0,3·2a_3-2a_2+a_1=0 $$

$$ k(k-1)a_k-(k-1)(k-2)a_{k-1}+a_{k-2}=0,k>3 ① $$

本方程的解应该为通解

∴ 需要找到两个线性无关的解

设 $ x_1(t) 和 x_2(t) $ 是满足初值条件 $ x_1(0)=0,x'_1(0)=1,x_2(0)=1,x'_2(0)=0 $ 的两个解

对于 $ x_1(t) $ ,显然有 $ a_0=0,a_1=1 $ ,代入①式，得

$a_2=0,a_3=\frac{-1}{1·2·3},a_4={-1}{3·4}=\frac{-2}{1·2·3·4},a_5=\frac{-5}{1·2·3·4·5}…a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)} $

即 $ x_1(t)=t-\frac{t^3}{3!}-\frac{2}{4!}t^4-\frac{5}{5!}t^5-\frac{18}{6!}t^6-… $

对于 $ x_2(t) $ ,显然有 $ a_0=0,a_1=1 $ ,代入①式，得

$a_2=0,a_3=\frac{-1}{1·2·3},a_4={-1}{3·4}=\frac{-2}{1·2·3·4},a_5=\frac{-5}{1·2·3·4·5}…a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)} $

即 $ x_2(t)=1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-… $

∴方程的通解为 $$ x=c_1x_1(t)+c_2x(t)=c_1(t-\frac{t^3}{3!}-\frac{2}{4!}t^4-\frac{5}{5!}t^5-\frac{18}{6!}t^6-…)+c_2(1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-…) $$
