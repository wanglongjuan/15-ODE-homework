1.求解下列方程：

（2）$xx''-(x')^2+(x')^3=0$

（4）$x''+\sqrt{1-(x')^2}=0$

（6）$x''-\frac1tx'+(x')^2=0$

（2）

解：

令$x'=y$,则方程化为$$xy\frac{dy}{dx}-y^2+y^3=0$$

得到$$y=0,\frac{dy}{dx}=\frac{y-y^2}{x}$$

积分得到$$\frac{y}{1-y}=c_1x$$

即$$\frac{x'}{1-x'}=c_1x$$

再次积分得到$$x+\frac{1}{c_1}ln\mid x \mid =t+c_2$$

当$y=x'=0$时，得到$x=c$

综上，方程的解为$$x+\frac{1}{c_1}ln\mid x \mid =t+c_2,x=c$$

其中$c_1,c_2,c$为任意常数.

（4）

解：

令$x'=y$,则方程化为$$y\frac{dy}{dx}+\sqrt{1-y^2}=0$$

当$y=x'=\pm 1$时,得到$x=\pm t+c_1$

当$y\neq \pm 1$时，方程化为$$\frac{ydy}{\sqrt{1-y^2}}=-dx$$

积分得到$$\sqrt{1-y^2}=x+c$$

即$$x'=\pm \sqrt{1-(x+c)^2}$$

再次积分得到

①$x'=\sqrt{1-(x+c)^2}$时，$x=\sin(t+c_1)-c$

②$x'=\sqrt{1-(x+c)^2}$时，$x=\cos(t+c_1)-c$

综上，方程的解为$$x=\pm t+c_1，x=\sin(t+c_1)-c，x=\cos(t+c_1)-c$$

其中$c_1,c$为任意常数.

（6）

解：

当$x'\neq 0$,则方程两边同除$x'$,得到$$\frac{x''}{x'}-\frac1t+x'=0$$

得到$$d(ln \mid x' \mid -ln \mid t \mid +x)=0$$

积分得到$$ln \mid \frac{x'}{t}\mid +t=c_1$$
 
即$$x'=t(e^{-x+c_1})$$

再次积分得到$$e^x=e^{c_1}(\frac{t^2}{2})+c_2$$

因为$c_1,c_2$为任意常数，所以上式可化为$$e^x=c_1(t^2+c_2)$$

当$x'=0$时，得到$x=c$

综上，方程的解为$$e^x=c_1(t^2+c_2),x=c$$

其中$c_1,c_2,c$为任意常数.

2.用幂级数解法求解下列方程：

（2）$(1-t)x''+x=0$

（2）

解：

设$x=a_0+a_1t+a_2t^2+a_3t^3+…+a_nt^n+…$是方程的解，则

$$x'=a_1+2a_2t+3a_3t^2+…+na_nt^{n-1}+…$$

$$x''=2a_2+3 \cdot 2a_3t+…+n(n-1)a_nt^{n-2}+…$$

将$x,x',x''$的表达式代入原方程，整理后将各项系数等于零，得到

$$
\begin{cases}
a_0+2a_2=0\\
a_1-2a_2=0\\
n(n-1)a_n-(n-1)(n-2)a_{n-1}+a_{n-2}=0
\end{cases}
$$

设$x_1(t)和x_2(t)为满足初值条件x_1(0)=0,x'_1(0)=1,x_2(0)=1,x'_2(0)=0的两个解.$

对于$x_1(t)，有a_0=0，a_1=1$代入得到

$$a_2=0,a_3=\frac{-1}{3\cdot 2},a_4=\frac{-1}{4 \cdot 3}=\frac{-2}{4 \cdot 3 \cdot 2},
a_5=\frac{-5}{5\cdot 4 \cdot 3 \cdot 2}$$

$$a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)}$$

即$$x_1(t)=t-\frac{t^3}{3!}-\frac{2}{4!}t^4-\frac{5}{5!}…$$

对于$x_2(t),有a_0=1,a_1=0$代入得到

$$a_2=-\frac12,a_3=\frac{-1}{3\cdot 2},a_4=\frac{-1}{4 \cdot 3 \cdot 2},…$$

$$a_n=\frac{(n-1)(n-2)a_{n-1}-a_{n-2}}{n(n-1)}$$

即$$x_2(t)=1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-…$$

得到通解
$$
\begin{aligned}
x&=c_1x_1(t)+c_2x_2(t)\\\
&=c_1(t-\frac{t^3}{3!}-\frac{2}{4!}t^4-\frac{5}{5!}…)+c_2(1-\frac{1}{2!}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4…)
\end{aligned}
$$

2.将下面的初值问题化为与之等价的一阶方程组的初值问题：

（3）$\begin{cases}
x''+5y'-7x+6y=e^t\\
y''-2y+13y'-15x=\cos t
\end{cases}$

（3）

解：

令$w_1=x,w_2=x',w_3=y,w_4=y'$,则方程组化为

$$
\left\{
 \begin{matrix}
  w_1 \\
  w_2 \\
  w_3 \\
  w_4
  \end{matrix}
  \right\}'=
  \left\{
 \begin{matrix}
  0 & 1 & 0 & 0 \\
  7 & 0 & -6 & -5 \\
  0 & 0 & 0 & 1 \\
  15 & 0 & 2 & -13
  \end{matrix}
  \right\}
  \left\{
 \begin{matrix}
  w_1 \\
  w_2 \\
  w_3 \\
  w_4
  \end{matrix}
  \right\}+
  \left\{
 \begin{matrix}
  0 \\
  e^t \\
  0 \\
  \cos t
  \end{matrix}
  \right\}
$$

初值条件为$w_1(0)=1,w_2(0)=0,w_3(0)=0,w_4(0)=1$.

$$\phi_0(t)

$$
\begin{aligned}
