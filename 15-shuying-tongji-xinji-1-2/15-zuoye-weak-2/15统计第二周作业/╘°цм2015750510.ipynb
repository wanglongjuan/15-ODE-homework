1.验证下列各函数是相应微分方程的解。  
（7）$y^2+1,y'=y^2-(x^2+1)y+2x$  
解： ∵ $y=x^2+1,y'=2x$ ,   
   ∴ $y^2-（x^2+1）y+2x=(x^2+1)^2-(x^2+1)(x^2+1)+2x=2x=y'$,    
   因此 $y=x^2+1$是方程$y'=y^2-(x^2+1)y+2x$的解  
   
（8）$y=-g(x)/f(x),y'=[f'(x)/g(x)]y^2-g'(x)/f(x)$  
解：∵ $y=-g(x)/f(x)$,  
∴ $y'= -[g'(x)f(x)-g(x)f'(x)]/f^2(x)=g(x)f'(x)/f^2(x)-g'(x)/f(x)$,  
又 $[f'(x)/g(x)]y^2-g'(x)/f(x)=[f'(x)/g(x)]*[-g(x)/f(x)]^2-g'(x)/f(x)=g(x)f'(x)/f^2(x)-g'(x)/f(x)=y'$    
∴$ y=-g(x)/f(x)$是方程的解

（6）$xdy/dx-y+$$\sqrt(x^2-y^2)=0$  
解：$将方程改写为dy/dx-y/x+$$\sqrt(1-(y/x)^2)=0$,  
这是齐次方程。  
以$y/x=u及dy/dx=xdu/dx+u$代入，  
$则原方程变为xdu/dx=-\sqrt(1-u^2)$,  
分离变量，得到$-du/\sqrt(1-u^2)=dx/x,$  
两边积分，得$arcsin(u)+ln|x|=c,$  
代回原来的变量，得$arcsin(y/x)+ln|x|=c,$  
另外，$u^2=1,即y^2=x^2$也是方程的解  
所以，原方程的解为$arcsin(y/x)+ln|x|=c或y^2=x^2$

（7）$dy/dx=(2x^3+3xy^2+x)/(3x^2y+3y^3-y)$  
解：方程可改写为$d(y^2+1)/d(x^2-1)=[2(x^2-1)+3(y^2+1)]/[3(x^2-1)+2(y^2+1)],$  
取变换$t=x^2-1,s=y^2+1,化为齐次方程ds/dt=(2t+3s)/(3t+2s).$  
再另$s=wt,得tdw/dt=(2-2w^2)/(3+2w),dt/t=(3+2w)/2(1-w^2)dw=(1/4)(5/(1-w)+1/(1+w))dw,$  
积分得$-(5/4)ln|1-w|+(1/4)ln|1+w|=ln|t|+c,(1+w)=ct^4(1-w)^5,$  
即$t+s=c(t-s)^5,$  
通解为$x^2+y^2=c(x^2-y^2-2)^5$,其中c为任意常数。
