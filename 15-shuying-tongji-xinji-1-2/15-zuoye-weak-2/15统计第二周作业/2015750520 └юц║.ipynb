$3.验证下列各函数是相应微分方程的解$

$（7）y=x^2+1,y'=y^2-(x^2+1)y+2x$

```python
var('x,y')
y=function('y')(x)
g=function('y')(x)
y=x^2+1
de=diff(y,x,1)
g=y^2-(x^2+1)*y+2*x
show(de)
eqn=de-g==0
show(eqn)

```

$由上面的计算可以知道将y=x^2+1及其倒数代入第二个式子，可使得两边相等$ 

$（8）y=\frac{-g(x)}{f(x)},y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f（x)}$

```python
var('y,g,f,x')
y=function('y')(x)
g=function('g')(x)
f=function('f')(x)
Y=function('Y')(x)
y=-g(x)/f(x)
de1=diff(y,x,1)
de2=diff(g,x,1)
de3=diff(f,x,1)
Y=(de3/g)*y^2-de2/f
eqn=de1-Y==0
show(eqn)

```

$将第一个式子及其倒数代入第二个式子发现式子成立$
