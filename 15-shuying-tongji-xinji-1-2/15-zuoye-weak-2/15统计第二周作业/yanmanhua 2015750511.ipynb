##   $3、验证下列各函数是相应微分方程的解$

###  $(7)、y=x^2+1,  y’=y^2-(x^2+1)y+2x $

```python
var('x')
y=function('y')(x)
y=x^2+1
de=diff(y,x,1)==y^2-(x^2+1)*y+2*x
show(de)
```


### $由上式可得，函数 y=x^2+1是相应微分方程 y’=y^2+(x^2+1)Y+2x$

##  $(8)、y= -\frac{\mathrm g(x)}{\mathrm f(x)} ,y’= \frac{\mathrm f'(x}{\mathrm g(x)}y^2- \frac{\mathrm g'(x)}{\mathrm f(x)}$


```python
var('x')
f=function('f')(x)
g=function('g')(x)
y=-g/f
G=diff(g,x,1)
F=diff(f,x,1)
de=diff(y,x,1)==(F/g)*y^2-(G/f)
show(de)


```

### $由上式可得，函数y=\frac{\mathrm g(x)}{\mathrm f(x)}是相应微分方程 y’= \frac{\mathrm f'(x}{\mathrm g(x)}y^2- \frac{\mathrm g'(x)}{\mathrm f(x)}的解$

```python
var('x')
y=function('y')(x)
de=diff(y,x,1)-2*x==0
show(de)
show(desolve(de,y))
```

### $(2)求通过点（1，4）的特解$

#### $    所以它的特解为y=x^2+3$

### $(3)求出与直线 y=2x+3 相切的解$

###  $ 解：$

#### $ 即切点为（1，5），代入通解，得c=4$
#### $ 所以与直线y=2x+3相切的解为y=x^2+4$

### $(4)求出满足条件\int\limits_0^1y\,\mathrm{d}x=2的解$

###  $ 解：$
#### $ 由条件\int\limits_0^1y\,\mathrm{d}x=2可知(\left.\frac{x^3}{3}+cx)\right|_0^1=\frac{\mathrm1}{\mathrm3}+c=2$
#### $ 所以c=\frac{\mathrm5}{\mathrm3}$
#### $ 所以满足条件\int\limits_0^1y\,\mathrm{d}x=2的解为y=x^2+\frac{\mathrm5}{\mathrm3}$

### $(6)\frac{\mathrm dy}{\mathrm dx}=\frac{\mathrm y^6-2x^2}{\mathrm 2xy^5+x^2y^2}$

### $解：$
#### $\frac{\mathrm dy}{\mathrm dx}=\frac{\mathrm (y^3)^2-2x^2}{\mathrm y^2(2xy^3+x^2)}=\frac{\mathrm dy^3}{\mathrm x}=\frac{\mathrm 3[(y^3)^2-2x^2]}{\mathrm 2xy^3+x^2}$
#### $ 令y^3=u,则原方程化为\frac{\mathrm du}{\mathrm dx}=\frac{\mathrm 3u^2-6x^2}{\mathrm 2xu+x^2}=\frac{\mathrm {\frac{\mathrm 3u^2}{\mathrm x^2}}-6}{\mathrm {\frac{\mathrm 2u}{\mathrm x}}+1}，这是齐次方程，令$
#### $\frac{\mathrm u}{\mathrm x}=z,则\frac{\mathrm du}{\mathrm dx}=z+x\frac{\mathrm dz}{\mathrm dx}，所以\frac{\mathrm 3z^2-6}{\mathrm 2z+1}=z+x\frac{\mathrm dz}{\mathrm dx},两边积分得(z-3)^7(z+2)^3=x^5c$
#### $即(y^3-3x)^7(y^3+2x)^3=x^5c$
####  $又因为y^3=3x或y^3=-2x包含在通解中$
#### $当c=0时，故原方程的解为(y^3-3x)^7(y^3+2x)^3=x^{15} c$
