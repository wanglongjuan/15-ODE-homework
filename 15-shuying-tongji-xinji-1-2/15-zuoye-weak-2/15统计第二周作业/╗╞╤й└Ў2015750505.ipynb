$即证y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$

$1.求下列方程的解：$

$(6)x\frac{dy}{dx}-y+（x^2-y^2)^\frac{1}{2}=0$

$解：将原式变形，得$

$\frac{dy}{dx}=\frac{y}{x}-[（1-(\frac{y}{x})^2)]^\frac{1}{2} （将该式子定义为@）$

$令u=\frac{y}{x},即y=ux,两边求积分，得$

$\frac{dy}{dx}=x\frac{du}{dx}+u$

$u=\frac{y}{x}，\frac{dy}{dx}=x\frac{du}{dx}+u代入@中，有$

$x\frac{du}{dx}+u=u-（1-u^2)^\frac{1}{2}$

$x\frac{du}{dx}=-（1-u^2)^\frac{1}{2}$

$当u≠1时，即x^2≠y^2时$

$分离变量，得\frac{du}{-(1-u^2)^\frac{1}{2}}=\frac{dx}{x},$

$两边积分，得$

$-\arcsin u=ln|x|+c$

$将u=\frac{y}{x}代入，并且移项得，$

$\arcsin \frac{y}{x}+ln|x|=c,c为任意常数.$

$当y^2=x^2时，$

$原式为x\frac{dy}{dx}-y=0$

$同理，将变量分离，再两边求积分得,$

$y^2=x^2$

$综合以上，原方程的解为y^2=x^2及\arcsin \frac{y}{x}+ln|x|=c,c为任意常数.$

$(6)\frac{dy}{dx}=\frac{y^6-2x^2}{2xy^5+x^2y^2};$

$解：令u=y^3,则原式为\frac{du}{dx}={3u^2-6x^2}{2xu+x^2},$

$推出(\frac{7}{5})(\frac{dv}{v-3}+(\frac{3}{5})(\frac{dv}{v+2}）=\frac{dx}{x},$

$两边同时积分得，$

$(v-3)^7(v+2)^3=cx^5.$

$代回原变量，并且化简原方程的解为（y^3-3x)^7(y^3+2x^2)^3=cx^15.$

$(7).\frac{dy}{dx}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}.$

$解：原方程可以变形为;\frac{ydy}{xdx}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1},$

$令u=x^2-1,v=y^2+1,则方程可化为\frac{du}{dv}=\frac{2u+3v}{3u+2v},$

$再令w=\frac{u}{v},即v=uw,则有\frac{du}{dv}=u\frac{dw}{du}+w,$

$因此，u\frac{dw}{du}=\frac{2-2w^2}{3+2w},$

$分离变量并且化简，得\frac{4}{u}du=(\frac{5}{1-w}+\frac{1}{1+w})dw,$

$两边积分，得u^4(1-w^5)=c(1+w),$

$代回原变量即得方程的解为$

$(x^2-y^2-2)^5=c(x^2+y^2),其中c为任意常数.$
