3.验证下列各函数是相应微分方程的解：  
（7）$y=x^2+1,y'=y^2-(x^2+1)y+2x$；


解：$∵y=x^2+1,y'=2x$   
$∴y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)(x^2+1)+2x=2x=y'$      
因此y=x^2+1是方程y'=y^2-(x^2+1)y+2x的解。


（8）$y=-g(x)/f(x),y'=f'(x)/g(x)y^2-g'(x)/f(x)$；


解：∵$y=-g(x)/[f(x)]$

$=[g(x)f'(x)]/[f^2(x)]-g'(x)/[f(x)]=y'$  
因此，$y=-g(x)/[f(x)]是方程y'=f'(x)/[g(x)]y^2-g'(x)/[f(x)]$的解。
