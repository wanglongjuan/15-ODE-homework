(6)因为$x\frac{dy}{dx}-y+\sqrt{x^2-y^2}=0$

则$xdy-ydx+\sqrt{x^2-y^2}dx=0$

当x^2不等于y^2时，

$\frac{xdy-ydx}{\sqrt{x^2-y^2}}+dx=0$

即$\frac{d(\frac{y}{x})}{\sqrt{1-(\frac{y}{x})^2}}+\frac{1}{x}dx=0$

两边积分得 $arcsin{\frac{y}{x}}+\ln{|x|}=c$

其中c为任意常数

而当$x^2=y^2$时，其也是方程的解。

所以，原方程的解为 $arcsin{\frac{y}{x}}+\ln{|x|}=c$ 或$x^2=y^2$
