 $验证下列各函数是相应微分方程的解$

$（7）$
$y=x^2+1,y＇=y^2-(x^2+1)y+2x$

$解：因为y=x^2+1，所以y＇=2x$

  $将y=x^2+1代入y^2-(x^2+1)y+2x$
  
  $得 (x^2+1)^2-（x^2+1)（x^2+1）+2x=2x= y＇$
  
  $所以y=x^2+1是方程的解$
  
  

$（8）$
$y=-\frac{g(x)}{f(x)},y＇=-\frac{f＇(x)}{g(x)}y^2-\frac{g＇(x)}{f(x)}$

$解：因为y=-\frac{g(x)}{f(x)}，所以y＇=-\frac{g＇(x)f(x)-g(x)f＇(x)}{f^2(x)}=\frac{g(x)f＇(x)}{f^2(x)}-\frac{g＇(x)}{f(x)}$

$将y=-\frac{g(x)}{f(x)}代入-\frac{f＇(x)}{g(x)}y^2-\frac{g＇(x)}{f(x)}$

$得\frac{f＇(x)}{g(x)}(-\frac{g(x)}{f(x)})^2-\frac{g＇(x)}{f(x)}=\frac{g(x)f＇(x)}{f^2(x)}-\frac{g＇(x)}{f(x)}=y＇$

$所以y=-\frac{g(x)}{f(x)}是方程的解$

(6)
$x\frac{dy}{dx}-y+(x^2-y^2)^\frac{1}{2}=0$

$解:两边同时除以x,得\frac{dy}{dx}-\frac{y}{x}+(1-\frac{y^2}{x^2})^\frac{1}{2}=0$

$令u=\frac{y}{x}$

$y=ux$

$因为\frac{dy}{dx}=u+ x\frac{du}{dx}$

$将u=\frac{y}{x}代入\frac{dy}{dx}-\frac{y}{x}+(1-\frac{y^2}{x^2})^\frac{1}{2}$

$化简：\frac{dy}{dx}-u+(1-u^2)^\frac{1}{2}=0$

$即 x\frac{du}{dx}=-(1-u^2)^\frac{1}{2}$

$-\frac{dx}{x}=\frac{du}{(1-u^2)^\frac{1}{2}}$

$两边同时积分$

$-ln|x|+c=arc\sin u$

$arc\sin u+ln|x|=c$

$当u^2=1时，y=x,-x$

$所以方程的解为arc\sin u+ln|x|=c和y=x,-x$

$\frac{7}{5}\frac{dv}{v-3}+\frac{3}{5}\frac{dv}{v+2}=\frac{dx}{x}$

$(v-3)^7(v+2)^3=cx^5$

$（7）.\frac{dy}{dx}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$

$解：等式两边同乘\frac{y}{x},原方程可变为$

$\frac{ydy}{xdx}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$

$令u=x^2-1,v=y^2+1,$

$方程可化为\frac{dv}{du}=\frac{2u+3v}{3u+2v}$

$再令w=\frac{v}{u},所以v=wu$

$又因为\frac{dv}{du}=w+u\frac{dw}{du}$

$u\frac{dw}{du}=\frac{2-2w^2}{3+2w}$

$分离变量后得 \frac{4}{u}du=(\frac{5}{1-w}+\frac{1}{1+w})dw$

$两边同时积分得u^4(1-w)^5=c(1+w)$

$故解为(x^2-y^2-2)^5=c(x^2+y^2)$
