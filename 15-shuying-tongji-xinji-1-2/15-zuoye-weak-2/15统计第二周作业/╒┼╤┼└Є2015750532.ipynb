$（7）y=x^2+1,y'=y^2-(x^2+1)y+2x$

  $解：因为y=x^2+1,所以y'=2x$
  
  $所以y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)(x^2+1)+2x=2x=y'$
     
  $所以y=x^2+1是方程y'=y^2-(x^2+1)y+2x的解。$
      
 $（8）y=-\frac{g(x)}{f(x)},y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$
 
  $ 解：因为y=-\frac{g(x)}{f(x)}$
  
  $所以y'=-\frac{g'(x)f(x)-g(x)f'(x)}{f^2(x)}=\frac{g(x)f'(x)}{f^2(x)}-\frac{g'(x)}{f(x)}$
       
  $又因为\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}=\frac{f'(x)}{g(x)}(-\frac{g(x)}{f(x)})^2-\frac{g'(x)}{f(x)}=\frac{(g(x)f'(x)}{f^2(x)}-\frac{g'(x)}{f(x)}=y'$
       
  $所以,y=-\frac{g(x)}{f(x)}是方程y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)} 的解。$

$（5）绘出（2），（3），（4）中的解的图形。$

$解：（1）因为\frac{dy}{dx}=2x,所以dy=2xdx$

$所以c=\frac{5}{3}$

$所以满足条件\int_0^1 \mathrm y\,\mathrm{d}x=2的解为y=x^2+\frac{5}{3}$

$2、作适当的变量求解下列方程$

$（6）\frac{dy}{dx}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$

$（7）\frac{dy}{dx}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$

$解：(6)令u=y^3,则方程可变形为\frac{du}{dx}=\frac{3u^2-6x^2}{2xu+x^2}$

$再令v=\frac{u}{x},即u=vx,则上面方程化为x\frac{dv}{dx}+v=\frac{3v^2-6}{2v+1}$

$整理得，\frac{7}{5}\frac{dv}{v-3}+\frac{3}{5}\frac{dv}{v+2}=\frac{dx}{x}$

$两边同时积分，得(v-3)^7(v+2)^3=cx^5(其中c为正常数）$

$代回原变量，再简化一下，可得原方程的解为(y^3-3x)^7(y^3+2x)^3=cx^(15)$

$(7)原方程可以变形为\frac{ydy}{xdx}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$

$令u=x^2-1,v=y^2+1,则方程可化为\frac{dv}{du}=\frac{2u+3v}{3u+2v}$

$再令w=\frac{v}{u},即v=uw,则有方程可化为\frac{dv}{du}=w+u\frac{dw}{du},可得u\frac{dw}{du}=\frac{2-2w^2}{3+2w}$

$分离变量并简化得\frac{4}{u}du=(\frac{5}{1-w}+\frac{1}{1+w})dw$

$两边积分，得u^4(1-w)^5=c(1+w)$

$代回原变量,可得原方程的解为(x^2-y^2-2)^5=c(x^2+y^2)（其中c为任意常数）$
