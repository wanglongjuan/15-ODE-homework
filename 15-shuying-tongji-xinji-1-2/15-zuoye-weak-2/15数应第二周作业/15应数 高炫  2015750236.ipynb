# 1.求下列方程的解：

$(1)x\frac{dy} {dx}-y+\sqrt{x^2-y^2}=0$

$将方程变形为xdy-ydx+\sqrt{x^2-y^2}dx=0$

$当y^2\not=x^2时,容易验证y^2=x^2也是方程的解，$

可化得
$$
\frac{d(\frac{y} {x}} {\sqrt{1-(\frac{y} {x}^2}}+\frac{1} {x}dx=0
$$

$两边积分可得\arcsin（\frac{y} {x})+ln{x}=c,c为任意常数$

$所以原方程的解为y^2=x^2或\arcsin(\frac{y} {x})+ln[x]=c$

$(2)\tan y dx-\cot x dy=0$

$当\tan \not=0时，即y\not=k\pi(k为整数或零）时$

$分离变量得\cot y dy=\tan x dx$

$两边同时积分的ln|\sin x|=-ln|\cos x|+c_1,c_1为任意常数$

$即\sin y \cos x=c,c为任意正常数$

$另外，当\tan=0时，即y=k\pi时，易验证也为方程的解$

$综上，原方程的解为\sin y \cos x=c或y=k\pi(k为整数或零）$

# 证明题

$设y_1(t)是\frac{dy} {dt}=f(y),y(0)=y_0的解，证明y_2(t)=y_1(t-t_0)是\frac{dy} {dt}=f(y),y(t_0)=y_0的解。$

$证：因为y是以t为自变量的函数，且y'=f(y)$

$所以y_1'(t)=f(y_1(t))$

$t-t_0也满足上式成立$

$所以，y_1'(t-t_0)=f(y_1(t-t_0))$

$即y_2'(t)=f(y_2(t))也是方程\frac{dy} {dt}=f(y)的解$

$又因为y(0)=y_0$

$所以，对函数y_2(t)来说，当t=t_0时上式成立，即y(t_0)=y_0$
