$3.验证下列各函数是相应微分方程的解：$

$（7）y=x^2+1,y'=y^2-(x^2+1)y+2x;$

$(8)y=-\frac{g(x)}{f(x)},y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}.$

$解：(7)左边：$$$y'=2x$$

$右边：$$$=(x^2+1)-(x^2+1)(x^2+1)+2x$$

$$=2x$$

$所以方程两边相等即y=x^2+1是方程的解。$

$(8)左边：$$$y'=\frac{-[g'(x)f(x)-g(x)f'(x)]}{f^2(x)}$$

$$=\frac{g(x)f'(x)-g'(x)f(x)}{f^2(x)}$$

$右边$$$=\frac{f'(x)g^2(x)}{g(x)f^2(x)}-\frac{g'(x)}{f(x)}$$

$$=\frac{g(x)[f'(x)g(x)-g'(x)f(x)]}{f^2(x)g(x)}$$

$$=\frac{g(x)f'(x)-g'(x)f(x)}{f^2(x)}$$

$所以方程两边相等即y=-\frac{g(x)}{f(x)}是方程的解。$
