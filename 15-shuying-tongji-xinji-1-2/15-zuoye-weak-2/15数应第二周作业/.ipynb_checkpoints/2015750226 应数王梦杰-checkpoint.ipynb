一、验证下列各函数是相应微分方程的解：
(7)$y=x^2+1,y'=y^2-(x^2+1）y+2x$

(8)$y=-\frac{g(x)}{f(x)},y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$

解;（7）对$y=x^2+1$方程求导可得$$y=2x$$

将$y=x^2+1$平方可得$$y^2=x^4+2x^2+1$$

再将上式代入$y'=y^2-(x^2+1)y+2x$可得$$y^2-(x^2+1)y+2x=2x=y'$$

故方程$y=x^2+1$是微分方程$y'=y^2-(x^2+1)y+2y$的解

解;(8)对$y=-\frac{g(x)}{f(x)}$【此式记为（1）式】
求导可得$$y'=-\frac{g'(x)f(x)-f'(x)g(x)}{f^2(x)}$$
记为（3）式

将1式平方可得3式$$y^2=\frac{g^2(x)}{f^2(x)}$$

将3式代入$y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$ 【此式记为（2）式】
可知$$\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}=-\frac{g'(x)f(x)-f'(x)g(x)}{f^2(x)}=y'$$

故方程1式是微分方程2式的解

P42 一、求下列方程的解：
（6）$x\frac{dy}{dx}=y+\sqrt{x^2-y^2}$

(7)$tanydx-cotxdy=0$




（6）将等式左侧的x除到等式右侧可得$$\frac{dy}{dx}=\frac{y}{x}-\sqrt{1-(\frac{y}{x}）^2}$$【记为(1)式】

再将（1）式中的$\frac{y}{x}=m$   $$\frac{dy}{dx}=m+x\frac{dm}{dx}$$

则（1）式可化为$$\frac{dy}{dx}=m-\sqrt{1-m^2}$$

分离变量可得：$$\frac{dm}{\sqrt{1-m^2}}=-\frac{dx}{x}$$

对两边分别积分可得：$$arcsin m=ln|x|+c$$
即为$$arcsin\frac{y}{x}=ln|x|+c,y^2=x^2$$

(7)$$tanydx-cotxdy=0$$ 通过移项，将tany移到右项，cotx移到左项
可得：$$cotydy=tanxdx$$

对两边同时积分可得：$$ln|sinx|=-ln|cosx|+a$$,则$$ln\frac{sinx}{cosx}=a$$
即为$$sinxcosx=e^a=c$$,【其中c为常数】其中y=kπ,k=0,1,-1.....

课外题（3）
设y1(t)是$$\frac{dy}{dt}=f(y),y(0)=y。$$的解，证明$$y2(t)=y1(t-t。）$$是$$\frac{dy}{dx}=f(y),y(t。）=y。$$的解

证明：$$\frac{dy1(t)}{dt}=f(y1(t)),y1(0)=y。$$

因为$$\frac{d(t-t。）}{dt}=f(y1(t-t。））=f(y2(t))$$

$$y2(t。）=y1(t。-t。）=y1(0)=y。$$

即$$\frac{dy2(t)}{dt}=f(y2(t)),y2(t。）=y。$$

所以$$y2(t)=y1(t-t。）$$是$$\frac{dy}{dt}=f(y),y(t。）=y。$$的解
