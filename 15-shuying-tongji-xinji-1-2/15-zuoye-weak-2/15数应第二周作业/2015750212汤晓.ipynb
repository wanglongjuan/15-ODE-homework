**1、验证下列各函数函数y是相应微分方程的解**


（1）y=x^2+1,$\frac{\mathrm d y}{\mathrm d x}=y^2-(x^2+1)y+2x$

```python
y(x)=x^2+1
eqn=y^2-(x^2+1)*y+2*x
f1=diff(y,x)
f2=eqn.subs(y==x^2+1)
show(f1)
show(f2)
```

**经过计算可以得到f1=f2，即可证明$y=x^2+1$是微分方程$\frac{\mathrm d y}{\mathrm d x}=y^2-(x^2+1)y+2x$的解,得证**

经过计算可以得到：
$$f1=\frac{\mathrm d y}{\mathrm d x}=\frac{f'(x)g(x)-g'(x)f(x)}{f^2(x)}$$
将y的函数代入到微分方程的右边：
$$f2=\frac{f'(x)}{g(x)}\frac{g^2(x)}{f^2(x)}-\frac{g'(x)}{f(x)}=\frac{f'(x)g(x)}{f^2(x)}-\frac{f(x)g'(x)}{f^2(x)}=\frac{f'(x)g(x)-f(x)g'(x)}{f^2(x)}$$
**经过计算可以得到f1=f2，即可证明其是微分方程的解，得证**

**3、求下列方程的解**

（1）$x \frac{\mathrm d y}{\mathrm d x}-y+\sqrt{x^2-y^2}=0$

解：

$x \frac{\mathrm d y }{\mathrm d x}=y-\sqrt{x^2-y^2}$

$\frac{\mathrm d y }{\mathrm d x}=\frac{y}{x}-\sqrt{1-(\frac{y}{x})^2}$

令u=$\frac{y}{x}$

y=ux

$\frac{\mathrm d y}{\mathrm d x}=\frac{\mathrm d u}{\mathrm d x}x+u$

综上可知：

$\frac{\mathrm d u}{\mathrm d x}x+u=u-\sqrt{1-u^2}$

$\frac{\mathrm d u}{\mathrm d x}=\frac{-\sqrt{1-u^2}}{x}$

$\frac{1}{\sqrt{1-u^2}} du=\frac{-1}{x}$

两边同时取积分：

$\int\ \frac{1}{\sqrt{1-u^2}}\,\mathrm{d}u=\int\ \frac{-1}{x}\,\mathrm {d}x$

$arcsin \frac{y}{x}=-ln|x|+c$

同时注意考虑到$\sqrt{1-u^2}=0$的情况：

即可得到解为y^2=x^2

综上可知道其解为



$\begin{cases}
arcsin \frac{y}{x}+ln|x|=c\\
y^2=x^2
\end{cases}$

**4、作适当的变量变换求解下列方程：**

(1)$\frac{\mathrm d y}{\mathrm d x}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$

解：

令u=y^3

$\frac{\mathrm d u}{\mathrm d x}=\frac{\mathrm d u }{\mathrm d y} \frac{\mathrm d y}{\mathrm d x}=\frac{3y^6-6x^2}{2xy^3+x^2}=\frac{3u^2-6x^2}{2xu+x^2}=\frac{ 3 (\frac{u}{x})^2 -6}{2 (\frac{u}{x})+1}$

令$v= \frac{u}{x}$

u=vx

$\frac{\mathrm d u}{\mathrm d x}=\frac{\mathrm d v}{\mathrm d x} x +v$

总结可得到： 

$\frac{\mathrm d v}{\mathrm d x} x+v=\frac{3v^2-6}{2v+1}$

$\frac{\mathrm d v}{\mathrm d x}=\frac{v^2-v-6}{2v+1} \frac{1}{x}$

$\frac{2v+1}{v^2-v-6}dv=\frac{1}{x}dx$

$\int\ \frac{7}{5} \frac{1}{v-3}\,\mathrm {d}x+\int\ \frac{3}{5} \frac{1}{v+2}\,\mathrm {d}v=\int\ \frac{1}{x}\,\mathrm {d}x$

$\frac{7}{5} ln|v-3|+\frac{3}{5} ln|v+2|=ln|x|+c$

$ln (v-3)^7+ln(v+2)^3=ln|x|^5+c$

即可得到其通解为:

$(v-3)^7(v+2)^3=cx^5$
