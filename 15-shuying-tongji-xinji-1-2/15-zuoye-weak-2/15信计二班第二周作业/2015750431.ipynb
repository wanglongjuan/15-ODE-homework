### P27.3.验证下列各函数是相应微分方程的解。

#### （7）$ y = x^2+1 ,y'=y^2 - (x^2+1) y + 2x ;$

解：将函数$y=x^2+1$代入等式两边

左边 $$y'=2x$$

右边$$y^2+(x^2+1)y+2x=(x^2+1)^2-(x^2)(x^2)+2x=2x$$

故函数$y=x^2+1$为该微分方程的解。

#### (8)$y=-\frac{g(x)}{f(x)}$ ,  $y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}.$

解：将函数$y=-\frac{g(x)}{f(x)}$代入等式两边

 左边 $$y'=-\frac{g'(x)f(x)-g(x)f'(x)}{f^2(x)}$$

右边$$\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}=\frac{f'(x)g(x)}{f^2(x)}-\frac{g'(x)}{f(x)}=-\frac{g'(x)f(x)-g(x)f'(x)}{f^2(x)}$$

故函数$y=-\frac{g(x)}{f(x)}$为该微分方程的解。

### P42.1.求下列方程的解：

#### (6)$x\frac{\mathrm d y}{\mathrm d x}-y+\sqrt{x^2-y^2}=0$

解：

等式两边同时除以一个$x$,移项即有

$\frac{\mathrm d y}{\mathrm d x}=\frac{y}{x}-\sqrt{1-(\frac{y}{x})^2}$

$\frac{\mathrm d u}{\mathrm d x}=\frac{g(u)-u}{x}$

$\frac{\mathrm d u}{g(u)-u}=\frac{\mathrm d x}{x}，\frac{\mathrm d u}{-\sqrt{1-u^2}}=\frac{\mathrm d x}{x}$

对两边同时积分，并整理得

$ln|x|+\arcsin{\frac{y}{x}}+c=0$

#### (7)$\tan y  \mathrm d x-\cot x   \mathrm d y=0$

解：

移项，将含有同一变量的移至同一边得：

$\frac{\mathrm d x}{\cot x}=\frac{\mathrm d y}{\tan y}$

对两边求积分有,最终得：

$-\ln |\cos x|+\ln |\sin y|+c=0$

$y=k\pi是方程的特解，k=1,2……$

### P43.2.作适当的变量变换求解下列方程：

#### (6)$\frac{\mathrm d y}{\mathrm d x}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$;

解：令$u=y^3$,则$\frac{\mathrm d u}{\mathrm d x}=\frac{\mathrm d u}{\mathrm d y}\frac{\mathrm d y}{\mathrm d x}$

$\frac{\mathrm d u}{\mathrm d x}=\frac{3(\frac{u}{x})^2-6}{2\frac{u}{x}+1}$

$\frac{\mathrm d v}{\mathrm d x}=\frac{g(v)-v}{x}$

$\frac{\mathrm d v}{g(v)-v}=\frac{\mathrm d x}{x}$

最终解得：

$(y^3-3x)^7(y^3+2x)^3=cx^{15}$

$\frac{\mathrm d w}{g(w)-w}=\frac{\mathrm d u}{u}$
