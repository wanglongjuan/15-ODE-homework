P27  第三题、验证下列各函数是相应微分方程的解：

(7)$y=x^2+1,y'=y^2-(x^2+1)y+2x$;

(8)$y=-\frac{\mathrm g(x)}{\mathrm f(x)},y'=\frac{\mathrm f'(x)}{\mathrm g(x)}y^2-\frac{\mathrm g'(x)}{\mathrm f(x)}$.

解：（7）因为$y=x^2+1，y'=2x$

所以$y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)(x^2+)+2x=2x=y'$

所以该方程的解为$y=x^2+1$

(8)因为$y=-\frac{\mathrm g(x)}{\mathrm f(x)}$

所以$y'=-\frac{\mathrm g'(x)f(x)-g(x)f'(x)}{\mathrm f^2(x)}=\frac{\mathrm g(x)f'(x)}{\mathrm f^2(x)}-\frac{\mathrm g'(x)}{\mathrm f(x)}$

又$\frac{\mathrm f'(x)}{\mathrm g(x)}y^2-\frac{\mathrm g'(x)}{\mathrm f(x)}=\frac{\mathrm g(x)f'(x)}{\mathrm f^2(x）}-\frac{\mathrm g'(x)}{\mathrm f(x)}=y'$

所以该方程的解为：$y=-\frac{\mathrm g(x)}{\mathrm f(x)}$.

P42 第一题、求下列方程的解：

(6)$x\frac{\mathrm dy}{\mathrm dx}-y+\sqrt{x^2-y^2}=0$;

(7)$tanydx-cotxdy=0$.

解：(6)将原方程变形为$xdy-ydx+\sqrt{x^2-y^2}dx=0$

当$y^2≠x^2$时，两边同除以$\sqrt{x^2-y^2}$得

$$\frac{\mathrm xdy-ydx}{\mathrm {\sqrt{x^2-y^2}}}+dx=0$$

即有$$\frac{\mathrm d(\frac{\mathrm y}{\mathrm x})}{\mathrm {\sqrt{1-(\frac{\mathrm y}{\mathrm x})^2}}}+\frac{\mathrm 1}{\mathrm x}dx=0$$

两边积分得 $arcsin \frac{\mathrm y}{\mathrm x}+ln |x|=c ,c$为任意常数。

当$y^2=x^2$时，可以容易验证$y^2=x^2$也是方程的解.

所以原方程的解为：$y^2=x^2$  或  $arcsin \frac{\mathrm y}{\mathrm x}+ln|x|=c$.

(7)当$tany≠0$时，即$y≠kπ,k=0,±1,±2,…$时分离变量得  $cotydy=tanxdx$

两边同时积分得 $ln|siny|=-ln|cosx|+{c}1,{c}1$为任意常数.

即有 $sinycosx=c,c$为任意正常数.

当$y=kπ,k=0,±1,±2,…$时，容易验证也是方程的解.

所以原方程的解为：$sinycosx=c$  或  $y=kπ,k=0,±1,±2…$.

第二题、 做适当的变量变换求解下列方程：

（6）$\frac{\mathrm dy}{\mathrm dx}=\frac{\mathrm y^6-2x^2}{\mathrm 2xy^5+x^2y^2}$

(7)$\frac{\mathrm dy}{\mathrm dx}=\frac{\mathrm 2x^3+3xy^2+x}{\mathrm 3x^2y+2y^3-y}$

解：(6)令$u=y^3$, 则方程可变形为$\frac{\mathrm du}{\mathrm dx}=\frac{\mathrm 3u^2-6x^2}{\mathrm 2xu+x^2}$

再令$v=\frac{\mathrm u}{\mathrm x}$,即u=vx,则上面方程化为$$x\frac{\mathrm dv}{\mathrm dx}+v=\frac{\mathrm 3v^2-6}{\mathrm 2v+1}$$

所以有  $$\frac{\mathrm 7dv}{\mathrm 5(v-3)}+\frac{\mathrm 3dv}{\mathrm 5(v+2)}=\frac{\mathrm dx}{\mathrm x}$$

两边同时积分得 $(v-3)^7(v+2)^3=cx^5 ,  c为正常数$.

代入$v=\frac{\mathrm u}{\mathrm x}=\frac{\mathrm y^3}{\mathrm x},$  化简原方程的解为：

$$(y^3-3x)^7(y^3+2x)=cx^{15}$$.

(7)原方程可变形为：

$$\frac{\mathrm ydy}{\mathrm xdx}=\frac{\mathrm 2x^2+3y^2+1}{\mathrm 3x^2+2y^2-1}$$

也即为：$$\frac{\mathrm dy^2}{\mathrm dx^2}=\frac{\mathrm 2x^2+3y^2+1}{\mathrm 3x^2+2y^2-1}$$

令$u=x^2-1$,$v=y^2+1$,则上面方程化为$$\frac{\mathrm dv}{\mathrm du}=\frac{\mathrm 2u+3v}{\mathrm 3u+2v}$$

再令$w=\frac{\mathrm v}{\mathrm u}$,即$v=uw$,则上面方程化为$$u\frac{\mathrm dw}{\mathrm du}+w=\frac{\mathrm 2+3w}{\mathrm 3+2w}$$

分离变量并且积分得：  $c(1+w)=u^4(1-w)^5$

把 $w=\frac{\mathrm v}{\mathrm u},v=y^2+1,u=x^2-1,$ 代入得

$$c(x^2+y^2)=(x^2-y^2-2)^5$$  $c$为任意常数.
