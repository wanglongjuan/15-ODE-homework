1、验证下列各函数是相应微分方程的解： 

（1）$y=x^2+1$,$y'=y^2-(x^2+1)y+2x$; 

（2）$y=-\frac{g(x)}{f(x)}$,$y'=\frac{f'(x)y^2}{g(x)}-\frac{g'(x)}{f(x)}$;

解：（1）因为 $y=x^2+1$,$y'=2x$

所以$$y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)(x^2+1)+2x=2x=y'$$

因此$y=x^2+1$是方程$y'=y^2-(x^2+1)y+2x$的解。

（2）因为$y=-\frac{g(x)}{f(x)}$

所以$$y'=-\frac{g'(x)g(x)-g(x)f'(x)}{f^2(x)}=\frac{g(x)f'(x)}{f^2(x)}-\frac{(g'(x)}{f(x)}$$

而且$$\frac{f'(x)y^2}{g(x)}-\frac{g'(x)}{f(x)}=\frac{f'(x)}{g(x)}[\frac{-g(x)}{f(x)}]^2-\frac{g'(x)}{f(x)}=\frac{g(x)f'(x)}{f^2(x)}-\frac{g'(x)}{f(x)}=y'$$

所以$y=-\frac{g(x)}{f(x)}$是方程 $y'=\frac{f'(x)y^2}{g(x)}-\frac{g'(x)}{f(x)}$ 的解。

2、给定一阶微分方程$dy=2x$, 

（1）求出它的通解；

（2）求通过点（1,4）的特解；

（3）求出与直线$y=2x+3$相切的解；

（4）求出满足条件$\int0^1 y{\rm d}x=2$的解； 

（5）绘出（2）（3）（4）中的解的图形。

解:
（1）由$\frac{dy}{dx}=2x$,得$dy=2xdx$,
方程两边积分,既得$$y=x^2+c$$ 这时c为任意常数。 

（2）把x=1,y=4代入$y=x^2+c$,得c=3
所以过点（1,4）的特解为$$y=x^2+3$$. 

（3）因为与直线相切，
所以方程组$$y=x^2+c$$ $$y=2x+3$$有且只有唯一一组解

即$x^2+c=2x+3$有唯一解，故c=4
因此，与直线$y=2x+3$相切的解是$y=x^2+4$.

（4）因为$$\int0^1 y{\rm d}x=\int0^1 （x^2+c）{\rm d}x=\left.[x^3/3+cx]\right|_0^1=\frac{1}{3}+c=2$$
 所以c=\frac{5}{3},即满足条件$\int0^1 y{\rm d}x=2$的解为$$y=x^2+\frac{5}{3}$$

2、作适当的变量变换求解下列方程：

（1）$\frac{dy}{dx}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$

（2）$\frac{dy}{dx}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$

解：
（1）令$u=y^3$
则方程可变形为$$\frac{du}{dx}=\frac{3u^2-6x^2}{2xu+x^2}$$
再令$v=\frac{u}{x}$,即$u=vx$
则上面方程化为$$\frac{xdv}{dx}+v=\frac{3v^2-6}{2v+1}$$
整理得$$\frac{7dv}{5(v-3)}+\frac{3dv}{5(v+2)}=\frac{dx}{x}$$
两边同时积分，得$（v-3）^7(v+2)^3=cx^5$ 这里c为正常数。 
   
代入$v=\frac{u}{x}=\frac{y^3}{x},并且化简得原方程的解为$$（y^3-3x）^7(y^3+2x)=cx^15$$.

（2）原方程可变形为$$\frac{ydy}{xdx}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$$
即$$\frac{dy^2}{dx^2}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$$
令$u=x^2-1,v=y^2+1$,则上面方程化为$$\frac{dv}{du}=\frac{2u+3v}{3u+2v}$$
再令$w=\frac{v}{u}$,即$v=uw$
则上面方程化为 
$$\frac{udw}{du}+w=\frac{2+3w}{3+2w}$$ 
分离变量并且积分，得$c·(1+w)=u^4(1-w)^5$
把$w=\frac{v}{u}$,$v=y^2+1$,$u=x^2-1$,代入得$$c(x^2+y^2)=(x^2-y^2-2)^5$$ 这里c是任意常数。
