(7)$y=x^2+1,y'=y^2-(x^2+1)y+2$  

解：因为$y=x^2+1,y'=2x$,所以$y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)(x^2+1)+2x=2x=y'$  
所以$y=x^2+1$是方程$y'=y^2-(x^2+1)y+2x$的解

(8)$y=\frac{\mathrm -g(x)}{\mathrm f(x)},y'=\frac{\mathrm f'(x)}{\mathrm g(x)}y^2-\frac{\mathrm g'(x)}{\mathrm f(x)}$

解： 因为$y=-\frac{\mathrm g(x)}{\mathrm f(x)}$,所以$y'=-\frac{\mathrm g'(x)f(x)-g(x)f'(x)}{\mathrm f^2(x)}=\frac{\mathrm g(x)f'(x)}{\mathrm f^2(x)}-\frac{\mathrm g'(x)}{\mathrm f(x)}$   
$\frac{\mathrm f'(x)}{\mathrm g(x)}y^2-\frac{\mathrm g'(x)}{\mathrm f(x)}=\frac{\mathrm f'(x)}{\mathrm g(x)}(\frac{\mathrm -g(x)}{\mathrm f(x)})^2-\frac{\mathrm g'(x)}{\mathrm f(x)}=\frac{\mathrm g(x)f'(x)}{\mathrm f^2(x)}-\frac{\mathrm g'(x)}{\mathrm f(x)}=y'$  
则有$y=-\frac{\mathrm g(x)}{\mathrm f(x)}$是方程$y'=\frac{\mathrm f'(x)}{\mathrm g(x)}y^2-\frac{\mathrm g'(x)}{\mathrm f(x)}$的解

(5)绘出（2），（3），（4）中的解的图形.

解：（1）由$\frac{\mathrm dy}{\mathrm dx}=2x$得          $dy=2xdx$

所以$c=4$,即与直线$y=2x+3$相切的解是$y=x^2+4$

(4)因为$\int_0^1\mathrm y\,\mathrm {d}x=\int_0^1\mathrm (x^2+c)\,\mathrm {d}x=\frac{\mathrm 1}{\mathrm 3}+c=2$  
所以$c=\frac{\mathrm 5}{\mathrm 3}$,则满足条件的解为$y=x^2+\frac{\mathrm 5}{\mathrm 3}$

（6）$x\frac{\mathrm dy}{\mathrm dx}-y+\sqrt{x^2-y^2}=0$

解：当$y^2≠x^2$时，两边同除以$\sqrt{x^2-y^2}$得  
$$ \frac{\mathrm xdy-ydx}{\mathrm {\sqrt{x^2-y^2}}}+dx=0$$  
即$$\frac{\mathrm {d(\frac{\mathrm y}{\mathrm x})}}{\mathrm {\sqrt{1-(\frac{\mathrm y}{\mathrm x})^2}}}+\frac{\mathrm dx}{\mathrm x}=0$$  
两边积分得$arcsin\frac{\mathrm y}{\mathrm x}+ln|x|=c$  （c为任意常数 ）   
当$y^2=x^2$时，容易验证$y^2=x^2$也是方程的解   
所以，原方程的解为$$y^2=x^2$或$arcsin\frac{\mathrm y}{\mathrm x}+ln|x|=c$$

(6)$\frac{\mathrm dy}{\mathrm dx}=\frac{\mathrm y^6-2x^2}{\mathrm 2xy^5+x^2y^2}$  

解：令$u=y^3$则方程可化为$$\frac{\mathrm du}{\mathrm dx}=\frac{\mathrm 3u^2-6x^2}{\mathrm 2xu+x^2}$$   
再令$v=\frac{\mathrm u}{\mathrm x}$即$u=vx$,则方程可化为$$x\frac{\mathrm dv}{\mathrm dx}+v=\frac{\mathrm 3v^2-6}{\mathrm 2v+1}$$  
整理得$$\frac{\mathrm 7dv}{\mathrm 5(v-3)}+\frac{\mathrm 3dv}{\mathrm 5(v+2)}=\frac{\mathrm dx}{\mathrm x}$$  
两边同时积分得 $(v-3)^7(v+2)^3=c·x^5$ （c为正常数）  
代入$v=\frac{\mathrm u}{\mathrm x}=\frac{\mathrm y^3}{\mathrm x}$,  
得原方程的解为$$(y^3-3x)^7(y^3+2x)=c·x^{15}$$

(7)$\frac{\mathrm dy}{\mathrm dx}=\frac{\mathrm 2x^3+3xy^2+x}{\mathrm 3x^2y+2y^3-y}$

解：原方程可化为$$\frac{\mathrm ydy}{\mathrm xdx}=\frac{\mathrm 2x^2+3y^2+1}{\mathrm 3x^2+2y^2-1}$$  
即$$\frac{\mathrm dy^2}{\mathrm dx^2}=\frac{\mathrm 2x^2+3y^2+1}{\mathrm 3x^2+2y^2-1}$$  
令$$u=x^2-1,v=y^2+1$$,  
则方程可化为$$\frac{\mathrm dv}{\mathrm du}=\frac{\mathrm 2u+3v}{\mathrm 3u+2v}$$   
再令$w=\frac{\mathrm v}{\mathrm u}$即$v=wu$,  
则方程可化为$$u·\frac{\mathrm dw}{\mathrm du}+w=\frac{\mathrm 2+3w}{\mathrm 3+2w}$$  
求得$$c·(1+w)=u^4(1-w)^5$$  
把$w=\frac{\mathrm v}{\mathrm u},v=y^2+1,u=x^2-1$代入  
得 $$c(x^2+y^2)=(x^2-y^2-2)^5$$ (c为任意常数）
