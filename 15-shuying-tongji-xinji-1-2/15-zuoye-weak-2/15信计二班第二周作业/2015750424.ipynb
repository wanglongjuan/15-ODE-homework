P27 第三题 验证下列各函数是相应微分方程的解：

(7)$y=x^2+1,y'=y^2-(x^2+1)y+2$

(8)$y=\frac{\mathrm -g(x)}{\mathrm f(x)},y'=\frac{\mathrm f'(x)}{\mathrm g(x)}y^2-\frac{\mathrm g'(x)}{\mathrm f(x)}$

解：(7)$因为y=x^2+1,y'=2x$

所以$y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)(x^2+1)+2x=2x=y'$

所以$y=x^2+1$是方程$y'=y^2-(x^2+1)y+2x$的解

(8) 因为$y=-\frac{\mathrm g(x)}{\mathrm f(x)}$

所以$y'=-\frac{\mathrm g'(x)f(x)-g(x)f'(x)}{\mathrm f^2(x)}=\frac{\mathrm g(x)f'(x)}{\mathrm f^2(x)}-\frac{\mathrm g'(x)}{\mathrm f(x)}$

$\frac{\mathrm f'(x)}{\mathrm g(x)}y^2-\frac{\mathrm g'(x)}{\mathrm f(x)}=\frac{\mathrm f'(x)}{\mathrm g(x)}(\frac{\mathrm -g(x)}{\mathrm f(x)})^2-\frac{\mathrm g'(x)}{\mathrm f(x)}=\frac{\mathrm g(x)f'(x)}{\mathrm f^2(x)}-\frac{\mathrm g'(x)}{\mathrm f(x)}=y'$

则有$y=-\frac{\mathrm g(x)}{\mathrm f(x)}$是方程$y'=\frac{\mathrm f'(x)}{\mathrm g(x)}y^2-\frac{\mathrm g'(x)}{\mathrm f(x)}$的解

  第四题    给定一阶微分方程$\frac{\mathrm dy}{\mathrm dx}=2x$,

(2)求通过点（1，4）的特解；

(3)求出与直线$y=2x+3$相切的解；

(4)求出满足条件

(5)绘出（2），（3），（4）中的解的图形.

解：（1）由$\frac{\mathrm dy}{\mathrm dx}=2x$得          $dy=2xdx$

方程两边积分可得$y=x^2+c$,c为任意常数.

所以方程$\frac{\mathrm dy}{\mathrm dx}=2x$的通解为$y=x^2+c$,c为任意常数.

(2)把$x=1,y=4$代入$y=x^2+c$得$c=3$

所以过点（1,4）的特解为$y=x^2+3$.

(3)因为与直线相切，所以方程组$ \begin{cases}y=x^2+c\\y=2x+3\end{cases}$

有且只有唯一一组解，即$x^2+c=2x+3$有唯一解

所以$c=4$,即与直线$y=2x+3$相切的解是$y=x^2+4$

(4)因为$\int_0^1\mathrm y\,\mathrm {d}x=\int_0^1\mathrm (x^2+c)\,\mathrm {d}x=\frac{\mathrm 1}{\mathrm 3}+c=2$

所以$c=\frac{\mathrm 5}{\mathrm 3}$

则满足条件的解为$y=x^2+\frac{\mathrm 5}{\mathrm 3}$

再令$v=\frac{\mathrm u}{\mathrm x}$即$u=vx$,则方程可化为$x\frac{\mathrm dv}{\mathrm dx}+v=\frac{\mathrm 3v^2-6}{\mathrm 2v+1}$

整理得$\frac{\mathrm 7dv}{\mathrm 5(v-3)}+\frac{\mathrm 3dv}{\mathrm 5(v+2)}=\frac{\mathrm dx}{\mathrm x}$

两边同时积分得 $(v-3)^7(v+2)^3=c·x^5$ （c为正常数）

代入$v=\frac{\mathrm u}{\mathrm x}=\frac{\mathrm y^3}{\mathrm x}$,简化方程得原方程的解为$(y^3-3x)^7(y^3+2x)=c·x^{15}$
