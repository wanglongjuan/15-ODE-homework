## P42.1求下列方程的解
### $(6)x\frac {dy}{dx}-y+\sqrt{x^2-y^2}=0$.


 解：
 $$ x\frac{\mathrm d y}{\mathrm d x}-y+\sqrt{x^2-y^2}=0 $$

 $$x\mathrm dy-y\mathrm dx+\sqrt{x^2-y^2}\mathrm dx=0 $$
 
 $$当 y^2 \neq x^2  时， \frac{x\mathrm dy-y\mathrm dx}{\sqrt{x^2-y^2}}+\mathrm dx=0 $$

   $$ -\frac{\mathrm d(\frac{y}{x})}{\sqrt{1-(\frac{y}{x})^2}}+\frac{1}{x}\mathrm dx=0 $$
   
 $$两边积分得 -\arcsin\frac{y}{x}+\ln{|x|}=c $$

 $$当 y^2 = x^2  时， y^2 = x^2 也是方程解$$

$$所以，解为  y^2 = x^2  和  \arcsin\frac{y}{x}=\ln{|x|}+c $$

## 2.做适当的变量变换求解下列方程：
### (6) $ \frac{dy}{dx} = \frac{y^6-2x^2}{2xy^5+x^2y^2}$

解：

$$ 令u=y^3,则方程为 \frac{\mathrm du}{\mathrm dx}=\frac{3u^2-6x^2}{2xu+x^2}$$

$$再令 v=\frac{u}{x}, 即 x\frac{\mathrm dv}{\mathrm dx}+v=\frac{3v^2-6}{2v+1}$$

$$ 得 \frac{7}{5}\frac{\mathrm dv}{v-3}+\frac{3}{5}\frac{\mathrm dv}{v+2}=\frac{\mathrm dx}{x}$$

$$ 两边积分 (v-3)^7(v+2)^3=cx^5 $$

$$ 代回得 (y^3-3x)^7(y^3+2x)^3=cx^{15} 为解$$


### (7)  $ \frac{\mathrm dy}{\mathrm dx} =  \frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$ 

解：

$$ 先同时除以 x 得 \frac{dy}{dx}= \frac{2x^2+3y^2+1}{3x^2+2y^2-1} $$

$$ 令 u=x^2-1,v=y^2+1,则 \frac{\mathrm dv}{\mathrm du}=\frac{2u+3v}{3u+2v}$$

$$ 再令 w=\frac{v}{u},则\frac{\mathrm dv}{\mathrm du}=w+u\frac{\mathrm dw}{\mathrm du}$$

$$ 所以 u\frac{\mathrm dw}{\mathrm du}=\frac{2-2w^2}{3+2w}$$

$$ 分离变量得 \frac{4}{u}\mathrm du=(\frac{5}{1-w}+\frac{1}{1+w})\mathrm dw $$

$$ 两边积分得 u^4(1-w)^5=c(1+w)$$

$$ 代入原式得$$

$$ 解为(x^2-y^2-2)^5=c(x^2+y^2) $$
