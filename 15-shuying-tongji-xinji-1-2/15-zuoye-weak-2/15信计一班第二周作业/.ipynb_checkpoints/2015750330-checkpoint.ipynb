给定一阶微分方程dy/dx=2x
(1)求出它的通解
解：由dy/dx=2x，得dy=2xdx,方程两边积分，既得y=x^2+c,其中c为任意常数
    所以方程dy/dx=2x的通解为y=x^2+c,c为任意常数
(2)求通过点（1，4)的特解
解：将点（1，4）带入得c=3
    特解为y=x^2+3
(3)求出与直线y=2x+3相切解
解：x^2+c=2x+3只有一个解
        c=4
    所以相切解为y=x^2+4
(4)求出满足条件∫ydx=2的解
解：∫ydx=∫(x^2+c)=x^3/3+cx
    有题意得，c=5/3
    所以y=x^2+5/3
(5)绘出(2),(3),(4)中的解的图形


```python
var('x')
f(x)=x^2+3
plot(f(x,-1,1))
```

```python
验证y=-g(x)/f(x)是y'=[f'(x)/g(x)]y^2-g'(x)/f(x)的解
var('x')
f=function('f')(x)
g=function('g')(x)
y=-g/f
y1=diff(y,x)
f1=diff(f,x)
g1=diff(g,x)
s=(f1/g)*y^2-g1/f-y1
show(s)
```

```python
验证y=x^2+1是y'=y^2-(x^2+1）y+2x的解
var('x')
y=x^2+1
y1=diff(y,x)
s=y^2-(x^2+1)*y+2*x-y1
show(s)
```

    (v-3)^7(v+2)^3=c*x^3,其中c为正常数

(2)dy/dx=2x^3+3xy^2+x/3x^2y+2y^3-y
解：ydy/xdx=2x^2+3y^2+1/3x^2+2y^2-1
    令u=x^2-1,v=y^2+1
    dv/du=2u+3v/3u+2v
    再令w=v/u
    dv/du=w+udw/du
    udw/du=2-2w^2/3+2w
    分离变量得4du/u=(5/(1-w)+1/(1+w)dw
    u^4(1-w)^5=c(1+w)
    (x^2-y^2-2)^5=c(x^2+y^2),其中c为任意常数
