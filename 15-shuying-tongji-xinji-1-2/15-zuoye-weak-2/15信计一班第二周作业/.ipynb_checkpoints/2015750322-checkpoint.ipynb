    则$$\frac{du}{dx}=\frac{\sqrt{x^2(1-u^2)}}{x}$$

    $$arcsin\frac{y}{x}=sgn x*ln|x|+c$$

**P42.2(6):**$$ $$
    解：$$\frac{dy}{dx}=\frac{(y^3)^2-2x^2}{y^2(2xy^3+x^2)}则\frac{dy^3}{dx}=\frac{3((y^3)^2-2x^2)}{2xy^3+x^2}$$
    $$令y^3=u则原方程化为\frac{du}{dx}=\frac{3u^2-6x^2)}{2xu+x^2}=\frac{\frac{3u^2)}{x^2}-6}{\frac{2u)}{x}+1}$$
    $$这是齐次方程，令\frac{u}{x}=z,则\frac{du}{dx}=z+x\frac{dz}{dx}$$
    $$所以\frac{3z^2-6}{2z+1}=z+x\frac{dz}{dx},x\frac{dz}{dx}=\frac{z^2-z-6}{2z+1}(1)$$
    $$当z^2-z-6=0时，得z=3或z=-2是(1)方程的解，即y^3=3x或y^3=-2x^2是方程的解$$
    $$当z^2-z-6!=0时，变量分离\frac{2z+1}{z^2-z-d}dz=\frac{dx}{x}$$
    $$两边积分得：(z-3)^7(z+2)^3=x^5c,$$
    $$即(y^3-3)^7(y^3+2)^3=x^5c,又因为当c=0时，y^3=3x或y^3=-2x^2包含在通解中$$
    $$故原方程的解为(y^3-3)^7(y^3+2)^3=x^5c$$

**P42.2(7):**$$ $$
    解：原方程化为：$$\frac{dy}{dx}=\frac{x(2x^2+3y^2+1)}{y(3x^2+2y^2-1)}$$
    则$$\frac{dy^2}{d^2 x}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$$
    令$$y^2=u ,x^2=v$$
    则$$\frac{du}{dv}=\frac{2v+3u+1}{3v+2u-1}$$(1)
    方程组$$\begin{cases}2v+3u+1=0\\3v+2u-1=0\end{cases}$$的解为(1,-1)
    令  $$a=v-1,b=u-1$$
    则有$$\begin{cases}2a+3b=0\\2b+3a=0\end{cases}$$
    从而方程(1)化为$$\frac{db}{da}=\frac{2+3\frac{b}{a}}{3+2\frac{b}{a}}$$
    令$$t=\frac{b}{a}$$
    则有$$\frac{db}{da}=t+a\frac{dt}{da}$$
    所以$$t+a\frac{dt}{da}=\frac{2+3t}{3+2t},a\frac{dt}{da}=\frac{2-2t^2}{3+2t}$$
    $$当2-2t^2=0时，得y=x^2-2或y=-x^2是原方程的解$$
    $$当2-2t^2!=0时，分离变量得\frac{3+2t}{2-2t^2}dt=\frac{da}{a}两边积分得y^+x^2=(y^2-x^2+2)^5c$$
    $$另外y^=x^2-2或y=-x^2包含在其通解中，故原方程的解为y^+x^2=(y^2-x^2+2)^5c$$

$$\frac {\mathrm dy_2(t)}{\mathrm dt} = \frac {\mathrm dy_1(t-t_0)}{\mathrm dt}= \frac{\mathrm dy_1(t-t_0)}{\mathrm d(t-t_0)}=f(y_1(t-t_0)) = f(y_2)$$  所以 $$ y_2 = y_1(t-t_0)$$也为  $$ \frac{\mathrm dy}{\mathrm dt} = f(y)$$的解。由 $$y_1(0)= y_0$$得 即$$ t=0$$ 时  $$y_1(0)=y_0$$ 所以$$ t-t_0=0$$时  $$y_1(t-t_0)=y_0$$ 所以 $$ y_2(t_0)=y_0$$ 所以 $$ y_2(t_0)=y_1(t-t_0)$$也是方程的解
