4.给定一阶微分方程dy/dx=2x
1)方程两边同乘以dx得dy=2xdx
∫dy=∫2xdx
即y=x^2+c
2)把y=1，x=4代入得c=3
∴通解为y=x^2+3
3）∵与直线相切
   ∴方程组y=x^2+c只有一个解
           y=2x+3
   ∴c=4
   ∴与直线y=2x+3相切的解为y=x^2+4
4）∵∫ydx=2（0~1）
∴c=5/3
∴满足条件的解为y=x^2+5/3
