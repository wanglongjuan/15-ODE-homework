3.(7)$ y=x^2+1 $ ，$ y^{\prime}=y^2-(x^2+1)y+2x $

解：∵  $ y=x^2+1 $ 

∴ $ y^{\prime}=2x $ 

又∵ $ y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)^2+2x=2x=y^{\prime} $

∴ $ y=x^2+1 $ 是方程 $ y^{\prime}=y^2-(x^2+1)y+2x $ 的解


(8)$ y=-g(x)/f(x) $ ,$ y^{\prime}=f(x)^{\prime}y^2/g(x)-g(x)^{\prime}/f(x) $

解：∵ $ y=-g(x)/f(x)$

∴ $ y^{\prime}=-(g(x)^{\prime}f(x)-g(x)f(x)^{\prime})/f(x)^2=g(x)f(x)^{\prime}/f(x)^2-g(x)^{\prime}/f(x) $

又∵ $ f(x)^{\prime}y^2/g(x)-g(x)^{\prime}/f(x)=(-g(x)/f(x))^2*f(x)^{\prime}/g(x)-g(x)^{\prime}/f(x)=g(x)f(x)^{\prime}/f(x)^2-g(x)^{\prime}/f(x)=y^{\prime} $

∴ $ y=-g(x)/f(x) $ 是方程 $ y^{\prime}=f(x)^{\prime}y^2/g(x)-g(x)^{\prime}/f(x) $的解

1.解：(6)原方程可化为 $ xdy-ydx+\sqrt(x^2-y^2)dx=0 $

当 $ y^2≠x^2 $时

两边同时除以 $ \sqrt(x^2-y^2) $ 得到 $ (xdy-ydx)/\sqrt(x^2-y^2)+dx=0 $

即 $ d(y/x)/\sqrt(1-(y/x)^2)+1dx/x=0 $

两边积分得到 $ arsin(y/x)+ln|x|=c $ c为任意常数

而且，当 $ y^2=x^2 $ 时，可得 $ y^2=x^2 $ 也是方程的解

∴方程的解为 $ y^2=x^2 $ 和 $ arcsin(y/x)+ln|x|=c $

2.解：(6)令 $ u=y^3 $ 则方程可化为 $ du/dx=(3u^2-6x^2)/(2xu+x^2) $

再令 $ v=u/x,即u=vx $ 则上方程可化为 $ xdv/dx+v=(3v^2-6)/(2v+1)$

整理得到 $ 7/5*dv/(v-3)+3/5*dv/(v+2)=dx/x $

两边同时积分，得 $ (v-3)^7(v+2)^3=cx^5 $ c为正常数

代回原变量，然后化简原方程的解为 $ (y^3-3x)^7(y^3+2x)^3=c*x^15 $

(7)原方程可以化为 $ ydy/xdx=(2x^2+3y^2+1)/(3x^2+2y^2-1) $

令 $ u=x^2-1,v=y^2+1,则原方程可化为dv/du=(2u+3v)/(3u+2v) $

再令 $ w=v/u,即v=uw,则dv/du=w+udw/du,则dw/du=(2-2w^2)/(3+2w) $

分离变量化简得 $ 4du/u=(5/(1-w)+1/(1+w))dw $

两边积分得 $ u^4*(1-w)^5=c(1+w) $

代回原变量，得到方程的解为 $ (x^2-y^2-2)^5=c(x^2+y^2) $ c为任意常数
