1.求下列方程的解：

（6）$x\frac{dy}{dx}-y+\sqrt{x^2-y^2}=0$

解：$$x\frac{dy}{dx}-y+\sqrt{x^2-y^2}=0$$

同除x，得：$$\frac{dy}{dx}-\frac{y}{x}+\sqrt{1-\frac{y^2}{x^2}}=0$$

令$u=\frac{y}{x}$ 

则$$\frac{dy}{dx}=u+x\frac{du}{dx}$$

带回原式得$$x\frac{du}{dx}=-\sqrt{1-u^2}$$

则$$x\frac{du}{\sqrt{1-u^2}}=-\frac{dx}{dx}$$

则$$arcsinu=-ln|x|+c$$

解得$u^2=1$，则方程为$$y=\pm{x}$$

（7）$tanydx-cotxdy=0$；
解：$$\frac{dy}{tany}=\frac{dx}{cotx}$$

则$$\frac{cosydy}{siny}=\frac{sinxdx}{cosx}$$

积分得$$∫\frac{cosydy}{siny}=∫\frac{sinxdx}{cosx}$$

化简得$$∫\frac{dsiny}{siny}=-∫\frac{dcosx}{cosx}$$
$$siny=\frac{C}{cosx}$$
通解为$$y=arcsin(\frac{C}{cosx})$$
