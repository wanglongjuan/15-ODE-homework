3.验证下列各函数是相应微分方程的解     
   $（7）y=x^2+1, y\prime=y^2-(x^2+1)y+2x$

解：$y=x^2+1$可得$y\prime=2x$    
又$y^2-(x^2+1)y+2x=2x$    
故函数是相应微分方程的解

$(8)y=-\frac{g(x)}{f(x)},y\prime=\frac{f\prime(x)}{g(x)}y^2-\frac{g\prime(x)}{f(x)}$

解：$y=-\frac{g(x)}{f(x)}$    
可求得$y\prime=\frac{g(x)f\prime(x)-g\prime(x)f(x)}{f^2(x)}$             
又$\frac{f\prime(x)}{g(x)}y^2-\frac{g\prime(x)}{f(x)}=\frac{g(x)f\prime(x)-g\prime(x)f(x)}{f^2(x)}$    
故函数是相应微分方程的解

4.给定一阶微分方程$\frac{{\mathrm d}y}{{\mathrm d}x}=2x$

(1)求出它的通解

解：由题得到${\mathrm d}y=2x{\mathrm d}x$         
即$y=x^2+c$

（2）求通过点（1,4）的特解

解：把(1,4)带入方程$y=x^2+c$                
得到$c=3$               
特解为$y=x^2+3$

（3）求出与直线$y=2x+3$相切的解

解：若要与直线相切，可得     
$x^2+c=2x+3$有唯一解     
得到$c=4$       
故所求$y=x^2+4$

（4）求出满足条件$\int_0^1ydx=2$的解

解：      $y=x^2+c$                   
              即 $\int_0^1ydx=\frac{1}{3}x^3+cx|_0^1$          
              解得$\frac{1}{3}+c=2$            
              $c=\frac{5}{3}$    
              所求解$y=x^2+\frac{5}{3}$

1.求下列方程的解         
(6)$x\frac{dy}{dx}-y+\sqrt{(x^2-y^2)}=0$

解：$x\frac{{\mathrm d}y}{{\mathrm d}x}=y-\sqrt{x^2-y^2}$     
令$\frac{y}{x}=u$
可得到$\frac{{\mathrm d}y}{{\mathrm d}x}=u-\frac{{\mathrm d}u}{{\mathrm d}x}$     
即$\frac{{\mathrm d}u}{{\mathrm d}x}=\frac{\sqrt{x^2(1-u^2)}}{x}$    
分离变量可得到$\frac{{\mathrm d}u}{\sqrt{1-u^2}}=-\frac{1}{x}{\mathrm d}x$   
${\arcsin u}+In|x|=c$
另还有解$y^2=x^2$

(7)$\tan y{\mathrm d}x-\cot x{\mathrm d}y=0$

解：$\frac{{\mathrm d}y}{\tan y}=\frac{{\mathrm d}x}{\cot x}$      
   可得${\cot y}{\mathrm d}y={\tan x}{\mathrm d}x$       
   即$In|{\sin y}|=-In|{\cos x}|+C$

2.作适当的变量变换求解下列方程：

(6)$\frac{{\mathrm d}y}{{\mathrm d}x}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$

解：$\frac{{\mathrm d}y}{{\mathrm d}x}=\frac{(y^3)^2-2x^2}{y^3(2xy^3+x^2)}$                
即$\frac{{\mathrm d}y^3}{{\mathrm d}x}=\frac{3((y^3)^2-2x^2)}{2xy^3+x^2}$     
令$u=y^3$           
得到$\frac{{\mathrm d}u}{{\mathrm d}x}=\frac{3u^2-6x^2}{2xu+x^2}=\frac{\frac{3u^2}{x^2}-6}{2{\frac{u}{x}+1}}$      
再令$z=\frac{u}{x}$    
得到$\frac{3z^2-6}{2z+1}=z+x\frac{{\mathrm s}z}{{\mathrm d}x}$    
$x\frac{{\mathrm d}z}{{\mathrm d}x}=\frac{z^2-z-6}{2z+1}$
当：     
$z^2-z-6=0$,得到$z=3,z=-2$     
即$y^3=3x$或者$y^3=-2x$
当：    
$z^2-z-6 ≠ 0$,分离变量，得到   
$\frac{2z+1}{z^2-z-6}{{\mathrm d}z}=\frac{1}{x}{\mathrm d}x$       
两边积分，有$(y^3-3x)^7(y^3+2x)^3=x^5c$    
又之前求得$y^3=3x,y^3=-2x$包含在通解中$c=0$时    
故可求得：    
$(y^3-3x)^7(y^3+2x)^3=x^{15}c$ 

(7)$\frac{{\mathrm d}y}{{\mathrm d}x}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$

解：将原方程化为：$\frac{{\mathrm d}y}{{\mathrm d}x}=\frac{x(2x^2+3y^2+1)}{y(3x^2+2y^2-1)}$    
故$\frac{{\mathrm d}y^2}{{\mathrm d}x^2}=\frac{(2x^2+3y^2+1)}{(3x^2+2y^2-1)}$
令$y^2=u,x^2=v$         
得到$\frac{{\mathrm d}u}{{\mathrm d}v}=\frac{2v+3u+1}{3v+2u-1}$     
即，建立方程组 
$  2v+3u+1=0，
     3v+2u-1=0$     
解为$u=1,v=-1$
那么   
再令，$z=v-1,y=u+1$   
则$2z+3y=0,3z+2y=0$   
$\frac{{\mathrm d}y}{{\mathrm d}z}=\frac{2+3\frac{y}{z}}{3+2\frac{y}{z}}$    
又，令$t=\frac{y}{z}$       
即$\frac{{\mathrm d}y}{{\mathrm d}z}=t+\frac{{\mathrm d}t}{{\mathrm d}z}$        
$z\frac{{\mathrm d}t}{{\mathrm d}z}=\frac{2-2t^2}{3+2t}$   
当： 
$2-2t^2=0$时，得$t=1$或者$t=-1$  
得$y=x^2-2$或者$y=-x^2$是原方程的解  
$2-2t^2≠0$时，分离变量，可得 
$\frac{3+2t}{2-2t^2}=\frac{1}{z}{\mathrm d}z$  
两边积分可得$y^2+x^2=(y^2-x^2+2)^5c$  
另外  
$y^2=x^2-2$或者$y^2=-x^2$包含在通解中，得到   
$y^2+x^2=(y^2-x^2+2)^5c$

题目：设$y_1(t)$是$\frac{{\mathrm d}y}{{\mathrm d}t}=f(y),y(0)=y_0$的解，证明$y_2(t)=y_1(t-t_0)$是$\frac{{\mathrm d}y}{{\mathrm d}t}=f(y),y(t_0)=y_0$的解

解：$\frac{{\mathrm d}y_2(t)}{{\mathrm d}t}=\frac{{\mathrm d}y_1(t-t_0)}{{\mathrm d}t}=\frac{{\mathrm d}y_1(t-t_0)}{{\mathrm d}(t-t_0)}$      
得$f(y_1(t-t_0))=f(y_2(t))$     
那么有$y_2(t)=y_1(t-t_0)$也是$\frac{{\mathrm d}y}{{\mathrm d}t}=f(y)$的解   
$y_1(0)=y_0$可以得到$t=0$时$y_1(0)=y_0$  
$t-t_0=0$时$y_1(t-t_0)=y_0$    
故$y_2(t)=y_1(t-t_0)$是所给方程的解
