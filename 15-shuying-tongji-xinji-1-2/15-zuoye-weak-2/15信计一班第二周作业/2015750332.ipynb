
3、验证下列各函数是相应微分方程的解:

（7）$ y=x^2+1 $，$ y^{\prime}= y^2-(x^2+1)y+2x $。

解：因为$ y=x^2+1 $

所以$ y^{\prime}=2x $

所以$ y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)*(x^2+1)+2x=2x=y^{\prime} $

所以$ y=x^2+1 $ 是 $ y^{\prime}=y^2-(x^2+1)y+2x $的解

（8）$ y=-\frac{g(x)}{f(x)} $，$ y^{\prime}=\frac{f^{\prime}(x)}{g(x)}y^2-\frac{g^{\prime}(x)}{f(x)} $


解：因为$ y=-\frac{g(x)}{f(x)} $

所以$ y^{\prime}=-\frac{g^{\prime}(x)f(x)-g(x)f^{\prime}(x)}{f^2(x)}
=\frac{g(x)f^{\prime}(x)}{f^2(x)}-\frac{g^{\prime}(x)}{f(x)} $

又因为$ \frac{f^{\prime}(x)}{g(x)}y^2-\frac{g^{\prime}(x)}{f(x)}
=\frac{f^{\prime}(x)}{g(x)}(-\frac{g(x)}{f(x)})^2-\frac{g^{\prime}(x)}{f(x)}
=\frac{g(x)f^{\prime}(x)}{f^2(x)}-\frac{g^{\prime}(x)}{f(x)}
=y^{\prime} $

所以$ y=-\frac{g(x)}{f(x)} $ 是$ y^{\prime}=\frac{f^{\prime}(x)}{g(x)}y^2-\frac{g^{\prime}(x)}{f(x)} $ 的解

（6）$\frac{\mathrm dy}{\mathrm dx}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$;

再令$v=\frac{u}{x}$,即$x\frac{\mathrm dv}{\mathrm dx}+v=\frac{3v^2-6}{2v+1}$

得$\frac{7}{5}\frac{\mathrm dv}{v-3}+\frac{3}{5}\frac{\mathrm dv}{v+2}=\frac{\mathrm dx}{x}$

两边积分$(v-3)^7(v+2)^3=cx^5$

代回得$(y^3-3x)^7(y^3+2x)^3=cx^{15}$为解
