1、假设A是$n\times n$矩阵，试证：

（2）对任意整数k，都有

$$(expA)^t=expkA$$

(当k是负整数时，规定$（expA)^k=[(expA)^{-1}]^{-k}.)$

证明：
由（1）知$exp(c_1A+c_2A)=expc_1A\cdot expc_2A;$

设$c_1=c_2=1$，有：

$$exp(2A)=exp(A+A)=expA\cdot expA=(expA)^2$$

$$exp(3A)=exp(2A+A)=exp2A\cdot expA=(expA)^3$$

$$……$$

$$exp(kA)=exp[(k-1)A+A]=(expA)^k$$

当k=0时；

$$(expA)^k=E,expkA=E$$

即等式成立

当k=1时，等式左边显然等于右边

由数学归纳法可知，等式对于所有的非负整数均成立。

所以当k为负整数时，有

$$(expA)^k[(expA)^{-1}]^{-k}=exp(-k)(-A)=expkA$$

这就是说当k为负整数时。等式同样成立。

综上可得：对于任意整数k，都有

$$(expA)^k=expkA$$

3、设$\Phi(t)$为方程$x’=Ax$(A为$n\times n$常数矩阵)的标准基解矩阵（即$\Phi(0)=E$），证明：

$$\Phi(t)\Phi^{-1}(t_0)=\Phi(t-t_0)$$

其中$t_0$为某一值。

证明：由题意可知A为$n\times n$矩阵，故方程$x’=Ax$的基解矩阵满足

$$\frac{d\Phi(t)}{dt}=A\cdot\Phi(t)$$

求得：$$\Phi(t)=e^{At}$$

所以$\Phi(t)$是指数型，由其特点可知

$$\Phi^{-1}(t)=e^{-At}$$

故

$$\Phi(t)\Phi^{-1}(t_0)=e^{At}e^{-At_0}=\Phi(t-t_0)$$

即证

解：特征方程为：$\lambda^3-8=0$
