5.设$\Phi(t)$为方程$x'=Ax$($A$为$n\times n$常数矩阵)的标准基解矩阵(即$\Phi(0)=E$),证明：$$\Phi(t)\Phi^{-1}(t_0)=\Phi(t-t_0),$$其中$t_0$为某一值.


解：由题意

$$\frac{\mathbf{\phi(t)}}{dt}=\mathbf{A}\mathbf{\phi(t)}$$

分离变量可得

$$\frac{\mathbf{\phi(t)}}{d\mathbf{\phi(t)}}=\mathbf{A}dt$$

积分得

$$\mathbf{\phi(t)}=e^{\mathbf{A}t}$$

又由

$$\mathbf{\phi(t)}\mathbf{\phi't)}=\mathbf{E}$$

$$\mathbf{\phi't)}=e^{\mathbf-{A}t}$$

故

$$\mathbf{\phi(t)}\mathbf{\phi'(t)}=e^{\mathbf{A}(t-t_0)}=\mathbf{\phi(t-{t_0})}$$
