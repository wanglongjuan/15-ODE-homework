1.考虑方程组$x'=A(T)x$

其中A(t)是区间a≤t≤b上的连续n×n矩阵，他的元为$a_{ij}(t)(i,j=1,2,...,n)$

(1).如果$x_1(t),x_2(t),...,x_n(t)$是方程组的任意n个解，那么他的朗斯基行列式$W[x_1(t),x_2(t),...,x_n(t)]≡W(t)$满足下列一阶线性微分方程

$W'=[a_{11}(t)+a_{22}(t)+...+a_{nn}(t)]W$

(2).解上面线性微分方程，证明下列公式：

$W(t)=W(t_0)e^{\int^t_{t_0}[a_{11}(s)+a_{22}(s)+...+a_{nn}(s)]dt}$

解：（1）$将x_1(t),x_2(t),…x_n(t)用向量的形式表示，并写出它们的朗斯基行列式为$

$$W(t)=
\begin{vmatrix}
x_{11}(t)&x_{12}(t)&…&x_{1n}(t)\\
x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\
…&…&…&…\\
x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)
\end{vmatrix},$$

$对W(t)求导得$

$\begin{aligned}
\frac{\mathrm d W(t)}{\mathrm d t}&=\frac{\mathrm d}{\mathrm d t}
\begin{vmatrix}
x_{11}(t)&x_{12}(t)&…&x_{1n}(t)\\
x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\
…&…&…&…\\
x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)
\end{vmatrix}\\
&=\begin{vmatrix}
x'_{11}(t)&x'_{12}(t)&…&x'_{1n}(t)\\
x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\
…&…&…&…\\
x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)
\end{vmatrix}
+
\begin{vmatrix}
x_{11}(t)&x_{12}(t)&…&x_{1n}(t)\\
x'_{21}(t)&x'_{22}(t)&…&x'_{2n}(t)\\
…&…&…&…\\
x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)
\end{vmatrix}
+…+\begin{vmatrix}
x_{11}(t)&x_{12}(t)&…&x_{1n}(t)\\
x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\
…&…&…&…\\
x'_{n1}(t)&x'_{n2}(t)&…&x'_{nn}(t)
\end{vmatrix}
\end{aligned}.\quad\quad\quad\quad\quad\quad\quad\quad\quad$①

$$由于
\begin{aligned}
x'_{11}&=a_{11}x_{11}+a_{12}x_{21}+…+a_{1n}x{n1},\\
x'_{12}&=a_{11}x_{12}+a_{12}x_{22}+…+a_{1n}x{n2},\\
……\\
x'_{1n}&=a_{11}x_{1n}+a_{12}x_{2n}+…+a_{1n}x{nn},\\
\end{aligned}.$$

$则由行列式的性质有$

$\begin{aligned}
\begin{vmatrix}
x'_{11}(t)&x'_{12}(t)&…&x'_{1n}(t)\\
x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\
…&…&…&…\\
x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)
\end{vmatrix}
&=
\begin{vmatrix}
a_{11}x_{11}(t)&a_{11}x_{21}(t)&…&a_{11}x{1n}(t)\\
x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\
…&…&…&…\\
x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)
\end{vmatrix}\\
&+\begin{vmatrix}
a_{12}x_{21}(t)&a_{12}x_{22}(t)&…&a_{12}x{2n}(t)\\
x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\
…&…&…&…\\
x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)
\end{vmatrix}\\
&+…+
\begin{vmatrix}
a_{1n}x_{n1}(t)&a_{1n}x_{n2}(t)&…&a_{1n}x{nn}(t)\\
x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\
…&…&…&…\\
x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)
\end{vmatrix}\\
&=a_{11}(t)W(t).
\end{aligned}$

$同理①式右端第2行列式等于a_{22}（t）W(t),第 3行列式等于a_{33}(t)W(t)，…第n行列式等于a_{nn}(t)W(t).因此，可得$

$\quad\frac{\mathrm d W(t)}{\mathrm d t}=W'(t)=[a_{11}(t)+a_{22}(t)+…+a_{nn}(t)]W(t).$


(2)$将上面的一阶线性微分方程变形，可得$

$\quad\frac{\mathrm d W}{\omega}=[a_{11}(t)+a_{22}(t)+…+a_{nn}(t)]\mathrm d t.$

$对上式从t_0到t积分可得$

$\quad\ln\mid W(t)\mid-\ln\mid W(t_0)\mid=\int_{t_0}^t\ [a_{11}(s)+a_{22}(s)+…+a_{nn}(s)]\mathrm d s,$

$再用指数函数对两边作用，可得$

$\quad W(t)=W(t_0)e^{\int_{t_0}^t\ [a_{11}(s)+a_{22}(s)+…+a_{nn}(s)]\mathrm d s},其中t_0,t\in[a,b].$

2.设A(t)为区间a≤t≤b上连续n×n实矩阵，$\phi(t)$为方程$x'=A(t)x$的基解矩阵，而$x=\varphi(t)$为其一解，试证:

(1)对于方程$y'=-A^T(t)y$的任一解$y=\psi(y)$必有$\psi^T(t)\phi(t)$=常数

(2)$\psi(t)$为方程$y'=-A^T(t)y$的基解矩阵的充要条件是存在非奇异的常数矩阵C,使$\psi^T（t）\phi(t)=C.$

解：（1）$由于\varphi(t)为方程x'=A(t)x的解，则有$

$\varphi'(t)=
\begin{pmatrix}
\varphi'_1(t)\\
\varphi'_2(t)\\
…\\
\varphi'_n(t)
\end{pmatrix}
=\begin{pmatrix}
a_{11}&a_{12}&…&a_{1n}\\
a_{21}&a_{22}&…&a_{2n}\\
…&…&…&…\\
a_{n1}&a_{n2}&…&a_{nn}
\end{pmatrix}
\begin{pmatrix}
\varphi_1(t)\\
\varphi_2(t)\\
…\\
\varphi_n(t)
\end{pmatrix}=
\begin{pmatrix}
\displaystyle\sum_{k=1}^{n}a_{1k}\varphi_k\\
\displaystyle\sum_{k=1}^{n}a_{2k}\varphi_k\\
…\\
\displaystyle\sum_{k=1}^{n}a_{nk}\varphi_k\\
\end{pmatrix}$

$同理，由于\Psi(t)为方程y'=-A^TY的解,得$

$\Psi'(t)=
\begin{pmatrix}
\psi'_1(t)\\
\psi'_2(t)\\
…\\
\psi'_n(t)
\end{pmatrix}
=\begin{pmatrix}
a_{11}&a_{21}&…&a_{n1}\\
a_{12}&a_{22}&…&a_{n2}\\
…&…&…&…\\
a_{1n}&a_{2n}&…&a_{nn}
\end{pmatrix}
\begin{pmatrix}
\psi_1(t)\\
\psi_2(t)\\
…\\
\psi_n(t)
\end{pmatrix}=
-\begin{pmatrix}
\displaystyle\sum_{k=1}^{n}a_{k1}\psi_k\\
\displaystyle\sum_{k=1}^{n}a_{k2}\psi_k\\
…\\
\displaystyle\sum_{k=1}^{n}a_{kn}\psi_k\\
\end{pmatrix}$

$而对\Psi^T(t)\varphi(t)求导得$

$\begin{aligned}
\frac{\mathrm d (\Psi^T(t)\varphi(t))}{\mathrm d t}&=\frac{\mathrm d \Psi^T(t)}{\mathrm d t}\varphi(t)+\Psi^T(t)\frac{\mathrm d \varphi(t)}{\mathrm d t}\\
&=-(\displaystyle\sum_{k=1}^{n}a_{k1}\psi_k,\displaystyle\sum_{k=1}^{n}a_{k2}\psi_k,…,\displaystyle\sum_{k=1}^{n}a_{kn}\psi_k)
\begin{pmatrix}
\varphi_1(t)\\
\varphi_2(t)\\
…\\
\varphi_n(t)
\end{pmatrix}+(\varphi_1(t),\varphi_2(t),…,\varphi_n(t))
\begin{pmatrix}
\displaystyle\sum_{k=1}^{n}a_{1k}\varphi_k\\
\displaystyle\sum_{k=1}^{n}a_{2k}\varphi_k\\
…\\
\displaystyle\sum_{k=1}^{n}a_{nk}\varphi_k\\
\end{pmatrix}\\
&=-[\displaystyle\sum_{k=1}^{n}a_{k1}\psi_k\varphi_1+\displaystyle\sum_{k=1}^{n}a_{k2}\psi_k\varphi_2+…+\displaystyle\sum_{k=1}^{n}a_{kn}\psi_k\varphi_n]+[\displaystyle\sum_{k=1}^{n}\psi_1 a_{1k}\varphi_k+\displaystyle\sum_{k=1}^{n}\psi_2 a_{2k}\varphi_k+…+\displaystyle\sum_{k=1}^{n}\psi_n a_{nk}\varphi_k]\\
&=-\displaystyle\sum_{k=1}^{n}\displaystyle\sum_{i=1}^{n}(a_{ki}\psi_k\varphi_i)+\displaystyle\sum_{k=1}^{n}\displaystyle\sum_{i=1}^{n}(\psi_i a_{ik}\varphi_k)\\
&=-\displaystyle\sum_{k=1}^{n}\displaystyle\sum_{i=1}^{n}(a_{ki}\psi_k\varphi_i)+\displaystyle\sum_{i=1}^{n}\displaystyle\sum_{k=1}^{n}(a_{ik}\psi_i\varphi_k)\\
&=0
\end{aligned}.$

$∴\Psi^T(t)\varphi(t)=常数.$

（2）$充分性：存在非奇异矩阵C，使得\Psi^T(t)\phi(t)=C,由于\phi(t)是方程x'=A(t)x的基解矩阵，故\phi^{-1}(t)存在.$

$\quad ∴\Psi^T(t)=C·\Phi^{-1}(t)且\det(\Psi^T(t))=\det(C·\Phi^{-1}(t))\neq 0.∴\Psi(t)是基解矩阵.$

$必要性：\Psi(t)是基解矩阵，则\Psi(t)的每一个向量\psi_1(t),\psi_2(t),…\psi_n(t)与\Psi(t)的每一个列向量\varphi_1(t),\varphi_2(t),…\varphi_n(t)满足\varphi_i^T\varphi_j(t)=常数,i=1,2,…n,$

$\quad j=1,2,…n.则\Psi^T(t)\phi是常数矩阵，记为c,另外由于\Psi(t)和\phi(t)是基解矩阵，∴\det(\Psi^T(t))\neq 0,\det(\phi(t))\neq 0.$

$\quad则\det C=\det(\phi(t))\det(\Psi^T(t))\neq 0,即矩阵C是非奇异的.$
