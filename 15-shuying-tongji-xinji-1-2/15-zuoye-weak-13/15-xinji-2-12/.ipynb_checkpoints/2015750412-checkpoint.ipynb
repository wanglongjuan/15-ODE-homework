x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)

x_{11}(t)&x_{12}(t)&…&x_{1n}(t)\\
x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\

x'_{n1}(t)&x'_{n2}(t)&…&x'_{nn}(t)

x'_{1n}&=a_{11}x_{1n}+a_{12}x_{2n}+…+a_{1n}x{nn},\\
\end{aligned}.$$

所以：

$\begin{aligned}
\begin{vmatrix}

3.(1)

$因为\varphi(t)是方程x'=A(t)x的解，所以：$

$$\varphi'(t)=
\begin{pmatrix}
a_{11}&a_{12}&…&a_{1n}\\
a_{21}&a_{22}&…&a_{2n}\\
…&…&…&…\\
a_{n1}&a_{n2}&…&a_{nn}
\end{pmatrix}
\begin{pmatrix}
\varphi_1(t)\\
\varphi_2(t)\\
…\\
\varphi_n(t)
\end{pmatrix}=
\begin{pmatrix}
\displaystyle\sum_{k=1}^{n}a_{1k}\varphi_k\\
\displaystyle\sum_{k=1}^{n}a_{2k}\varphi_k\\
…\\
\displaystyle\sum_{k=1}^{n}a_{nk}\varphi_k\\
\end{pmatrix}$$

同理:

$$\Psi'(t)=
\begin{pmatrix}
a_{11}&a_{21}&…&a_{n1}\\
a_{12}&a_{22}&…&a_{n2}\\
…&…&…&…\\
a_{1n}&a_{2n}&…&a_{nn}
\end{pmatrix}
\begin{pmatrix}
\psi_1(t)\\
\psi_2(t)\\
…\\
\psi_n(t)
\end{pmatrix}=
-\begin{pmatrix}
\displaystyle\sum_{k=1}^{n}a_{k1}\psi_k\\
\displaystyle\sum_{k=1}^{n}a_{k2}\psi_k\\
…\\
\displaystyle\sum_{k=1}^{n}a_{kn}\psi_k\\
\end{pmatrix}$$

所以：

$\begin{aligned}
\frac{\mathrm d (\Psi^T(t)\varphi(t))}{\mathrm d t}&=-(\displaystyle\sum_{k=1}^{n}a_{k1}\psi_k,\displaystyle\sum_{k=1}^{n}a_{k2}\psi_k,…,\displaystyle\sum_{k=1}^{n}a_{kn}\psi_k)
\begin{pmatrix}
\varphi_1(t)\\
\varphi_2(t)\\
…\\
\varphi_n(t)
\end{pmatrix}+(\varphi_1(t),\varphi_2(t),…,\varphi_n(t))
\begin{pmatrix}
\displaystyle\sum_{k=1}^{n}a_{1k}\varphi_k\\
\displaystyle\sum_{k=1}^{n}a_{2k}\varphi_k\\
…\\
\displaystyle\sum_{k=1}^{n}a_{nk}\varphi_k\\
\end{pmatrix}\\
&=-[\displaystyle\sum_{k=1}^{n}a_{k1}\psi_k\varphi_1+\displaystyle\sum_{k=1}^{n}a_{k2}\psi_k\varphi_2+…+\displaystyle\sum_{k=1}^{n}a_{kn}\psi_k\varphi_n]+[\displaystyle\sum_{k=1}^{n}\psi_1 a_{1k}\varphi_k+\displaystyle\sum_{k=1}^{n}\psi_2 a_{2k}\varphi_k+…+\displaystyle\sum_{k=1}^{n}\psi_n a_{nk}\varphi_k]\\
&=-\displaystyle\sum_{k=1}^{n}\displaystyle\sum_{i=1}^{n}(a_{ki}\psi_k\varphi_i)+\displaystyle\sum_{k=1}^{n}\displaystyle\sum_{i=1}^{n}(\psi_i a_{ik}\varphi_k)\\
&=-\displaystyle\sum_{k=1}^{n}\displaystyle\sum_{i=1}^{n}(a_{ki}\psi_k\varphi_i)+\displaystyle\sum_{i=1}^{n}\displaystyle\sum_{k=1}^{n}(a_{ik}\psi_i\varphi_k)\\
&=0
\end{aligned}$

$所以\Psi^T(t)\varphi(t)=常数$

3.(2)

充分性：

存在非奇异矩阵C，使得$\Psi^T(t)\phi(t)=C,由于\phi(t)是方程x'=A(t)x的基解矩阵，故\phi^{-1}(t)存在$

所以$\quad \Psi^T(t)=C·\Phi^{-1}(t)且\det(\Psi^T(t))=\det(C·\Phi^{-1}(t))\neq 0$

所以$\Psi(t)是基解矩阵$

必要性：

$因为\Psi(t)是基解矩阵$

$所以\Psi(t)的每一个向量\psi_1(t),\psi_2(t),…\psi_n(t)与\Psi(t)的每一个列向量\varphi_1(t),\varphi_2(t),…\varphi_n(t)满足\varphi_i^T\varphi_j(t)=常数$

$其中\quad i=1,2,…n,j=1,2,…n$

所以$\Psi^T(t)\phi是常数矩阵，记为c$

又因为$\Psi(t)和\phi(t)是基解矩阵$

所以$\det(\Psi^T(t))\neq 0,\det(\phi(t))\neq 0$

$则\det C=\det(\phi(t))\det(\Psi^T(t))\neq 0,即矩阵C是非奇异的.$
