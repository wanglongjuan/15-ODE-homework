二、考虑方程组

$$x'=A(t)x……………………(5.15)$$

$其中A(t)是区间a\leq t\leq b$

是连续$nxn$矩阵,它的元为$a_ij(t)(i,j=1,2,3……n)$.

(1)如果$x_1(t）,x_2(t)…，x_n(t)$

是(5.15)的任意n个解，那么它们的朗斯基行列式

$[Wx_1(t),x_2(t),…,x_n(t)]=W(t)$，

满足下面的一阶线性微分方程

$$W'=[a_{11}(t)+a_{22}(t)+…+a_{nn}(t)]W;$$

(提示：利用行列式的微分公式，求出$W'$的表达式。）


(2)解上面的一阶线性微分方程，证明下面的公式：

$$
W(t)=W(t_0)e^{\int_t^{t_0}{[a_{11}(s)+a_{22}(s)+…+a_{nn}(s)]}{d}s},t_0,t\in[a,b]
$$

三、设A(t)为区间$a\leq t\leq b$上的连续$n*n$实矩阵，$\psi(t)$为方程x'=A(t)x的基解矩阵，而$x=\phi(t)$为其一解。试证：

（1)对于方程$y'=-A^T(t)y$的任一解$y=\psi(t)$必有$\psi^T(t)$$\phi(t)=常数$;

 (2)$\psi(t)$为方程$y'=-A^T(t)y$的基解矩阵的充要条件是存在非奇异的常数矩阵C，使
 
 $$\psi^T(t)\Phi(t)=C$$


解 由题可知

因$\phi(t)是x'=A(t)x$的解，则

$$\phi'(t)=
\left[
\begin{matrix}
\phi'_1(t)\\
\phi'_2(t)\\
.......\\
\phi'_n(t)
\end{matrix}
\right]
$$

$$=
\left[
\begin{matrix}
a_{11}&a_{12}&....&a_{1n}\\
a_{21}&a_{22}&....&a_{2n}\\
....&....&....&....&\\
a_{n1}&a_{n2}&....&a_{nn}\\
\end{matrix}
\right]\left[
\begin{matrix}
\phi_1(t)\\
\phi_2(t)\\
.......\\
\phi_n(t)
\end{matrix}
\right]
$$

$$
=\left[
\begin{matrix}
\displaystyle\sum_{k=1}^{n} a_{1k}\phi_k\\
\displaystyle\sum_{k=1}^{n} a_{2k}\phi_k\\
.......\\
\displaystyle\sum_{k=1}^{n} a_{nk}\phi_k
\end{matrix}
\right]
$$

同理

$y'=-A^T(t)y$

则

$$\psi'(t)=
\left[
\begin{matrix}
\psi'_1(t)\\
\psi'_2(t)\\
.......\\
\psi'_n(t)
\end{matrix}
\right]
$$

$$=-
\left[
\begin{matrix}
a_{11}&a_{12}&....&a_{1n}\\
a_{21}&a_{22}&....&a_{2n}\\
....&....&....&....&\\
a_{n1}&a_{n2}&....&a_{nn}\\
\end{matrix}
\right]\left[
\begin{matrix}
\psi_1(t)\\
\psi_2(t)\\
.......\\
\psi_n(t)
\end{matrix}
\right]
$$

$$
=-\left[
\begin{matrix}
\displaystyle\sum_{k=1}^{n} a_{1k}\psi_k\\
\displaystyle\sum_{k=1}^{n} a_{2k}\psi_k\\
.......\\
\displaystyle\sum_{k=1}^{n} a_{nk}\psi_k
\end{matrix}
\right]
$$

对$\psi^T(t)\phi(t)求导数得$:

$$
\frac{d(\psi^T(t)\phi(t))}{dt}=\frac{d\psi^T(t)}{dt}\phi(t)+\frac{d\phi(t)}{dt}\psi^T(t)
$$

$$
=-(\displaystyle\sum_{k=1}^{n} a_{1k}\psi_k,\displaystyle\sum_{k=1}^{n} a_{2k}\psi_k,......,\displaystyle\sum_{k=1}^{n} a_{nk}\psi_k)\left[
\begin{matrix}
\phi_1(t)\\
\phi_2(t)\\
.......\\
\phi_n(t)
\end{matrix}
\right]
+(\psi_1(t),\psi_2(t),....,\psi_n(t))\left[
\begin{matrix}
\displaystyle\sum_{k=1}^{n} a_{1k}\phi_k\\
\displaystyle\sum_{k=1}^{n} a_{2k}\phi_k\\
.......\\
\displaystyle\sum_{k=1}^{n} a_{nk}\phi_k
\end{matrix}
\right]
$$

$$
=-[\displaystyle\sum_{k=1}^{n} a_{1k}\psi_k\phi_1+\displaystyle\sum_{k=1}^{n} a_{2k}\psi_k\phi_2+......+\displaystyle\sum_{k=1}^{n} a_{nk}\psi_k\phi_n]+
$$

$$
\displaystyle\sum_{k=1}^{n} \psi_1a_{1k}\phi_k+
\displaystyle\sum_{k=1}^{n} \psi_2a_{2k}\phi_k+
.......+\displaystyle\sum_{k=1}^{n} \psi_na_{nk}\phi_k=0
$$

$\psi^T(t)$$\phi(t)=常数$

(2)充分性：
存在非奇异矩阵C，使$\psi^T(t)\phi(t)=C$,

由于$\phi(t)$是方程$$x'=A(t)x$$的基解矩阵，故$\phi^{-1}(t)$存在，


所以$$\psi^T(t)=C\phi^{-1}(t)且det(\psi^T(t))=det(C\phi^{-1}(t))\neq 0$$

因此，$\psi(t)$是基础解组

必要性：
$\psi(t)$是基础解组，则$\psi(t)$的每一组列向量

$$\psi_1(t),\psi_2(t),\psi_3(t).........\psi_n(t)$$与
$\phi(t)$的每一组列向量$$\phi_1(t),\phi_2(t),\phi_3(t).........\phi_n(t)$$满足$$\psi^T_i(t)\phi_j(t)=常数,i,j=1,2....n$$

则$\psi^T(t)\phi(t)$是常数矩阵，记为c.

$$\phi(t),\psi(t)是基解矩阵，从而det（\phi(t)\neq 0）,det(\psi^T（t）\neq 0)$$

则$detC\neq 0$

即矩阵是非奇异的
