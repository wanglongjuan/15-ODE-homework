2.考虑方程组$$x'=A(t)x,\quad(5.15)$$
其中$A(t)$是区间$a\leq t\leq b$上的连续$n\times n$矩阵,它的元为$a_{ij}(t)(i,j=1,2,…,n)$.

(1)如果$x_1(t）,x_2(t),…，x_n(t)$是(5.15)的任意$n$个解，那么它们的朗斯基行列式$W[x_1(t),x_2(t),…,x_n(t)]\equiv W(t)$满足下面的一阶线性微分方程
$$W'=[a_{11}(t)+a_{22}(t)+…+a_{nn}(t)]W;$$
(提示：利用行列式的微分公式，求出$W'$的表达式。）

(2)解上面的一阶线性微分方程，证明下面的公式：
$$W(t)=W(t_0)e^{\int_{t_0}^t{[a_{11}(s)+a_{22}(s)+…+a_{nn}(s)]}\,\mathrm {d}s},t_0,t\in[a,b]$$

解(1)将$x_1(t),x_2(t),…，x_n(t)$用向量表示，它们的朗斯基行列式为

\begin{vmatrix}

x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\
…&…&…&…\\
x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)\\

$$

x_{n1}'(t)&x_{n2}'(t)&…&x_{nn}'(t)\\

$$

x_{1n}'&=a_{11}x_{1n}+a_{12}x_{2n}+…+a_{1n}x_{nn}\\
\end{aligned}
$$
由行列式的性质有$$
\begin{vmatrix}
x_{11}'(t)&x_{12}'(t)&…&x_{1n}'(t)\\

同理可得，(a)式右端第2个行列式等于$a_{22}(t)W(t)$,第3个行列式等于$a_{33}(t)W(t)$,…,第$n$个行列式等于$a_{nn}(t)W(t)$

3.设$A(t)$为区间$a\leq t\leq b$上的连续$n\times n$实矩阵，$\Phi(t)$为方程$x'=A(t)x$的基解矩阵，而$x=\varphi(t)$为其一解。试证：

(1)对于方程$y'=-A^T(t)y$的任一解$y=\psi(t)$必有$\Psi^T(t)\varphi(t)=常数$;

(2)$\Psi(t)$为方程$y'=-A^T(t)y$的基解矩阵的充要条件是存在非奇异的常数矩阵C，使$\Psi^T(t)\Phi(t)=C$.

解(1)由$\varphi(t)$是$x'=A(t)x$的解得$$\varphi'(t)=
\begin{pmatrix}
\varphi'_1(t)\\
\varphi'_2(t)\\
\vdots\\
\varphi'_n(t)
\end{pmatrix}
=
\begin{pmatrix}
a_{11}(t)&a_{12}(t)&…&a_{1n}(t)\\
a_{21}(t)&a_{22}(t)&…&a_{2n}(t)\\
…&…&…&…\\
a_{n1}(t)&a_{n2}(t)&…&a_{nn}(t)
\end{pmatrix}
\begin{pmatrix}
\varphi_1(t)\\
\varphi_2(t)\\
\vdots\\
\varphi_n(t)
\end{pmatrix}
=
\begin{pmatrix}
\displaystyle\sum_{k=1}^{n}a_{1k}\varphi_k\\
\displaystyle\sum_{k=1}^{n}a_{2k}\varphi_k\\
\vdots\\
\displaystyle\sum_{k=1}^{n}a_{nk}\varphi_k
\end{pmatrix}
$$
同理，$\psi(t)$是方程$y'=-A^T(t)y$的解, 所以$$\psi'(t)=
\begin{pmatrix}
\psi'_1(t)\\
\psi'_2(t)\\
\vdots\\
\psi'_n(t)
\end{pmatrix}
=
\begin{pmatrix}
a_{11}(t)&a_{12}(t)&…&a_{1n}(t)\\
a_{21}(t)&a_{22}(t)&…&a_{2n}(t)\\
…&…&…&…\\
a_{n1}(t)&a_{n2}(t)&…&a_{nn}(t)
\end{pmatrix}
\begin{pmatrix}
\psi_1(t)\\
\psi_2(t)\\
\vdots\\
\psi_n(t)
\end{pmatrix}
=
\begin{pmatrix}
\displaystyle\sum_{k=1}^{n}a_{k1}\psi_k\\
\displaystyle\sum_{k=1}^{n}a_{k2}\psi_k\\
\vdots\\
\displaystyle\sum_{k=1}^{n}a_{kn}\psi_k
\end{pmatrix}
$$
对$\psi^T(t)\varphi(t)$求导得$$
\begin{aligned}
\frac{d(\Psi^T(t)\varphi(t))}{dt}&=\frac{d\Psi^T(t)}{dt}\varphi(t)+\Psi^T(t)\frac{d\varphi(t)}{dt}\\
&=-(\displaystyle\sum_{k=1}^{n}a_{k1}\psi_k,\displaystyle\sum_{k=1}^{n}a_{k2}\psi_k,…,\displaystyle\sum_{k=1}^{n}a_{kn}\psi_k)
\begin{pmatrix}
\varphi_1(t)\\
\varphi_2(t)\\
\vdots\\
\varphi_n(t)
\end{pmatrix}
+(\psi_1(t),\psi_2(t),…,\psi_n(t))
\begin{pmatrix}
\displaystyle\sum_{k=1}^{n}a_{1k}\varphi_k\\
\displaystyle\sum_{k=1}^{n}a_{2k}\varphi_k\\
\vdots\\
\displaystyle\sum_{k=1}^{n}a_{nk}\varphi_k
\end{pmatrix}\\
&=-[\displaystyle\sum_{k=1}^{n}a_{k1}\psi_k\varphi_1+\displaystyle\sum_{k=1}^{n}a_{k2}\psi_k\varphi_2+…+\displaystyle\sum_{k=1}^{n}a_{kn}\psi_k\varphi_n]+[\displaystyle\sum_{k=1}^{n}\phi_1a_{1k}\varphi_k+\displaystyle\sum_{k=1}^{n}\phi_2a_{2k}\varphi_k+…+\displaystyle\sum_{k=1}^{n}\phi_na_{nk}\varphi_k]\\
&=-\displaystyle\sum_{k=1}^{n}\displaystyle\sum_{i=1}^{n}(a_{ki}\psi_k\varphi_i)+\displaystyle\sum_{k=1}^{n}\displaystyle\sum_{i=1}^{n}\psi_i(a_{ik}\varphi_k)\\
&=-\displaystyle\sum_{k=1}^{n}\displaystyle\sum_{i=1}^{n}(a_{ki}\psi_k\varphi_i)+\displaystyle\sum_{i=1}^{n}\displaystyle\sum_{k=1}^{n}(a_{ik}\psi_i\varphi_k)\\
&=0
\end{aligned}
$$
所以$\Psi^T(t)\varphi(t)=常数$

(2)充分性：存在非奇异的常数矩阵C，使$\Psi^T(t)\Phi(t)=C$

由于$\Phi(t)$为方程$x'=A(t)x$的基解矩阵

故$\Phi^{-1}(t)$存在

所以$\Psi^T(t)=C\bullet \Phi^{-1}(t)$且$det(\Psi^T(t))=det(C\bullet \Phi^{-1}(t))\neq0$

因此$\Psi(t)$为方程$y'=-A^T(t)y$的基解矩阵

必要性：$\Psi(t)$为方程$y'=-A^T(t)y$的基解矩阵

则$\Psi(t)$的每个列向量$\Psi_1(t),\Psi_2(t),…,\Psi_n(t)$和$\Phi(t)$的每个列向量$\Phi_1(t),\Phi_2(t),…,\Phi_n(t)$满足$\psi_i^T(t)\phi_j(t)=常数，i,j=1,2,…,n$

则$\Psi^T(t)\Phi(t)$是常数矩阵，记为C

又由于$\Psi(t)，\Phi(t)$分别是基解矩阵

从而$det(\Psi(t))\neq0,det(\Phi(t))\neq0$

则$detC=det(\Psi(t))\bullet det(\Phi(t))\neq0$

即矩阵C是非奇异的
