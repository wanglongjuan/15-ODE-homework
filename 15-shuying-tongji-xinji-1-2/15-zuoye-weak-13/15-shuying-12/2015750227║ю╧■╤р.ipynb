$$W'=[a_{11}(t)+a_{22}(t)+…+a_{nn}(t)] \equiv W;$$

(提示：利用行列式的微分公式，求出$W'$的表达式。）


(2)解上面的一阶线性微分方程，证明下面的公式：

三、设A(t)为区间$a\leq t\leq b$上的连续$n*n$实矩阵，$\psi(t)$为方程x'=A(t)x的基解矩阵，而$x=\phi(t)$为其一解。试证：

（1)对于方程$y'=-A^T(t)y$的任一解$y=\psi(t)$必有$\psi^T(t)$$\phi(t)=常数$;

 (2)$\psi(t)$为方程$y'=-A^T(t)y$的基解矩阵的充要条件是存在非奇异的常数矩阵C，使$\psi^T(t)\Phi(t)=C$


解：
（1）由于$x=\phi(t)$为x'=A(t)x的一个解
所以
$$
\phi'(t)=
\begin{pmatrix}
\phi'_1(t)\\
\phi'_2(t)\\
…      \\
\phi'_n(t)\\
\end{pmatrix}
=
A(t)x
=
\begin{pmatrix}
\quad{a_{11}} \quad{a_{12}} … \quad{a_{1n}}\\
\quad{a_{21}} \quad{a_{22}} … \quad{a_{2n}}\\
…                     …             …\\
\quad{a_{n1}} \quad{a_{n2}} … \quad{a_{nn}}\\
\end{pmatrix}
\begin{pmatrix}
\phi'_1(t)\\
\phi'_2(t)\\\
…      \\
\phi'_n(t)\\\
\end{pmatrix}
=
\begin{pmatrix}
\displaystyle\sum_{i=1}^n a_{1i} \phi_i\\
\displaystyle\sum_{i=1}^n a_{2i} \phi_i\\\
…      \\
\displaystyle\sum_{i=1}^n a_{ni} \phi_i\\\
\end{pmatrix}
$$

由于$y=\psi(t)$为方程$y'=-A^T(t)y$的一个解
易得
$$
\psi'(t)
=
\begin{pmatrix}
\phi'_1(t)\\
\phi'_2(t)\\
…      \\
\phi'_n(t)\\
\end{pmatrix}
=
-A^T(t) \psi(t)
=
-
\begin{pmatrix}
\quad{a_{11}} \quad{a_{21}} … \quad{a_{n1}} \\
\quad{a_{12}} \quad{a_{22}} … \quad{a_{n2}} \\
…                     …             …    \\
\quad{a_{1n}} \quad{a_{2n}} … \quad{a_{nn}}\\
\end{pmatrix}
\begin{pmatrix}
\psi'_1(t)\\
\psi'_2(t)\\
…      \\
\psi'_n(t)\\
\end{pmatrix}
=
-
\begin{pmatrix}
\displaystyle\sum_{i=1}^n a_{i1} \psi_i\\
\displaystyle\sum_{i=1}^n a_{i2} \psi_i\\\
…      \\
\displaystyle\sum_{i=1}^n a_{in} \psi_i\\\
\end{pmatrix}
$$

因为要求证$\psi^T(t)\phi(t)=常数$
故证明$\psi^T(t)\phi(t)$的导数为0
$$\frac{d\psi^T(t)\phi(t)}{dt}=
\frac{d\psi^T(t)}{dt}\phi(t)+\frac{d\phi(t)}{dt}\psi^T(t)
$$
综上可知

$$
上式=
-
\begin{pmatrix}
\displaystyle\sum_{i=1}^n a_{1i} \phi_i,\displaystyle\sum_{i=1}^n a_{2i} \phi_i,………
\end{pmatrix}
\begin{pmatrix}
\phi_1(t)\\
\phi_2(t)\\\
…      \\
\phi_n(t)\\\
\end{pmatrix}
+
\begin{pmatrix}
\psi_1(t),\psi_2(t),………，\psi_n(t)\\
\end{pmatrix}
\begin{pmatrix}
\displaystyle\sum_{i=1}^n a_{i1} \psi_i\\
\displaystyle\sum_{i=1}^n a_{i2} \psi_i\\\
…      \\
\displaystyle\sum_{i=1}^n a_{in} \psi_i\\\
\end{pmatrix}
$$
$$
=
-\displaystyle\sum_{i=1}^n \displaystyle\sum_{t=1}^n {\psi_i a_{it} \phi_t}+\displaystyle\sum_{i=1}^n \displaystyle\sum_{t=1}^n {\psi_i a_{ti} \phi_t}
=0
$$
$$故\psi^T(t)\phi(t)=常数$$

(2)
必要性：
由（1）可知

$$\psi^T(t)\Phi(t)=C,C为常数$$

$$由题意易知，\psi(t)，\phi(t)均为基解矩阵$$

$$由定理2可知，det\psi(t)\neq 0,即矩\psi(t)的行列式不为0$$

$$det\phi(t)\neq 0,即矩\phi(t)的行列式不为0$$

$$detC=det(\phi(t))det(\psi(t)) \neq 0$$

$$故C为非奇异矩阵$$

充分性：

$$由于存在非奇异矩阵C，\psi^T(t)\Phi(t)=C,C为常数$$

$$\psi^T(t)=C \Phi^T(t)$$

$$det(\psi^T(t))=det(C) det(\Phi^T(t)) \neq 0$$

$$由定理2 ，所以\psi(t)为基解矩阵$$
