一、考虑方程组

$$x'=A(t)x$$

其中A(t)是区间$a\leq t\leq b$是连续$nxn$矩阵,它的元为$a_ij(t)(i,j=1,2,3……n)$.

(1)如果$x_1(t）,x_2(t)…，x_n(t)$是(5.15)的任意n个解，那么它们的朗斯基行列式$[Wx_1(t),x_2(t),…,x_n(t)]=W(t)$，满足下面的一阶线性微分方程

$$W'=[a_{11}(t)+a_{22}(t)+…+a_{nn}(t)]W;$$

(提示：利用行列式的微分公式，求出$W'$的表达式。）


(2)解上面的一阶线性微分方程，证明下面的公式：

$$
W(t)=W(t_0)e^{\int_t^{t_0}{[a_{11}(s)+a_{22}(s)+…+a_{nn}(s)]}{d}s},t_0,t\in[a,b]
$$

解：$x_1(t）,x_2(t)…，x_n(t)$的朗斯基行列式为
$$
W(t)=\begin{vmatrix}
x_{11}(t)&x_{12}(t)&…&x_{1n}(t)\\
x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\
…&…&…&\\
x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)\\
\end{vmatrix}
$$
对$W(t)$求导可得到:
$$
W'(t)=\frac{d}{dt}\begin{vmatrix}
x_{11}(t)&x_{12}(t)&…&x_{1n}(t)\\
x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\
…&…&…&\\
x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)\\
\end{vmatrix}
=\begin{vmatrix}
x_{11}'(t)&x_{12}'(t)&…&x_{1n}'(t)\\
x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\
…&…&…&\\
x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)\\
\end{vmatrix}
+\begin{vmatrix}
x_{11}(t)&x_{12}(t)&…&x_{1n}(t)\\
x_{21}'(t)&x_{22}'(t)&…&x_{2n}'(t)\\
…&…&…&\\
x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)\\
\end{vmatrix}
+…+
\begin{vmatrix}
x_{11}(t)&x_{12}(t)&…&x_{1n}(t)\\
x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\
…&…&…&\\
x_{n1}'(t)&x_{n2}'(t)&…&x_{nn}'(t)\\
\end{vmatrix}
\quad (1)
$$

又因为
$$
\begin{aligned}
x_{11}'&=a_{11}x_{11}+a_{12}x_{21}+…+a_{1n}x_{n1}\\
x_{12}'&=a_{11}x_{12}+a_{12}x_{22}+…+a_{1n}x_{n2}\\
x_{13}'&=a_{11}x_{13}+a_{12}x_{23}+…+a_{1n}x_{n3}\\
                         ………\\
x_{1n}'&=a_{11}x_{1n}+a_{12}x_{2n}+…+a_{1n}x_{nn}\\
\end{aligned}
$$

由行列式的性质可得：

$$
\begin{vmatrix}
x_{11}'(t)&x_{12}'(t)&…&x_{1n}'(t)\\
x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\
…&…&…&\\
x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)\\
\end{vmatrix}
=
\begin{vmatrix}
a_{11}x_{11}(t)&a_{11}x_{12}(t)&…&a_{11}x_{1n}(t)\\
x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\
…&…&…&\\
x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)\\
\end{vmatrix}
+\begin{vmatrix}
a_{12}x_{21}(t)&a_{12}x_{22}(t)&…&a_{12}x_{2n}(t)\\
x_{21}'(t)&x_{22}'(t)&…&x_{2n}'(t)\\
…&…&…&\\
x_{n1}(t)&x_{n2}(t)&…&x_{nn}(t)\\
\end{vmatrix}
+…+
\begin{vmatrix}
a_{1n}x_{n1}(t)&a_{1n}x_{n2}(t)&…&a_{1n}x_{nn}(t)\\
x_{21}(t)&x_{22}(t)&…&x_{2n}(t)\\
…&…&…&\\
x_{n1}'(t)&x_{n2}'(t)&…&x_{nn}'(t)\\
\end{vmatrix}
$$

同理可得，(1）式右端第2,3，……n个行列式分别等于$a_{22}(t)W(t)$,$a_{33}(t)W(t)$,……$a_{nn}(t)W(t)$,因而可得到：
$$
\frac{dW(t)}{dt}W'(t)=[a_{22}(t)+a_{22}(t)+…+a_{nn}(t)]W(t)
$$

(2)可将第一问中的一节线性微分方程化为：
$$
\frac{dW}{w}=[a_{22}(t)+a_{22}(t)+…+a_{nn}(t)]dt
$$

从$t_0到t$积分可得：
$$
ln|W(t)|-ln|W(t_0|=\int_{t_0}^t{[a_{11}(s)+a_{22}(t)+…+a_{nn}(s)]}{d}s
$$
对两边用指数函数可得到：
$$
W(t)=W(t_0)e^{\int_{t_0}^t{[a_{11}(s)+a_{22}(t)+…+a_{nn}(s)]}{d}s}
$$





二、设A(t)为区间$a\leq t\leq b$上的连续$nxn$实矩阵，$\psi(t)$为方程x'=A(t)x的基解矩阵，而$x=\phi(t)$为其一解。试证：

（1)对于方程$y'=-A^T(t)y$的任一解$y=\psi(t)$必有$\psi^T(t)\phi(t)=常数$;

 (2)$\psi(t)$为方程$y'=-A^T(t)y$的基解矩阵的充要条件是存在非奇异的常数矩阵C，使$\psi^T(t)\Phi(t)=C$
 
 解：由题意可知$\phi(t)$是$x'=A(t)x$的解，所以有：

$$
 \phi'(t)=
\left[
\begin{matrix}
\phi_1'(t)\\
\phi_2'(t)\\
\vdots\\
\phi_n'(t)\\
\end{matrix}
\right]
= 
\left[
\begin{matrix}
a_{11}\quad a_{12}\quad a_{13}…a_{1n}\\
a_{21}\quad a_{22}\quad a_{23}…a_{2n}\\
…\quad  …\quad     …  \quad…\\
a_{n1}\quad a_{n2}\quad a_{n3}…a_{n4}\\
\end{matrix}
\right]
\left[
\begin{matrix}
\phi_1(t)\\
\phi_2(t)\\
\vdots\\
\phi_n(t)\\
\end{matrix}
\right]
=\left[
\begin{matrix}
\displaystyle\sum_{k=1}^n a_{1k}\phi_k\\
\displaystyle\sum_{k=1}^n a_{2k}\phi_k\\
\vdots\\
\displaystyle\sum_{k=1}^n a_{nk}\phi_k\\
\end{matrix}
\right]
\quad (1)
$$
同理，因为$\psi(t)$是方程$y'=-A^Ty$的解,所以有
$$
 \psi'(t)=
\left[
\begin{matrix}
\psi_1'(t)\\
\psi_2'(t)\\
\vdots\\
\psi_n'(t)\\
\end{matrix}
\right]
=
\left[
\begin{matrix}
a_{11}\quad a_{12}\quad a_{13}…a_{1n}\\
a_{21}\quad a_{22}\quad a_{23}…a_{2n}\\
…\quad  …\quad     …  \quad…\\
a_{n1}\quad a_{n2}\quad a_{n3}…a_{n4}\\
\end{matrix}
\right]
\left[
\begin{matrix}
\psi_1(t)\\
\psi_2(t)\\
\vdots\\
\psi_n(t)\\
\end{matrix}
\right]
=\left[
\begin{matrix}
\displaystyle\sum_{k=1}^n a_{1k}\psi_k\\
\displaystyle\sum_{k=1}^n a_{2k}\psi_k\\
\vdots\\
\displaystyle\sum_{k=1}^n a_{nk}\psi_k\\
\end{matrix}
\right]
\quad (2)
$$

求导$\psi^T(t)\phi(t)$得到：

$$
\frac{d(\psi^T(t)\phi(t)}{dt}=\frac{d\psi^T(t)}{dt}\phi(t)+\psi^T(t)\frac{d\phi(t)}{dt}
$$
将（1）（2）代入可得：
$$
\frac{d(\Psi^T(t)\Phi(t)}{dt}=\frac{d\Psi^T(t)}{dt}\Phi(t)+\Psi^T(t)\frac{d\Phi(t)}{dt}=0
$$
即$\Psi^T(t)\phi(t)=常数$

（2）
证明：

必要性
因为$\Psi(t)$是基解矩阵，则$\Psi(t)$的每个列向量$\psi_1(t),\psi_2(t),……\psi_n(t)$与$\Phi(t)$的每一个列向量$\phi_1(t),\phi_2(t),……\phi_n(t)$
满足$\phi_j(t)\psi^T(t)=常数i,j=1,2,3……n$,则$\Psi^T(t)\Phi(t)$是常数矩阵，记为A，另外因为$\Phi(t)$,$\Psi(t)$分别为基解矩阵
，所以$det(\Phi(t))\neq0$,$det(\Psi^T(t)\neq0$所以$detA=det(\Phi(t))det(\Psi^T(t)\neq0$,即可证明矩阵C是非奇异的。

充分性
若存在非奇异的矩阵C,使得$\Psi^T(t)\Phi(t)=C$,因为$\Phi(t)$是$x'=A(t)x$的基解矩阵，
故$\Phi^{-1}(t)$存在，所以$\Psi^T(t)=C\Phi^{-1}(t)$且$det(\Psi^T(t)=det(C\Phi^{-1}(t))\neq0$,故$\Psi(t)$是基解矩阵。
