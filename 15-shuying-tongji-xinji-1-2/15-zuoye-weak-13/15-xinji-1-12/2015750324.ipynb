2.考虑方程组$$x'=A(t)x$$其中$A(t)$是区间$a \leq t \leq b$上的连续$n \times n$矩阵，他的元为$a_{ij}(t)(i,j=1,2,\cdots，n)$

（1）如果$x_1(t),x_2(t),\cdots，x_n(t)$是（5.15）的任意n个解，那么他们的朗斯基行列式$W[x_1(t),x_2(t),\cdots，x_n(t)] \equiv W(t)$满足下面的一阶线性微分方程$$W'=[a_{11}(t)+a_{22}(t)+\cdots+a_{nn}(t)]W$$

(提示：利用行列式的微分公式，求出$W'$的表达式.)

（2）解上面的一阶线性微分方程，证明下面的公式：
$$W(t)=W(t_0)e^{\int_{t_0}^t[a_{11}(s)+a_{22}(s)+\cdots+a_{nn}(s)]ds},t_0,t \in [a,b]$$


解：

（1）用向量的形式表示$x_1(t),x_2(t),…，x_n(t)$，他们的朗斯基行列式
$$
W(t)=\left[
\begin{matrix}
x_{11}(t) & x_{12}(t) & \cdots & x_{1n}(t) \\
x_{21}(t) & x_{22}(t) & \cdots & x_{2n}(t) \\
\cdots & \cdots & \cdots & \cdots \\
x_{n1}(t) & x_{n2}(t) & \cdots & x_{nn}(t)
\end{matrix}
\right]
$$

对$W(t)$求导，得到
$$
\begin{align}
\frac{dW(t)}{dt}
& =\frac{d}{dt}
\left[
\begin{matrix}
x_{11}(t) & x_{12}(t) & \cdots & x_{1n}(t) \\
x_{21}(t) & x_{22}(t) & \cdots & x_{2n}(t) \\
\cdots & \cdots & \cdots & \cdots \\
x_{n1}(t) & x_{n2}(t) & \cdots & x_{nn}(t)
\end{matrix}
\right]\\\
& =
\left[
\begin{matrix}
x'_{11}(t) & x'_{12}(t) & \cdots & x'_{1n}(t) \\
x_{21}(t) & x_{22}(t) & \cdots & x_{2n}(t) \\
\cdots & \cdots & \cdots & \cdots \\
x_{n1}(t) & x_{n2}(t) & \cdots & x_{nn}(t)
\end{matrix}
\right]\\\
& +
\left[
\begin{matrix}
x_{11}(t) & x_{12}(t) & \cdots & x_{1n}(t) \\
x'_{21}(t) & x'_{22}(t) & \cdots & x'_{2n}(t) \\
\cdots & \cdots & \cdots & \cdots \\
x_{n1}(t) & x_{n2}(t) & \cdots & x_{nn}(t)
\end{matrix}
\right]\\\
& +\cdots+
\left[
\begin{matrix}
x_{11}(t) & x_{12}(t) & \cdots & x_{1n}(t) \\
x_{21}(t) & x_{22}(t) & \cdots & x_{2n}(t) \\
\cdots & \cdots & \cdots & \cdots \\
x'_{n1}(t) & x'_{n2}(t) & \cdots & x'_{nn}(t)
\end{matrix}
\right]
\end{align}
$$
由
$$
\begin{align}
& x'_{11}=a_{11}x_{11}+a_{12}x_{21}+\cdots+a_{1n}x_{n1},\\\
& x'_{12}=a_{11}x_{12}+a_{12}x_{22}+\cdots+a_{1n}x_{n2},\\\
& \cdots \cdots \cdots \cdots \\\
& x'_{1n}=a_{11}x_{1n}+a_{12}x_{2n}+\cdots+a_{1n}x_{nn},
\end{align}
$$

得到
$$
\begin{align}
\left[
\begin{matrix}
x'_{11}(t) & x'_{12}(t) & \cdots & x'_{1n}(t) \\
x_{21}(t) & x_{22}(t) & \cdots & x_{2n}(t) \\
\cdots & \cdots & \cdots & \cdots \\
x_{n1}(t) & x_{n2}(t) & \cdots & x_{nn}(t)
\end{matrix}
\right]
& =
\left[
\begin{matrix}
a_{11}x_{11}(t) & a_{11}x_{12}(t) & \cdots & a_{11}x_{1n}(t) \\
x_{21}(t) & x_{22}(t) & \cdots & x_{2n}(t) \\
\cdots & \cdots & \cdots & \cdots \\
x_{n1}(t) & x_{n2}(t) & \cdots & x_{nn}(t)
\end{matrix}
\right]\\\
& +
\left[
\begin{matrix}
a_{12}x_{11}(t) & a_{12}x_{12}(t) & \cdots & a_{12}x_{1n}(t) \\
x_{21}(t) & x_{22}(t) & \cdots & x_{2n}(t) \\
\cdots & \cdots & \cdots & \cdots \\
x_{n1}(t) & x_{n2}(t) & \cdots & x_{nn}(t)
\end{matrix}
\right]\\\
& +\cdots+
\left[
\begin{matrix}
a_{1n}x_{11}(t) & a_{1n}x_{12}(t) & \cdots & a_{1n}x_{1n}(t) \\
x_{21}(t) & x_{22}(t) & \cdots & x_{2n}(t) \\
\cdots & \cdots & \cdots & \cdots \\
x_{n1}(t) & x_{n2}(t) & \cdots & x_{nn}(t)
\end{matrix}
\right]\\\
& =a_{11}(t)W(t)
\end{align}
$$

同理得到
$$\frac{dW(t)}{dt}=W'(t)=[a_{11}(t)+a_{22}(t)+\cdots+a_{nn}(t)]W(t)$$
命题得证.

（2）

证明：

上式可变形为$$\frac{dW}{w}=[a_{11}(t)+a_{22}(t)+\cdots+a_{nn}(t)]dt$$

对上式从$t_0到t$求积分，得到
$$ln \mid W(t) \mid-ln \mid W(t_0) \mid=\int_{t_0}^t[a_{11}(s)+a_{22}(s)+\cdots+a_{nn}(s)]ds$$

即为$$W(t)=W(t_0)e^{\int_{t_0}^t[a_{11}(s)+a_{22}(s)+\cdots+a_{nn}(s)]ds}$$

证毕.

3.设$A(t)$为区间$a \leq t \leq b$上的连续$n \times n$实矩阵，$\Phi(t)$为方程$x'=A(t)x$的基解矩阵，而$x'=\phi(t)$为其一解，试证：

（1）对于方程$y'=-A^T(t)y$的任一解$y=\Psi(t)$必有$\Psi^T(t)\phi(t)=常数$

（2）$\Psi(t)$为方程$y'=-A^T(t)y$的基解矩阵的充要条件是存在非奇异的常数矩阵C，使$\Psi^T(t)\Phi(t)=C$

证明：

（1）由$\phi(t)$为方程$x'=A(t)x$的解，得到
$$
\phi'(t)=
\left[
\begin{matrix}
\phi'_1(t) \\
\phi'_2(t) \\
\vdots \\
\phi'_n(t)
\end{matrix}
\right]
=
\left[
\begin{matrix}
a_{11}(t) & a_{12}(t) & \cdots & a_{1n}(t) \\
a_{21}(t) & a_{22}(t) & \cdots & a_{2n}(t) \\
\cdots & \cdots & \cdots & \cdots \\
a_{n1}(t) & a_{n2}(t) & \cdots & a_{nn}(t)
\end{matrix}
\right]
\left[
\begin{matrix}
\phi_1(t) \\
\phi_2(t) \\
\vdots \\
\phi_n(t)
\end{matrix}
\right]
=
\left[
\begin{matrix}
\sum_{k=0}^n a_{1k}\phi_k \\
\sum_{k=0}^n a_{2k}\phi_k \\
\vdots \\
\sum_{k=0}^n a_{nk}\phi_k 
\end{matrix}
\right]
$$

同理，由$\Phi(t)$为方程$y'=-A^T(t)y$的解，得到
$$
\Psi'(t)=
\left[
\begin{matrix}
\psi'_1(t) \\
\psi'_2(t) \\
\vdots \\
\psi'_n(t)
\end{matrix}
\right]
=
-\left[
\begin{matrix}
a_{11}(t) & a_{12}(t) & \cdots & a_{1n}(t) \\
a_{21}(t) & a_{22}(t) & \cdots & a_{2n}(t) \\
\cdots & \cdots & \cdots & \cdots \\
a_{n1}(t) & a_{n2}(t) & \cdots & a_{nn}(t)
\end{matrix}
\right]
\left[
\begin{matrix}
\psi_1(t) \\
\psi_2(t) \\
\vdots \\
\psi_n(t)
\end{matrix}
\right]
=
\left[
\begin{matrix}
\sum_{k=0}^n a_{1k}\psi_k \\
\sum_{k=0}^n a_{2k}\psi_k \\
\vdots \\
\sum_{k=0}^n a_{nk}\psi_k 
\end{matrix}
\right]
$$

对$\Psi^T(t)\phi(t)$求导，得到
$$
\begin{align}
\frac{d(\Psi^T(t)\phi(t))}{dt} & =\frac{d\Psi^T(t)}{dt}\phi(t)+\Psi^T(t)\cdot \frac{d\phi(t)}{dt}\\\
& =-(\sum_{k=1}^n a_{k1}\psi_k,\sum_{k=1}^n a_{k2}\psi_k,\cdots,\sum_{k=1}^n a_{kn}\psi_k) \cdot 
\left[
\begin{matrix}
\phi_1(t) \\
\phi_2(t) \\
\vdots \\
\phi_n(t)
\end{matrix}
\right]
+(\psi_1(t),\psi_2(t),\cdots,\psi_n(t))\cdot
\left[
\begin{matrix}
\sum_{k=1}^n a_{k1}\psi_k \\
\sum_{k=1}^n a_{k2}\psi_k \\
\vdots \\
\sum_{k=1}^n a_{kn}\psi_k
\end{matrix}
\right]\\\
& =-[\sum_{k=1}^n a_{k1}\psi_k\phi_1,\sum_{k=1}^n a_{k2}\psi_k\phi_2,\cdots,\sum_{k=1}^n a_{kn}\psi_k\phi_n]
+[\sum_{k=1}^n \psi_1a_{1k}\phi_k,\sum_{k=1}^n \psi_2a_{2k}\phi_k,\cdots \sum_{k=1}^n \psi_na_{nk}\phi_k,]\\\
& =-\sum_{k=1}^n \sum_{i=1}^n (a_{ki}\psi_k\phi_i)+\sum_{k=1}^n \sum_{i=1}^n (\psi_ia_{ki}\phi_k)\\\
& =-\sum_{k=1}^n \sum_{i=1}^n (a_{ki}\psi_k\phi_i)+\sum_{i=1}^n \sum_{k=1}^n (a_{ki}\psi_i\phi_k)\\\
& =0
\end{align}
$$

所以有$\Psi^T(t)\phi(t)=常数$

（2）

充分性.

存在非奇异矩阵$C,使得\Psi^T (t)\Phi(t)=C$,由于$\Phi(t)$是方程$y'=-A^T (t)y$的基解矩阵，
所以$\Phi^{-1}(t)$存在，故$\Psi^T (t)=C\Phi^{-1}(t)$，并且有$det(\Psi^T (t))=det(C\Phi^{-1} (t)) \neq 0$.
因此，$\Phi(t)$是基解矩阵.

必要性.

$\Phi(t)$是基解矩阵，则$\Phi(t)$的每一个列向量$\phi_1(t),\phi_2(t),\cdots,\phi_n(t)$与$\Phi(t)$的每一个列向量$\phi_1(t),\phi_2(t),\cdots,\phi_n(t)$满足$\psi^T_i(t)\phi_j(t)=常数，i=1,2,\cdots,n,j=1,2,\cdots,n.$

即$\Psi^T(t)\Phi(t)是常数矩阵，记为c.$

又由于$\Phi(t),\Psi(t)$是基解矩阵,得到$det(\Phi(t)) \neq 0,det(\Psi(t)) \neq 0$,所以$detC=det(\Phi(t))\cdot det(\Psi(t))$

即$C$是非奇异的.

证毕.
