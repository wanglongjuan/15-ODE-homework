$P89页6、证明格朗沃尔(Gronwall)不等式：设K为非负常数，f(t)和g(t)为在区间α\leq t\leqβ上的连续非负函数，且满足不等式$
$f(t)\leq K+\int_{α}^{t} \mathrm f(s)g(s)\,\mathrm{d}s,α\leq t\leqβ,$
$则有f(t)\leq Kexp(\int_{α}^{t} \mathrm g(s)\,\mathrm{d}s),α\leq t\leqβ.并由此证明定理1的命题5。$

$证：(1)K>0时，令ω(t)=K+\int_{α}^{t} \mathrm f(s)g(s)\,\mathrm{d}s$

$则ω'(t)=f(t)g(t)\leq g(t)ω(t),由ω(t)>0,可得\frac{ω'(t)}{ω(t)}\leq g(t)$

$两边从α到t积分,得lnω(t)-lnω(α)\leq \int_{α}^{t} \mathrm g(s)\,\mathrm{d}s$

$即有\frac{ω(t)}{ω(α)}\leq exp(\int_{α}^{t} \mathrm g(s)\,\mathrm{d}s),ω(α)=K>0$

$所以ω(t)\leq Kexp(\int_{α}^{t} \mathrm g(s)\,\mathrm{d}s)$

$有f(t)\leqω(t)\leq Kexp(\int_{α}^{t} \mathrm g(s)\,\mathrm{d}s),α\leq t\leqβ$

$(2)K=0时,对任意ε>0,由于f(t)\leq\int_{α}^{t} \mathrm f(s)g(s)\,\mathrm{d}s$

$所以f(t)\leq ε+\int_{α}^{t} \mathrm f(s)g(s)\,\mathrm{d}s$

$由(1)，得f(t)\leqεexp(\int_{α}^{t} \mathrm g(s)\,\mathrm{d}s)$

$当ε趋于0+时，有f(t)\leq0.又因为f(t)\geq0,可得f(t)\equiv0,从而f(t)\leq Kexp(\int_{α}^{t} \mathrm g(s)\,\mathrm{d}s),α\leq t\leqβ$

$由(1)(2)知，格朗沃尔不等式成立。$

$用格朗沃尔不等式证明定理1的命题5$

$设ψ(x)是积分方程y=y_0+\int_{x_0}^{x} \mathrm f(x,y)\,\mathrm{d}x,x_0\leq x\leq x_0+h的定义于x_0\leq x\leq x_0+h的另一连续解,$
$则φ(x)=ψ(x)(x_0\leq x\leq x_0+h)$

$证：设φ(t),ψ(t)是初值问题x'=f(t,x),x(t_0)=x_0的两个解，则有φ(t)=x_0+\int_{t_0}^{t} \mathrm f(ξ,φ(ξ))\,\mathrm{d}ξ,$
$ψ(t)=x_0+\int_{t_0}^{t} \mathrm f(ξ,ψ(ξ))\,\mathrm{d}ξ$

$于是，|φ(t)-ψ(t)|\leq\int_{t_0}^{t} \mathrm |f(ξ,φ(ξ))-f(ξ,ψ(ξ))|\,\mathrm{d}ξ\leq L\int_{t_0}^{t} \mathrm |φ(ξ)-ψ(ξ)|\,\mathrm{d}ξ,其中L为利普希茨常数$

$由格朗沃尔不等式K=0且0\leq|φ(t)-ψ(t)|\leq Kexp(\int_{t_0}^{t} \mathrm L\,\mathrm{d}ξ)=0$

$于是φ(t)=ψ(t).即定理1的命题5成立。$

$8、如果函数f(x,y)于带域α\leq x\leqβ上连续且关于y满足利普希茨条件，则方程(3,1)满足条件y(x_0)=y_0的解于整个区间[α,β]上存在且唯一.试证明之。$
$(提示：用逐步逼近法，取(x,y)属于R,M=max|f(x,y_0)|)$

$证:设x∈[α,β]，M=max|f(x,y_0)|,h=β-α,b=Mh及有某常数L使得在域R={(x,y)|α\leq x\leqβ,|y-y_0|\leq b}中有$
$|f(x,y_1)-f(x,y_2)|\leq L|y_1-y_2|,(x,y_1),(x,y_2)∈R$

$构造逐步逼近函数序列 $$\begin{cases}φ_0(x)=y_0\\φ_n(x)=y_0+\int_{x_0}^{x} \mathrm f(ξ,φ_{n+1}(ξ))\,\mathrm{d}x,α\leq x\leqβ,n=1,2...\end{cases}$

$因为|φ_1(x)-y_0|=|\int_{x_0}^{x} \mathrm f(ξ,y_0)\,\mathrm{d}x|\leq M(x-x_0)\leq Mh\leq b$

$可证|φ_n(x)-φ_{n-1}(x)|\leq\frac{M}{n!}L^{n+1}(β-α)^n$

$因为有|φ_{n+1}(x)-φ_n(x)|\leq|\int_{x_0}^{x} \mathrm |f(ξ,φ_n(ξ))-f(ξ,φ_{n-1}(ξ))|\,\mathrm{d}ξ|$

$\leq L|\int_{x_0}^{x} \mathrm |φ_n(ξ)-φ_{n-1}(ξ)|\,\mathrm{d}ξ|$

$\leq L|\int_{x_0}^{x} \mathrm {\frac{M}{n!}L^{n-1}|ξ-x_0|^n}\,\mathrm{d}ξ|$

$\leq\frac{M}{(n+1)!}L^n|x-x_0|^{n+1}$

$\leq\frac{M}{(n+1)!}L^n(β-α)^{n+1}$

$由于级数\sum(k从1到正无穷)\frac{M}{k!}L^{k-1}(β-α)^k收敛，知级数φ_n(x)=φ_0(x)+\sum(k从1到n)[φ_k(x)-φ_{k-1}(x)]$

$在区间[α,β]一致收敛，于是存在极限lim(n趋于正无穷时)φ_n(x)=φ(x),它在区间[α,β]连续,且φ(x)=lim(n趋于正无穷时)φ_n(x)$
$=y_0+lim(n趋于正无穷时)\int_{x_0}^{x} \mathrm f(ξ,φ_{n-1}(ξ))\,\mathrm{d}x=y_0+\int_{x_0}^{x} \mathrm f(x,lim(n趋于正无穷时)φ_{n-1}(ξ))\,\mathrm{d}ξ=y_0+\int_{x_0}^{x} \mathrm f(ξ,φ(ξ))\,\mathrm{d}ξ$

$故φ(x)是积分方程也就是微分方程的解。$

$设φ(x),ψ(x)均满足积分方程y=y_0+\int_{x_0}^{x} \mathrm f(ξ,y(ξ))\,\mathrm{d}ξ,α\leq x_0\leqβ,α\leq x\leqβ$

$当x\geq x_0时，利用格朗沃尔不等式，由|φ(x)-ψ(x)|\leq\int_{x_0}^{x} \mathrm |f(s,φ(s))-f(s,ψ(s))|\,\mathrm{d}s\leq\int_{x_0}^{x} \mathrm |φ(s)-ψ(s)|\,\mathrm{d}s$

$有|φ(x)-ψ(x)|\leq0,即φ(x)\equivψ(x)(α\leq x_0\leq x\leqβ)$

$同理可证，当α\leq x\leq x_\leqβ时，有φ(x)\equivψ(x)$

$即积分方程也就是微分方程的解是唯一的。$

$10、给定积分方程φ(x)=f(x)+λ\int_a^b \mathrm K(x,ξ)φ(ξ)\,\mathrm{d}ξ,其中f(x)是[a,b]上的已知连续函数，K(x,ξ)是a\leq x\leq b,$
$a\leq ξ\leq b上的已知连续函数.证明当|λ|足够小时(λ是常数)，方程在[a,b]上存在唯一的连续解。(提示：作逐步逼近函数序列φ_0(x)=f(x),$
$φ_{n+1}(x)=f(x)+λ\int_a^b \mathrm K(x,ξ)φ_n(ξ)\,\mathrm{d}ξ(n=0,1,2...))$

$证：作逐步逼近函数序列，取φ_0(x)=f(x)$

$令φ_1(x)=f(x)+λ\int_a^b \mathrm K(x,ξ)φ_0(ξ)\,\mathrm{d}ξ,...φ_{n+1}(x)=f(x)+λ\int_a^b \mathrm K(x,ξ)φ_n(ξ)\,\mathrm{d}ξ,...(1)$

$令x∈[a,b],M=max|f(x)|,x∈[a,b],ξ∈[a,b],L=max|K(x,ξ)|>0$

$则|φ_1(x)-φ_0(x)|=|λ\int_a^b \mathrm K(x,ξ)φ_0(ξ)\,\mathrm{d}ξ|\leq|λ|\int_a^b \mathrm |K(x,ξ)||f(ξ)|\,\mathrm{d}ξ$
$\leq|λ|ML(b-a)$

$假设|φ_k(x)-φ_{k-1}(x)|\leq|λ|M^kL^k(b-a)^k成立，则有|φ_{k+1}(x)-φ_k(x)|=|λ\int_a^b \mathrm K(x,ξ)(φ_k(x)-φ_{k-1}(x))\,\mathrm{d}ξ|$

$\leq|λ|\int_a^b \mathrm |K(x,ξ)||φ_k(x)-φ_{k-1}(x)|\,\mathrm{d}ξ$

$\leq|λ|\int_a^b \mathrm L^{k+1}|λ|^kM(b-a)^k\,\mathrm{d}ξ=|λ|^{k+1}L^{k+1}M(b-a)^{k+1}$

$由数学归纳法可知，对任意n∈N,有|φ_n(x)-φ_{n-1}(x)|\leq|λ|^nL^nM(b-a)^{k+1}$

$下证{φ_n(x)}在[a,b]上一致收敛.考虑级数φ_0(x)+\sum(k从1到正无穷)(φ_k(x)-φ_{k-1}(x))$

$只需证该级数在[a.b]上一致收敛即可.$

$由于|φ_k(x)-φ_{k-1}(x)|\leq|λ|^kL^kM(b-a)^k$

$当|λ|<\frac{1}{(b-a)L}时，级数\sum(k从1到正无穷)|λ|^kL^kM(b-a)^k收敛，由此可知级数$
$φ_0(x)+\sum(k从1到正无穷)(φ_k(x)-φ_{k-1}(x))在[a,b]上一致收敛，即{φ_n(x)}在[a,b]上一致收敛.$

$不妨设其极限函数为φ*(x)，且由φ_n(x)的连续性知φ*(x)在[a,b]上连续，对迭代列(1)，当n趋于正无穷时两边取极限，$
$得φ*(x)=f(x)+λ\int_a^b \mathrm K(x,ξ)φ*(x)\,\mathrm{d}ξ(|λ|<\frac{1}{(b-a)L}),即当|λ|<\frac{1}{(b-a)L}时，积分方程必存在连续解.$

$下证当|λ|<\frac{1}{(b-a)L}时，(*)方程的解的唯一性.用反证法.$

$设另有解φ`(x)且φ`(x)≠φ*(x),即φ`(x)=f(x)+λ\int_a^b \mathrm K(x,ξ)φ`(x)\,\mathrm{d}ξ$

$记x∈[a,b],M`=max|φ*(x)-φ`(x)|>0,|φ*(x)-φ`(x)|=|λ\int_a^b \mathrm K(x,ξ)(φ*(ξ)-φ`(ξ))\,\mathrm{d}ξ|$

$\leq|λ|\int_a^b \mathrm L|φ*(ξ)-φ`(ξ)|\,\mathrm{d}ξ=|λ|LM`(b-a)$

$即有M`\leq|λ|LM`(b-a),由于M`>0,所以，|λ|\geq\frac{1}{(b-a)L}与|λ|<\frac{1}{(b-a)L}矛盾.唯一性得证。$
