## 一.求初值问题 \begin{cases}\frac{dy}{dx}=x^2-y^2,R:|x+1|≤1,|y|≤1,\\y(-1)=0\end{cases}的解的存在区间，并求第二次近似解，给出在解的存在区间的误差估计。

$(x,y)∈R,可知M=max|f(x,y)|=4,h=min(1,\frac{1}{4})=\frac{1}{4}$

$f(x,y)的利普希茨常数为L=2$

$因为\frac{\partial f}{\partial y}=|-2y|\leq2$

$所以φ_0(x)=0,φ_1(x)=\int_{-1}^{x} \mathrm (t^2-φ^2_0(x))=\frac{x^3}{3}+\frac{1}{3}$

$φ_2(x)=\int_{-1}^{x} \mathrm (t^2-φ^2_1(x))\,\mathrm{d}t=\frac{x^3}{3}-\frac{x}{9}-\frac{x^4}{18}-\frac{x^7}{63}+\frac{11}{42}$

$由误差估计公式，得|φ_2(x)-φ(x)|\leq\frac{4·2^2}{3!}(\frac{1}{4})^3=\frac{1}{24}$

$所以，初值问题的解的存在区间为-\frac{5}{4}\leq x\leq-\frac{3}{4}$

$第二次近似解为y_2=\frac{x^3}{3}-\frac{x}{9}-\frac{x^4}{18}-\frac{x^7}{63}+\frac{11}{42}$

$在解的存在区间的误差为|y-y_2|\leq\frac{1}{24}$


## 二.证明格朗沃尔(Gronwall)不等式：设K为非负常数，f(t)和g(t)为在区间α\leq t\leqβ上的连续非负函数，且满足不等式
$当K=0时,对任意ε>0,由于f(t)\leq\int_{α}^{t} \mathrm f(s)g(s)\,\mathrm{d}s$

$所以f(t)\leq ε+\int_{α}^{t} \mathrm f(s)g(s)\,\mathrm{d}s$

$可知f(t)\leqεexp(\int_{α}^{t} \mathrm g(s)\,\mathrm{d}s)$

$当ε趋于0+时，有f(t)\leq0.又因为f(t)\geq0,可得f(t)\equiv0,从而f(t)\leq Kexp(\int_{α}^{t} \mathrm g(s)\,\mathrm{d}s),α\leq t\leqβ$
$f(t)\leq K+\int_{α}^{t} \mathrm f(s)g(s)\,\mathrm{d}s,α\leq t\leqβ,$
$则有f(t)\leq Kexp(\int_{α}^{t} \mathrm g(s)\,\mathrm{d}s),α\leq t\leqβ.

$当K>0时，令ω(t)=K+\int_{α}^{t} \mathrm f(s)g(s)\,\mathrm{d}s$

$则ω'(t)=f(t)g(t)\leq g(t)ω(t),由ω(t)>0,可得\frac{ω'(t)}{ω(t)}\leq g(t)$

$两边从α到t积分,得lnω(t)-lnω(α)\leq \int_{α}^{t} \mathrm g(s)\,\mathrm{d}s$

$即有\frac{ω(t)}{ω(α)}\leq exp(\int_{α}^{t} \mathrm g(s)\,\mathrm{d}s),ω(α)=K>0$

$所以ω(t)\leq Kexp(\int_{α}^{t} \mathrm g(s)\,\mathrm{d}s)$

$有即f(t)\leqω(t)\leq Kexp(\int_{α}^{t} \mathrm g(s)\,\mathrm{d}s),α\leq t\leqβ$

$综上格朗沃尔不等式成立。$

$用格朗沃尔不等式证明定理1的命题5$

$设ψ(x)是积分方程y=y_0+\int_{x_0}^{x} \mathrm f(x,y)\,\mathrm{d}x,x_0\leq x\leq x_0+h的定义于x_0\leq x\leq x_0+h的另一连续解,$
$则φ(x)=ψ(x)(x_0\leq x\leq x_0+h)$

$证：设φ(t),ψ(t)是初值问题x'=f(t,x),x(t_0)=x_0的两个解，则有φ(t)=x_0+\int_{t_0}^{t} \mathrm f(ξ,φ(ξ))\,\mathrm{d}ξ,$
$ψ(t)=x_0+\int_{t_0}^{t} \mathrm f(ξ,ψ(ξ))\,\mathrm{d}ξ$

$于是，|φ(t)-ψ(t)|\leq\int_{t_0}^{t} \mathrm |f(ξ,φ(ξ))-f(ξ,ψ(ξ))|\,\mathrm{d}ξ\leq L\int_{t_0}^{t} \mathrm |φ(ξ)-ψ(ξ)|\,\mathrm{d}ξ,其中L为利普希茨常数$

$由格朗沃尔不等式K=0且0\leq|φ(t)-ψ(t)|\leq Kexp(\int_{t_0}^{t} \mathrm L\,\mathrm{d}ξ)=0$

$于是φ(t)=ψ(t).即定理1的命题5成立。$

## 三.如果函数f(x,y)于带域α\leq x\leqβ上连续且关于y满足利普希茨条件，则方程(3,1)满足条件y(x_0)=y_0的解于整个区间[α,β]上存在且唯一.试证明。

$设x∈[α,β]，M=max|f(x,y_0)|,h=β-α,b=Mh及有某常数L使得在域R={(x,y)|α\leq x\leqβ,|y-y_0|\leq b}中有$
$|f(x,y_1)-f(x,y_2)|\leq L|y_1-y_2|,(x,y_1),(x,y_2)∈R$

$构造逐步逼近函数序列 $
$\begin{cases}φ_0(x)=y_0\\φ_n(x)=y_0+\int_{x_0}^{x} \mathrm f(ξ,φ_{n+1}(ξ))\,\mathrm{d}x,α\leq x\leqβ,n=1,2...\end{cases}$

$因为|φ_1(x)-y_0|=|\int_{x_0}^{x} \mathrm f(ξ,y_0)\,\mathrm{d}x|\leq M(x-x_0)\leq Mh\leq b$

$则|φ_n(x)-φ_{n-1}(x)|\leq\frac{M}{n!}L^{n+1}(β-α)^n$

$而|φ_{n+1}(x)-φ_n(x)|\leq|\int_{x_0}^{x} \mathrm |f(ξ,φ_n(ξ))-f(ξ,φ_{n-1}(ξ))|\,\mathrm{d}ξ|$

$\leq L|\int_{x_0}^{x} \mathrm |φ_n(ξ)-φ_{n-1}(ξ)|\,\mathrm{d}ξ|$

$\leq L|\int_{x_0}^{x} \mathrm {\frac{M}{n!}L^{n-1}|ξ-x_0|^n}\,\mathrm{d}ξ|$

$\leq\frac{M}{(n+1)!}L^n|x-x_0|^{n+1}$

$\leq\frac{M}{(n+1)!}L^n(β-α)^{n+1}$

$由于级数\sum(k从1到正无穷)\frac{M}{k!}L^{k-1}(β-α)^k收敛，知级数φ_n(x)=φ_0(x)+\sum(k从1到n)[φ_k(x)-φ_{k-1}(x)]$

$在区间[α,β]一致收敛，于是存在极限lim(n趋于正无穷时)φ_n(x)=φ(x),它在区间[α,β]连续,且φ(x)=lim(n趋于正无穷时)φ_n(x)$
$=y_0+lim(n趋于正无穷时)\int_{x_0}^{x} \mathrm f(ξ,φ_{n-1}(ξ))\,\mathrm{d}x=y_0+\int_{x_0}^{x} \mathrm f(x,lim(n趋于正无穷时)φ_{n-1}(ξ))\,\mathrm{d}ξ=y_0+\int_{x_0}^{x} \mathrm f(ξ,φ(ξ))\,\mathrm{d}ξ$

$故φ(x)是微分方程的解。$

$设φ(x),ψ(x)均满足积分方程y=y_0+\int_{x_0}^{x} \mathrm f(ξ,y(ξ))\,\mathrm{d}ξ,α\leq x_0\leqβ,α\leq x\leqβ$

$当x\geq x_0时，利用格朗沃尔不等式，由|φ(x)-ψ(x)|\leq\int_{x_0}^{x} \mathrm |f(s,φ(s))-f(s,ψ(s))|\,\mathrm{d}s\leq\int_{x_0}^{x} \mathrm |φ(s)-ψ(s)|\,\mathrm{d}s$

$有|φ(x)-ψ(x)|\leq0,即φ(x)\equivψ(x)(α\leq x_0\leq x\leqβ)$

$同理可证，当α\leq x\leq x_\leqβ时，有φ(x)\equivψ(x)$

$即积分方程也就是微分方程的解是唯一的。$

## 四、给定积分方程
## $\varphi(x)=f(x)+\lambda\int_a^b K(x,\xi)\varphi(\xi)\mathrm d\xi \quad(*)$
## 其中f(x)是[a,b]上的已知连续函数,$K(x,\xi)是a\leq x\leq b,a\leq \xi\leq b$上的已知连续函数。证明当|$\lambda$|足够小时（$\lambda是常数$）,（ * ）在[a,b]上存在唯一的连续解。

$构造逼近函数序列，设\varphi_0(x)=f(x)$

$令\varphi_1(x)=f(x)+\lambda\int_a^bK(x,\xi)\varphi_0(ζ)dζ,......\varphi_{n+1}=f(x)+\lambda\int_a^bK(x,\xi)\varphi_n(ζ)dζ...$

$令M=\max_{\substack{
x\in[a,b] \\
}}
\mid f(x) \mid \quad L=\max_{\substack{
x\in [a,b] \\
\xi\in[a,b]
}}
\mid K(x,\xi)\mid $

$则有|\varphi_1(x)-\varphi_0(x)|=|\lambda \int_a^b K(x,\xi)\varphi_0(\xi)d\xi|\leq|\lambda||\int_a^bK(x,\xi)||f(\xi)|d\xi$

$设|\varphi_k(x)-\varphi_{k-1}(x)|\leq|\lambda|^kL^kM(b-a)^{k}成立$

$有|\varphi_{k+1}(x)-\varphi_k(x)|=|\lambda\int_a^bK(x,\xi)(\varphi_k(\xi)-\varphi_{k-1}(\xi))d\xi|\leq |\lambda|\int_a^b|K(x,\xi)||\varphi_k(\xi)-\varphi_{k-1}(\xi))|d\xi\leq|\lambda|\int_a^bL^{k+1}|\lambda|^kM(b-a)^kd\xi=|\lambda|^{k+1}L^{k+1}M(b-a)^{k+1}$

$由数学归纳法知，对任意的n\in N$

$有|\varphi_n(x)-\varphi_{n-1}(x)|\leq ML^n|\lambda|^n(b-a)^n$

$证｛\varphi_n(x)｝在[a,b]上一致收敛，考虑级数\varphi_0(x)+\displaystyle\sum_{k=1}^{\infty}(\varphi_k(x)-\varphi_{k-1}(x))$

$即需证明该级数在[a,b]上致收敛$

$因为|\varphi_k(x)-\varphi_{k-1}(x)|\leq ML^k|\lambda|^k(b-a)^k$

$当|\lambda|<\frac{1}{(b-a)L}时$

$级数\displaystyle\sum_{k=1}^{\infty} ML^k|\lambda|^k(b-a)^k收敛$

$可知当|\lambda|<\frac{1}{(b-a)L}时$

$级数\varphi_0(x)+\displaystyle\sum_{k=1}^{\infty}(\varphi_k(x)-\varphi_{k-1}(x))在[a,b]上一致收敛$

$即｛\varphi_n(x)｝在[a,b]上一致收敛$

$设其极限函数为\varphi^*(x)$

$即当x\in [a,b]时，\varphi_n(x)当n \to\infty时一致收敛于\varphi^*(x)$

$且由\varphi_n(x)的连续性可知\varphi^*(x)在[a,b]上连续$

$对迭代列①，当n \to\infty时两边取极限$

$得\varphi^*(x)=f(x)+\lambda \int_a^b K(x,\xi)\varphi^*(x)d\xi(|\lambda|<\frac{1}{(b-a)L})$

$即\lambda|<\frac{1}{(b-a)L}当时，积分方程必存在连续解$

$证当\lambda|<\frac{1}{(b-a)L}时（*）方程的解的唯一性$

$设另有解\varphi'(x)且，即\varphi'(x)=f(x)+\lambda \int_a^b K(x,\xi)\varphi'(x)d\xi$

$记M'=\max_{\substack{
x\in [a,b] \\
}}
\mid \varphi^*(x)-\varphi'(x)\mid>0$

$\mid \varphi^*(x)-\varphi'(x)\mid=\mid\lambda \int_a^b K(x,\xi)(\varphi^*(\xi)-\varphi'(\xi))\mid\leq|\lambda|\int_a^bL\mid\varphi^*(\xi)-\varphi'(\xi)\mid d\xi=\mid\lambda\mid M'L(b-a)$

$即有M'\leq\mid\lambda\mid M'L(b-a)$

$由于M'>0,所以\mid\lambda\mid\geq\frac{1}{(b-a)L}与|\lambda|<\frac{1}{(b-a)L}矛盾$

$所以解唯一$
