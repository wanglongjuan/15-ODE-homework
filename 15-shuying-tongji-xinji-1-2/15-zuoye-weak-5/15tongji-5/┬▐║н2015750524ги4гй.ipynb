**$3 求初值问题$**
\begin{cases}
\frac{\mathrm {d} y}{\mathrm {d} x}=x^2-y^2 & \quad \text{R:$|x+1|\leq1$,$|y|\leq1$, } \\
y(-1)=0\\
\end{cases}
$的解的存在区间，并求第二次近似解，给出在解的存在区间的误差估计.$ 


 **$ 解:$**
$$ M=\max \limits_{(x,y) \in R}|f(x,y)|=4 \\
   h=min\{a,\frac{b}{M}\}即h=min\{1,\frac{1}{4}\}=\frac{1}{4} \\ 
   在R上函数f(x,y)=x^2-y^2的利普希茨常数可取L=2，\frac{\partial f}{\partial y}=|-2y|\leq 2=L $$
   $$ \varphi_0(x)=0, $$
   $$ \varphi_1(x)=\int_{-1}^x \mathrm{\Big( t^2-\varphi_0^2(t)\Big)}\mathrm{d}t=\frac{x^3}{3}+\frac{1}{3}, $$
   $$ \varphi_2(x)=\int_{-1}^x \mathrm{\Big( t^2-\varphi_1^2(t)\Big)}\mathrm{d}t\\
   =\frac{x^3}{3}-\frac{x}{9}-\frac{x^4}{8}-\frac{x^7}{63}+\frac{11}{42} , $$
   $$ \varphi_2(x)-\varphi(x)\leq \frac{4*2^2}{3!}*(\frac{1}{4})^3=\frac{1}{24} $$
   $$此初值问题的解的存在区间为{-\frac{5}{4}}\leq {x}\leq{-\frac{3}{4}},\\
   第二次近似解为y_2=\frac{x^3}{3}-\frac{x}{9}-\frac{x^4}{8}-\frac{x^7}{63}+\frac{11}{42} ,\\
   在解的存在区间的误差估计为|y-y_2|\leq \frac {1}{24}$$
   
 

**$8.如果函数f(x,y)于带域{\alpha}\leq{x}\leq{\beta}上连续且关于y满足利普希茨条件，则方程\frac{\mathrm {d} y}{\mathrm {d} x}=f(x,y)满足条件y(x_0)=y_0的解于整个区间[\alpha,\beta]上存在且唯一，试证明之. $**
**$提示：用逐步逼近法，取M=\max \limits_{x \in [\alpha,\beta]}|f(x,y_0)|   $**

**$ 解:$**
$$
\begin{aligned}
y_0(x)&=y_0,\\
y_1(x)&=y_0+\int_{x_0}^x \mathrm{ f(\xi,y_0(\xi))}\,\mathrm{d}\xi,\\
\cdots\\
y_n(x)&=y_0+\int_{x_0}^x \mathrm{ f(\xi,y_{n-1}(\xi))}\,\mathrm{d}\xi,\quad ①\\
\cdots
\end{aligned}
$$ 
$y_n(x)在区间{\alpha}\leq{t}\leq{\beta}上存在且连续$
$$y_0(x)+\displaystyle\sum_{k=1}^{\infty}[y_k(x)-y_{k-1}(x)],x \in[\alpha,\beta]$$
$取M=\max \limits_{x \in [\alpha,\beta]}|f(x,y_0)|$
$$|y_1(x)-y_0(x)|\leq\int_{x_0}^x \mathrm{| f(\xi,y_0(\xi))|}\,\mathrm{d}\xi \leq M|x-x_0|,$$
$$|y_2(x)-y_1(x)|\leq\int_{x_0}^x \mathrm{| f(\xi,y_1(\xi))-f(\xi,y_0(\xi))|}\,\mathrm{d}\xi $$
$利用利普希茨条件得：$
 $$
\begin{aligned}
|y_2(x)-y_1(x)|&\leq L\int_{x_0}^x \mathrm{| y_1(\xi)-y_0(\xi)|}\,\mathrm{d}\xi \\
&\leq L\int_{x_0}^x \mathrm{M|\xi-x_0|}\,\mathrm{d}\xi=\frac{ML}{2!}(x-x_0)^2
\end{aligned}
$$
$由数学归纳法得：$
$$|y_k(x)-y_{k-1}(x)|\leq \frac{ML^{k-1}}{k!}|x-x_0|^k \leq \frac{ML^{k-1}}{k!}(\beta-\alpha
)^k$$
$其右端是正向收敛级数$
$$\displaystyle\sum_{k=1}^{\infty}[\leq \frac{ML^{k-1}}{k!}(\beta-\alpha
)^k]的一般项$$
$$据Weierstrass判别法知级数在x \in [\alpha,\beta]上一致收敛，所以函数列\{y_n(x)\}在[\alpha,\beta]一致收敛$$
$$设：\lim_{x \to \infty}y-n(x)=y(x)$$
$$易知y_n在[\alpha,\beta]连续，且\{f(x,y_n(x))\}在[\alpha,\beta]上一致收敛于f(x,y_n(x))$$
$对①式两边取极限：$
$$y(x)=y_0+\int_{x_0}^x \mathrm{ f(\xi,y_0(\xi))}\,\mathrm{d}\xi\\
解的存在性得证.
$$
$下证解的唯一性.$
$$
 我们证明 \psi(x) 也是序列 \{y(x)_n\}_{n=0}^{\infty} 的一致收敛极限函数. 从
$$
$$
\begin{aligned}
y_0(x)&=y_0,\\
y_n(x)&=y_0+\int_{x_0}^x \mathrm{ f(\xi,y_{n-1}(\xi))}\,\mathrm{d}\xi,\quad ①\\
\psi(x)& =y_0+\int_{x_0}^x f(\xi,\psi(\xi))\, \mathrm d\xi,
\end{aligned}
$$

可以进行如下估计

$$
\begin{aligned}
|y_0(x)-\psi(x)| & \leq \int_{x_0}^x|f(\xi,\psi_0(\xi))|\leq M(x-x_0)\\
|y_1(x)-\psi(x)| & \leq \int_{x_0}^x |f(\xi,y_0(x)(\xi))-f(\xi,\psi(\xi))|\,\mathrm d\xi\\
& \leq L\int_{x_0}^x |y_0(x)(\xi)-\psi(\xi)|\,\mathrm d\xi\\
& \leq ML\int_{x_0}^x(\xi-x_0)\,\mathrm d \xi\\
& =\frac{ML}{2!}(x-x_0)^2
\end{aligned}
$$

现设 
$$ 
|y_{n-1}(x)-\psi(x)|\leq \frac{ML^{n-1}}{n!}(x-x_0)^n,
$$
则有

$$
\begin{aligned}
|y_n(x)-\psi(x)| & \leq \int_{x_0}^x |f(\xi,y_{n-1}(\xi))-f(\xi,\psi(\xi))|\,\mathrm d\xi\\
& \leq L\int_{x_0}^x|y_{n-1}(\xi)-\psi(\xi)|\,\mathrm d\xi\\
& \leq\frac{ML^n}{n!}\int_{x_0}^x(\xi-x_0)^n\,\mathrm d\xi\\
& =\frac{ML^n}{(n+1)!}(x-x_0)^{n+1}
\end{aligned}
$$  

故由数学归纳法得知，对于所有的正整数 $n$, 有下面的估计式

$$
|y_n(x)-\psi(x)|\leq\frac{ML^n}{(n+1)!}(x-x_0)^{n+1}
$$


因此, 在 ${\alpha}\leq{x}\leq{\beta}$ 有

$$
|y_n(x)-\psi(x)|\leq \frac{ML^n}{(n+1)!}(x-x_0)^{n+1}
$$

$\frac{ML^n}{(n+1)!}(\beta-\alpha)^{n+1}$ 是收敛级数的公项，故 $n\to \infty$ 时，$\frac{ML^n}{(n+1)!}(\beta-\alpha)^{n+1}\to 0$, 因而 ${y_n(x)}$ 在 ${\alpha}\leq{x}\leq{\beta}$ 上一致收敛于 $\psi(x)$, 根据极限的唯一性，即得

$$
y(x)=\psi(x),{\alpha}\leq{x}\leq{\beta}.
$$

**$6.证明格朗沃尔（Gronwall）不等式:$**
**$设K为非负常数，f(t)和g(t)为在区间{\alpha}\leq{t}\leq{\beta}上的连续非负函数，且满足不等式$**
**$$ f(t)\leq {K+\int_a^t \mathrm{ f(s)g(s)}\,\mathrm{d}s},{\alpha}\leq{t}\leq{\beta}, $$**
**$则有$**
**$$ f(t)\leq {K exp \Big(\int_a^t \mathrm{ g(s)}\,\mathrm{d}s}\Big),{\alpha}\leq{t}\leq{\beta}, $$**
**$并由此证明定理1的命题5$**


**$ 证明:$**
$(1)K>0时，令$
$$w(t)\leq {K+\int_a^t \mathrm{ f(s)g(s)}\,\mathrm{d}s}$$
   $$则w'(t)=f(t)g(t)，由w(t)>0得\frac{w'(t)}{w(t)}\leq g(t)$$
   $对其两边从\alpha到t积分：$
   $$ln w(t)-ln w(\alpha)\leq \int_{\alpha}^t \mathrm{g(s)}\,\mathrm{d}s  $$
   $$\frac{w(t)}{w(\alpha)} \leq exp\Big(\int_{\alpha}^t \mathrm{g(s)}\,\mathrm{d}s \Big),w(\alpha)=K>0$$
   $$w(t) \leq K*exp\Big(\int_{\alpha}^t \mathrm{g(s)}\,\mathrm{d}s \Big)$$
   $$即：f(t) \leq w(t)\leq K*exp\Big(\int_{\alpha}^t \mathrm{g(s)}\,\mathrm{d}s \Big),{\alpha}\leq{t}\leq{\beta}$$
$(2)K=0时，对任意\varepsilon>0，$
$$由于f(t) \leq\int_a^t \mathrm{ f(s)g(s)}\,\mathrm{d}s,$$
$$所以f(t)\leq \varepsilon +\int_a^t \mathrm{ f(s)g(s)}\,\mathrm{d}s $$
$$由（1）有f(t) \leq \varepsilon exp\Big(\int_{\alpha}^t \mathrm{g(s)}\,\mathrm{d}s \Big)$$
$$当\varepsilon \rightarrow 0^{+}时，有f(t) \leq 0 $$
$$因f(t)\geq 0,有f(t)=0\\
f(t) \leq K exp\Big(\int_{\alpha}^t \mathrm{g(s)}\,\mathrm{d}s \Big),{\alpha}\leq{t}\leq{\beta}\\ 
综上，格朗沃尔不等式成立。
$$
$由此证明定理1的命题5:$
$$设\varphi（t）,\psi(t)是初值问题x'=f(t,x),x(t_0)=x_0的两个解\\
有\varphi（t）x_0+\int_{t_0}^t \mathrm{f(\xi,\varphi(\xi))}\,\mathrm{d}\xi \\
\psi(t)=x_0+\int_{t_0}^t \mathrm{f(\xi,\psi(\xi))}\,\mathrm{d}\xi $$
 $$
\begin{aligned}
|\varphi（t）-\psi(t)|&\leq \int_{t_0}^t \mathrm{|f(\xi,\varphi(\xi))-f(\xi,\psi(\xi))|}\,\mathrm{d}\xi\\
&\leq L\int_{t_0}^t \mathrm{|\varphi(\xi)-\psi(\xi)|}\,\mathrm{d}\xi
\end{aligned}
$$   
$$其中L为利普希茨条件，由格朗沃尔不等式知\\
0\leq |\varphi（t）-\psi(t)| \leq 0 \\
有\varphi（t）=\psi(t)\\
证毕$$ 
   
   
   
   

**$10.给定积分方程 $**
**$$ \varphi(x)=f(x)+\lambda \int_a^b \mathrm{ K(x,\xi)\varphi(\xi)}\,\mathrm{d}\xi  $$**
**$其中f(x)是[a,b]上的已知连续函数，K(x,\xi)是{a}\leq{x}\leq{b},{a}\leq{\xi}\leq{b}上的已知连续函数.证明当|\lambda| 足够小时(\lambda是常数)，积分方程在[a,b]上存在唯一的连续解.$**
**$提示：作逐步逼近函数序列$**
**$$ \varphi_0(x)=f(x), $$**
**$$ \varphi_{n+1}(x)=f(x)+\lambda \int_a^b \mathrm{ K(x,\xi)\varphi_n(\xi)}\,\mathrm{d}\xi (n=0,1,2\cdots)$$**

**$ 解:$**
$作逐步逼近序列$
$$
\begin{aligned}
\varphi_0(x)& =f(x),\\
\varphi_1(x)&=f(x)+f(x)+\lambda \int_a^b \mathrm{ K(x,\xi)\varphi_0(\xi)}\,\mathrm{d}\xi,\\
&\cdots\\
\varphi_{n+1}(x)&=f(x)+f(x)+\lambda \int_a^b \mathrm{ K(x,\xi)\varphi_n(\xi)}\,\mathrm{d}\xi,\\
&\cdots\\
\end{aligned}
$$
$令M=\max \limits_{x \in [a,b]}|f(x)| ,L=\max _{\substack{x\in [a,b]\\ \xi \in [a,b]\\}}\mid k(x,\xi)\mid>0 $
$$
\begin{aligned}
|\varphi_1(x)-\varphi_0(x)|&=|\lambda \int_a^b \mathrm{ K(x,\xi)\varphi_0(\xi)}\,\mathrm{d}\xi| \\
&\leq|\lambda |\int_a^b \mathrm{| K(x,\xi)||f(\xi)|}\,\mathrm{d}\xi\\
&\leq|\lambda |ML(b-a)
\end{aligned}
$$
$假设|\varphi_{k}(x)-\varphi_{k-1}(x)|\leq|\lambda |^k ML^k(b-a)^k成立，则有：$
$$
\begin{aligned}
|\varphi_{k+1}(x)-\varphi_{k}(x)|&=|\lambda \int_a^b \mathrm{ K(x,\xi)(\varphi_k(\xi)-\varphi_{k-1}(\xi))}\,\mathrm{d}\xi| \\
&\leq|\lambda |\int_a^b \mathrm{| K(x,\xi)||\varphi_k(\xi)-\varphi_{k-1}(\xi)|}\,\mathrm{d}\xi\\
&\leq|\lambda |\int_a^b \mathrm{L^{k+1}|\lambda|^kM(b-a)^k}\,\mathrm{d}\xi\\
&=|\lambda |^{k+1} ML^{k+1}(b-a)^{k+1}
\end{aligned}
$$
$由数学归纳法知，对任意n \in N有：$
$$|\varphi_{n}(x)-\varphi_{n-1}(x)|\leq|\lambda |^n ML^n(b-a)^n$$
$下证\{|\varphi_n(x)\}在[a,b]上一致收敛$
$$考虑级数\varphi_0(x)+\displaystyle\sum_{k=1}^{\infty} (\varphi_k(x)-\varphi_{k-1}(x))\\
其前n项和为：\varphi_0(x)+\displaystyle\sum_{k=1}^{n} (\varphi_k(x)-\varphi_{k-1}(x))=\varphi_n(x)\\
要证\{|\varphi_n(x)\}在[a,b]上一致收敛，只需证级数\varphi_0(x)+\displaystyle\sum_{k=1}^{\infty} (\varphi_k(x)-\varphi_{k-1}(x))在[a,b]上一致收敛\\
|\varphi_{k}(x)-\varphi_{k-1}(x)|\leq|\lambda |^k ML^k(b-a)^k\\
当|\lambda|<\frac{1}{(b-a)L}时，级数\displaystyle\sum_{k=1}^{\infty} |\lambda |^k ML^k(b-a)^k收敛\\
即此时，级数\varphi_0(x)+\displaystyle\sum_{k=1}^{\infty} (\varphi_k(x)-\varphi_{k-1}(x))在[a,b]上一致收敛\\
设极限函数\varphi^*(x),即x\in [a,b]时,当n\rightarrow\infty时\varphi_n(x)一致收敛于\varphi^*(x)\\
则知\varphi^*(x)在[a,b]上连续\\
对前逐步逼近序列两边取极限\\
\varphi^*(x)=f(x)+f(x)+\lambda \int_a^b \mathrm{ K(x,\xi)\varphi^*(\xi)}\,\mathrm{d}\xi\\
\Big(|\lambda|<\frac{1}{(b-a)L}\Big)\\
即|\lambda|<\frac{1}{(b-a)L}时，积分方程必存在连续解得证.
$$
$
下证|\lambda|<\frac{1}{(b-a)L}时，\varphi(x)=f(x)+\lambda \int_a^b \mathrm{ K(x,\xi)\varphi(\xi)}\,\mathrm{d}\xi 的解的唯一性.$
$$
反证法：设另有解\tilde{\varphi}(x)且\tilde{\varphi}(x)\neq \varphi^*(x)\\
\tilde{\varphi}(x)=f(x)+f(x)+\lambda \int_a^b \mathrm{ K(x,\xi)\tilde{\varphi}(\xi)}\,\mathrm{d}\xi\\
记Q=\max \limits_{x \in [a,b]}|\varphi^*(x)-\tilde{\varphi}(x)| >0\\
$$
$$
\begin{aligned}
|\varphi^*(x)-\tilde{\varphi}(x)|&=|\lambda \int_a^b \mathrm{ K(x,\xi)(\varphi^*(\xi)-\tilde{\varphi}(\xi))}\,\mathrm{d}\xi| \\
&\leq|\lambda |\int_a^b \mathrm{| K(x,\xi)||\varphi^*(\xi)-\tilde{\varphi}(\xi)|}\,\mathrm{d}\xi\\
&=|\lambda |QL(b-a)
\end{aligned}
$$
$$即有0\leq Q\leq|\lambda |QL(b-a),有|\lambda|\geq\frac{1}{(b-a)L}，与|\lambda|<\frac{1}{(b-a)L}矛盾，唯一性得证.$$
