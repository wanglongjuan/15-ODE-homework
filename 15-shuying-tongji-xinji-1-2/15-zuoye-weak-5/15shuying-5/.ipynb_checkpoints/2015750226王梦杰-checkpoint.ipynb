8 如果函数f(x,y)于带域$\alpha\leq x\leq \beta$上连续且关于y满足利普希茨条件，则方程（3.1）满足条件$y(x_0)=y_0$的解于整个区间[a,b]上存在且唯一，试证明之
提示：用逐步紧逼法，取$M=\max\limits_{x\in {[\alpha,\beta]}}\{f(x,y_0)\}$

证明：
由命题1知，方程（3.1）满足条件$y(x_0)=y_0$的解等价于求积分方程
$$
y=y_0+\int_{t_0}^t f(x,y)\,\mathrm dx\
$$
$$x\in {[\alpha,\beta]}$$

证明积分方程的唯一性即可
取
$$y_0(x)=y_0$$
$$
y_1(x)=y_0+\int_{t_0}^t f(\varepsilon,y_0(\varepsilon))\,\mathrm d\varepsilon\
$$
$$
y_2(x)=y_0+\int_{t_0}^t f(\varepsilon,y_1(\varepsilon))\,\mathrm d\varepsilon\
$$
$$....................$$
$$
y_n(x)=y_0+\int_{t_0}^t f(\varepsilon,y_{n-1}(\varepsilon))\,\mathrm d\varepsilon\
$$
$$....................$$【记为（1）式】
易见$y_n(x)$在$\alpha\leq x\leq \beta$上存在且连续，然后证明函数列${y_n(x)}$在$\alpha\leq x\leq \beta$上一致连续
考察级数
$$ 
y_0(x)+\displaystyle\sum_{k=1}^\infty[y_k(x)-y_{k-1}(x)],x\in {[\alpha,\beta]}

$$

$$
$$\leq M|x-x_0)|$$【记为（3）式】

|y_2(x)-y_1(x)|\leq \int_{x_0}^x |f(\varepsilon,y_1(\varepsilon))-f(\varepsilon,y_0(\varepsilon))|\,\mathrm d\varepsilon\

由（3）式以及利普希茨条件得
$$
|y_2(x)-y_1(x)|\leq L\int_{x_0}^x |y_1(\varepsilon)-y_0(\varepsilon)|\,\mathrm d\varepsilon\
$$
$$
\leq L\int_{x_0}^x M|\varepsilon-x_0|\,\mathrm d\varepsilon\

$$
=\frac{ML}{2!}(x-x_0)^2

唯一性得证

6 证明格朗沃尔不等式
设K为非负常数，f(t)和g(t)为在区间$\alpha\leq t\leq \beta$上的连续非负函数，且满足不等式
$$
f(t)\leq K+\int_\alpha^t f(s)g(s)\,\mathrm ds\
$$
$$\alpha\leq t\leq \beta$$
则有
$$
f(t)\leq Kexp(\int_\alpha^t g(s)\,\mathrm ds),\alpha\leq t\leq \beta
$$
并由此证明定理1的命题5

证明  设$$
\varphi(x)=K+\int_\alpha^t f(s)g(s)\,\mathrm ds\
$$
当K>0时
对$\varphi(x)$进行求导得
$$
\varphi'(t)=f(t)g(t)\leq g(t)\varphi(t)
$$
即为
$$
\frac{\varphi'(t)}{\varphi(t)}\leq g(t)
$$
其中$\frac{\varphi'(t)}{\varphi(t)}$是$ln\varphi（t）$的导数

因此对上式从$\alpha$到t积分得
$$
ln\varphi(t)-ln\varphi(\alpha)\leq \int_\alpha^t g(s)\,\mathrm ds\
$$
变形得
$$
\frac{\varphi(t)}{\varphi(\alpha)}\leq exp(\int_\alpha^t g(s)\,\mathrm ds)
$$
令$\varphi(\alpha)=K$
可得
$$
\varphi(t)\leq Kexp(\int_\alpha^t g(s)\,\mathrm ds)
$$
则
$$
f(t)\leq \varphi(t)\leq Kexp(\int_\alpha^t g(s)\,\mathrm ds)
$$
上式得证
当K=0时
$$
f(t)\leq \int_\alpha^t f(s)g(s)\,\mathrm ds\
$$
当任意的$m>0$时
$$
f(t)\leq \int_\alpha^t f(s)g(s)\,\mathrm ds\ \leq m+\int_\alpha^t f(s)g(s)\,\mathrm ds\
$$
由上式$K>0$得
$$
f(t)\leq mexp(\int_\alpha^t g(s)\,\mathrm ds)
$$
当m趋近于0时，得
$$
f(t)\leq 0
$$
又因为$0\leq f(t)$
因此$$f(t)=0$$
则
$$
f(t)\leq Kexp(\int_\alpha^t g(s)\,\mathrm ds)
$$
$$\alpha\leq t\leq \beta$$
综上，格朗沃尔不等式成立
由此证明定理1的命题5
证明：设$\varphi_0(t)$和$\varphi_1(t)$是初值问题$$x'=f(t,x),x(t_0)=x_0$$的两个解
则有
$$
\varphi_0(t)=x_0+\int_{t_0}^t f(\varepsilon,\varphi_0(\varepsilon))\,\mathrm d\varepsilon\
$$
$$
\varphi_1(t)=x_0+\int_{t_0}^t f(\varepsilon,\varphi_1(\varepsilon))\,\mathrm d\varepsilon\
$$
则
$$
|\varphi_0(t)-\varphi_1(t)|\leq \int_{t_0}^t {f(\varepsilon,\varphi_0(\varepsilon))-f(\varepsilon,\varphi_1(\varepsilon))}\,\mathrm d\varepsilon\
$$
$$\leq L\int_{t_0}^t |\varphi_0(\varepsilon)-\varphi_1(\varepsilon)|\,\mathrm d\varepsilon\
$$
其中L为利普希茨条件，由格朗沃尔不等式得
$$0\leq |\varphi_0(t)-\varphi_1(t)|\leq 0$$
因此
$$
\varphi_0(t)=\varphi_1(t)
$$
命题5得证

10 给定积分方程
$$
\varphi(x)=f(x)+\lambda\int_a^b K(x,\varepsilon)\varphi(\varepsilon)\,\mathrm d\varepsilon\
$$     $【*】$
其中f(x)是$[\alpha,\beta]$上的已知连续函数，$K(x,\varepsilon)$是$a\leq x\leq b$,$a\leq \varepsilon\leq b$上的已知连续函数。
证明当$|\lambda|$足够小时（$\lambda$是常数），$（*）$在$[\alpha,\beta]$上存在唯一的连续解
提示：作逐步逼近函数序列
$$
\varphi_0(x)=f(x)
$$
$$
\varphi_{n+1}(x)=f(x)+\lambda\int_a^b K(x,\varepsilon)\varphi_n(\varepsilon)\,\mathrm d\varepsilon\
$$  
$(n=0,1,2.....)$

解
用逐步逼近法，取
$$
\varphi_0(x)=f(x)
$$
$$
\varphi_1(x)=f(x)+\lambda\int_a^b k(x,\varepsilon)\varphi_0(\varepsilon)\,\mathrm d\varepsilon\
$$  
$$..............$$
$$
\varphi_{n+1}(x)=f(x)+\lambda\int_a^b k(x,\varepsilon)\varphi_n(\varepsilon)\,\mathrm d\varepsilon\
$$
$$..............     【记为（1）式】$$
令$M=\max\limits_{x\in {[a,b]}}\{|f(x)|\}$
$$
L=\max\limits_{{x\in {[a,b]}}{\varepsilon\in {[a,b]}}}\{|k(x,\varepsilon)|\}>0
$$
$$
|\varphi_1(x)-\varphi_0(x)|=|\lambda\int_a^b k(x,\varepsilon)\varphi_0(\varepsilon)\,\mathrm d\varepsilon|
$$
$$
\leq |\lambda|\int_a^b |k(x,\varepsilon)||f(\varepsilon)|\,\mathrm d\varepsilon
$$
$$
\leq |\lambda|ML(b-a)
$$
假设$$|\varphi_k(x)-\varphi_{k-1}(x)|\leq |\lambda|^kML^k(b-a)^k$$
成立，则有
$$
|\varphi_{k+1}(x)-\varphi_k(x)|=|\lambda\int_a^b k(x,\varepsilon)(\varphi_k(\varepsilon)-\varphi_{k-1}(\varepsilon))\,\mathrm d\varepsilon|
$$
$$
\leq |\lambda|\int_a^b |k(x,\varepsilon)||(\varphi_k(\varepsilon)-\varphi_{k-1}(\varepsilon))|\,\mathrm d\varepsilon
$$
$$
\leq |\lambda|\int_a^b L^{k+1}|\lambda|^kM(b-a)^{k+1}\,\mathrm d\varepsilon
$$
$$
= |\lambda|^{k+1}ML^{k+1}(b-a)^{k+1}
$$
由数学归纳法知，对任意的$n\in{N}$,有
$$
|\varphi_n(x)-\varphi_{n-1}(x)|\leq |\lambda|^nML^n(b-a)^n
$$
下面证明${\varphi_n(x)}$在$[a,b]$上一致收敛，考虑级数
$$ 
\varphi_0(x)+\displaystyle\sum_{k=1}^\infty[\varphi_k(x)-\varphi_{k-1}(x)],x\in {[\alpha,\beta]}    【记为（2）式】
$$ 
该级数的前n项和为$\varphi_n(x)$
所以要证明序列${\varphi_n(x)}$在$[a,b]$上一致收敛，只需证明级数（2）在$[a,b]$上一致收敛即可，由于
$$
|\varphi_k(x)-\varphi_{k-1}(x)|\leq |\lambda|^kML^k(b-a)^k
$$
当$|\lambda|<\frac{1}{(b-a)L}|$时，级数$\displaystyle\sum_{k=1}^\infty|\lambda|^kML^k(b-a)^k$收敛，

由此可知当$|\lambda|<\frac{1}{(b-a)L}|$时，级数（2）在$[a,b]$上一致收敛。设极限函数为$\varphi_*(x)$,即$x\in{[a,b]}$时，当n趋近于无穷大时，$\varphi_n(x)$一致收敛于$\varphi_*(x)$,

则由$\varphi_n(x)$的连续性知$varphi_*(x)$在$[a,b]$上连续
对迭代序列（1），当n趋近于无穷大时两边取极限，得
$$
\varphi_*(x)=f(x)+\lambda\int_a^b k(x,\varepsilon)\varphi_*(\varepsilon)\,\mathrm d\varepsilon\
$$
即证明当$|\lambda|<\frac{1}{(b-a)L}$时，积分方程必存在连续解。
利用反证法

设另有解$\bar{\varphi}(x)$且与$\varphi_*(x)$不相等，即
$$
\bar{\varphi}(x)=f(x)+\lambda\int_a^b k(x,\varepsilon)bar{\varphi}(\varepsilon)\,\mathrm d\varepsilon\
$$
记$Q=\max\limits_{x\in {[a,b]}}|\bar{\varphi}(x)-\varphi_*(x)|>0$
$$
|\bar{\varphi}(x)-\varphi_*(x)|=|\lambda\int_a^b k(x,\varepsilon)(\bar{\varphi}(\varepsilon)-\varphi_*(\varepsilon))\,\mathrm d\varepsilon|
$$
$$
\leq |\lambda|\int_a^b L|\bar{\varphi}(x)-\varphi_*(x)|\,\mathrm d\varepsilon
$$
$$
=|\lambda|QL(b-a)
$$
即有$Q\leq |\lambda|QL(b-a)$ 由于$Q>0$,所以$|\lambda|>=\frac{1}{(b-a)L}$

与$|\lambda|<\frac{1}{(b-a)L}$矛盾，唯一性得证
