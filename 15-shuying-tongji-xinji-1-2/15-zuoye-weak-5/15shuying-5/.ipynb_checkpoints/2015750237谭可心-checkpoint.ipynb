$3$、求初值问题

$$\begin{cases} \frac{\mathrm{dy}}{\mathrm{dx}}=x^2-y^2,R:\mid x+1\mid\le 1,\mid y\mid\le 1\\y(-1)=0\end{cases}$$

的解的存在区间，并求第二次近似解，给出在解的存在区间的误差估计。

解：$M=max{\mid x^2-y^2\mid}=4,h=min(a,\frac{b}{M})=min(1,\frac{1}{4})=\frac{1}{4}$

$\therefore \mid x-x_0\mid=\mid x-(-1)\mid=\mid x+1\mid\le \frac{1}{4}$

$\therefore$解的存在区间为$-\frac{5}{4}\le x\le -\frac{1}{4}$

令$\varphi_0(x)=0$

$\varphi_1(x)=\int^x_{-1}{(t^2-{\varphi_0}^2(t))}\mathrm{dt}=\frac{x^3}{3}+\frac{1}{3}$

$\varphi_2(x)=\int^x_{-1}{(t^2-{\varphi_1}^2(t))}\mathrm{dt}=\frac{x^3}{3}-\frac{x}{9}-\frac{x^4}{18}-\frac{x^7}{63}+\frac{11}{42}$

$\therefore$第二次近似解为$y_2=\frac{x^3}{3}-\frac{x}{9}-\frac{x^4}{18}-\frac{x^7}{63}+\frac{11}{42}$

又$\because \mid \frac{\partial f(x,y)}{\partial y}\mid=\mid -2y\mid \le 2=L$

$\therefore \mid \varphi_n(x)-\varphi(x)\mid \le \frac{ML^n}{(n+1)！}h^{n+1}=\frac{4*2^2}{3!}*(\frac{1}{4})^3=\frac{1}{24}$

$\therefore$解的存在区间的误差估计为$\mid y-y_2\mid \le \frac{1}{24}$

$8$、如果函数$f(x,y)$于带域$\alpha \le x \le \beta$上连续且关于$y$满足利普希茨条件，则方程$\frac{\mathrm{dy}}{\mathrm{dx}}=f(x,y)$满足条件$y(x_0)=y_0$的解于整个区间$[\alpha,\beta]$上存在且唯一，试证明之。（提示：用逐步逼近法，取$M=max_{\substack{x\in[\alpha,\beta]}}\mid f(x,y_0)\mid.$）

证明：由教材中命题1知方程$\frac{\mathrm{dy}}{\mathrm{dx}}=f(x,y)$满足条件$y(x_0)=y_0$的解等价于求积分方程$y=y_0+\int^x_{x_0}{f(x,y)\mathrm{dx},x\in [\alpha,\beta]}$的连续解，因此只要证明上述积分方程的解的存在唯一性即可。

先证存在性：

取$y_0(x)=y_0，y_1(x)=y_0+\int^x_{x_0}{f(\xi,y_0(\xi))}\mathrm{d\xi}，\cdots，y_n(x)=y_0+\int^x_{x_0}f(\xi,y_{n-1}(\xi))\mathrm{d\xi}，\cdots$①

易见$y_n(x)$在$\alpha\le x\le \beta$上存在且连续，然后证明函数列${y_n(x)}$在$\alpha\le x\le \beta$上一致收敛。

考察级数$y_0(x)+\sum^{\infty}_{k=1}[y_k(x)-y_{k-1}(x)],x\in [\alpha,\beta]$②

取$M=max_{x\in [\alpha,\beta]}\mid f(x,y_0)\mid$，由①式有$\mid{y_1(x)-y_0(x)}\mid \le \int^x_{x_0}{\mid{f(\xi,y_0(\xi))}\mid\mathrm{d\xi}}\le M\mid{x-x_0}\mid$③

及$\mid{y_2(x)-y_1(x)}\mid \le \int^x_{x_0}{\mid{f(\xi,y_1(\xi))-f(\xi,y_0(\xi))}\mid\mathrm{d\xi}}$

利用利普希茨条件及③式得

$\mid{y_2(x)-y_1(x)}\mid \le L\int^x_{x_0}{\mid{y_1(\xi)-y_0(\xi)}\mid}\mathrm{d\xi}\le L\int^x_{x_0}{M\mid{\xi-x_0}\mid}\mathrm{d\xi}=\frac{ML}{2!}(x-x_0)^2$

则由数学归纳法可知$\mid{y_k(x)-y_{k-1}(x)}\mid \le \frac{ML^{k-1}}{k!}\mid{x-x_0}\mid^k \le \frac{ML^{k-1}}{k!}(\beta-\alpha)^k$

上式右端是正项收敛级数$\sum^{\infty}_{k=1}\frac{ML^{k-1}}{k!}(\beta-\alpha)^k$的一般项。

根据Weierstrass判别法，知级数②在$\alpha\le x\le \beta$上一致收敛，因而函数列${y_n(x)}$在$[\alpha,\beta]$上一致收敛

设$\lim_{n\rightarrow \infty}{y_n(x)}=y(x)$，易知$y(x)$在$[\alpha,\beta]$上连续，且${f(x,y_n(x))}$在$[\alpha,\beta]$上一致收敛于$f(x,y_x))$

$\therefore$对①式两边取极限，即得$y(x)=y_0+\int^x_{x_0}{f(\xi,y(\xi))}\mathrm{d\xi}$，解的存在性得证。

再证唯一性：

设$\varphi(x)$是另一个连续解，则从$y_0(x)=y_0，y_n(x)=y_0+\int^x_{x_0}f(\xi,y_{n-1}(\xi))\mathrm{d\xi},\varphi(x)=y_0+\int^x_{x_0}f(\xi,\varphi(\xi))\mathrm{d\xi}$可以进行如下估计

$\mid{y_0(x)-\varphi(x)}\mid \le \int^x_{x_0}{\mid{f(\xi,\varphi(\xi))}\mid \mathrm{d\xi}}\le M\mid{x-x_0}\mid$

$\mid{y_1(x)-\varphi(x)}\mid \le \int^x_{x_0}{\mid{f(\xi,y_0(\xi))-f(\xi,\varphi(\xi))}\mid\mathrm{d\xi}}\le L\int^x_{x_0}{\mid{y_0(\xi)-\varphi(\xi)}\mid}\mathrm{d\xi}\le L\int^x_{x_0}{M\mid{\xi-x_0}\mid}\mathrm{d\xi}=\frac{ML}{2!}(x-x_0)^2$

现设$\mid{y_{n-1}(x)-\varphi(x)}\mid\le \frac{ML^{k-1}}{k!}(x-x_0)^k$，

则有$\mid{y_k(x)-\varphi(x)}\mid \le \int^x_{x_0}{\mid{f(\xi,y_{k-1}(\xi))-f(\xi,\varphi(\xi))}\mid\mathrm{d\xi}}\le L\int^x_{x_0}{\mid{y_{k-1}(\xi)-\varphi(\xi)}\mid}\mathrm{d\xi}\le \frac{ML^k}{k!}\int^x_{x_0}{(\xi-x_0)}\mathrm{d\xi}=\frac{ML^k}{(k+1)!}(x-x_0)^{k+1}$

故由数学归纳法得知，对于所有的正整数k，有下面的估计式

$\mid{y_k(x)-\varphi(x)}\mid \le \frac{ML^{k}}{(k+1)!}(x-x_0)^{k+1}$

$\therefore$在$[\alpha,\beta]$上有$\mid{y_n(x)-\varphi(x)}\mid \le \frac{ML^k}{(k+1)!}h^{n+1}$

$\frac{ML^k}{(k+1)!}h^{n+1}$是收敛级数的公项，故$n\rightarrow \infty$时，$\frac{ML^k}{(k+1)!}h^{n+1}\rightarrow 0$

$\therefore {y_n(x)}$在$\alpha \le x\le \beta$上一致收敛于$\varphi(x)$

根据极限的唯一性即得$y(x)=\varphi(x),x\in [\alpha,\beta]$

综上，结论即证。

$6$、证明格朗沃尔（Gronwall）不等式：

设K为非负常数，$f(t)$和$g(t)$为在区间$\mid \alpha \le t \le \beta$上的连续非负常数，且满足不等式

$$f(t)\le K+\int^t_\alpha{f(s)g(s)\mathrm{ds}},\alpha\le t\le \beta,$$

则有

$$f(t)\le Kexp(\int^t_\alpha{g(s)\mathrm{ds}}),\alpha\le t\le \beta.$$

并由此证明定理1的命题5：设$\psi(x)$是积分方程$y=y_0+\int^{x}_{x_0}{f(x,y)\mathrm{dx}},x_0\le x\le x_0+h$的定义于$x_0\le x\le x_0+h$上的另一个连续解，则$\varphi(x)=\psi(x)(x_0\le x\le x_0+h)$。

（1）证明：当$K>0$时，令$\omega(t)=K+\int^t_\alpha{f(s)g(s)}\mathrm{ds}$，则$\omega'(t)=f(t)g(t)\le g(t)\omega(t)$，

由$\omega(t)>0$可得$\frac{\omega'(t)}{\omega(t)}\le g(t)$，对其两边从$\alpha$到$t$积分得$\ln{\omega(t)}-\ln{\omega(\alpha)}\le \int^t_\alpha{g(s)}\mathrm{ds}$，

即$\frac{\omega(t)}{\omega(\alpha)}\le exp(\int^t_\alpha{g(s)}\mathrm{ds}),\omega(\alpha)=K>0$，

$\therefore \omega(t)\le K*exp(\int^t_\alpha{g(s)}\mathrm{ds})$，

即$f(t)\le \omega(t)\le Kexp(\int^t_\alpha{g(s)}\mathrm{ds}),\alpha\le t\le \beta$

当$K=0$时，对任意$\varepsilon>0$，由于$f(t)\le \int^t_\alpha{f(s)g(s)}\mathrm{ds}$，则$f(t)\le \varepsilon+\int^t_\alpha{f(s)g(s)}\mathrm{ds}$

由上面结论有$f(t)\le \varepsilon exp(\int^t_\alpha{g(s)}\mathrm{ds})$

当$\varepsilon \rightarrow0^+$时有$f(t)\le 0$

又因为$0\le f(t)$，即得$f(t)=0$，则$f(t)\le Kexp(\int^t_\alpha{g(s)}\mathrm{ds}),\alpha\le t\le \beta$.

综上所述，格朗沃尔不等式成立。

（2）证明：设$\varphi(t),\psi(t)$是初值问题$x'=f(t,x),x(t_0)=x_0$的两个解，

则有$\varphi(t)x_0+\int^t_{t_0}{f(\xi,\varphi(\xi))}\mathrm{d\xi}\\ \psi(t)=x_0+\int^t_{t_0}{f(\xi,\psi(\xi))}\mathrm{d\xi}$

那么$\mid \varphi(t)-\psi(t)\mid \le \int^t_{t_0}{\mid f(\xi,\varphi(\xi))-f(\xi,\psi(\xi))\mid}\mathrm{d\xi}\le L\int^t_{t_0}{\mid \varphi(\xi)-\psi(\xi)\mid}\mathrm{d\xi}$，

其中L为利普希茨条件，由格朗沃尔不等式知$0\le \mid \varphi(t)-\psi(t)\mid \le 0$，因而有$\varphi(t)=\psi(t)$，定理1的命题5得证。

$10$、给定积分方程

$$\varphi(x)=f(x)+\lambda\int^b_a{K(x,\xi)\varphi(\xi)\mathrm{d\xi}}，( * )$$

其中$f(x)$是$[a,b]$上的已知连续函数，$K(x,\xi)$是$a\le x\le b,a\le \xi\le b$上的已知连续函数。证明当$\mid \lambda\mid$足够小时（$\lambda$是常数），$(*)$在$[a,b]$上存在唯一的连续解。（提示：作逐步逼近函数序列$\varphi_0(x)=f(x),\varphi_{n+1}(x)=f(x)+\lambda\int^b_a{K(x,\xi)\varphi_n(\xi)}\mathrm{d\xi}(n=0,1,2\cdots)$.）

证明：作逐步逼近序列，取$\varphi_0(x)=f(x)$

令$\varphi_1(x)=f(x)+\lambda \int^b_a{k(x,\xi)\varphi_0(\xi)}\mathrm{d\xi}\\ \cdots \cdots\\ \varphi_{n+1}(x)=f(x)+\lambda \int^b_a{k(x,\xi)\varphi_n(\xi)}\mathrm{d\xi}\\ \cdots \cdots$①

令$M=max_{\substack{x\in[a,b]}}\mid{f(x)}\mid,L=max_{x\in[a,b],\xi \in[a,b]}\mid{k(x,\xi)}\mid >0$

$\mid{\varphi_1(x)-\varphi_0(x)}\mid=\mid{\lambda \int^b_a{k(x,\xi)\varphi_0(\xi)}\mathrm{d\xi}}\mid \\ \le \mid{\lambda}\mid\int^b_a{\mid{k(x,\xi)}\mid \mid{f(\xi)}\mid}\mathrm{d\xi}\\ \le \mid{\lambda}\mid ML(b-a)$

假设$\mid{\varphi_k(x)-\varphi_{k-1}(x)}\mid=\mid{\lambda \int^b_a{k(x,\xi)(\varphi_k(x)-\varphi_{k-1}(x))}\mathrm{d\xi}}\mid$

$\le \mid{\lambda}\mid\int^b_a{\mid{k(x,\xi)}\mid \mid{\varphi_k(x)-\varphi_{k-1}(x)}\mid}\mathrm{d\xi}$

$\le \mid{\lambda}\mid\int^b_a{L^{k+1}\mid{\lambda}\mid^kM(b-a)^k}\mathrm{d\xi}$

$=\mid{\lambda}\mid^{k+1}L^{k+1}M(b-a)^{k+1}$

由数学归纳法知，对任意的$n\in N$有$\mid{\varphi_n(x)-\varphi_{n-1}(x)}\mid \le ML^n\mid{\lambda}\mid^n(b-a)^n$

下面证明${\varphi_n(x)}$在$[a,b]$上一致收敛

考虑级数$\varphi_0(x)+\sum^{\infty}_{k=1}{(\varphi_k(x)-\varphi_{k-1}(x))}$②

该级数的前$n$项和为$\varphi_0(x)+\sum^n_{k=1}{(\varphi_k(x)-\varphi_{k-1}(x))}=\varphi_n(x)$

$\therefore$要证序列${\varphi_n(x)}$在$[a,b]$上一致收敛，只需证明级数②在$[a,b]$上一致收敛即可。

$\because \mid{\varphi_k(x)-\varphi_{k-1}(x)}\mid \le ML^k\mid{\lambda}\mid(b-a)^k$

当$\mid{\lambda}\mid<\frac{1}{(b-a)L}$时，级数$\sum^\infty_{k=1}{ML^k\mid{\lambda}\mid^k(b-a)^k}$收敛，

由此可知当$\mid{\lambda}\mid<\frac{1}{(b-a)L}$时，级数②在$[a,b]$上一致收敛

设极限函数为$\varphi^*(x)$，即$x\in[a,b]$时，当$n\rightarrow \infty$时，$\varphi_n(x)$一致收敛于$\varphi^*$

则由$\varphi_n(x)$的连续性知$\varphi^*(x)$在$[a,b]$上连续，对迭代序列①，当$n\rightarrow \infty$时两边取极限

得$\varphi^*(x)=f(x)+\lambda \int^b_a{k(x,\xi)\varphi^*(\xi)}\mathrm{d\xi}(\mid{\lambda}\mid<\frac{1}{(b-a)L})$

即证明了当$\mid{\lambda}\mid<\frac{1}{(b-a)L}$时，积分方程必存在连续解

下面证明当$\mid{\lambda}\mid<\frac{1}{(b-a)L}$时，$(*)$方程的解的唯一性，用反证法

设另有解$\psi(x)$且$\psi(x)\neq \varphi^*(x)$，即$\psi(x)=f(x)+\lambda \int^b_a{k(x,\xi)\psi(\xi)}\mathrm{d\xi}$

记$Q=max_{x\in[a,b]}{\mid{\varphi^*(x)-\psi(x)}\mid}>0$

$\mid{\varphi^*(x)-\psi(x)}\mid=\mid{\lambda \int^b_a{k(x,\xi)(\varphi^*(\xi)-\psi(\xi))}\mathrm{d\xi}}\mid$

$\le \mid{\lambda}\mid \int^b_a{L(\varphi^*(\xi)-\psi(\xi))}\mathrm{d\xi}$

$=\mid{\lambda}\mid QL(b-a)$

即有$Q\le \mid{\lambda}\mid QL(b-a)$

由于$Q>0$，所以$\mid{\lambda}\mid \ge \frac{1}{(b-a)L}$，与$\mid{\lambda}\mid <\frac{1}{(b-a)L}$矛盾

唯一性得证。
