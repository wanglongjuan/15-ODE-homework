$\phi_{n+1}(x)=f(x)+\lambda \int_a^b k(x,\varsigma) \phi_n(\varsigma)d\varsigma$

$……$

令$M=max_{x \in [a,b]} \mid f(x) \mid ,L=max \mid k(x,\varsigma) \mid >0$

所以


$
\begin{aligned}
\mid \phi_1(x)-\phi_0(x) \mid &=\mid \lambda \int_a^b k(x,\varsigma) \phi_0(\varsigma)d\varsigma \mid\\\
& \leq \mid \lambda \mid ML(b-a)
\end{aligned}
$

假设$\mid \phi_k(x)-\phi_{k-1}(x)\mid \leq {\mid \lambda \mid }^kL^kM(b-a)^k$成立，可得

$
\begin{aligned}
\mid \phi_{k+1}(x)-\phi_k(x)\mid &=\mid \lambda \int_a^b k(x,\varsigma) (\phi_k(\varsigma)-\phi_{k-1}(\varsigma))d\varsigma \mid\\\
&\leq \int_a^b L^{k+1}{\mid \lambda \mid }^kM(b-a)^k d\varsigma={\mid \lambda \mid }^{k+1}L^{k+1}M(b-a)^{k+1}
\end{aligned}
$

由归纳法即可得，对任意的$n \in N$有$\mid \phi_n(x)-\phi_{n-1}(x)\mid \leq {\mid \lambda \mid }^nL^nM(b-a)^n$

级数$\phi_0(x)+\sum_{k=1}^\infty(\phi_k(x)-\phi_{k-1}(x))$的前n项和为：

$\phi_0(x)+\sum_{k=1}^n(\phi_k(x)-\phi_{k-1}(x))=\phi_n(x)$
