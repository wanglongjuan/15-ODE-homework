3.求初值问题

$$
\begin{cases}
\frac{dy}{dx}=x^2-y^2\ ,R:\mid x+1\mid \leq1\ ,\mid y\mid \leq1,\\
y(-1)=0 
\end{cases}
$$

的解的存在区间，并求第二次近似解，给出在解的存在区间的误差估计.

解：

$M=max\mid f(x,y)\mid =4$,h为$a=1$及$\frac{b}{M}=\frac14$中的最小者，故，$h=\frac14$,

在R上，$f(x,y)=x^2-y^2$的利普希茨常数可取为$L=2$，因为$\frac{\partial f}{\partial y}=\mid -2y \mid \leq2=L$

$$
\begin{aligned}
\phi_0(x)&=0\\\
\phi_1(x)&=\int_{-1}^x(t^2-\phi_0^2(t))dt=\frac{x^3}{3}+\frac13\\\
\phi_2(x)&=\int_{-1}^x(t^2-\phi_1^2(t))dt\\\
&=\frac{x^3}{3}-\frac x9-\frac{x^4}{18}-\frac{x^7}{63}+\frac{11}{42}\\\
\end{aligned}
$$

由误差估计公式可得，$\mid \phi_2(x)-\phi(x)\mid \leq\frac{{4}\cdot{2^2}}{3!}\cdot(\frac14)^3=\frac{1}{24}$

所以，

已知解的存在区间为$-\frac54\leq x \leq -\frac34$,

第二次近似解为$y_2=\frac{x^3}{3}-\frac x9-\frac{x^4}{18}-\frac{x^7}{63}+\frac{11}{42}$,

解的存在区间的误差估计为$\mid y-y_2 \mid \leq \frac{1}{24}$.

6.证明格朗沃尔（Gromwall）不等式：

设K为非负常数，$f(t)$和$g(t)$为在区间$\alpha \leq t \leq \beta$上的连续非负函数，且满足不等式

$$f(t) \leq K+\int_\alpha^tf(s)g(s)ds,\alpha \leq t \leq \beta$$

则有$$f(t)\leq Kexp(\int_\alpha^tg(s)ds)\ ,\alpha \leq t \leq \beta$$

并由此证明教材中定理1的命题5.


证明：

（1）

$K>0$时，

令$\omega(t)=K+\int_\alpha^tf(s)g(s)ds,$

则，$\omega \prime(t)=f(t)g(t) \leq g(t)\omega(t)$,

由$\omega(t)>0$可得

$\frac{\omega\prime(t)}{\omega(t)}\leq g(t)$,

$ln\omega(t)-ln\omega(\alpha)\leq \int_\alpha^tg(s)ds$,

即，$\frac{\omega(t)}{\omega(\alpha)}\leq exp(\int_\alpha^tg(s)ds)，\omega(\alpha)=K>0$,

所以$\omega(t)\leq K \cdot exp(\int_\alpha^tg(s)ds)$,

即，$f(t)\leq \omega(t) \leq Kexp(\int_\alpha^tg(s)ds),\ \alpha \leq t \leq \beta$

（2）

$K=0$时，

对任意$\varepsilon >0$，

因为$f(t)\leq \int_\alpha^tf(s)g(s)ds$,

所以$f(t)\leq \varepsilon+\int_\alpha^tf(s)g(s)ds$.

由（1）可知，$f(t)\leq \varepsilon exp(\int_\alpha^tg(s)ds)$

当$\varepsilon→0^+$时，$f(t)\leq 0$

又因为$f(t)\geq 0$,

即得$f(t)=0$,

所以$f(t)\leq Kexp(\int_\alpha^tg(s)ds),\ \alpha \leq t \leq \beta$

综上，格朗沃尔不等式成立。

用格朗沃尔不等式证明教材中定理1的命题5.

证明：

设$\phi(t),\psi(t)$是初值问题$x\prime=f(t,x),\ x(t_0)=x_0$的两个解，则，

$$
\begin{aligned}
\phi(t)&=x_0+\int_{t_0}^tf(\varsigma,\phi(\varsigma))d\varsigma,\\
\psi(t)&=x_0+\int_{t_0}^tf(\varsigma,\psi(\varsigma))d\varsigma,
\end{aligned}
$$


$$
\begin{aligned}
\mid \phi(t)-\psi(t) \mid &\leq \int_{t_0}^t \mid f(\varsigma,\phi(\varsigma))-f(\varsigma,\psi(\varsigma)) \mid d\varsigma\\\
&\leq L\int_{t_0}^t \mid \phi(\varsigma)-\psi(\varsigma) \mid d\varsigma
\end{aligned}
$$

其中L为利普希茨条件，由格朗沃尔不等式知
$0 \leq \mid \phi(\varsigma)-\psi(\varsigma) \mid \leq 0$

所以$\phi(t)=\psi(t)$

定理1的命题5得证.

8.如果函数$f(x,y)$于带域$\alpha \leq x \leq \beta$上连续且关于y满足利普希茨条件，则教材中方程（3.1）满足条件$y(x_0)=y_0$的解于整个区间$[\alpha  ,\beta]$上存在且唯一.试证明之.

（提示：用逐步逼近法，取$M=max_{x\in [\alpha\beta]}\mid f(x,y_0) \mid$.）

证明：
由定理1可知，方程（3.1）满足条件$y(x_0)=y_0$的解等价于求积分方程$y=y_0+\int_{x_0}^xf(x,y)dx,\ x \in [\alpha,\beta]$的连续解。

取
$$
\begin{aligned}
y_0(x)&=y_0\\\
y_1(x)&=y_0+\int_{x_0}^xf(\varsigma,y_0(\varsigma))d\varsigma\\\
&…\\\
y_n(x)&=y_0+\int_{x_0}^xf(\varsigma,y_{n-1}(\varsigma))d\varsigma\\\
&…\\\
\end{aligned}
$$

易见，$y_n(x)在\alpha \leq x \leq \beta $上存在且连续.再证明函数列${y_n(x)}在\alpha \leq x \leq \beta$上一致连续.

考察级数$y_0(x)+\sum_{k=1}^\infty[y_k(x)-y_{k-1}(x)],x\in [\alpha,\beta].$

取$M=max_{x\in [\alpha,\beta]}\mid f(x,y_0)\mid$

由上式可得

$\mid y_1(x)-y_0(x)\mid \leq \int_{x_0}^x \mid f(\varsigma,y_0(\varsigma))\mid d\varsigma \leq M\mid x-x_0 \mid$

$\mid y_2(x)-y_1(x)\mid \leq \int_{x_0}^x \mid f(\varsigma,y_1(\varsigma))-f(\varsigma,y_0(\varsigma))\mid d\varsigma$

由利普希茨条件得

$\mid y_2(x)-y_1(x)\mid \leq L\int_{x_0}^x \mid y_1(\varsigma)-y_0(\varsigma)\mid d\varsigma \leq L\int_{x_0}^x M\mid \varsigma-x_0 \mid d\varsigma=\frac{ML}{2!}(x-x_0)^2.$

由数学归纳法$\mid y_k(x)-y_{k-1}(x) \mid \leq \frac{ML^{k-1}}{k!} \mid (x-x_0)\mid ^k \leq \frac{ML^{k-1}}{k!}(\beta-\alpha)^k$

由Weierstrass判别法得，$\alpha \leq x \leq \beta$上级数一致收敛，

所以，函数列${y_n(x)}在[\alpha,\beta]$上一致收敛.

设$\lim_{n→+\infty}y_n(x)=y(x)$，

所以，${y(x)}在[\alpha,\beta]$上连续

且${f(x,y_n(x))}在[\alpha,\beta]上一致收敛于f(x,y(x))$

取极限得$y(x)=y_0+\int_{x_0}^xf(\varsigma,y(\varsigma))d\varsigma$

10.给定积分方程$$\phi(x)=f(x)+\lambda \int_a^b K(x,\varsigma) \phi(\varsigma)d\varsigma$$

其中$f(x)是[\alpha,\beta]$上的已知连续函数，$K(x,\varsigma)是a \leq x \leq b,a \leq \varsigma \leq b$上的已知连续函数.证明当$\mid \lambda \mid足够小时（\lambda是常数）$，（*）在$[\alpha,\beta]$上存在唯一的连续解.

（提示：做逐步逼近函数序列

$$\phi_0=f(x),$$

$$\phi_{n+1}(x)=f(x)+\lambda \int_a^b K(x,\varsigma) \phi_n(\varsigma)d\varsigma     (n=0,1,2…)$$

证明：

做逐步逼近函数序列

$\phi_0=f(x),$

令

$$\phi_1(x)=f(x)+\lambda \int_a^b k(x,\varsigma) \phi_0(\varsigma)d\varsigma$$

$$……$$

$$\phi_{n+1}(x)=f(x)+\lambda \int_a^b k(x,\varsigma) \phi_n(\varsigma)d\varsigma$$

$$……$$

令$M=max_{x \in [a,b]} \mid f(x) \mid ,L=max \mid k(x,\varsigma) \mid >0,$

$$
\begin{aligned}
|\phi_1(x)-\phi_0(x) | &=| \lambda \int_a^b k(x,\varsigma) \phi_0(\varsigma)d\varsigma |\\\
& \leq | \lambda | \int_a^b |d k(x,\varsigma) | \  | f(\varsigma) | d\varsigma\\\
& \leq | \lambda | ML(b-a)
\end{aligned}
$$

假设$\mid \phi_k(x)-\phi_{k-1}(x)\mid \leq {\mid \lambda \mid }^kL^kM(b-a)^k$成立，则有
$$
\begin{aligned}
\mid \phi_{k+1}(x)-\phi_k(x)\mid &=\mid \lambda \int_a^b k(x,\varsigma) (\phi_k(\varsigma)-\phi_{k-1}(\varsigma))d\varsigma \mid\\\
&\leq \mid \lambda \mid \int_a^b \mid k(x,\varsigma) \mid \  \mid\phi_k(\varsigma)-\phi_{k-1}(\varsigma) \mid d\varsigma\\\
&\leq \mid \lambda \mid\int_a^b L^{k+1}{\mid \lambda \mid }^kM(b-a)^k d\varsigma\\\
&={\mid \lambda \mid }^{k+1}L^{k+1}M(b-a)^{k+1}
\end{aligned}
$$

由归纳法可知，对任意的$n \in N$,

有，$\mid \phi_n(x)-\phi_{n-1}(x)\mid \leq {\mid \lambda \mid }^nL^nM(b-a)^n$

证明一致收敛：

考虑级数,$\phi_0(x)+\sum_{k=1}^\infty(\phi_k(x)-\phi_{k-1}(x))$

该级数的前n项和为:$\phi_0(x)+\sum_{k=1}^n(\phi_k(x)-\phi_{k-1}(x))=\phi_n(x)$

由$\mid \phi_k(x)-\phi_{k-1}(x)\mid \leq {\mid \lambda \mid }^kL^kM(b-a)^k$

可得，当$\mid \lambda \mid<\frac{1}{(b-a)L}$时，

级数$\sum_{k=1}^\infty{\mid \lambda \mid }^kL^kM(b-a)^k$收敛，

所以，$\mid \lambda \mid<\frac{1}{(b-a)L}时,\phi_0(x)+\sum_{k=1}^\infty(\phi_k(x)-\phi_{k-1}(x))在[a,b]上一致收敛$

设极限函数为$\phi^*(x),即x \in [a,b]$时，

当$n→\infty $时，$\phi_n(x)一致收敛于\phi^*,且由\phi_n(x)的连续性可得\phi^*(x)在[a,b]上连续$

得,对于$\phi_{n+1}(x)=f(x)+\lambda \int_a^b k(x,\varsigma) \phi_n(\varsigma)d\varsigma$，

当$n→\infty $时，两边取极限

得,$\phi^*(x)=f(x)+\lambda \int_a^b k(x,\varsigma) \phi^*(\varsigma)d\varsigma$

$(\mid \lambda \mid<\frac{1}{(b-a)L})$

即，当$\mid \lambda \mid<\frac{1}{(b-a)L}$时，积分方程存在连续解.

再证明解的唯一性，反证法.

设另有解$\phi_0^*(x)\neq \phi_0^*(x)$,即有

$\phi_0^*(x)=f(x)+\lambda \int_a^b k(x,\varsigma) \phi_0^*(\varsigma)d\varsigma$

记$Q=max_{x\in [a,b]}\mid \phi^*(x)-\phi_0^*(x)\mid>0$

$$
\begin{aligned}
\mid \phi^*(x)-\phi_0^*(x)\mid &=\mid \lambda \int_a^b k(x,\varsigma) (\phi^*(\varsigma)-\phi_0^*(\varsigma))d\varsigma \mid\\\
&\leq \mid \lambda \mid \int_a^b L \mid\phi^*(\varsigma)-\phi_0^*(\varsigma) \mid d\varsigma\\\
&=\mid \lambda \mid QL(b-a)
\end{aligned}
$$

即有，$Q\leq \mid \lambda \mid QL(b-a)$,

由于$Q>0，可知\mid \lambda \mid \geq \frac{1}{(b-a)L},与\mid \lambda \mid<\frac{1}{(b-a)L}$矛盾，

唯一性得证.
