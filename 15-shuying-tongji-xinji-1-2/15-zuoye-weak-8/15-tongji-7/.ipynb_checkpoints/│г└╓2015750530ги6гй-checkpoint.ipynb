$1.设x(t)和y(t)是区间a\leq t\leq b上的连续函数，证明：如果在区间a\leq t\leq b上有\frac{x(t)}{y(t)}\neq常数，或者\frac{y(t)}{x(t)}\neq常数，则x(t)和y(t)在区间a\leq$t\leqb上线性无关。（提示：用反证法）


$证明：$


$假设x(t)和y(t)在区间a\leq$t\leq b$上线性无关，则存在不全为零的数A、B,使得$


$$Ax(t)+By(t)=0$$


$即有$


$$\frac{x(t)}{y(t)}=-\frac{B}{A}$$


$而\frac{B}{A}必为常数.$


$又由题意\frac{x(t)}{y(t)}\neq常数，两个相矛盾.$


$因而假设不成立。$



$2.证明非齐次线性微分方程的叠加原理：设x_1(t),x_2(t)分别是非齐次线性微分方程$ 


$                    \quad \frac{(dx)^n}{dt^n}+a_1(t)\frac{(dx)^{n-1}}{dt^{n-1}}+\cdots+a_n(t)x=f_1(t)$ 


$                    \quad \frac{(dx)^n}{dt^n}+a_1(t)\frac{(dx)^{n-1}}{dt^{n-1}}+\cdots+a_n(t)x=f_2(t)$ 


$的解，则x_1(t)+x_2(t)是方程\frac{(dx)^n}{dt^n}+a_1(t)\frac{(dx)^{n-1}}{dt^{n-1}}+\cdots+a_n(t)x=f_1(t)+f_2(t)的解$ 


$证明：x_1(t),x_2(t)分别是非齐次线性微分方程$  


$                    \quad \frac{(dx)^n}{dt^n}+a_1(t)\frac{(dx)^{n-1}}{dt^{n-1}}+\cdots+a_n(t)x=f_1(t)$  


$                    \quad \frac{(dx)^n}{dt^n}+a_1(t)\frac{(dx)^{n-1}}{dt^{n-1}}+\cdots+a_n(t)x=f_2(t)$


$的解，所以满足$  

$                    \quad \frac{(dx_1(t)^n}{dt^n}+a_1(t)\frac{(dx_1(t))^{n-1}}{dt^{n-1}}+\cdots+a_n(t)x_1(t)=f_1(t)$ 


$                    \quad \frac{(dx_2(t)^n}{dt^n}+a_1(t)\frac{(dx_2(t))^{n-1}}{dt^{n-1}}+\cdots+a_n(t)x_2(t)=f_2(t)$


$将以上两式相加，并且利用导数的运算法则有$  


$                    \quad \frac{(d(x_1(t)+x_2(t))^n}{dt^n}+a_1(t)\frac{d(x_1(t)+x_2(t))^{n-1}}{dt^{n-1}}+\cdots+a_n(t)(x_1(t)+x_2(t))=f_1(t)+f_2(t)$  


$即x_1(t）+x_2(t)是方程\frac{(dx)^n}{dt^n}+a_1(t)\frac{(dx)^{n-1}}{dt^{n-1}}+\cdots+a_n(t)x=f_1(t)+f_2(t)的解$  



$3.已知齐次线性微分方程的基本解组x_1,x_2,求下列方程对应的非其次线性微分方程的通解：$

$(1)x''-x=\cos t,x_1=e^t,x_2=e^{-t};$

$解：应用常数变易法，令$

$$x=c_1(t)e^t+c_2(t)e^{-t}$$

$将它代入方程，则有$

$$
\begin{cases}
c_1'(t)e^t+c_2'(t)e^{-t}=0.\\
c_1'(t)e^t-c_2'(t)e^{-t}=\cos t.
\end{cases}$$

$解得$

$\begin{cases}
c_(1)'=\frac{1}{2}e^{-t}\cos t.\\
c_2'(t)=-\frac{1}{2}e^t\cos t.
\end{cases}$$

$积分，得$

$$c_1(t)=-\frac{1}{4}e^{-t}(\cos t-\sin t)+c_1$$

$$c_2(t)=-\frac{1}{4}e^t(\cos t+\sin t)+c_2.$$

$因而，通解为：$

$$x(t)=c_1e^t+c_2e^{-t}-\frac{1}{2}\cos t$$



$(3)x''+4x=t\sin {2t},x_1=\cos {2t},x_2=\sin {2t};$

$解:应用常数变易法,令$

$$x=c_1(t)\cos {2t}+c_2(t)\sin {2t}.$$

$将它代入方程，则有$

$$\begin{cases}
c_1'(t)\cos {2t}+c_2'(t)\sin {2t}=0.\\
2c_2'(t)\cos {2t}-2c_1'\sin {2t}=t\sin {2t}.
\end{cases}$$

$解得：$

$$\begin{cases}
c_1'(t)=\frac{1}{4}t(\cos {4t}-1)\\
c_2'(t)=\frac{1}{4}t\sin {4t}.
\end{cases}$$

$积分，得$

$$\begin{cases}
c_1(t)=\frac{1}{64}\cos {4t}+\frac{1}{16}t\sin {4t}-\frac{1}{8}t^2+r_1.\\
c_2(t)=\frac{1}{64}\sin {4t}-\frac{1}{16}t\cos {4t}+r_2.
\end{cases}$$

$因此，方程的通解为：$

$$x=c_1\cos {2t}+c_2\sin {2t}-\frac{1}{8}t^2\cos{2t}+\frac{1}{16}t\sin {2t}.$$

$(5)t^2x"-tx'+x=64t+34t^2,x_1=t,x_2=tln t;$

$解：原方程可化为：$

$$x"-\frac{1}{t}x'+\frac{1}{t^2}x=\frac{6}{t}+34$$

$令方程的解为:$

$$x=c_1(t)t+c_2(t)tln t$$

$代入方程，得$

$$\begin{cases}
c_1'(t)t+c_2'(t)tln t=0.\\
c_1'(t)+c_2'(ln t+1)=\frac{6}{t}+34.
\end{cases}$$

$解得：$

$$\begin{cases}
c_1'(t)=-ln t(\frac{6}{t}+34)\\
c_2'(t)=\frac{6}{t}+34..
\end{cases}$$

$积分，得$

$$\begin{cases}
c_1(t)=-34(tln t-t)-3(ln t)^2+c_1.\\
c_2(t)=34t+6ln t+c_2.
\end{cases}$$

$因此，方程的通解为：$

$$x=c_1t+c_2tln t+34t^2+3t(ln t)^2,(其中c_1,c_2为任意常数.)$$


$6.假设x_1(t)\neq 0是二阶其次线性微分方程$

$$x"+a_1(t)x'+a_2(t)x=0.$$

$的解，这里a_1(t)和a_2(t)于区间[a,b]上连续，试证：$

$$ $$

$$x_1(x_2"+a_1x_2'+a_2x_2)=0$$

$(2)$

$因为x_1,x_2为方程的解，则由刘维尔公式$

$$
\begin{vmatrix}
