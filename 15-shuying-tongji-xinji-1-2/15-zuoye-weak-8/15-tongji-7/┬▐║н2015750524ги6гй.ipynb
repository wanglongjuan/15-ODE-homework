 1.设x(t)和y(t)是区间$a\leq t\leq b$上的连续函数，证明：如果在区间$a\leq t\leq b$上有$\frac{x(t)}{y(t)}\neq$常数或$\frac{y(t)}{x(t)}\neq$常数，则x(t)和y(t)在区间$a\leq t\leq b$上线性无关.（提示：用反证法）
 
 $证明：\quad假设x(t)和y(t)在区间[a,b]上线性相关，则存在不全为零的常数C_1和C_2使C_1x(t)+C_2y(t)\equiv 0,t\in[a,b].$
 $\quad\quad\quad若C_1\neq 0,则\frac{y(t)}{x(t)}=-\frac{C_1}{C_2}=常数.$
 $\quad\quad\quad若C_1\neq 0,则\frac{x(t)}{y(t)}=-\frac{C_2}{C_1}=常数.$
 $\quad但是在这两种情况下，都与已知条件相矛盾.所以假设不成立，即证x(t)和y(t)在区间a\leq t\leq b上线性无关.$

2.证明非齐次线性微分方程的叠加原理：设$x_1(t),x_2(t)$分别是非齐次线性微分方程
 $
\begin{aligned}
\quad\quad \frac{\mathrm d^n x}{\mathrm d t^n}+a_1(t)\frac{\mathrm d^{n-1} x}{\mathrm d t^{n-1}}+…+a_n(t)x &=f_1(t),\\
\frac{\mathrm d^n x}{\mathrm d t^n}+a_1(t)\frac{\mathrm d^{n-1} x}{\mathrm d t^{n-1}}+…+a_n(t)x &=f_2(t)\\
\end{aligned}
$的解，则$x_1(t)+x_2(t)$是方程$\frac{\mathrm d^n x}{\mathrm d t^n}+a_1(t)\frac{\mathrm d^{n-1} x}{\mathrm d t^{n-1}}+…+a_n(t)x =f_1(t)+f_2(t)$的解

$证明：\quad x_1(t),x_2(t)分别是非齐次线性微分方程$
$
\begin{aligned}
\quad\quad \quad\frac{\mathrm d^n x}{\mathrm d t^n}+a_1(t)\frac{\mathrm d^{n-1} x}{\mathrm d t^{n-1}}+…+a_n(t)x &=f_1(t),\\
\frac{\mathrm d^n x}{\mathrm d t^n}+a_1(t)\frac{\mathrm d^{n-1} x}{\mathrm d t^{n-1}}+…+a_n(t)x &=f_2(t)\\
\end{aligned}
$
$\quad的解，所以$
$
\begin{aligned}
\quad\quad\quad \frac{\mathrm d^n x_1(t)}{\mathrm d t^n}+a_1(t)\frac{\mathrm d^{n-1} x_1(t)}{\mathrm d t^{n-1}}+…+a_n(t)x_1(t) &=f_1(t),\\
\frac{\mathrm d^n x_2(t)}{\mathrm d t^n}+a_1(t)\frac{\mathrm d^{n-1} x_2(t)}{\mathrm d t^{n-1}}+…+a_n(t)x_2(t) &=f_2(t).\\
\end{aligned}
$
$\quad 将上述两个式子想加，利用导数的运算法则得$
$\quad \frac{\mathrm d^n (x_1(t)+x_2(t))}{\mathrm d t^n}+a_1(t)\frac{\mathrm d^{n-1} (x_1(t)+x_2(t))}{\mathrm d t^{n-1}}+…+a_n(t)(x_1(t)+x_2(t)) =f_1(t)+f_2(t).$
$\quad∴x_1(t)+x_2(t)是方程\frac{\mathrm d^n x}{\mathrm d t^n}+a_1(t)\frac{\mathrm d^{n-1} x}{\mathrm d t^{n-1}}+…+a_n(t)x =f_1(t)+f_2(t)的解.命题得证。$

 3.已知齐次线性微分方程的基本解组$x_1,x_2$,求下列方程对应的非齐次线性微分方程的通解：
 
 (1)$x''-x=\cos t,x_1=e^t,x_2=e^{-t};$
 
 $解：应用常数变易法，令$
 $\quad x=C_1(t)e^t+C_2(t)e^{-t}$
 
 $代入方程得$
 $\begin{aligned}
\quad\quad\quad e^\prime C^\prime_1(t)+e^{-t}C^\prime_2(t) & =0\\
\quad \quad e^\prime C^\prime_1(t)-e^{-t}C^\prime_2(t) & =\cos t\\
\end{aligned}$
$解得\quad C^\prime_1(t)=\frac{1}{2}\cos t·e^{-t},C^\prime_2(t)=-\frac{1}{2}\cos t·e^t.$
$\begin{aligned}
则\quad\quad C_1(t)&=\frac{\sin t-\cos t}{4}·e^{-t}+\gamma_1,\\
C_2(t)&=-\frac{\sin t+\cos t}{4}·e^t+\gamma_2,\\
\end{aligned}$
$∴原方程通解为x=\gamma_1e^t+\gamma_2e^{-t}-\frac{1}{2}\cos t.$

(3)$x''+4x=t\sin 2t,x_1=\cos 2t,x_2=\sin 2t;$

$解：应用常数变易法,令$
$\quad x=C_1(t)\cos 2t+C_2(t)\sin 2t$
$代入方程可得$
$
\begin{aligned}
\quad\quad\quad & \cos 2t C^\prime_1(t)+\sin 2t C^\prime_2(t) =0\\
和\quad \quad & -2\sin 2t C^\prime_1(t)+2\cos 2t C^\prime_2(t) =t\sin t\\
\end{aligned}
$
$解得\quad C^\prime_1(t)=-\frac{1}{2}t\sin t\sin 2t,C^\prime_2(t)=\frac{1}{2}t\sin t\cos 2t.$
$
\begin{aligned}
则\quad\quad C_1(t)&=\frac{1}{4}[\frac{1}{3}t\sin 3t+\frac{1}{9}\cos 3t-t\sin t-\cos t]+C_1,\\
C_2(t)&=\frac{1}{4}[-\frac{1}{3}t\cos 3t+\frac{1}{9}\sin 3t+t\cos t-\sin t]+C_2\\
\end{aligned}
$
$∴原方程通解为x=C_1\cos 2t+C_2\sin 2t+\frac{t\sin t}{3}-\frac{2\cos t}{9}.$

(5)$t^2x''-tx'+x=6t+34t^2,x_1=t,x_2=t\ln t.$

$解：方程可变为x''-\frac{1}{t}x'+\frac{1}{t^2}x=\frac{6}{t}+34.$
$应用常数变易法,令$
$\quad x=tC_1(t)+t\ln tC_2(t)$
$代入方程可得$
$
\begin{aligned}
\quad\quad\quad & t C^\prime_1(t)+t\ln t C^\prime_2(t) =0\\
和\quad \quad & C^\prime_1(t)+(\ln t+1)C^\prime_2(t) =(\frac{6}{t}+34)\\
\end{aligned}
$
$解得\quad C^\prime_1(t)=-\ln t(\frac{6}{t}+34)-34,C^\prime_2(t)=(\frac{6}{t}+34).$
$
\begin{aligned}
则\quad\quad C_1(t)&=-34(t\ln t)-3(\ln t)^2+C_1,\\
C_2(t)&=34t+6\ln t+C_2\\
\end{aligned}
$
$∴原方程通解为x=tC_1+t\ln tC_2+34t^2+3t(\ln t)^2.$
