因为$x_1(t),x_2(t)$分别是非齐次线性微分方程

$\frac{d^nx}{dt^n}+a_1(t)\frac{d^{n-1}x}{dt^{n-1}}+\cdots+a_n(t)x=f_1(t)$和$\frac{d^nx}{dt^n}+a_1(t)\frac{d^{n-1}x}{dt^{n-1}}+\cdots+a_n(t)x=f_2(t)$的解

所以有

$\frac{d^nx_1(t)}{dt^n}+a_1(t)\frac{d^{n-1}x_1(t)}{dt^{n-1}}+…+a_n(t)x_1(t)=f_1(t)$...........①

$\frac{d^nx_2(t)}{dt^n}+a_1(t)\frac{d^{n-1}x_2(t)}{dt^{n-1}}+…+a_n(t)x_2(t)=f_2(t)$...........②

①+②得

$\frac{d^n(x_1(t)+x_2(t))}{dt^n}+a_1(t)\frac {d^{n-1}(x_1(t)+x_2(t))}{dt^{n-1}}+…+a_n(t)(x_1(t)+x_2(t))=f_1(t)+f_2(t)$

所以$x_1(t)+x_2(t)$是

$\frac{d^nx}{dt^n}+a_1(t)\frac{d^{n-1}x}{dt^{n-1}}+\cdots+a_n(t)x=f_1(t)+f_2(t)$的解

3.（1）

令$x=C_1(t)e^t+C_2(t)e^{-t}$,代入得：

$$e'C'_1(t)+e^{-t}C'_2(t)=0$$

$$e'C'_1(t)-e^{-t}C'_2(t)=\cos t$$

解得:

$$C'_1(t)=\frac{1}{2}\cos t e^{-t}$$

$$C'_2(t)=-\frac{1}{2}e^t\cos t$$

所以
$$C_1(t)=\frac{\sin t-\cos t}{4}e^{-t}+γ_1$$

$$C_2(t)=-\frac{1}{4}(\sin t+\cos t)e^t+γ_2$$

方程通解为$x=γ_1e^t+γ_2e^{-t}-\frac{1}{2}\cos t$

3.（3）

令$x=C_1(t)\cos 2t+C_2(t)\sin 2t$,代入得

$$\cos 2t C'_1(t)+\sin 2t C'_2(t)=0$$

$$-2\sin 2tC'_1(t)+2\cos 2t C'(t)=t\sin t$$

解得:

$$C'_1(t)=-\frac{1}{2}t\sin t\sin 2t$$

$$C'_2(t)=\frac{1}{2}2\sin t\cos 2t$$

所以

$$C_1(t)=\frac{1}{4}[\frac{1}{3}t\sin 3t+\frac{1}{9}\cos 3t-t\sin t-\cos t]+C_1$$

$$C_2(t)=\frac{1}{4}[-\frac{1}{3}t\cos 3t+\frac{1}{9}\sin 3t+t\cos t-\sin t]+C_2$$

方程通解为$x=C_1\cos 2t+C_2\sin 2t+\frac{t}{3}\sin t-\frac{2}{9}\cos t$

3.（5）

先将方程变形为$x''-\frac{1}{t}x'+\frac{1}{t^2}x=\frac{6}{t}+34$

令$x=tC_1(t)+t\ln t C_2(t)$，代入得

$$tC'_1(t)+t \ln tC'_2(t)=0$$

$$C'_1(t)+(\ln t+1)C'_2(t)=\frac{6}{t}+34$$

解得

$$C'_1(t)=-\ln t(\frac{6}{t}+34)$$

$$C'_2(t)=\frac{6}{t}+34$$

所以

$$C_1(t)=-34(t\ln t)-3(\ln t)^2+C_1$$

$$C_2(t)=34t+6\ln t+C_2$$

方程的通解为$x=C_1t+C_2t\ln t+34t^2+3t(\ln t)^2$

6.（1）

①充分性:

$$
W'[x_1,x_2]=
\begin{vmatrix}
x'_1&x'_2\\x'_1&x'_2
\end{vmatrix}
+
\begin{vmatrix}
x_1&x_2\\x''_1&x''_2
\end{vmatrix}
=
\begin{vmatrix}
x_1&x_2\\x''_1&x''_2
\end{vmatrix}
$$

$$
W'[x_1,x_2]+a_1(t)W[x_1,x_2]=
\begin{vmatrix}
x_1&x_2\\x''_1&x''_2
\end{vmatrix}
+a_1(t)
\begin{vmatrix}
x_1&x_2\\x'_1&x'_2
\end{vmatrix}
=0
$$ 

又因为$x_1(t)\neq0$是方程的解，所以有

$$
\begin{vmatrix}
x_1&x_2\\-a_2(t)x_1&x''_2+a_1(t)x'_2
\end{vmatrix}
=x_1
\begin{vmatrix}
1&x_2\\-a_2(t)&x''_2+a_1(t)x'_2
\end{vmatrix}
=0
$$

即可得$x''_2+a_1(t)x'_2+a_2(t)x_2=0$，所以$x_2(t)$是已知方程的解

②必要性

因为$W[x_1,x_2]$是方程的解$x_1(t),x_2(t)$的朗斯基行列式，

所以

$$
W'[x_1,x_2]=
\begin{vmatrix}
x_1&x_2\\x''_1&x''_2
\end{vmatrix}
+
\begin{vmatrix}
x_1&x_2\\x''_1+a_2(t)x&x''_2+a_2(t)x_2
\end{vmatrix}
=
\begin{vmatrix}
1&x_2\\-a_1(t)x'_1&a_1(t)x'_2
\end{vmatrix}
=-a_1(t)W[x_1,x_2]
$$

综上所述，$x_2(t)$为方程的解的充要条件是
$W'[x_1,x_2]+a_1W[x_1,x_2]=0$

6.（2）

因为$x_1,x_2$为方程的解，则
$$
\begin{vmatrix}
x_1&x_2\\x'_1&x'_2
\end{vmatrix}
=W(t_0)e^{-\int^t_{t_0}a_1(s)ds}
$$

即:$x_1x'_2-x'_1x_2=W(t_0)e^{-\int^t_{t_0}a_1(s)ds}$

两边同乘$\frac{1}{x_1^2}$，则有:

$\frac{d\frac{x_2}{x_1}}{dt}=\frac{W(t_0)}{x_1^2}e^{-\int^t_{t_0}a_1(s)ds}$

解得:$x_2=(c_1\int\frac{1}{x^2}e^{-\int^t_{t_0}a_1(s)ds}dt+c_2)x_1$

令$c_1=1,c_2=0$，得：

$x_2=x_1\int\frac{1}{x_1^2}e^{-\int^t_{t_0}a_1(s)ds}dt$

所以原方程的通解为：$x=x_1[c_1\int\frac{1}{x_1^2}exp(-\int^t_{t_0}a_1(s)ds)dt+c_2]$
其中$c_1,c_2$为常数，$t_0,t∈[a,b]$
