$1.求下列方程的解$

$（4）dy/dx-(n/x)y=e^xx^n$

$解：令e^xx^n=0则dy/dx=(n/x)y$

$分离变量：dy/y=(n/x)dx$

$两边同时积分得：y=ce^{∫(n/x)dx}$

$令y=c（x）e^{∫(n/x)dx}$

$两边同时微分得：dy/dx=dc(x)/dx*e^{∫(n/x)dx}+c(x)(n/x)e^{∫(n/x)dx}$

$带入原方程得：dc(x)/dx*e^{∫(n/x)dx}+c(x)(n/x)e^{∫(n/x)dx}=c(x)(n/x)e^{∫(n/x)dx}+e^xx^n$

即：
$dc(x)/dx=e^xx^ne^{∫(n/x)dx}$

$积分得：c(x）=∫e^xx^ne^{∫(n/x)dx}dx+c(c为常数）$

$得方程的通解为：y=(∫e^xx^ne^{∫(n/x)dx+c)}*e^{∫((n/x)dx)}=x^n(e^x+c)$

$课堂作业：$

$\int_{x_0}^{x}\mathrm p(t)\,\mathrm{d}t$

$的变上限求导$

$P(x+△x)-P(x)=\int_{x_0}^{x+△x}\mathrm p(t)\,\mathrm{d}t-\int_{x_0}^{x}\mathrm p(t)\,\mathrm{d}t=\int_{x_0}^{x+△x}\mathrm p(t)\,\mathrm{d}t$

$F'(x)等于对[(P(x+△x)-P(x))/△x]在△x趋于0时求极限$

$由定积分第一中值定理可知：P'(x)=p(x)$

$故：\int_{x_0}^{x}\mathrm p(t)\,\mathrm{d}t=p(x)$
