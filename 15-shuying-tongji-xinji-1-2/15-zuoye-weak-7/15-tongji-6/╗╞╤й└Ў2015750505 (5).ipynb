$$\bf习题3.3$$

$1.假设函数f(x,y)及\frac{∂f}{∂y}都在区域G内连续，又y=φ(x,x_0,y_0)是方程（3.1)满足初值条件φ（x_0,x_0,y_0)=y_0的解，试证\frac{∂f}{∂y_0}存在且连续，并写出其表达式.$

$解：$

$记初值为（x_0,y_0+∆y_0)的解为y=\psi(x,x_0,y_0+∆y_0),其中|y_0|≤α,α足够小，则有$

$$\phi=y_0+\int_{x_0}^x\mathrm f(s,\phi(s))\,\mathrm{d}s,\psi=y_0+∆y_0+\int_{x_0}^x\mathrm f(s,\psi(s))\,\mathrm{d}s.$$

$\psi-\phi=∆y_0+\int_{x_0}^x\mathrm [f(s,\psi(s))-f(s,\phi(s))]\,\mathrm{d}s$

$\quad\quad\quad=∆y_0+\int_{x_0}^x\mathrm [\frac{∂f(s,\phi+θ(\psi-\phi))}{∂y}(\psi(s)-\phi(s))]\,\mathrm{d}s,其中θ∈[0,1],$
$由\frac{∂f}{∂y}、\psi,\phi的连续性，有$

$${∂f(s,\phi+θ(\psi-\phi))}{∂y}=\frac{∂f(s,\phi)}{∂y}+r$$

$其中r满足当∆y_0→0时，r→0，且∆y_0=0时，r=0.$

$参照书本证明:$

$当∆y_0≠0时，有$

$$\frac{\psi-\phi}{∆y_0}=1+\int_{x_0}^x\mathrm [\frac{∂f(s,\phi)}{∂y}+r]\frac{\psi-\phi}{∆y_0}\,\mathrm{d}s$$

$即z=\frac{\psi-\phi}{∆y_0}是微分方程初值问题：$

$$\begin{cases}
 \frac{dz}{dx}=[\frac{∂f(s,\phi)}{∂y}+r]z\\
z(x_0)=1
 \end{cases}的解$$
 
$所以$

$$\frac{∂\phi}{∂y_0}=exp{\int_{x_0}^x\mathrm [\frac{f(s,\phi(s))}{∂y}]\,\mathrm{d}s}$$

$$它是x，x_0,y_0的连续函数.$$

$$即证$$




$2.假设函数P（x)和Q(x)于区间[α，β]上连续，y=\phi(x,x_0,y_0)是方程$

$$\frac{dy}{dx}=Q(x)y+Q(x)$$

$的解，y_0=\phi(x_0,x_0,y_0).试求\frac{∂\phi}{∂x_0},\frac{∂\phi}{∂y_0},\frac{∂\phi}{∂x},并从解的表达式出发，利用对参数求导的方法，检验所得结果.$

$解：$

$由题目，可知：$

$$f(x,y)=P(x)y+Q(x).$$

$所以$

$$\frac{∂f}{∂y}=P(x).$$

$利用公式，有$

$$\frac{∂\phi}{∂x_0}=-f(x_0,y_0)exp{\int_{x_0}^x\mathrm [\frac{f(x,\phi)}{∂y}\,\mathrm{d}x)]}$$

$$\quad\quad\quad=-[P(x_0)+Q(x_0)]exp{\int_{x_0}^x\mathrm P(x)\,\mathrm{d}x}$$

$同理，$

$$\frac{∂\phi}{∂y_0}=exp{\int_{x_0}^x\mathrm [\frac{f(x,\phi)}{∂y}]\,\mathrm{d}x}$$

$$\quad=exp{\int_{x_0}^x\mathrm P(x)\,\mathrm{d}x}$$

$$\frac{∂\phi}{∂x}=f(x,\phi(x,x_0,y_0))$$

$即用\phi(x,x_0,y_0)代替y$

$所以，上式为：$

$$\quad=P(x)\phi(x,x_0,y_0)+Q(x).$$

$$\bf验证：$$

$该方程解的通式为：$

$$y={e}^{\int\mathrm{P(x)}\,\mathrm{d}x}(\int\mathrm{Q(x)}{e}^{\int\mathrm {-P(x)}\,\mathrm{d}x}\,\mathrm{d}x+c)$$

$于是$

$\phi(x,x_0,y_0)=y_0e^{\int_{x_0}^x\mathrm P(s)\,\mathrm{d}s}+\int_{x_0}^x\mathrm Q(s)e^{\int_s^x\mathrm P(t)\,\mathrm{d}t}\,\mathrm{d}s$

$直接对参数求导，得$

$$\frac{∂\phi(x,x_0,y_0)}{∂x_0}=-P(x_0)y_0e^{\int_{x_0}^x\mathrm P(s)\,\mathrm{d}s}-Q(x_0){\int_{x_0}^x\mathrm P(s)\,\mathrm{d}s}$$

$$\quad=-[P(x_0)y_0+Q(x_0)]e^{\int_{x_0}^x\mathrm P(s)\,\mathrm{d}s},$$

$同理$

$\quad\quad\quad\quad\quad\quad\quad\quad\quad\quad\frac{∂\phi(x,x_0,y_0)}{∂y_0}=e^{\int_{x_0}^x\mathrm P(s)\,\mathrm{d}s},$

$$\frac{∂\phi(x,x_0,y_0)}{∂x}=P(x_0)y_0e^{\int_{x_0}^x\mathrm P(s)\,\mathrm{d}s}+Q(x)+\int_{x_0}^x\mathrm P(s)Q(s)e^{\int_s^x\mathrm P(t)\,\mathrm{d}t}\,\mathrm{d}s$$

$$\quad\quad\quad\quad=P(x)[y_0e^{\int_{x_0}^x\mathrm P(s)\,\mathrm{d}s}+\int_{x_0}^x\mathrm Q(s)e^{\int_s^x\mathrm P(t)\,\mathrm{d}t}\,\mathrm{d}s]+Q(x).$$

$\quad\quad\quad\quad\quad\quad\quad\quad\quad\quad\quad\quad\quad\quad=P(x)\phi(x,x_0,y_0)+Q(x).$

$经验证，其解与可微性定理中的公式一致.$

$$即证$$



$3.给定方程$

$$\frac{dy}{dx}=\sin (\frac{y}{x}),$$

$试求\frac{∂y(x,x_0,y_0)}{∂x_0},\frac{∂y(x,x_0,y_0)}{∂y_0}在x_0=1,y_0=0时的表达式.$

$解:$

$由题目,可知：$

$$f（x,y)=\sin (\frac{y}{x})$$

$则$

$$\frac{∂f}{∂y}=\frac{1}{x}\cos (\frac{y}{x})$$

$由解对初值的可微性定理，有：$

$$\frac{∂y(x,x_0,y_0)}{∂x_0}=-f(x_0,y_0)exp{\int_{x_0}^x\mathrm [\frac{f(x,\phi)}{∂y}\,\mathrm{d}x)]}$$

$$=-\sin (\frac{y_0}{x_0})exp({\int_{x_0}^x\mathrm [\frac{1}{s}\cos (\frac{y}{s})]\,\mathrm{d}s)}$$

$$\frac{∂y(x,x_0,y_0)}{∂y_0}=exp({\int_{x_0}^x\mathrm [\frac{1}{s}\cos (\frac{y}{s})]\,\mathrm{d}s)}$$

$当x_0=1,y_0=0时，$

$$\frac{∂y(x,x_0,y_0)}{∂x_0}|_{x_0=1,y_0=0}=0$$

$$\frac{∂y(x,x_0,y_0)}{∂y_0}|_{x_0=1,y_0=0}=exp{\int_1^x\mathrm [\frac{1}{s})]\,\mathrm{d}s}=|x|$$

$所以$

$$在x_0=1,y_0=0时\frac{∂y}{∂x_0}=0,\frac{∂y}{∂y_0}=|x|$$

$4.设y=\phi(x,x_0,y_0,λ)是初值问题$

$$\frac{dy}{dx}=\sin (λxy),\phi(x_0,x_0,y_0,λ)=y_0$$

$的饱和解，这里λ是参数，求\frac{∂\phi}{∂_0},\frac{∂\phi}{∂y_0}在（x，0，0，1）处的表达式。$

$解：$

$由题意，有$

$$f(x,y)=\sin (λxy)$$

$所以$

$$\frac{∂f}{∂y}=λx\cos (λxy).$$

$$由解对初值的可微性定理，有：$

$$\frac{∂\phi}{∂x_0}=-f(x_0,y_0)exp{\int_{x_0}^x\mathrm [\frac{f(s,\phi)}{∂y}\,\mathrm{d}s)]}$$

$$=-\sin (λx_0y_0)exp{\int_{x_0}^x\mathrm λs\cos (λs\phi)}\,\mathrm{d}s)$$

$$\frac{∂\phi}{∂y_0}=exp{\int_{x_0}^x\mathrm [\frac{f(s,\phi)}{∂y}\,\mathrm{d}s]}=exp{\int_{x_0}^x\mathrm λs\cos (λs\phi)}\,\mathrm{d}s$$

$当x_0=0,y_0=0,λ=1时，则有$

$$\frac{∂\phi}{∂x_0}|_{(x,0,0,1)}=0$$

$$\frac{∂\phi}{∂y_0}|_{(x,0,0,1)}=exp(\frac{x^2}{2})$$
