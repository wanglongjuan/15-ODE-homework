$2.求解下列常系数微分方程。$

$(12)x''-6x'+5x=e^{2t}$

$解：其特征方程为：\lambda^2+6\lambda+5=0,求得\lambda的值为：-1，-5.$

$所以齐次线性微分方程的通解就为：x=c_1e^{-t}+c_2e^{-5t}$

$由f(t)=e^{2t}得\lambda=2,而\lambda=2不是特征方程的解，所以设\tilde {x}=Ae^{2t},将其带入方程可得：$

$4Ae^{2t}+12Ae^{2t}+5Ae^{2t}=e^{2t},求得A=\frac{1}{21}$

$所该方程的解为：x=c_1e^{-t}+c_2e^{-5t}+\frac{1}{21}e^{2t}$

$~~~~~~~~~~$

$(14)x''+x=\sin t-\cos 2t$

$解：其特征方程为：\lambda^2+1=0,求得\lambda的值为：\pm i.$

$所以齐次线性微分方程的通解就为：x=c_1\cos t+c_2\sin t$

$令f_1(t)=\sin t,f_2(t)=-\cos 2t,f(t)=f_1(t)+f_2(t)$

$f_1(t)=\sin t时，\pm i是特征方程的根，所以设\tilde {x_1}=t(a\cos t+b\sin t)可得：$

$\tilde {x''}=(2b-ta)\cos t-(2a+tb)\sin t,所以原方程为：$

$x''+x=2b\cos t-2a\sin t=\sin t,求得：b=0.a=-\frac{1}{2},所以\tilde {x_1}=-\frac{1}{2}t\cos t.$

$f_2(t)=-\cos 2t时，\pm 2i不是特征方程的根，所以设\tilde {x_2}=c\cos 2t+d\sin 2t$

$带入原方程，通过比较系数法可求得：c=\frac{1}{3},d=0,所以\tilde {x_2}=\frac{1}{3}t\cos 2t$

$所以该方程的解为：x=c_1\cos t+c_2\sin t-\frac{1}{2}t\cos t+\frac{1}{3}t\cos 2t$


$~~~~~~~~~~~~$

$~~~~~~~~~~~~$

$(16)x''+9x=t\sin 3t$

$解：其特征方程为：\lambda^2+9=0,解得\lambda的值为：\pm 3i$

$所以该齐次线性方程的通解就为：x=c_1\cos 3t+c_2\sin 3t.$

$f(t)=t\sin 3t,\pm 3i是特征方程的根，所以设\tilde {x}=t[(at+c)\cos 3t+(bt+d)\sin 3t],可求得：$

$\tilde {x''}=(-12at-9dt-9bt^2-6c)\sin 3t+(-9ct-9at^2+6d+2a+6bt)\cos 3t$

$所以\tilde {x''}+9\tilde{x}=(-12at-6c)\sin 3t+(6d+2a+6b-t)\cos 3t=t\sin 3t$

$通过比较系数法可求得：c=0,b=0,a=\frac{-1}{12},d=\frac{1}{36}$

$所以该方程的解为：x=c_1\cos 3t+c_2\sin 3t-\frac{1}{12}t^2\cos 3t+\frac{1}{36}t\sin 3t$

$4.求下列初值问题的解：$

$(1)x''+9x=6e^{3t},x(0)=x'(0)=0;$

$解：令x(t)为方程的一个特解，则有L[x(t)]=X(s).$

$则x''=s^2X(s)-sx(0)-x'(0)$

$所以原方程就可以化为：s^2X(s)-sx(0)-x'(0)+9X(s)=\frac{6}{s-3} Re s>3$

$可求得X(s)=\frac{6}{(s-3)(s^2+9)}=\frac{1}{3}e^{3t}-\frac{1}{3}(\frac{s}{s^2+9}-\frac{3}{s^2+9})$

$通过查表可得该线性方程的解为：x=\frac{1}{3}(e^{3t}-\cos 3t+\sin 3t)$


$~~~~~~~~~~~~$

$~~~~~~~~~~~~~$


$(2)x^{(4)}+x=2e^t,x(0)=x'(0)=x''(0)=x'''(0)=1.$

$解：令x(t)为该方程的一个特解，则有L[x(t)]=X(s).$

$利用拉普拉斯变换可将原方程化为：L[x^{(4)}]+L[x]=L[2e^t]$

$s^4X(s)-s^3x(0)-s^2x'(0)-sx''(0)-x'''(0)+X(s)=\frac{2}{s-1} Re s>1$

$通过运算可得X(s)=\frac{1}{s-1}$

$通过查表可得该线性方程的解为：x=e^t$
