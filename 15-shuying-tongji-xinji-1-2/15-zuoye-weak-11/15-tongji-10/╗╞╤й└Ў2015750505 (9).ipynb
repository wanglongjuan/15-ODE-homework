$$常微分作业$$

$2.求解下列常系数线性微分方程:$

$（12）x"+6x'+5x=e^{2t};$

$解:先求其次微分方程x"+6x'+5x=0的通解:$

$该特征方程为:$

$$\lambda^2+6\lambda+5=0$$

$即$
$$（\lambda+1）(\lambda+5)=0$$

$得到$

$$\lambda=-1或者\lambda=-5$$

$所以其次方程的通解为:$

$$ x=c_1e^{-1}+c_2e^{-5}\quad其中c_1,c_2为任意常数$$

$现求非其次方程的一个特解$

$这里f(t)=e^{2t},\lambda=2$

$然而\lambda=2不是特征方程的解，所以特解形如$

$$\tilde{x}=Ae^{2t}.$$

$将它代入原方程化简得$

$$21A=1.$$

$A=\frac{1}{21}$

$$从而\tilde{x}=\frac{1}{21}e^{2t}$$

$所以，原方程的通解为:$

$$x=c_1e^{-1}+c_2e^{-5}+\frac{1}{21}e^{2t}.$$






$(14)x"+x=\sin t-\cos{2t};$

$解:先求其次微分方程x"+x=0的通解:$

$该特征方程为:$

$$\lambda^2+1=0$$

$即,特征根为:$

$$\lambda_1=i,\lambda_2=-i.$$

$所以，齐次线性微分方程的通解为:$

$$x=c_1\cos t+c_2\sin t.$$

$由于f(t)中出现两个不同的\beta,所以分两部分求解：$

$(1)对于方程x"+x=\sin t，其\beta=\pm i,\alpha=0，它是特征根，所以其通解形如:$

$$\tilde{x}=t(A\cos t+B\sin t).$$

$代入原方程，解得:$

$$A=-\frac{1}{2},B=0$$

$所以，$

$$\tilde{x}=-\frac{1}{2}t\cos t.$$

$(2)对于方程x"+x=-\cos {2t},其\beta=\pm 2i,\alpha=0，它不是特征根，所以其通解形如:$

$$\tilde{x}=A\cos {2t}+B\sin {2t}.$$

$代入原方程，解得:$

$$A=-\frac{1}{3},B=0$$

$所以，$

$$\tilde{x}=\frac{1}{3}\cos {2t}.$$

$综合以上，原方程的通解为：$

$$x=c_1\cos t+c_2\sin t-\frac{1}{2}t\cos t+\frac{1}{3}\cos {2t}.$$



$(16)x"+9=t\sin {3t};$

$解:先求其次微分方程x"+9=0的通解:$

$该特征方程为:$

$$\lambda^2+9=0$$

$即,特征根为:$

$$\lambda_1=3i,\lambda_2=-3i.$$

$设方程的一个特解为:$

$$\tilde{x}=t[(At+B)\cos {3t}+(Et+F)\sin {3t}].$$

$代入方程，得到：$

$$A=-\frac{1}{12},F=\frac{1}{36},B=E=0,$$

$即方程的一个特解为:$

$$\tilde{x}=-\frac{1}{12}t^2\cos {3t}+\frac{1}{36}\sin {3t}.$$

$于是，方程的通解为:$

$$x=c_1\cos {3t}+c_2\sin {3t}-\frac{1}{12}t^2\cos {3t}+\frac{1}{36}\sin {3t},其中c_1,c_2为任意常数.$$



$4.求下组初值问题的解:$

$(1)x"+9x=6e^{3t},x(0)=x'(0)=0;$

$解：$

$对方程两边作拉普拉斯变换，得到：$

$$s^2X(s)-sx_0-x_0'+9X(s)=6\frac{1}{s-3}$$

$把x(0)=x'(0)=0代入上式，得：$

$$s^2X(s)+9X(s)=\frac{6}{s-3}$$

$因此：$

$$X(s)=\frac{1}{3}[\frac{1}{s-3}-\frac{s}{s^2+9}]-\frac{1}{s^2+9}.$$

$查拉普拉斯变换表，可得：$

$$x(t)=\frac{1}{3}e^{3t}-\frac{1}{3}\cos {3t}-\frac{1}{3}sin {3t}.$$

$$即解.$$



$(2）x^{(4)}+x=2e^t,x(0)=x'(0)=x"(0)=x"'(0)=1.$

$解：$

$对方程两边作拉普拉斯变换，得到：$

$$s^4X(s)-s^3x_0-s^2x_0'-sx_0"-x_0"'+X(s)=2\frac{1}{s-1}$$

$把x(0)=x'(0)=x"(0)=x"'(0)=1代入上式，得：$

$$s^4X(s)-s^3-s^2-s-1+X(s)=2\frac{1}{s-1}$$

$因此：$

$$X(s)=\frac{1}{s-1}$$

$查拉普拉斯变换表，可得：$

$$x(t)=e^t.$$

$$即解.$$
