它有根

\lambda_1=i,\lambda_2=-i

x=c_1\cos t+c_2\sin t

$$
\overline x=-\frac{1}{2}t\cos t

A=-\frac{1}{3},B=0
$$

即

$$
\overline x=-\frac{1}{3}t\cos 2t
$$

通解为

$$
x=c_1\cos t+c_2\sin t-\frac{1}{2}t\cos t+\frac{1}{3}\cos 2t

\overline x=-\frac{1}{12}t^2\cos 3t+\frac{1}{36}t\sin 3t

通解为

x=c_1\cos 3t+c_2\sin 3t-\frac{1}{12}t^2\cos 3t+\frac{1}{36t}\sin 3t
