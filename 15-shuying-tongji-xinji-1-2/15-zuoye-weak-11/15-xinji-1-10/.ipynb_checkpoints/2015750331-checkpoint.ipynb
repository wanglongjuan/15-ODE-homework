$一,求解下列常系数线性微分方程$

$1,$

$$x^{{\prime}{\prime}}+6x^{\prime}+5x=e^{2t}$$

$解,$

$ 对应齐次方程的特征方程为$

$$\lambda^2+6\lambda+5=0$$

$即得特征根为$

$$\lambda_1=-1$$

$$\lambda_2=-5$$

$令方程的一个特解为$

$$\overline{x}=Ae^2t$$

$代入方程则得到$

$$A=\frac{1}{21}$$

$即特解为$

$$\overline{x}=\frac {1}{21}e^2t$$

$于是方程的通解为$

$$x=c_1e^{-t}+c_2e^{-5t}+\frac {1}{21}e^{2t}$$

$2,$

$$\quad x^{{\prime}{\prime}}+x=sint-cos2t$$

$解,$

$\quad 对应的齐次方程的特征方程为$

$$\lambda^2+1=0$$

$即特征根为$

$$\lambda_1=i$$

$$\lambda_2=-i$$

$a\quad 对于方程$

$$x^{{\prime}{\prime}}+x=sint$$

$可以设其特解为$

$$\overline{x}=t(Asint+Bcost)$$

$代入方程则得到$

$$A=0，B=-\frac {1}{2}$$

$即方程$

$$x^{{\prime}{\prime}}+x=sint$$

$的一个特解为$

$$\overline{x}=-\frac {1}{2}tcost$$

$b\quad对于方程$

$$x^{{\prime}{\prime}}+x=-cos2t$$

$可以设其特解为$

$$\overline{x}=Acos2t+Bsin2t$$

$代入方程，则得到$

$$A=\frac{1}{3},B=0$$

$即方程$

$$x^{{\prime}{\prime}}+x=-cos2t$$

$的一个特解为$

$$\overline{x}=\frac{1}{3}cos2t$$

$于是原方程的通解为$

$$x=c_1cost+c_2sint-\frac {1}{2}tcost+\frac{1}{3}cos2t$$





$3$

$$\quad x^{{\prime}{\prime}}+9x=tsin3t$$

$解,$

$对应线性齐次方程的特征方程为$

$$\lambda^2+9=0$$

$即得到特征根为$

$$\lambda_1=3i$$

$$\lambda_2=-3i$$

$设方程的一个特解为$

$$\overline{x}=t[(At+B)cos3t+(Et+F)sin3t]$$

$代入方程得到$

$$A=-\frac{1}{12},F=\frac{1}{36},B=E=0$$

$即方程的一个特解为$

$$\overline{x}=-\frac{1}{12}t^2cos3t+\frac{1}{36}tsin3t$$

$于是方程的通解为$

$$x=c_1cos3t+c_2sin3t-\frac{1}{12}t^2cos3t+\frac{1}{36}tsin3t$$

$这里c_1,c_2是任意常数$$





$二, 求下组初值问题的解:$

$1,$

$$x^{{\prime}{\prime}}+9x=6e^{3t},x(0)=x^{\prime}(0)=0$$

$解,$

$对方程两边施行拉普拉斯变换得$

$$(s^2+9)X(s)=\frac{6}{s-3}$$

$由此得$

$$X(s)=\frac{1}{3}[\frac{1}{s-3}-\frac{s}{s^2+9}]-\frac{1}{s^2+9}$$

$对上式两端各项分别求出其原函数，则它们的和就是X(s)的原函数$

$$x(t)=\frac{1}{3}(e^{3t}-cos3t-sin3t)$$

$这就是所要求的解$



$2,$

$$\quad {x^{(4)}}+x=2e^t$$

$$x(0)=x^{\prime}(0)=x^{{\prime}{\prime}}(0)=x^{{\prime}{\prime}{\prime}}(0)=1$$

$解,$

$ 对方程两边施行拉普拉斯变换得$

$$s^4X(s)-s^3-s^2-s-1+X(s)=\frac{2}{s-1}$$

$由此得$

$$X(s)=\frac{1}{s-1}$$

$对上式右端求出其原函数，则X(s)的原函数为$

$$x(t)=e^t$$

$这就是所要求的解$
