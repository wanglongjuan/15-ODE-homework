## 二、证明格朗沃尔（Gronwall）不等式：
## 设K为非负数，f(t)和g(t)为在区间$\alpha\leq t\leq \beta$上的连续非负函数，且满足不等式
## $f(t)\leq K+\int_{\alpha}^t f(s)g(s)\mathrm ds,\alpha\leq t\leq \beta$

## $则有f(t)\leq Kexp(\int_{\alpha}^t g(s)\mathrm ds),\alpha\leq t\leq \beta$

## 并由此证明定理1的命题5.

## 证明：

1.)

$当K>0,令$

$\omega(t)=K+\int_\alpha^t f(s)g(s) \mathrm ds$

$则\omega'(t)=f(t)g(t)\leq g(t)\omega(t)$

$由\omega(t)>0得，\frac{\omega'(t)}{\omega(t)}\leq g(t)$

$两边从\alpha到t积分得$

$ln\omega(t)-ln\omega(\alpha)\leq \int_\alpha^t g(s)\mathrm ds$

$有\frac{\omega(t)}{\omega(\alpha)}\leq exp(\int_\alpha^t g(s) \mathrm ds)$

$\omega(\alpha)=K>0$

$则\omega(\alpha)\leq K exp(\int_\alpha^t g(s) \mathrm ds)$

$即f(t)\leq \omega(t)\leq K exp(\int_\alpha^t g(s) \mathrm ds),\alpha\leq t\leq \beta$

2.)

$当K=0,$

$对任意\varepsilon>0,由于f(t)\leq \int_\alpha^t f(s)g(s) \mathrm ds$

$所以f(t)\leq \varepsilon+\int_\alpha^t f(s)g(s) \mathrm ds$

$由1.)得，f(t)\leq \varepsilon exp(\int_\alpha^t g(s) \mathrm ds)$

$ 当\varepsilon \to 0~+时$

$f(t)\leq 0$

$又因为f(t)\geq 0$

$即得f(t)\equiv0$

$从而f(t)\leq K exp(\int_\alpha^t g(s) \mathrm ds),\alpha\leq t\leq \beta$

综上，即证得格朗沃尔不等式。

$设\varphi(t),\psi(t)$是初值问题$x'=f(t,x),x(t_0)=x_0的两个解，则有$

$\varphi(t)=x_0+\int_{t_0}^t f(\xi,\varphi(\xi)) \mathrm d\xi $

$\psi(t)=x_0+\int_{t_0}^t f(\xi,\psi(\xi)) \mathrm d\xi $

$则|\varphi(t)-\psi(t)|\leq \int_{t_0}^t |f(\xi,\varphi(\xi))-f(\xi,\psi(\xi))| \mathrm d\xi \leq L\int_{t_0}^t |\varphi(\xi)-\psi(\xi)| \mathrm d\xi $

其中L为利普希茨条件，由朗格沃尔不等式知

$0\leq |\varphi(t)-\psi(t)|\leq 0$

$所以\varphi(t)=\psi(t)$

即证得定理1的命题5.

## 三、如果函数$f(x,y)于带域\alpha\leq x\leq \beta$上连续且关于y满足利普希茨条件，则方程（3.1）满足条件$y(x_0)=y_0$的解于整个区间$[\alpha,\beta]$存在且唯一，试证明之。

## 证明：

$设M=max｜f(x,y_0)｜,h=β-α，b=Mh$

$及有常数L使得在域R=｛(x,y)|\alpha\leq x\leq \beta,|y-y_0|\leq b｝中有$

$|f(x,y_1)-f(x,y_2)|\leq L|y_1-y_2|,(x,y_1),(x,y_2)∈R$

构造逐步副近函数序列

$\left\{\begin{array}\ \varphi_0(x)=y_0\\ \varphi_0(x)=y_0+\int_{x_0}^x f(ζ,\varphi_{n-1}(ζ))dx,\alpha\leq x\leq \beta,n=1,2... \end{array}\right.$

$因 |\varphi_1(x)-y_0|=|\int_{x_0}^x f(ζ,y_0)dx|\leq \int_{x_0}^x |f(ζ,y_0)dx| \leq M(x-x_0)\leq Mh\leq b$

$可证 |\varphi_n(x)-\varphi_{n-1}(x)|\leq \frac{M}{n!}L^{n-1}(β-α)^n$


$|\varphi_{n+1}（x）-\varphi_n(x)|\leq |\int_{x_0}^x|f(ζ,\varphi_n(ζ))-f(ζ,\varphi_{n-1}(ζ))|dζ|\leq L|\int_{x_0}^x|\varphi_n(ζ)-\varphi_{n-1}(ζ)|dζ|\leq L|\int_{x_0}^x \frac{M}{n!}L^{n-1}|ζ-x_0|^ndζ|\leq \frac{M}{(n+1)!}L^n|x-x_0|^{n+1}\leq \frac{M}{(n+1)!}L^n(β-α)^{n+1}$

$因为级数\displaystyle\sum_{k=1}^{\infty}\frac{M}{k!}L^{k-1}(β-α)^k收敛$

$所以级数\varphi_n(x)=\varphi_0(x)+\displaystyle\sum_{k=1}^{\infty}[\varphi_k(x)-\varphi_{k-1}(x)]在区间[α,β]一致收敛$

$即存在极限\lim_{n \to \infty} \varphi_n(x)=\varphi(x)在区间[α,β]连续$

$且\varphi(x)=\lim_{n \to \infty}\varphi_n(x)=y_0+\lim_{n \to \infty}\int_{x_0}^xf(ζ,\varphi_{n-1}(ζ)dx=y_0+\int_{x_0}^xf(x,\lim_{n \to \infty}\varphi_{n-1}(\xi))dζ=y_0+\int_{x_0}^xf(\xi,\varphi(\xi))d\xi$

$所以解为\varphi(x)$

$设\varphi(ζ)，\psi(x)均满足方程y=y_0+\int_{x_0}^xf(ζ,y(ζ))dζ,\alpha≤ x_0≤ \beta,\alpha≤ x≤ \beta$

$当x≥x_0时$

$用格朗沃尔不等式$

$由|\varphi(ζ)-\psi(x)|≤\int_{x_0}^x|f(s,\varphi(s))-f(s,\psi(s))|ds≤L\int_{x_0}^x|\varphi(s)-\psi(s)|ds$

$得|\varphi(s)-\psi(s)|≤0，即\varphi(x)\equiv \psi(x)(\alpha≤ x_0≤ x≤ \beta)$

$同理，当\alpha≤ x_0≤ x≤ \beta时,\varphi(x)\equiv \psi(x)$

$即解是唯一的。$

## 四、给定积分方程
## $\varphi(x)=f(x)+\lambda\int_a^b K(x,\xi)\varphi(\xi)\mathrm d\xi \quad(*)$
## 其中f(x)是[a,b]上的已知连续函数,$K(x,\xi)是a\leq x\leq b,a\leq \xi\leq b$上的已知连续函数。证明当|$\lambda$|足够小时（$\lambda是常数$）,（ * ）在[a,b]上存在唯一的连续解。

## 证明：

$构造逼近函数序列，设\varphi_0(x)=f(x)$

$令\varphi_1(x)=f(x)+\lambda\int_a^bK(x,\xi)\varphi_0(ζ)dζ,......\varphi_{n+1}=f(x)+\lambda\int_a^bK(x,\xi)\varphi_n(ζ)dζ...$

$令M=\max_{\substack{
x\in[a,b] \\
}}
\mid f(x) \mid \quad L=\max_{\substack{
x\in [a,b] \\
\xi\in[a,b]
}}
\mid K(x,\xi)\mid $

$则有|\varphi_1(x)-\varphi_0(x)|=|\lambda \int_a^b K(x,\xi)\varphi_0(\xi)d\xi|\leq|\lambda||\int_a^bK(x,\xi)||f(\xi)|d\xi$

$设|\varphi_k(x)-\varphi_{k-1}(x)|\leq|\lambda|^kL^kM(b-a)^{k}成立$

$有|\varphi_{k+1}(x)-\varphi_k(x)|=|\lambda\int_a^bK(x,\xi)(\varphi_k(\xi)-\varphi_{k-1}(\xi))d\xi|\leq |\lambda|\int_a^b|K(x,\xi)||\varphi_k(\xi)-\varphi_{k-1}(\xi))|d\xi\leq|\lambda|\int_a^bL^{k+1}|\lambda|^kM(b-a)^kd\xi=|\lambda|^{k+1}L^{k+1}M(b-a)^{k+1}$

$由数学归纳法知，对任意的n\in N$

$有|\varphi_n(x)-\varphi_{n-1}(x)|\leq ML^n|\lambda|^n(b-a)^n$

$证｛\varphi_n(x)｝在[a,b]上一致收敛，考虑级数\varphi_0(x)+\displaystyle\sum_{k=1}^{\infty}(\varphi_k(x)-\varphi_{k-1}(x))$

$即只需证明该级数在[a,b]上致收敛$

$因为|\varphi_k(x)-\varphi_{k-1}(x)|\leq ML^k|\lambda|^k(b-a)^k$

$当|\lambda|<\frac{1}{(b-a)L}时$

$级数\displaystyle\sum_{k=1}^{\infty} ML^k|\lambda|^k(b-a)^k收敛$

$由此可知当|\lambda|<\frac{1}{(b-a)L}时$

$级数\varphi_0(x)+\displaystyle\sum_{k=1}^{\infty}(\varphi_k(x)-\varphi_{k-1}(x))在[a,b]上一致收敛$

$即｛\varphi_n(x)｝在[a,b]上一致收敛$

$设其极限函数为\varphi^*(x)$

$即当x\in [a,b]时，\varphi_n(x)当n \to\infty时一致收敛于\varphi^*(x)$

$且由\varphi_n(x)的连续性可知\varphi^*(x)在[a,b]上连续$

$对迭代列①，当n \to\infty时两边取极限$

$得\varphi^*(x)=f(x)+\lambda \int_a^b K(x,\xi)\varphi^*(x)d\xi(|\lambda|<\frac{1}{(b-a)L})$

$即\lambda|<\frac{1}{(b-a)L}当时，积分方程必存在连续解$

$证当\lambda|<\frac{1}{(b-a)L}时（*）方程的解的唯一性$

$设另有解\varphi'(x)且，即\varphi'(x)=f(x)+\lambda \int_a^b K(x,\xi)\varphi'(x)d\xi$

$记M'=\max_{\substack{
x\in [a,b] \\
}}
\mid \varphi^*(x)-\varphi'(x)\mid>0$

$\mid \varphi^*(x)-\varphi'(x)\mid=\mid\lambda \int_a^b K(x,\xi)(\varphi^*(\xi)-\varphi'(\xi))\mid\leq|\lambda|\int_a^bL\mid\varphi^*(\xi)-\varphi'(\xi)\mid d\xi=\mid\lambda\mid M'L(b-a)$

$即有M'\leq\mid\lambda\mid M'L(b-a)$

$由于M'>0,所以\mid\lambda\mid\geq\frac{1}{(b-a)L}与|\lambda|<\frac{1}{(b-a)L}矛盾$

$所以解唯一$
