10.给定积分方程$$φ(x)=f(x)+λ\int_a^b\mathrm K(x, ζ) ψ(ζ)\,\mathrm{d}ζ（1）$$

$f(x)是[a,b]上的已知连续函数，K(x, ζ)是a≤x≤,a≤ζ≤b上的已知连续函数.证明当|λ|足够小时（λ是常数），（1）在[a,b]上存在唯一的连续解.$

$解：做逐步逼近函数数列，取φ_0=f(x)，$

令$$ψ_1=f(x)+ λ\int_a^b\mathrm K(x, ζ) ψ_0(ζ)\,\mathrm{d}ζ,…\\ψ_{n+1}(x)=f(x)+ λ\int_a^b\mathrm K(x, ζ) ψ_n(ζ)\,\mathrm{d}ζ,…$$

$令M=\max \limits_{x\in[a,b]}\{|f(x)|\},L=\max \limits_{x\in[a,b]}\{K(x, ζ)\}>0$

$则|ψ_1(x)-ψ_0(x)|=| λ\int_a^b\mathrm K(x, ζ) ψ_0(ζ)\,\mathrm{d}ζ|≤|λ|\int_a^b\mathrm |K(x, ζ)|| f(ζ)| \,\mathrm{d}ζ≤|λ|ML(b-a),$

$假设|ψ_k(x)-ψ_{k-1}(x)| ≤|λ|kMLk(b-a)k成立，则有$

$|ψ_{k+1}(x)-ψ_k(x)|= | λ\int_a^b\mathrm K(x, ζ) (ψ_k(ζ)- ψ_{k-1}(ζ))\,\mathrm{d}ζ|≤| λ|\int_a^b\mathrm L^{k+1}| λ|^kM(b-a)^k\,\ {d}ζ=| λ|^{k+1} L^{k+1}M(b-a)^{k+1}$

由数学归纳法知，对任意的n∈N有$$|ψ_n(x)-ψ_{n-1}(x)| ≤|λ|^nML^n(b-a)^n.$$

$因为|ψ_k(x)-ψ_{k-1}(x)| ≤|λ|^kML^k(b-a)^k，所以当|λ|<\frac{1}{(b-a)L}时，$级数$∑_(k=1)^∞ML^k |λ|^k (b-a)^k收敛$，$由此知当|λ|<\frac{1}{(b-a)L}时，级数ψ_0(x)+∑_(k=1)^∞(ψ_k(x)- ψ_{k-1}(x))在[a,b]上一致收敛，即{ψ_n(x)}在[a,b]上一致收敛，设其极限函数为ψ*(x)，即当x∈[a,b]时，ψ_n(x)当n→∞时一致收敛于ψ*(x)，且由ψ_n(x)的连续性可知ψ*(x)在[a,b]上连续.$
