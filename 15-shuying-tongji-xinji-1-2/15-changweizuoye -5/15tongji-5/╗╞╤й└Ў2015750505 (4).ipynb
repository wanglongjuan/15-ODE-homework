$10.给定积分方程φ（x)=f(x)+λ\int_a^b\mathrm K(x,ε)φ_n(ε)\,\mathrm{d}ε,其中f(x)是[a,b]上的已知连续函数，K(x,ε)是a≤x≤b，a≤ε≤b上的
已知连续函数。证明当|λ|足够小时，（*）在[a,b]上存在唯一的连续解.$

$证明$

$作逐步逼近函数序列，取φ_0(x)=f(x),$

$$令φ_1(x)=f(x)+λ\int_a^b\mathrm K(x,ε)φ_0(ε)\,\mathrm{d}ε...,$$

$$φ_{n+1}(x)=f(x)+λ\int_a^b\mathrm K(x,ε)φ_n(ε)\,\mathrm{d}ε...,$$


$令M=max_{x \in [a,b]} \mid f(x) \mid ,L=max \mid k(x,\varsigma) \mid >0$

$$\begin{aligned}
\mid \phi_1(x)-\phi_0(x) \mid &=\mid \lambda \int_a^b k(x,\varsigma) \phi_0(\varsigma)d\varsigma \mid\\\
& \leq \mid \lambda \mid \int_a^b \mid k(x,\varsigma) \mid \  \mid f(\varsigma) \mid d\varsigma\\\
& \leq \mid \lambda \mid ML(b-a)
\end{aligned}$$

$假设$

$$\mid \phi_k(x)-\phi_{k-1}(x)\mid \leq {\mid \lambda \mid }^kL^kM(b-a)^k$$

$成立，则有$

$$\begin{aligned}
\mid \phi_{k+1}(x)-\phi_k(x)\mid &=\mid \lambda \int_a^b k(x,\varsigma) (\phi_k(\varsigma)-\phi_{k-1}(\varsigma))d\varsigma \mid\\\
&\leq \mid \lambda \mid \int_a^b \mid k(x,\varsigma) \mid \  \mid\phi_k(\varsigma)-\phi_{k-1}(\varsigma) \mid d\varsigma\\\
&\leq \int_a^b L^{k+1}{\mid \lambda \mid }^kM(b-a)^k d\varsigma\\\
&={\mid \lambda \mid }^{k+1}L^{k+1}M(b-a)^{k+1}
\end{aligned}$$

$由归纳法可知，对任意的$n \in N$有$

$$\mid \phi_n(x)-\phi_{n-1}(x)\mid \leq {\mid \lambda \mid }^nL^nM(b-a)^n$$

$$证一致收敛：$$

$考虑到级数$

$$\phi_0(x)+\sum_{k=1}^\infty(\phi_k(x)-\phi_{k-1}(x))$$

$级数的前n项和为：$

$$\phi_0(x)+\sum_{k=1}^n(\phi_k(x)-\phi_{k-1}(x))=\phi_n(x)$$

$要证{\phi_n(x)}在[a,b]上一致收敛，则证明$

$$\phi_0(x)+\sum_{k=1}^\infty(\phi_k(x)-\phi_{k-1}(x))$$

$在[a,b]上也一致收敛.$

$由$

$$\mid \phi_k(x)-\phi_{k-1}(x)\mid \leq {\mid \lambda \mid }^kL^kM(b-a)^k$$

$可知：$

$$当\mid \lambda \mid<\frac{1}{(b-a)L}时，级数\sum{\mid \lambda \mid }^kL^kM(b-a)^k收敛$$

$则\mid \lambda \mid<\frac{1}{(b-a)L}时,\phi_0(x)+\sum_{k=1}^\infty(\phi_k(x)-\phi_{k-1}(x))在[a,b]上一致收敛$

$设极限函数为：$

$$\phi^*(x)。$$

$即x \in [a,b]时，当n→\infty 时，\phi_n(x)一致收敛于\phi^*(x),且由\phi_n(x)的连续性可得\phi^*(x)在[a,b]上连续$

$对于\phi_{n+1}(x)=f(x)+\lambda \int_a^b k(x,\varsigma) \phi_n(\varsigma)d\varsigma，当n→\infty 时，两边取极限得$

$$\phi^*(x)=f(x)+\lambda \int_a^b k(x,\varsigma) \phi^*(\varsigma)d\varsigma$$

$$(\mid \lambda \mid<\frac{1}{(b-a)L})$$

$说明当\mid \lambda \mid<\frac{1}{(b-a)L}时，积分方程存在连续解.$

$$  $$

$$证解的唯一性$$

$用反证法：$

$设另有解\phi'^*(x)\neq \phi^*(x)所以有$

$$\phi'^*(x)=f(x)+\lambda \int_a^b k(x,\varsigma) \phi'^*(\varsigma)d\varsigma$$

$$Q=max_{x\in [a,b]}\mid \phi^*(x)-\phi'^*(x)\mid>0$$

$$\begin{aligned}
\mid \phi^*(x)-\phi'^*(x)\mid &=\mid \lambda \int_a^b k(x,\varsigma) (\phi^*(\varsigma)-\phi'^*(\varsigma))d\varsigma \mid\\\
&\leq \mid \lambda \mid \int_a^b L \mid\phi^*(\varsigma)-\phi'^*(\varsigma) \mid d\varsigma\\\
&=\mid \lambda \mid QL(b-a)
\end{aligned}$$

$因此，有$

$$Q\leq \mid \lambda \mid QL(b-a)$$

$由于Q>0,可知$

$$\mid \lambda \mid \geq \frac{1}{(b-a)L},与\mid \lambda \mid<\frac{1}{(b-a)L}$矛盾$$

$$即证$$
