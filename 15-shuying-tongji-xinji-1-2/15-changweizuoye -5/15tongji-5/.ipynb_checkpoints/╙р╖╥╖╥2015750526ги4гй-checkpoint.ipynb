$8.如果函数f(x,y)于带域\alpha\leq x\leq\beta上连续且关于y满足利普希茨条件，则方程\frac{dy}{dx}=f(x,y)满足条件y(x_0)=y_0的解于整个区间[\alpha,\beta]上存在且唯一.试证明之.$
