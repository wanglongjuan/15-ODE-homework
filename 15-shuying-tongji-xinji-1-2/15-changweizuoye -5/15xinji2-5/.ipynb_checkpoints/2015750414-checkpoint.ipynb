**6.证明格朗沃日不等式：设K是非负常数，f(t)和g(t)为在区间α≤t≤β上的连续非负函数，且满足不等式$$f(t)≤K+\int_{α}^t f(s)g(s) d s,α≤t≤β$$
则有$$f(t)≤Kexp(\int_{α}^t g(s) d s),α≤t≤β$$并由此证明定理1的命题5$ **

证明：(1)$$K>0时，令w(t)=K+\int_{α}^t f(s)g(s) d s$$则$$w'(t)=f(t)g(t)≤g(t)w(t)$$

由$$w(t)>0得\frac{w'(t)}{w(t)}≤g(t)$$两边从α到t积分得$$lnw(t)-lnw(α)≤\int_{α}^t g(s) d s$$

即有$$\frac{w(t)}{w(α)}≤exp(\int_{α}^t g(s) d s),w(α)=K>0$$

所以$$w(t)≤K exp(\int_{α}^t g(s) d s)$$即有$$f(t)≤w(t)≤K exp(\int_{α}^t g(s) d s),α≤t≤β$$

(2)$$K=0时$$对任意ε>0,由于$$f(t)≤\int_{α}^t f(s)g(s) d s$$所以$$f(t)≤ε+\int_{α}^t f(s)g(s) d s$$

由(1)可得$$f(t)≤εexp(\int_{α}^t g(s) d s)$$当ε趋向于0时，有$$f(t)≤0$$

又因为$$f(t)≥0$$即得$$f(t)≡0$$从而$$f(t)≤K exp(\int_{α}^t g(s) d s),α≤t≤β$$

由（1）（2）得，格朗沃日不等式成立

**证明定理1的命题5：**

设φ(t),ψ(t)是初值问题$$x'=f(t,x),x(t_0)=x_0$$的两个解，则有

$$φ(t)=x_0+\int_{t_0}^t f(ξ,φ(ξ)) d ξ,ψ(t)=x_0+\int_{t_0}^t f(ξ,ψ(ξ)) d ξ$$

于是$$|φ(t)-ψ(t)|≤\int_{t_0}^t |f(ξ,φ(ξ))-f(ξ,ψ(ξ))| d ξ≤L\int_{t_0}^t |φ(ξ)-ψ(ξ)| d ξ$$

由格朗沃日不等式知K=0且$$0≤|φ(t)-ψ(t)|≤K exp(\int_{t_0}^t L d ξ)=0$$

**$$8.如果函数f(x,y)于带域\alpha≤x≤\beta上连续且关于y满足利普希茨条件，则方程（3.1)满足条件y(x_0)=y_0的解于整个区间[\alpha,\beta]上存在且唯一。试证明。$$
$$（提示：用逐步逼近法，取M=max|f(x,y_0)|)$$**

解：设$$M=max|f(x,y_0)|,h=β-α,b=Mh$$及有某常数L使得在域$$R={(x,y)|α≤x≤β,|y-y_0|≤b}$$中有**

$$|f(x,y_1)-f(x,y_2)|≤L|y_1-y_2|,(x,y_1),(x,y_2)∈R$$

构造逐步逼近函数序列
$$\begin{cases}
φ_0(x)=y_0\\
φ_n(x)=y_0+\int_{x_0}^x f(ξ,φ_{n-1}(ξ)) d x
\end{cases}
$$
$$α≤x≤β,n=1,2,…$$

因为$$|φ_1(x)-y_0|=|\int_{x_0}^x f(ξ,y_0) d x|≤\int_{x_0}^x |f(ξ,y_0)| d x≤M(x-x_0)≤Mh≤b$$

可证：$$|φ_n(x)-φ_{n-1}(x)|≤\frac{ML^{n-1}(β-α)^n}{n!}$$

又$$|φ_{n+1}(x)-φ_n(x)|≤|\int_{x_0}^x |f(ξ,φ_n(ξ))-f(ξ,φ_{n-1}(ξ))| d ξ|≤\frac{ML^n(β-α)^{n+1}}{(n+1)!}$$

易知$$φ_n(x)=φ_0(x)+∑[φ_k(x)-φ_{n-1}(x)]$$在区间[α,β]一致收敛，于是存在极限φ(x)

$$φ(x)=y_0+\int_{x_0}^x f(ξ,φ(ξ)) d ξ$$所以φ(x)是微分方程的解

设φ(x)，ψ(x)均满足积分方程$$y=y_0+\int_{x_0}^x f(ξ,φ(ξ)) d ξ,α≤x_0≤β,α≤x≤β$$

当x≥x_0时，有格朗沃日不等式，有$$|φ(x)-ψ(x)|≤\int_{x_0}^x |f(s,φ(s))-f(s,ψ(s))| d s≤L\int_{x_0}^x |φ(x)-ψ(x)| d s$$

有$$|φ(x)-ψ(x)|≤0$$,即$$φ(x)≡ψ(x)（α≤x_0≤x≤β）$$同理可证$$α≤x≤x_0≤β$$时有$$φ(x)≡ψ(x)$$

即积分方程也就是微分方程的解是唯一。

**10.给定积分方程$$φ(x)=f(x)+λ \int_a^b \mathrm K(x,ε) \mathrm φ(x) \mathrm dε4...................(*)$$
其中f(x)是[a,b]上的已知连续函数，K(x,ε)是a≤x≤b,a≤ε≤b上的已知连续函数，证明当|λ|足够小时（λ是常数），(*)在[a,b]上存在唯一的连续解。**

证明：作逐步逼近函数序列，取$$φ_0(x)=f(x),令φ_1(x)=f(x)+λ\int_{a}^b K(x,ξ)φ_0(ξ) d ξ,…$$

$$φ_{n+1}(x)=f(x)+λ\int_{a}^b K(x,ξ)φ_n(ξ) d ξ,…$$

令$$M=max|f(x)|,L=max|K(x,ξ)|>0$$则$$|φ_1(x)-φ_0(x)|=|λ\int_{a}^b K(x,ξ)φ_0(ξ) d ξ|≤|λ|\int_{a}^b |K(x,ξ)||f(ξ)| d ξ≤|λ|ML;(b-a)$$

假设$$|φ_{k+1}(x)-φ_k(x)|=|λ\int_{a}^b K(x,ξ)(φ_k(ξ)-φ_{k-1}(ξ)) d ξ|≤|λ|^{k+1}L^{k+1}M(b-a)^{k+1}$$

由数学归纳法，对任意的n∈N,有$$|φ_n(x)-φ_{n-1}(x)|≤ML^n|λ|^n(b_a)^n$$

接下来证明$${φ_n(x)}在[a,b]$$上一致收敛

由于$$|φ_k(x)-φ_{k-1}(x)|≤ML^k|λ|^k(b-a)^k$$

当$$|λ|<\frac{1}{(b-a)L}$$时，级数$$∑ML^k|λ|^k(b-a)^k$$收敛

所以，当$$|λ|<\frac{1}{(b-a)L}$$时，级数$$φ_0(x)+∑ML^k|λ|^k(b-a)^k$$在[a,b]上一致收敛,即{φ_n(x)}在[a,b]上一致收敛

$设其极限函数为φ'(x),即当x∈[a,b]时，φ_n(x)当n→∞时一致收敛于φ'(x)，且由φ_n(x)的连续性可知φ'(x)在[a,b]上连续$

所以当$$|λ|<\frac{1}{(b-a)L}$$时，积分方程必存在连续解

设另有解φ"(x)且$$φ'(x)≠φ"(x)$$即$$φ"(x)=f(x)+λ\int_{a}^b K(x,ξ)φ"(x) d ξ$$

设$$M'=max|φ'(x)-φ"(x)|>0$$

$$|φ'(x)-φ"(x)|=|λ\int_{a}^b K(x,ξ)(φ'(x)-φ"(x)) d ξ|≤|λ|\int_{a}^b L|φ'(x)-φ"(x)| d ξ=|λ|M'L(b-a)$$

即有$$M'≤|λ|M'L(b-a)$$因为$$M'>0$$

所以$$|λ|≥\frac{1}{(b-a)L}$$与$$|λ|<\frac{1}{(b-a)L}$$矛盾，所以唯一性成立

综上，证毕
