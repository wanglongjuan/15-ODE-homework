P88.3、求初值问题
$$\begin{cases}\frac{dy}{dx}=x^2-y^2,R:|x+1|\le1,|y|\le1\\y(-1)=0\end{cases}$$的解的存在区间，并求第二次近似解，给出在解的存在区间的误差估计。

解:由已知可以得到$M=max|f(x,y)|=4$,此时有$(x,y)\in R$,h是$a=1$及$\frac{b}{M}=\frac{1}{4}$中的最小者，

故有$h=\frac{1}{4}$,在R上函数$f(x,y)=x^2-y^2$的利普希茨常数可取为$|\frac{\partial f}{\partial y}|=|-2y|\le2=L$

$\phi_0(x)=0$

$\phi_1(x)=y_0+\int ^x_{x_0}f(x,\phi_0(x))dx$

$=0+\int ^x_{-1}f(x,0))dx$

$=\int ^x_{-1}x^2dx$

$=\frac{1}{3}x^3|^x_{-1}$

$=\frac{1}{3}x^3+\frac{1}{3}$
               
$\phi_2(x)=y_0+\int ^x_{x_0}f(x,\phi_1(x))dx$

$=\int ^x_{-1}f(x,\phi_1(x))dx$
         
$=\int ^x_{-1}f(x,\frac{1}{3}x^3+\frac{1}{3})dx$
         
$=\int ^x_{-1}(x^2-(\frac{1}{3}x^3+\frac{1}{3})^2)dx$
         
$=\int ^x_{-1}(x^2-\frac{1}{9}x^6-\frac{2}{9}x^3+\frac{1}{9})dx$
         
$=(\frac{1}{3}x^3-\frac{1}{63}x^7-\frac{1}{18}x^4-\frac{1}{9})|^x_{-1}$
         
$=\frac{x^3}{3}-\frac{x^7}{63}-\frac{x^4}{18}-\frac{x}{9}+\frac{11}{42}$

由$|\phi_2(x)-\phi(x)|\le\frac{4*2^2}{3!}*(\frac{1}{4})^3=\frac{1}{24}$

因此，已知初值问题的解的存在区间为$-\frac{5}{4}\le x\le -\frac{3}{4}$

第二次近似解为$y_2=\frac{x^3}{3}-\frac{x^7}{63}-\frac{x^4}{18}-\frac{x}{9}+\frac{11}{42}$

所以在解的存在区间的误差估计为$|y-y_2|\le\frac{1}{24}$

                      

6.证明格朗沃尔(Gronwall)不等式：

设K为非负常数，$f(t)$和$g(t)$为在区间$\alpha\le t\le\beta$上的连续非负整数，且满足不等式

$$f(t)\le K+\int ^t_\alpha f(s)g(s)ds,\alpha\le t\le\beta$$

则有$$f(t)\le Kexp(\int ^t_\alpha g(s)ds),\alpha\le t\le\beta$$
并由此证明定理1的命题5.




证明：(1)K>0时，令$$\omega(t)=K+\int ^t_sf(s)g(s)ds$$

则可以得到$\omega'(t)=f(t)g(t)\le g(t)\omega(t)$,由$\omega(t)>0$

$$\frac{\omega'(t)}{\omega(t)}\le g(t)$$

对上面的不等式两边从$\alpha$到t进行积分得到

$$ln\omega(t)-ln\omega(\alpha)\le\int^t_\alpha g(s)ds$$

即$$\frac{\omega(t)}{\omega(\alpha)}\le exp(\int^t_\alpha g(s)ds),\omega(\alpha)=K>0$$

所以$$\omega(t)\le K*exp(\int^t_\alpha g(s)ds)$$

即就是$$f(t)\le\omega(t)\le Kexp(\int^t_\alpha g(s)ds,\alpha\le t\le\beta$$

(2)当$K=0$时，对任意$\varepsilon>0$,由于$f(t)\le\int^t_\alpha f(s)g(s)ds$,所以$f(t)\le\varepsilon+\int^t_\alpha f(s)g(s)ds$

由（1）可知$f(t)\le\varepsilon*exp(\int^t_\alpha g(s)ds)$,当$\varepsilon\rightarrow 0^+时，$\f(t)\le 0$

又因为$0\le f(t)$,得到$f(t)=0$,因此得到$$f(t)\le Kexp(\int^t_\alpha g(s)ds,\alpha\le t\le\beta$$

综上可以得到格朗沃尔不等式成立

证明定理1的命题5

设$\varphi(t),\psi(t)$是初值问题$$x'=f(t,x),x(t_0)=x_0$$的两个解，则有
$$\varphi(t)=x_0+\int^t_{t_o}f(\varepsilon,\varphi(\varepsilon))d\varepsilon$$
$$\psi(t)=x_0+\int^t_{t_o}f(\varepsilon,\psi(\varepsilon))d\varepsilon$$

那么可得$$|\varphi(t)-\psi(t)|\le\int^t_{t_0}|f(\varepsilon,\varphi(\varepsilon))-f(\varepsilon,\psi(\varepsilon))|d\varepsilon\le L\int^t_{t_0}|\varphi(\varepsilon)-\psi(\varepsilon)|d\varepsilon$$

其中L为利普希茨条件，由格朗沃尔不等式知$$0\le|\varphi(t)-\psi(t)|\le 0$$

因此得到$\varphi(t)=\psi(t)$即可证得定理1的命题5成立。


8.如果函数$f(x,y)$于带域$\alpha\le x\le\beta$上连续且关于y满足利普希茨条件，则方程（3.1）满足条件$y(x_0)=y_0$的解于整个区间[$\alpha,\beta$]上存在且唯一.试证明之

（提示：用逐步逼近法，取$M=max|f(x,y_0)|$有$x\in$[$\alpha,\beta$]

证明：要证明方程（3.1）满足条件$y(x_0)=y_0$的解于在整个区间[$\alpha,\beta$]上存在且唯一，

即就是要证明积分方程的解的存在唯一性即可。

取$$y_0(x)=y_0$$,

$$y_1(x)=y_o+\int^x_{x_0}f(\varepsilon,y_0(\varepsilon))d\varepsilon$$

$$\cdots$$

$$y_n(x)=y_0+\int^x_{x_0}f(\varepsilon,y_{n-1}(\varepsilon))d\varepsilon$$

$$\cdots (1)$$

显然$y_n(x)$在$\alpha\le x\le\beta$上存在且连续.

然后证明函数序列${y_n(x)}$在$\alpha\le x\le\beta$上一致收敛

考虑级数$$y_0(x)+\sum\limits^\infty_{k=1}[y_k(x)-y_{k-1}(x)],x\in[\alpha,\beta] (2)$$

取$M=max_{x\in[\alpha,\beta]}|f(x,y_0)|$,由(1)可知$$|y_1(x)-y_0(x)|\le\int^x_{x_0}|f(\varepsilon,y_0(\varepsilon))|d\varepsilon\le M|x-x_0|(3)$$
及$$|y_2(x)-y_1(x)|\le\int^x_{x_0}|f(\varepsilon,y_1(\varepsilon))-f(\varepsilon,y_0(\varepsilon))|d\varepsilon$$

由利普希茨条件及（3）式，得到$$|y_2(x)-y_1(x)|\le L\int^x_{x_0}|y_1(\varepsilon)-y_0(\varepsilon)|d\varepsilon\le L\int^x_{x_0}M|\varepsilon-x_0|d\varepsilon=\frac{ML}{2!}(x-x_0)^2$$

由数学归纳法得到
 
|$y_k(x)-y_{k-1}(x)$| $\le$ $\frac{ML^{k-1}}{k!}$ $|x-x_0|^k$ $\le$ $\frac{ML^{k-1}}{k!}$$(\alpha-\beta)^k$

上式右边是正项收敛级数$\sum\limits^\infty_{k=1}$ $\frac{ML^{k-1}}{k!}$ $(\alpha-\beta)^k$的一般项.根据Weierstrass判别法，
可知级数（2）在$\alpha\le x\le\beta$上一致收敛，

因而函数序列{$y_n(x)$}在$\alpha$$\le$ x$\le\beta$上一致收敛.

设$\lim_{n\rightarrow\infty}y_n(x)=y(x)$,易知$y(x)$在[$\alpha,\beta$]上连续，

且${f(x,y_n(x))}$在[$\alpha$,$\beta$]一致收敛于$f(x,y(x))$.

所以对（1)两边取极限，即得$y(x)=y_o+\int^x_{x_0}f(\varepsilon,y(\varepsilon))d\varepsilon$,解的存在性即可证得.

唯一性证明，由上面可得到$\frac{ML^{k-1}}{k!}$$(\alpha-\beta)^k$=$\frac{ML^n}{(n+1)!}$$h^{n+1}$是收敛级数的公项，故$n\rightarrow\infty$时，

$\frac{ML^n}{(n+1)!}h^{n+1}\rightarrow 0$因而${f(x,y_n(x))}$在[$\alpha,\beta$]满足唯一性，

即$y_k(x)=y_{k-1}(x),\alpha\le x\le\beta$证得唯一性


10.给定积分方程$$\varphi(x)=f(x)+\lambda\int^b_a K(x,\varepsilon)\varphi(\varepsilon)d\varepsilon (*)$$

其中$f(x)$是[a,b]上的已知连续函数，$K(x,\varepsilon)是$a$\le$ x$\le$ b$上的已知连续函数。证明当$|$\lambda$|足够小时

（$\lambda$是常数），（*）在[a,b]上存在唯一的连续解。

（提示：作逐步逼近函数序列
$$\varphi_0(x)=f(x)$$

$$\varphi_{n+1}(x)=f(x)+\lambda\int^b_a K(x,\varepsilon)\varphi(\varepsilon)d\varepsilon(n=0,1,2,\cdots))$$




证明：作逐步逼近序列，取
$$\varphi_0(x)=f(x)$$

$$\varphi_1(x)=f(x)+\lambda\int^b_a k(x,\varepsilon)\varphi_0(\varepsilon)d\varepsilon$$

$$\cdots$$

$$\varphi_{n+1}(x)=f(x)+\lambda\int^b_a k(x,\varepsilon)\varphi_n(\varepsilon)d\varepsilon$$

$$\cdots (1)$$

令$M=max_{x\in[a,b]}|f(x)|$,
$L=max_{x\in[a,b],\varepsilon\in[a,b]}|k(x,\varepsilon)|>o$

$$|\varphi_1(x)-\varphi_0(x)|=|\lambda\int^b_a k(x,\varepsilon)\varphi_0(\varepsilon)d\varepsilon| \le |\lambda|\int^b_a |k(x,\varepsilon)||f(\varepsilon)|d\varepsilon \le |\lambda|ML(b-a)$$

假设
$$|\varphi_k(x)-\varphi_{k-1}(x)|=|\lambda\int^b_a k(x,\varepsilon)(\varphi_k(\varepsilon)-\varphi_{k-1}(\varepsilon))d\varepsilon$$

$$\le |\lambda|\int^b_a|k(x,\varepsilon)||\varphi_k(\varepsilon)-\varphi_{k-1}(\varepsilon)|d\varepsilon$$

$$\le |\lambda|\int^b_a L^{k+1}|\lambda|^k M(b-a)^kd\varepsilon$$

$$=|\lambda|^{k+1}L^{k+1}M(b-a)^{k+1}$$,

由数学归纳法知，对任意的$n\in N$,有

$$|\varphi_n(x)-\varphi_{n-1}(x)| \le ML^n||\lambda|^n(b-a)^n  (2)$$

证明{$\varphi_n(x)$}在[a,b]上一致收敛。考虑级数

$$\varphi_0(x)+\sum\limits^\infty_{k=1}(\varphi_k(x)-\varphi_{k-1}(x))$$,

该级数的前n项和为$$\varphi_0(x)+\sum\limits^n_{k=1}(\varphi_k(x)-\varphi_{k-1}(x)）=\varphi_n(x)$$

所以要证序列{$\varphi_n(x)$}在[a,b]上一致收敛，只需要证明级数（2）在[a,b]上一致收敛即可。

由于$$|\varphi_k(x)-\varphi_{k-1}(x)| \le ML^k||\lambda|^k(b-a)^k,$$ 

当$|\lambda|<\frac{1}{(b-a)L}$时，级数$\sum\limits^\infty_{k=1}ML^k||\lambda|^k(b-a)^k$收敛，

由此可知当$|\lambda|<\frac{1}{(b-a)L}$时，级数（2）在[a,b]上一致收敛。

设极限函数为$\varphi^*(x)$,即$x\in[a,b]$时，当$n\rightarrow\infty$时，

$\varphi_n(x)$一致收敛于$\varphi^*(x)$，则由$\varphi_n(x)$的收敛性可知

$\varphi^*(x)$在[a,b]上连续，对于序列（1），

当$n\rightarrow\infty$时两边取极限，得到$$\varphi^*(x)=f(x)+\lambda\int^b_a k(x,\varepsilon)\varphi^*(\varepsilon)d\varepsilon$$

$$(|\lambda|<\frac{1}{(b-a)L})$$

即证明了当$|\lambda|<\frac{1}{(b-a)L}$时，积分方程必有连续解。

证明当$|\lambda|<\frac{1}{(b-a)L}$时，积分方程的解的唯一性，由反证法.

设另有一解为$\psi(x)$且$\psi(x)\neq\varphi^*(x)$,即就是

$$\psi(x)=f(x)+\lambda\int^b_a k(x,\varepsilon)\psi(\varepsilon)d\varepsilon$$

令$Q=max_{x\in[a,b]}|\varphi^*(x)-\psi(x)|>0$

$$|\varphi^*(x)-\psi(x)|=|\lambda\int^b_a k(x,\varepsilon)(\varphi^*(\varepsilon)-\psi(\varepsilon))d\varepsilon$$

$$\le |\lambda|\int^b_a L|\varphi^*(\varepsilon)-\psi(\varepsilon)|d\varepsilon$$
    
$$=|\lambda|QL(b-a)$$,

即有$Q \le|\lambda|QL(b-a)$,由于$Q>0$,所以$\frac{1}{(b-a)L} \le|\lambda|$,与$|\lambda|<\frac{1}{(b-a)L}$矛盾，因此可证得唯一性，
因此当$|\lambda|$足够小时（$\lambda$是常数），（*）在[a,b]上存在唯一的连续解。
