$\quad证明格朗沃尔（Gronwall)不等式：设K为非负常数，f(t)和g(t)为在区间\alpha\leq t\leq \beta上的连续非负函数，且满足不等式$$f(t)\leq K+\int_\alpha^t f(s)g(s)ds,\alpha\leq t \leq\beta$

$$则有f(t)\leq Kexp(\int_\alpha^t g(s)ds),\alpha\leq t \leq\beta.并由此证明定理1的命题5.$$


$$K>0时，令w(t)=K+\int_\alpha^t f(s)g(s)ds$$

$$w\prime(t)=f(t)g(t)\leq g(t)w(t)$$

$$w(t)>0    容易得\frac {w\prime(t)}{w(t)}\leq g(t)$$

$$积分得lnw(t)-lnw(\alpha)\leq\int_{\alpha}^{t}g(s)ds$$

$$\frac{w(t)}{w(\alpha)}\leq exp(\int_{\alpha}^{t}g(s)ds),w(\alpha)=K>0$$

$$所以w(t)\leq K*exp(\int_{\alpha}^{t}g(s)ds),有f(t)\leq w(t)\leq Kexp(\int_{\alpha}^{t}g(s)ds),\alpha\leq t\leq \beta$$

$$K=0时，任意\gamma>0,$$

$由于$

$$f(t)\leq\int_{\alpha}^{t}f{s}g(s)ds$$

$所以$

$$f(t)\leq\gamma+\int f(s)g(s)ds$$

$$又因为f(t)\leq\gamma exp(\int_{\alpha}^{t}g(s)ds).$$

$$当\gamma\rightarrow0时，有f(t)\leq 0$$

$$又因为f(t)\geq0,即得f(t)\equiv0 .$$

$$从而f(t)\leq Kexp(\int_{\alpha}^{t}g(s)ds),\alpha\leq t\leq \beta.$$


$$所以格朗沃尔不等式成立$$

$\quad给定积分方程\varphi(x)=f(x)+C\int_{a}^{b}K(x,\gamma)\varphi(\gamma)d\gamma,其中f(x)是[a,b]上的已知连续函数，
K(x,\gamma)是a\leq x\leq b,a\leq \gamma\leq b上的已知连续函数
证明当|C|足够小时（C是常数），上式在[a,b]上存在唯一的连续解$

$$\varphi_o(x)=f(x),令\varphi_1(x)=f(x)+C\int_{a}^{b}k(x,\gamma)\varphi_o(\gamma)d\gamma$$

$$\varphi_{n+1}(x)=f(x)+C\int_{a}^{b}k(x,\gamma)\varphi_n(\gamma)d\gamma$$

$$M=\max \limits_{x\in [a,b]}|f(x)|,L=\max\ limits_{\substack{x\in [a,b]\gamma\in[a,b]}}|k(x,\gamma)|>0$$

$$|\varphi_1(x)-\varphi_o(x)|\leq |C|ML(b-a)$$,

$$设|\varphi_k(x)-\varphi_{k-1}(x)|\leq|C|^kL^kM(b-a)^k成立$$

$$|\varphi_{k+1}(x)-\varphi_{k}(x)|\leq|C|^{k+1}L^{k+1}M(b-a)^{k+1}$$

$$所以对任意的n\in N,有$$

$$|\varphi_n(x)-\varphi_{n-1}(x)|\leq ML^n|C|^n(b-a)^n.$$

$证{\varphi_n(x)}在[a,b]上一致收敛$

$$\varphi_o(x)+\displaystyle\sum_{k=1}^{\infty}(\varphi_k(x)-\varphi_{k-1}(x))$$

$$\varphi_o(x)+\displaystyle\sum_{k=1}^{\infty}(\varphi_k(x)-\varphi_{k-1}(x))=\varphi_n(x).$$

$要证\varphi_n(x)在[a,b]上一致收敛，需证级数在[a,b]上一致收敛$

$$|\varphi_k(x)-\varphi_{k-1}(x)|\leq ML^k|C|^k(b-a)^k,当|C|<\frac {1}{(b-a)L}时$$

$$\displaystyle\sum_{k=1}^{\infty}ML^k|C|^k(b-a)^k收敛$$

$级数在[a,b]上一致收敛.$

$设极限函数为\varphi^*(x),当n\rightarrow_{\infty}时，\varphi_n(x)一致收敛于\varphi^*(x)$

$两边取极限得$

$$\varphi^*(x)=f(x)+C\int_{a}^{b}k(x,\gamma)\varphi^*(\gamma)d\gamma.(|C|<\frac {1}{(b-a)L})$$

$积分方程必存在连续解$

$证当|C|<\frac {1}{(b-a)L}时，方程解的唯一性$

$令有解$

$$\varphi^-(x) \neq \varphi^*(x)$$

$$\varphi^-(s)=f(x)+|C|\int_{a}^{b}k(x,\gamma)\varphi^-(\gamma)d\gamma$$

$$记Q=\max\limits_{x\in[a,b]}|\varphi^*(x)-\varphi^-(x)|>0$$

$$|\varphi^*(x)-\varphi^-(x)|\leq|C|\int_{a}^{b}L|\varphi^*(\gamma)-\varphi^-(\gamma)|d\gamma=|C|QL(b-a)$$

$即有Q\leq|C|QL(b-a),由于Q>0,所以|C|\leq\frac {1}{(b-a)L},与|C|<\frac {1}{(b-a)L}矛盾，唯一性可得证.$
