1 设$x(t)和y(t)是区间a\leq t\leq b上的连续函数$

$证明:如果在区间a\leq t\leq b上有\frac{x(t)}{y(t)}\neq 常数或\frac{y(t)}{x(t)}\neq 常数，则x(t)和y(t)在区间a\leq t\leq b上线性无关$

解 由题可知

利用反证法

假设$x(t)和y(t)在区间a\leq t\leq b上线性相关$

$$
则存在c_1和c_2(c_1,c_2不为0)
$$
使
$$
c_1x(t)+c_2y(t)=0
$$
可得
$$
\frac{x（t)}{y(t）}=-\frac{c_2}{c_1}
$$
与
$\frac{x(t)}{y(t)}\neq 常数矛盾$
因此
$x(t)和y(t)在区间a\leq t\leq b上线性无关$

2 $证明非齐次线性微分方程的叠加原理:设x_1(t),x_2(t)分别是非齐次线性微分方程$
$$
\frac{d^n{x}}{d{t^n}}+a_1(t）\frac{d^{n-1}{x}}{d{t^{n-1}}}+……+a_n(t）x=f_1(t)
$$
$$
\frac{d^n{x}}{d{t^n}}+a_1(t）\frac{d^{n-1}{x}}{d{t^{n-1}}}+……+a_n(t）x=f_2(t)
$$
的解，则$$x_1(t)+x_2(t)$$是方程
$$
\frac{d^n{x}}{d{t^n}}+a_1(t）\frac{d^{n-1}{x}}{d{t^{n-1}}}+……+a_n(t）x=f_1(t)+f_2(t)
$$
的解

解 由题可知
因$x_1(t)和x_2(t)是非齐次线性微分方程的解$
则
$$
\frac{d^n{x_1(t)}}{d{t^n}}+a_1(t）\frac{d^{n-1}{x_1(t)}}{d{t^{n-1}}}+……+a_n(t）x_1(t)=f_1(t)
$$
$$
\frac{d^n{x_2{t}}}{d{t^n}}+a_1(t）\frac{d^{n-1}{x_2(t)}}{d{t^{n-1}}}+……+a_n(t）x_2(t)=f_2(t)
$$
两式相加得
$$
\frac{d^n{x_1(t)}}{d{t^n}}+a_1(t）\frac{d^{n-1}{x_1(t)}}{d{t^{n-1}}}+……+a_n(t）x_1(t)+\frac{d^n{x_2{t}}}{d{t^n}}+a_1(t）\frac{d^{n-1}{x_2(t)}}{d{t^{n-1}}}+……+a_n(t）x_2(t)=f_1(t)+f_2(t)
$$
即为
$$
\frac{d^n{(x_1(t)+x_2(t))}}{d{t^n}}+a_1(t）\frac{d^{n-1}{(x_1(t)+x_2(t))}}{d{t^{n-1}}}+……+a_n(t）(x_1(t)+x_2(t))=f_1(t)+f_2(t)
$$
因此
$$x_1(t)+x_2(t)$$是方程
$$
\frac{d^n{x}}{d{t^n}}+a_1(t）\frac{d^{n-1}{x}}{d{t^{n-1}}}+……+a_n(t）x=f_1(t)+f_2(t)
$$
的解

3 已知齐次线性微分方程的基本解组$x_1,x_2$求下列方程对应的非齐次线性微分方程的通解

$$(1)x''-x=cost,x_1=e^t,x_2=e^{-t}$$
$$(3)x''+4x=tsint,x_1=cos2t,x_2=sin2t$$
$$(5)t^2x''-tx'+x=6t+34t^2,x_1=t,x_2=tlnt$$

解 （1）由题可知
令$$x=c_1(t)e^t+c_2(t)e^{-t}$$
将其代入方程，可得
$$
\begin{cases}
c'_1(t)e^t-c'_2(t)e^{-t}=cost,\\
c'_1(t)e^t+c'_2(t)e^{-t}=0
\end{cases}
$$
解得
$$
\begin{cases}
c'_1(t)=\frac{1}{2}coste^{-t},\\
c'_2(t)=-\frac{1}{2}coste^t
\end{cases}
$$
对其积分得
$$
c_1(t)=\frac{sint-cost}{4}e^{-t}+\gamma_1
$$
$$
c_2(t)=-\frac{sint+cost}{4}e^t+\gamma_2
$$
原方程的通解为
$$
x=\gamma_1e^t+\gamma_2e^{-t}-\frac{1}{2}cost
$$

(3)解 由题可知
令$$x=c_1(t)cos2t+c_2(t)sin2t$$
将其代入方程，可得
$$
\begin{cases}
-2c'_1(t)sin2t+2c'_2(t)cos2t=tsint,\\
c'_1(t)cos2t+c'_2(t)e^sin2t=0
\end{cases}
$$
解得
$$
\begin{cases}
c'_1(t)=-\frac{1}{2}tsintsin2t,\\
c'_2(t)=\frac{1}{2}tsintcos2t
\end{cases}
$$
对其积分得
$$
c_1(t)=\frac{1}{4}[\frac{1}{3}tsin3t+\frac{1}{9}cos3t-tsint-cost]+\gamma_1
$$
$$
c_2(t)=\frac{1}{4}[-\frac{1}{3}tsin3t+\frac{1}{9}cos3t+tcost-sint]+\gamma_2
$$
原方程的通解为
$$
x=\gamma_1cos2t+\gamma_2sin2t+\frac{t}{3}sint-\frac{2}{9}cost
$$

解 （5）由题可知
令$$x=c_1(t)t+c_2(t)tlnt$$
将其代入方程，可得
$$
\begin{cases}
c'_1(t)-c'_2(t)(1+lnt)=(\frac{6}{t}+34),\\
c'_1(t)t+c'_2(t)tlnt=0
\end{cases}
$$
解得
$$
\begin{cases}
c'_1(t)=\frac{6lnt}{t}-34lnt,\\
c'_2(t)=\frac{6}{t}+34
\end{cases}
$$
对其积分得
$$
c_1(t)=-34tlnt-3(lnt）^2+\gamma_1
$$
$$
c_2(t)=34t+6lnt+\gamma_2
$$
原方程的通解为
$$
x=\gamma_1t+\gamma_2tlnt+34t^2+3t(lnt)^2
$$

6.假设$x_1(t)\neq$0是二阶齐次线性微分方程组

$$x''+a_1(t)x'+a_2(t)x=0$$

的解，这里的$a_1(t)和a_2(t)$于区间[a,b]上连续，试证：

（1）$x_2(t)$为方程的解得充要条件是$W'[x_1,x_2]+a_1W[x_1，x_2]=0$;

(2)方程的通解可表为
$x=x_1[c_1 \int \mathrm {\frac{1}{x^2}exp(-\int_{t_0}^t \mathrm a_1(s)\,\mathrm {d}s)}\,\mathrm {d}t +c_2]$
$其中c_1,c_2为任意常数，t_0,t\in[a,b].$


解 由题可知
必要性
当$x_2(t)$为方程的解
则$$x_2(t)''+a_1(t)x_2(t)'+a_2(t)x_2(t)=0$$
$$
W'[x_1,x_2]=
\begin{vmatrix} 
x_1' & x_2' \\ 
x_1' & x_2'\\ 
\end{vmatrix} 
+\begin{vmatrix} 
x_1 & x_2 \\ 
x_1''& x_2''\\ 
\end{vmatrix} 
=\begin{vmatrix} 
x_1 & x_2 \\ 
x_1''& x_2''\\ 
\end{vmatrix} 
$$
所以

$$
W'[x_1,x_2]=
\begin{vmatrix} 
x_1 & x_2 \\ 
x_1''& x_2''\\ 
\end{vmatrix} 
$$

$$
=\begin{vmatrix} 
x_1 & x_2 \\ 
x_1''+a_2(t)x_1& x_2''+a_2(t)x_2\\ 
\end{vmatrix} 
$$

$$
=\begin{vmatrix} 
x_1 & x_2 \\ 
-a_1(t)x'_1& -a_1(t)x'_2\\ 
\end{vmatrix} 
$$

$$
=-a_1(t)\begin{vmatrix} 
x_1 & x_2 \\ 
x'_1&x'_2\\ 
\end{vmatrix} 
=-a_1W[x_1，x_2]
$$
因此
$W'[x_1,x_2]+a_1W[x_1，x_2]=0$
必要性得证

充分性
$$
W'[x_1,x_2]+a_1W[x_1，x_2]=
\begin{vmatrix} 
x_1 & x_2 \\ 
x_1''& x_2''\\ 
\end{vmatrix} +a_1(t)\begin{vmatrix} 
x_1 & x_2 \\ 
x'_1&x'_2\\ 
\end{vmatrix} 
$$

$$
=\begin{vmatrix} 
x_1 & x_2 \\ 
x_1''+a_1(t)x_1'& x_2''+a_1(t)x_2'\\ 
\end{vmatrix}=0 
$$

当$x_1(t)$为方程的解
则$$x_1(t)''+a_1(t)x_1(t)'+a_2(t)x_1(t)=0$$
即
$$x_1(t)''+a_1(t)x_1(t)'=-a_2(t)x_1(t)$$
$$
\begin{vmatrix} 
x_1 & x_2 \\ 
x_1''+a_1(t)x_1'& x_2''+a_1(t)x_2'\\ 
\end{vmatrix}
$$

$$=
\begin{vmatrix} 
x_1 & x_2 \\ 
-a_2(t)x_1(t)& x_2''+a_1(t)x_2'\\ 
\end{vmatrix}
$$

$$
=x_1\begin{vmatrix} 
1 & x_2 \\ 
-a_2(t）& x_2''+a_1(t)x_2'\\ 
\end{vmatrix}=0
$$

因此
$$x_2(t)''+a_1(t)x_2(t)'+a_2(t)x_2(t)=0$$
则
$x_2(t)$为方程的解
充分性得证

（2）解 由题可知
$$ 
\begin{vmatrix} 
x_1 & x_2 \\ 
x_1' & x_2'\\ 
\end{vmatrix} 
=w(t_0)e^{-\int_{t_0}^t \mathrm a_1(s)\,\mathrm d s}
$$ 
即
$$
x_1x_2'-x_1'x_2=w(t_0)e^{-\int_{t_0}^t \mathrm a_1(s)\,\mathrm d s}
$$

两边同时乘以$\frac{1}{x_1^2}$得
$$
\frac{x_1x_2'-x_1'x_2}{x_1^2}=\frac{w(t_0)e^{-\int_{t_0}^t \mathrm a_1(s)\,\mathrm d s}}{x_1^2}
$$

即得
$$
\frac{d(\frac{x_2}{x_1})}{dt}=\frac{w(t_0)e^{-\int_{t_0}^t \mathrm a_1(s)\,\mathrm d s}}{x_1^2}
$$

两边对t积分得
$$x_2=(c_1\int \mathrm {\frac{1}{x^2}e^{-\int_{t_0}^t \mathrm a_1(s)\,\mathrm d s}}\,\mathrm dt+c_2)x_1
$$

取
$c_1=1,c_2=0$
$$x_2=(\int \mathrm {\frac{1}{x^2}e^{-\int_{t_0}^t \mathrm a_1(s)\,\mathrm d s}}\,\mathrm dt)x_1
$$

$$ 
W(t)=
\begin{vmatrix} 
x_1 & x_2 \\ 
x_1' & x_2'\\ 
\end{vmatrix}
$$
$$
=e^{-\int_{t_0}^t \mathrm a_1(s)\,\mathrm d s}\neq 0
$$ 

从而原方程的通解为
$$
x=x_1[c_1 \int \mathrm {\frac{1}{x^2}exp(-\int_{t_0}^t \mathrm a_1(s)\,\mathrm {d}s)}\,\mathrm {d}t +c_2]$
$$
$其中c_1,c_2为任意常数$
