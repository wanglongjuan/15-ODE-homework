$1.设x(t)和y(t)是区间a\leq t\leq b上的连续函数，证明：如果在区间a\leq t\leq b上有\frac{x(t)}{y(t)}\neq常数或\frac{y(t)}{x(t)}\neq常数，则x(t)和y(t)在区间a\leq t\leq b上线性无关.（提示：用反证法.）$  
$证明：假设x(t),y(t)在区间a\leq t\leq b上线性相关$  
$则存在不全为零的常数\alpha,\beta,使得\alpha x(t)+\beta y(t)\equiv0,t\in[a,b].$  
$那么不妨设x(t)不为零，则有\frac{y(t)}{x(t)}=-\frac{\alpha}{\beta},显然-\frac{\alpha}{\beta}为常数，与题设矛盾,$  
$所以假设不成立，即证明了x(t),y(t)在区间a\leq t\leq b上线性无关.$  

$2.证明非齐次线性微分方程的叠加原理：设x_1(t),x_2(t)分别是非齐次线性微分方程$  
$$
\frac{d^nx}{dt^n}+a_1(t)\frac{d^{n-1}}{dt^{n-1}}+\cdots+a_n(t)x=f_1(t)\quad\quad(1)
$$
$$
\frac{d^nx}{dt^n}+a_1(t)\frac{d^{n-1}}{dt^{n-1}}+\cdots+a_n(t)x=f_2(t)\quad\quad(2)
$$  
$的解，则x_1(t)+x_2(t)是方程$  
$$
\frac{d^nx}{dt^n}+a_1(t)\frac{d^{n-1}x}{dt^{n-1}}+\cdots+a_n(t)x=f_1(t)+f_2(t)
$$  
$的解.$  
$证明：由题设条件可知x_1(t),x_2(t)分别是方程(1),(2)的解，$  
$则\quad\quad\frac{d^nx_1(t)}{dt^n}+a_1(t)\frac{d{n-1}x_1(t)}{dt^{n-1}}+\cdots+a_n(t)x_1(t)=f_1(t)\quad\quad(3)$  
$\frac{d^nx_2(t)}{dt^n}+a_1(t)\frac{d^{n-1}x_2(t)}{dt^{n-1}}+\cdots+a_n(t)x_2(t)=f_2(t)\quad\quad(4)$  
$那么由(3)和(4)相加，并利用导数的运算法则得：$  
$\frac{d^n(x_1(t)+x_2(t))}{dt^n}+a_1(t)\frac{d^{n-1}(x_1(t)+x_2(t))}{dt^{n-1}}+\cdots+a_n(t)(x_1(t)+x_2(t))=f_1(t)+f_2(t)$  
$此即说明x_1(t)+x_2(t)是方程\frac{d^nx}{dt^n}+a_1(t)\frac{d^{n-1}x}{dt^{n-1}}+\cdots+a_n(t)x=f_1(t)+f_2(t)的解.$  

$3.已知齐次线性微分方程的基本解组x_1,x_2,求下列方程对应的非齐次线性微分方程的通解：$  
$(1)x''-x=\cos t,x_1=e^t,x_2=e^{-t};$  
$解：应用常数易变法，令x(t)=c_1(t)e^t+c_2(t)e^{-t},将它代入方程，则有$  
$$
\begin{cases}
c'_1(t)e^t+c'_2(t)e^{-t}=0,\\
c'_1(t)e^t-c'_2(t)e^{-t}=\cos t,\\
\end{cases}
$$
$解之得$  
$$
\begin{cases}
c'_1(t)=\frac{1}{2}e^{-t}\cos t,\\
c'_2(t)=-\frac{1}{2}e^t\cos t,\\
\end{cases}
$$
$积分得c_1(t)=-\frac{1}{4}e^{-t}(\cos t-\sin t)+c_1;c_2(t)=-\frac{1}{4}e^t(\cos t+\sin t)+c_2,$  
$故所求通解为x(t)=c_1e^t+c_2e^{-t}-\frac{1}{2}\cos t.$  

$(3)x''+4x=t\sin 2t,x_1=\cos 2t,x_2=\sin 2t;$  
$令x=c_1(t)\cos 2t+c_2(t)\sin 2t,将它代入方程，可得关于c'_1(t)和c'_2(t)的方程组$  
$$
\begin{cases}
c'_1(t)\cos 2t+c'_2(t)\sin 2t=0,\\
-2c'_1(t)\sin 2t+2c'_2(t)\cos 2t=t\sin 2t.\\
\end{cases}
$$
$解得$  
$$
\begin{cases}
c'_1(t)=\frac{1}{4}t(\cos 4t-1),\\
c'_2(t)=\frac{1}{4}t\sin 4t.\\
\end{cases}
$$
$积分得$  
$$
\begin{cases}
c_1(t)=\frac{1}{64}\cos 4t+\frac{1}{16}t\sin 4t-\frac{1}{8}t^2+\gamma_1,\\
c_2(t)=\frac{1}{64}\sin 4t-\frac{1}{16}t\cos 4t+\gamma_2.\\
\end{cases}
$$
$于是方程的通解为x=c_1\cos 2t+c_2\sin 2t-\frac{1}{8}t^2\cos 2t+\frac{1}{16}t\sin 2t,其中c_1,c_2为任意常数.$  

$(5)t^2x''-tx'+x=6t+34t^2,x_1=t,x_2=t\ln t;$  
$解：原方程可化为\quad x''-\frac{1}{t}x'+\frac{1}{t^2}x=\frac{6}{t}+34.$  
$令方程的解为x=t·c_1(t)+t\ln t·c_2(t),将它代入方程，则可得关于c'_1(t)和c'_2(t)的方程组$  
$$
\begin{cases}
t·c'_1(t)+t\ln t·c'_2(t)=0,\\
c'_1(t)+(\ln t+1)c'_2(t)=\frac{6}{t}+34,\\
\end{cases}
$$
$解得$  
$$
\begin{cases}
c'_1(t)=-\ln t(\frac{6}{t}+34),\\
c'_2(t)=\frac{6}{t}+34,\\
\end{cases}
$$
$积分得$  
$$
\begin{cases}
c_1(t)=-34(t\ln t-t)-3(\ln t)^2+c_1\\
c_2(t)=34t+6\ln t+c_2\\
\end{cases}
$$
$于是方程的通解为x=c_1·t+c_2t\ln t+34t^2+3t(\ln t)^2，其中c_1,c_2为任意常数.$  

$6.假设x_1(t)\neq0是二阶齐次线性微分方程$  
$$
x''+a_1(t)x'+a_2(t)x=0\quad\quad\quad(*)
$$  
$的解，这里a_1(t)和a_2(t)于区间[a,b]上连续，试证：$  
$(1)x_2(t)为方程的解的充要条件是$  
$$
W'[x_1,x_2]+a_1W[x_1,x_2]=0;
$$
$的解，这里a_1(t)和a_2(t)于区间[a,b]上连续，试证：$  
$(2)方程的通解可表为$  
$$
x=x_1[c_1\int\frac{1}{x_1^2}\exp(-\int_{t_0}^ta_1(s)ds)dt+c_2],
$$
$其中c_1,c_2为任意常数，t_0,t\in[a,b].$  
$证明：（1）W'[x_1,x_2]+a_1W[x_1,x_2]=0$  
$\iff x_1x''_2-x''_1x_2+a_1x_1x'_2-a_1x'_1x_2=0$  
$\iff x_1x''_2-x''_1x_2+a_1x_1x'_2-a_1x'_1x_2+a_2x_1x_2-a_2x_1x_2=0$  
$\iff x_1(x''_2+a_1x'_2+a_2x_2)=0$  
$\iff x''_2+a_1x'_2+a_2x_2=0,(x_1\neq0)$  
$即x_2为(*)的解.$  
$(2)因为x_1,x_2为方程的解，则由刘维尔公式$  
$$ 
\begin{vmatrix} 
x_1 & x_2 \\ 
x_1' & x_2'\\ 
\end{vmatrix} 
=w(t_0)e^{-\int{t_0}^ta_1(s)ds}
$$ 
$即x_1,x'_2-x'_1x_2=W(t_0)e^{-\int_{t_0}^ta_1(s)ds}.$  
$两边都乘以\frac{1}{x^2}则有\frac{d(\frac{x_2}{x_1})}{dt}=\frac{W(t_0)}{x_1^2}e^{-\int_{t_0}^ta_1(s)ds},$  
$于是\frac{x_2}{x_1}=c_1\int\frac{1}{x_1^2}e^{-\int_{t_0}^ta_1(s)ds}dt+c_2$  
$即x_2=(c_1\int\frac{1}{x_1^2}e^{-\int_{t_0}^ta_1(s)ds}dt+c_2)x_1$  
$取c_1=1,c_2=0,得$  
$$
x_2=x_1\int\frac{1}{x_1^2}e^{-int_{t_0}^ta_1(s)ds}dt,
$$
$又$  
$$
W(t)=\begin{vmatrix}
x_1&x_2\\
x'_1&x'_2\\
\end{vmatrix}=e^{-\int_{t_0}^ta_1(s)ds}\neq0,
$$
$从而方程的通解可以表示为：$  
$x=x_1[c_1\int\frac{1}{x^2}\exp(-\int_{t_0}^ta_1(s)ds)dt+c_2],其中c_1,c_2为常数,t_0,t\in[a,b].$
