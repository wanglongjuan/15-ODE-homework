# 常微分作业六

**第131**

**第一题**

$设x(t)和y(t)是区间n\leq t \leq b上的连续函数，证明：如果在区间a \leq t \leq b上有\frac{x(t)}{y(t)}\neq常或\frac{y(t)}{x(t)}\neq常数，$

$则x(t)和y(t)在区间a \leq t \leq b上线性无关。（提示：用反证法）$


$解：反证法：假设x(t)和y(t)线性相关，则有不全为0的数c和d成立：cx(t)+dy(t)=0,$

$因为x(t)和y(t)是区间n\leq t \leq b上的连续函数，$

$所以x(t)\neq 0 y(t)\neq 0故c \neq 0(否则有d=0,与c,d不全为零矛盾),$

$所以有\frac{x(t)}{y(t)}=\frac{-d}{c},所以\frac{-d}{c}是常数，则与假设矛盾，$

$所以得证x(t)和y(t)在区间a \leq t \leq b上线性无关。$

**第二题**

$证明非齐次线性微分方程的叠加原理：设x_1(t),x_2(t)$分别是非齐次线性微分方程$$1.\frac{\mathrm d^nx}{\mathrm t^n}+a_1(t)\frac{\mathrm d^{n-1}x}{\mathrm dt^{n-1}}+\dots+a_n(t)x=f_1(t)$$,$$2.\frac{\mathrm d^nx}{\mathrm t^n}+a_1(t)\frac{\mathrm d^{n-1}x}{\mathrm dt^{n-1}}+\dots+a_n(t)x=f_2(t)$$

$的解，则x_1(t)+x_2(t)$是方程$$3.\frac{\mathrm d^nx}{\mathrm t^n}+a_1(t)\frac{\mathrm d^{n-1}x}{\mathrm dt^{n-1}}+\dots+a_n(t)x=f_1(t)+f_2(t)$$

$的解$.

$解：因为x_1(t)是方程1的解$，则有$$\frac{\mathrm d^nx_1 (t)}{\mathrm t^n}+a_1(t)\frac{\mathrm d^{n-1}x_1 (t)}{\mathrm dt^{n-1}}+\dots+a_n(t)x_1 (t)=f_1(t)$$

$x_2 (t)是微分方程2的解$，则有$$\frac{\mathrm d^nx_2 (t)}{\mathrm t^n}+a_1(t)\frac{\mathrm d^{n-1}x_2 (t)}{\mathrm dt^{n-1}}+\dots+a_n(t)x_2 (t)=f_2(t)$$

由两式相加得：$$\frac{\mathrm d^nx_1 (t)}{\mathrm t^n}+a_1(t)\frac{\mathrm d^{n-1}x_1 (t)}{\mathrm dt^{n-1}}+\dots+a_n(t)x_1 (t)+\frac{\mathrm d^nx_2 (t)}{\mathrm t^n}+a_1(t)\frac{\mathrm d^{n-1}x_2 (t)}{\mathrm dt^{n-1}}+\dots+a_n(t)x_2 (t)=f_1(t)+f_2(t)$$

$再将x_1(t)+x_2(t)带入方程3得：\frac{\mathrm d^n(x_1 (t)+x_2(t))}{\mathrm t^n}+a_1(t)\frac{\mathrm d^{n-1}(x_1 (t)+x_2(t))}{\mathrm dt^{n-1}}+\dots+a_n(t)(x_1 (t)+x_2(t))=f_1(t)+f_2(t)$

将其展开可得$$\frac{\mathrm d^n(x_1 (t)+x_2(t))}{\mathrm t^n}+a_1(t)\frac{\mathrm d^{n-1}(x_1 (t)+x_2(t))}{\mathrm dt^{n-1}}+\dots+a_n(t)(x_1 (t)+x_2(t))=\\ \frac{\mathrm d^nx_1 (t)}{\mathrm t^n}+a_1(t)\frac{\mathrm d^{n-1}x_1 (t)}{\mathrm dt^{n-1}}+\dots+a_n(t)x_1 (t)+\frac{\mathrm d^nx_2 (t)}{\mathrm t^n}+a_1(t)\frac{\mathrm d^{n-1}x_2 (t)}{\mathrm dt^{n-1}}+\dots+a_n(t)x_2 (t)$$

$即可证明非齐次线性微分方程的叠加原理。$

**第三题**

$已知齐次线性微分方程的基本解组x_1,x_2，求下列方程对应的非齐次线性微分方程的通解：$

$(1)x''-x=\cos t,x_1=e^t,x_2=e^{-t};$

$解：应用常数变易法，令x=c_1(t)e^t+c_2(t)e^{-t},将它带入方程可得关于c_1(t)和c_2(t)$的两个方程：$$e^tc_1'(t)+e^{-t}c_2'(t)=0\\ e^tc_1'(t)-e^{-t}c_2'(t)=\cos t$$

$解得c_1'(t)=\frac{1}{2}e^t\cos t,c_2'(t)=-\frac{1}{2}e^t\cos t,$

$对c_1'(t)和c_2'(t)$分别积分可得：$$c_1(t)=\frac{1}{4}e^{-t}(\sin t-\cos t)+\gamma_1,\\ c_2(t)=-\frac{1}{4}e^{t}(\sin t+\cos t)+\gamma_2$$

再带入通解的公式：$$x=\displaystyle\sum_{i=1}^{n}\gamma_ix_i(t)+ \displaystyle\sum_{i=1}^{n}x_i(t)(c_i(t)-\gamma_i)$$

$可得到其通解为：x=\gamma_1e^t+\gamma_2e^{-t}-\frac{1}{2}\cos t$

$(3)x''+4x=t\sin 2t,x_1=\cos 2t,x_2=\sin 2t;$

$解：应用常数变易法，令x=c_1(t)\cos 2t+c_2(t)\sin 2t,将它带入方程可得关于c_1(t)和c_2(t)$的两个方程：$$c_1'(t)\cos 2t+c_2'(t)\sin 2t=0,\\ -2\sin 2tc_1'(t)+2\cos 2t c_2'(t)=t\sin 2t$$

可解得$$c_1'(t)=-\frac{1}{4}t+\frac{1}{4}t\cos 4t\\c_2'(t)=\frac{1}{4}t\sin 4t$$

$对其分别积分可得$：$$c_1'(t)=\frac{1}{16}t\sin 4t+\frac{1}{64}\cos 4t-\frac{1}{8}t^2+\gamma_1,\\c_2'(t)=-\frac{1}{16}t\cos 4t+\frac{1}{64}\sin 4t+\gamma_2$$

再带入通解的公式：$$x=\displaystyle\sum_{i=1}^{n}\gamma_ix_i(t)+ \displaystyle\sum_{i=1}^{n}x_i(t)(c_i(t)-\gamma_i)$$

$可得到其通解为：x=\gamma_1\cos 2t+\gamma_2\sin 2t-\frac{1}{8}t^2\cos 2t+\frac{1}{16}t\sin 2t$

$(5)t^2x''-tx'+x=6t+34t^2,x_1=t,x_2=tlnt$

$解：将式子两边都除以t^2得：x''-\frac{1}{t}x'+\frac{1}{t^2}x=\frac{6}{t}+34$

$应用常数变易法，令x=c_1(t)t+c_2(t)tlnt,将它带入方程可得关于c_1(t)和c_2(t)$的两个方程：$$c_1'(t)t+c_2't)tlnt=0,\\c_1'(t)+(1+lnt)c_2'(t)=\frac{6}{t}+34$$

可解得$$c_1'(t)=-\frac{6}{t}lnt-34lnt,\\c_2'(t)=\frac{6}{t}+34$$

$再分别对c_1'(t)和c_2'(t)$求积分得：$$c_1(t)=4t^{-3}+12t^{-3}lnt+\gamma_1,\\c_2(t)=-\frac{9}{4}t^{-4}-9t^{-4}lnt+\gamma_2$$

再带入通解的公式：$$x=\displaystyle\sum_{i=1}^{n}\gamma_ix_i(t)+ \displaystyle\sum_{i=1}^{n}x_i(t)(c_i(t)-\gamma_i)$$

$可得到其通解为：x=t^2\gamma_1+t^3\gamma_2+\frac{7}{4}t^{-1}+3t^{-1}lnt$

**第六题**

$假设x_1(t)\neq 0是二阶齐次线性微分方程$ $$x''+a_1(t)x'+a_2(t)x=0$$

$的解，这里a_1(t)a_2(t)于区间[a,b]上连续，试证：$

$(1)x_2(t)为方程的解的充要条件是$ $$W'[x_1,x_2]+a_1W[x_1,x_2]=0;$$

$(2)方程的通解可表为$ $$x=x_1[c_1\int\frac{1}{x_1^2}(-\int_{t_0}^t a_1(s) \mathrm ds)\mathrm dt+c_2],$$

$其中c_1,c_2为任意常数，t_0,t\in [a,b]$

$解：必要性：W'[x_1,x_2]=\frac{\mathrm d}{\mathrm dt} \begin{vmatrix}  x_1(t) &x_2(t)\\ \\ x_1'(t)& x_2'(t)\\  \end{vmatrix}=\begin{vmatrix}  x_1'(t) &x_2'(t)\\ \\ x_1'(t)& x_2'(t)\\  \end{vmatrix}+ \begin{vmatrix}  x_1(t) &x_2(t)\\ \\ x_1''(t)& x_2''(t)\\  \end{vmatrix}$

$因为\begin{vmatrix}  x_1'(t) &x_2'(t)\\ \\ x_1'(t)& x_2'(t)\\  \end{vmatrix}=0，$

$所以W'[x_1,x_2]=\begin{vmatrix}  x_1(t) &x_2(t)\\ \\ x_1''(t)& x_2''(t)\\  \end{vmatrix}$

$将x''=-a_1(t)x'-a_2(t)x带入行列式得：$

$W'[x_1,x_2]=\begin{vmatrix}  x_1(t) &x_2(t)\\ \\-a_1(t)x_1'(t)-a_2(t)x_1'(t) & -a_1(t)x_2'(t)-a_2(t)x_2'(t)\\  \end{vmatrix}=\begin{vmatrix}  x_1(t) &x_2(t)\\ \\ -a_1(t)x_1'(t)& -a_1(t)x_2'(t)\\  \end{vmatrix}=-a_1(t)W[x_1,x_2]$

$充分性：W'[x_1,x_2]=\frac{\mathrm d}{\mathrm dt} \begin{vmatrix}  x_1(t) &x_2(t)\\ \\ x_1'(t)& x_2'(t)\\  \end{vmatrix}+a_1\begin{vmatrix}  x_1(t) &x_2(t)\\ \\ x_1'(t)& x_2'(t)\\  \end{vmatrix}$

$所以\frac{\mathrm d(x_1(t)x_2'(t)-x_2(t)x_1'(t))}{\mathrm dt}+a_1(x_1(t)x_2'(t)-x_2(t)x_1'(t))=0$

$即其可以化为：x_1(t)(x_2''(t)+a_1x_2'(t))-x_2(t)(x_1''(t)+a_1x_1'(t))=0$

$在前边加一个a_2x_1(t)x_2(t),后边减一个a_2x_1(t)x_2(t)将该式化为$

$x_1(t)(x_2''(t)+a_1x_2'(t)+a_2x_2(t))-x_2(t)(x_1''(t)+a_1x_1'(t)+a_2x_1(t))=0$

$又因为x_1(t)是方程的解，所以x_1''(t)+a_1x_1'(t)+a_2x_1(t)=0，$

$所以x_1(t)(x_2''(t)+a_1x_2'(t)+a_2x_2(t))=0$

$即证x_2(t)是方程的一个解.$

$(2)解：用常数变易法，设x=c(t)x_1为方程的解，带入方程可得：$

$c''x_1+c'(2x_1'+a_1x_1)=0$

$令y'=c',可得y'x_1+y(2x_1'+a_1x_1)=0,y'=-(\frac{2x_1'}{x_1}+a_1)y.有解：$

$c'=y'=c_1exp[-\int_{t_0}^t (\frac{2x_1'}{x_1}+a_1)dt]=\frac{c_1}{x_1^2}exp[-\int_{t_0}^t a_1(s)ds]$

$再积分得$ $$c=\int \frac{c_1}{x_1^2}exp[-\int_{t_0}^t a_1(s)ds]dt+c_2=c_1\int \frac{1}{x_1^2}exp[-\int_{t_0}^t a_1(s)ds]dt+c_2$$

$可得到方程的解为：$ $$x=x_1[c_1{x_1^2}exp[-\int_{t_0}^t a_1(s)ds]dt+c_2]$$

$其中c_1,c_2为任意常数，t_0,t\in[a,b].因解中有两个独立的任意常数，故解为二阶齐次线性微分方程的通解$
