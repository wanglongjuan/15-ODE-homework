1.证明：假设$x(t)$，$y(t)$在区间$a \leq t \leq b$上线性相关

则存在不全为零得常数$\alpha,\beta$，使得

$$
\alpha x(t)+\beta y(t) \equiv 0,t \in [a,b]
$$

不妨设$x(t)$不为零，则有

$$
\frac{y(t)}{x(t)}=-\frac{\alpha}{\beta}
$$

显然$-\frac{\alpha}{\beta}$为常数，矛盾,即得证。

2.证明：由题设条件可知$x_1(t)、x_2(t)$分别是方程(1)，(2)的解,
则

$$
\frac{d^{n}x_1(t)}{\mathrm d t^n}+a_1(t)\frac{d^{n-1}x_1(t)}{\mathrm d t^{n-1}}+\cdots+a_n(t)x_1(t)=f_1(t)
$$

$$
\frac{d^{n}x_2(t)}{\mathrm d t^n}+a_1(t)\frac{d^{n-1}x_2(t)}{\mathrm d t^{n-1}}+\cdots+a_n(t)x_2(t)=f_2(t)
$$

两式相加，得

$$
\frac{d^{n}(x_1(t)+x_2(t))}{\mathrm d t^n}+a_1(t)\frac{d^{n-1}(x_1(t)+x_2(t))}{\mathrm d t^{n-1}}+\cdots+a_n(t)(x_1(t)+x_2(t))=f_1(t)+f_2(t)
$$

说明$(x_1(t)+x_2(t))$是上式的解。

3.(1)令$x(t)=c_1(t)e^t+c_2(t)e^{-t}$,将其代入方程，则有

$$
\begin{cases}
c'_1(t)e^t+c'_2(t)e^{-t}=0\\
c'_1(t)e^t-c'_2(t)e^{-t}=\cos t
\end{cases}
$$

解得

$$
\begin{cases}
c'_1(t)=\frac{1}{2}e^{-t}\cos t\\
c'_2(t)=-\frac{1}{2}e^t\cos t
\end{cases}
$$

积分得

$$
\begin{cases}
c_1(t)=-\frac{1}{4}e^{-t}(\cos t - \sin t)+c_1\\
c_2(t)=-\frac{1}{4}e^t(\cos t + \sin t)+c_2
\end{cases}
$$

故通解为
$$
x(t)=c_1e^t +c_2e^{-t}-\frac{1}{2}\cos t
$$

(3)令$x(t)=c_1(t)\cos 2t+c_2(t)\sin 2t$,将其代入方程，则有

$$
\begin{cases}
c'_1(t)\cos 2t+c'_2(t)\sin 2t=0
\\
\\
-2c'_1(t)\sin 2t+2c'_2(t)\cos 2t=t\sin 2t
\end{cases}
$$

解得

$$
\begin{cases}
c'_1(t)=\frac{1}{4}(\cos 4t -1)
\\
\\
c'_2(t)=\frac{1}{4}\sin 4t
\end{cases}
$$

积分得

$$
\begin{cases}
c_1(t)=\frac{1}{64}\cos 4t + \frac{1}{16}t\sin 4t - \frac{1}{8}t^2 + \gamma_1
\\
\\
c_2(t)=\frac{1}{64}\sin 4t - \frac{1}{16}t\cos 4t + \gamma_2
\end{cases}
$$

通解为
$$c_1\cos 2t+c_2\sin 2t-\frac{1}{8}t^2\cos 2t+\frac{1}{16}t\sin 2t$$

(5)原方程可化为
$$
x''-\frac{1}{t}x'+\frac{1}{t^2}x=\frac{6}{t}+34
$$
令方程组的解为
$$
x=t \bullet c_1(t)+t\ln t \bullet c_2(t)
$$
可得

$$
\begin{cases}
t \bullet c'_1(t)+t\ln t \bullet c'_2(t)=0
\\
\\
c'_1(t)+(\ln t +1)c'_2(t)=\frac{6}{t}+34
\end{cases}
$$

解得

$$
\begin{cases}
c'_1(t)=-\ln t(\frac{6}{t}+34)
\\
\\
c'_2(t)=\frac{6}{t}+34
\end{cases}
$$

积分得

$$
\begin{cases}
c_1(t)=-34(t\ln t-t)-3(\ln t)^2 +c_1
\\
\\
c_2(t)=34t+6\ln t+c_2
\end{cases}
$$

通解为
$$
c_1 \bullet t+c_2 t\ln t+34t^2+3t(\ln t)^2
$$

6.证明：(1)
$W'[x_1,x_2]+a_1W[x_1,x_2]=0$

$\Leftrightarrow x_1x''_2-x''_1x_2+a_1x_1x'_2-a_1x'_1x_2=0$

$\Leftrightarrow x_1x''_2-x''_1x_2+a_1x_1x'_2-a_1x'_1x_2+a_2x_1x_2-a_1x_1x_2=0$

$\Leftrightarrow x_1(x''_2+a_1x'_2+a_2x_2)=0$

$\Leftrightarrow x''_2+a_1x'_2+a_2x_2=0   (x_1 \neq 0)$

即$x_2$为解

(2)因为$x_1,x_2$为方程的解，则由刘维尔公式得

$$
\begin{vmatrix}
x_1 & x_2 \\
x'_1 & x'_2\\
\end{vmatrix}
=W(t_0)e^{-\int_{t_0}^t a_1(s)\mathrm d s}
$$

即

$$
x_1x'_2-x'_1x_2=W(t_0)e^{-\int_{t_0}^t a_1(s)\mathrm d s}
$$

两边乘$\frac{1}{x^2_1}$,则有

$$
\frac{\mathrm d (\frac{x_2}{x_1})}{\mathrm d t}=\frac{W(t_0)}{x^2_1}e^{-\int_{t_0}^t a_1(s)\mathrm d s}
$$

则

$$
\frac{x_2}{x_1}=c_1 \int \frac{1}{x^2_1}e^{-\int_{t_0}^t a_1(s)\mathrm d s}\mathrm d t +c_2
$$

即

$$
x_2=x_1\int \frac{1}{x^2_1}e^{-\int_{t_0}^t a_1(s)\mathrm d s}\mathrm d t
$$

又
$$W(t)=
\begin{vmatrix}
x_1 & x_2 \\
x'_1 & x'_2\\
\end{vmatrix}
=e^{-\int_{t_0}^t a_1(s)\mathrm d s} \neq 0
$$

通解为
$$
x=x_1[c_1 \int \frac{1}{x^2_1} \exp(-\int_{t_0}^t a_1(s)\mathrm d s)\mathrm d t +c_2 ]
$$
