（8）$（x+2y）dy+xdy=0$

(9)$[xcos(x+y)+sin(x+y)]dy+xcos(x+y)dy=0$

(10)$(ycosx-xsinx)dx+(ysinx+xcosx)dy=0$

解：
（8）$M=x+2y,N=x$,$\frac{\partial M}{\partial y}\neq\frac{\partial N}{\partial x}$

$\frac{(\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x})}{N}=\frac{1}{x}$,  积分因子为$\mu=e^{\int\frac{1}{x}dx}=x$

两边乘$\mu$得恰当方程为$x(x+2y)dx+x^2dy=0$

所以通解为：$\int(x^2+2xy)dx+\int[x^2-\frac{\partial}{\partial y}\int(x^2+2xy)dx]dy=c\\\\\frac{x^3}{3}+x^3y=c$

即：$x^3+3x^2y=c$

(9)$M=xcos(x+y)+sin(x+y),N=xcos(x+y)$,

因为$\frac{\partial M}{\partial y}=\frac{\partial N}{\partial x}$，

所以方程的通解为：$\int[xcos(x+y)+sin(x+y)]dx+\int[xcos(x+y)-\frac{\partial}{\partial y}\int[xcos(x+y)+sin(x+y)]dx]dy=c$

即：$xsin(x+y)=c$

(10)$M=ycosx-xsinx,N=ysinx+xcosx$,,$\frac{\partial M}{\partial y}\neq\frac{\partial N}{\partial x}$

$\frac{(\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x})}{-M}=1$

积分因子为：$\mu=e^{\int dy}=e^y$

两边乘$\mu$得恰当方程：$e^y(ycosx-xsinx)dx+e^y(ysinx+xcosx)dy=0$,

所以通解为：$\int e^y(ycosx-xsinx)dx+\int [N-\frac{\partial}{\partial y}\int e^y(ycosx-xsinx)dx]dy=c$

即：$e^ysinx(y-1)+e^ycosx=c$

(11)两边乘$x^2y$得：$(4x^3y^2dx+2x^4ydy)+(3x^2y^5dx+5x^3ydy)=0$

$d(x^4y^2)+d(x^3y^5)=0$

所以方程通解为：$x^4y^2+x^3y^5=c$

4.设$f(x,y)$及$\frac{\partial f}{\partial y}$连续，试证方程$dy-f(x,y)dx=0$。

证：必要性：如果方程为线性方程，则有$\frac{\mathrm d y}{\mathrm d x}=P(x)y+Q(x)$,积分因子为$\mu(x)=e^{-\int P(x)dx}$,只与x有关。

充分性：如果方程有只与x有关的积分因子$\mu(x)$, 则恰当方程为$\mu(x)dy-\mu(x)f(x,y)dx=0$,

所以$f=-\int\frac{\mu'(x)}{\mu(x)}dy+Q(x)=-\frac{\mu'(x)}{\mu(x)}y+Q(x)=P(x)+Q(x)$.其中$P(x)=-\frac{\mu'(x)}{\mu(x)}$,

于是方程化为$dy-(P(x)y+Q(x))dx=0$

所以为一阶线性方程。

8.求出伯努利方程的积分因子.

解：$\frac{\mathrm d y}{\mathrm d x}=P(x)y+Q(x)y^n$,$y\neq0$.

两边乘$y^{-n}$,令$z=y^{-n}$,

$\frac{\mathrm d z}{\mathrm d x}=(1-n)P(x)z+(1-n)Q(x)$,积分因子为：$\mu=e^{-\int(1-n)P(x)dx}=e^{(n-1)\int P(x)dx}$.

所以积分因子为：$\mu=e^{-\int(1-n)P(x)dx}=e^{(n-1)\int P(x)dx}$.
