2.求下列方程的解：

(8)$(x+2y)\mathrm d x-x\mathrm d y = 0$

(9)$[x\cos{(x+y)}+\sin {(x+y)}]\mathrm d x+x\cos{(x+y)}\mathrm d y = 0$

(10)$(y\cos x-x\sin x)\mathrm d x+(y\sin x+x\cos x)\mathrm d y = 0$

(11)$x(4y\mathrm d x+2x\mathrm d y)+y^3(3y\mathrm d x+5x\mathrm d y) = 0$

解：$(8这里M=x+2y,N=x,\frac{\partial M}{\partial y}=2,\frac{\partial N}{\partial x}=1,方程不是恰当的$

$∵\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac{1}{x}∴方程有积分因子μ=e^{\int{\frac{1}{x}\mathrm d x}}= x$

$以μ=x乘方程两边，得到x^2\mathrm d x + 2xy\mathrm d x + x^2\mathrm d y = 0$

$即\mathrm d (x^2y+\frac{1}{3}x^3) = 0$

$因而，方程的通解为x^3+3x^2y = c$

(9)$这x\cos(x+y)+\sin (x+y),N=x\cos(x+y),\frac{\partial M}{\partial y}=-x\sin (x+y)+\cos(x+y),\frac{\partial N}{\partial x}=\cos(x+y)-x\sin (x+y)$

$方程是恰当的，并且方程可以写\mathrm d (x\sin (x+y)) = 0$

$∴方程的通解为x\sin(x+y) = c$

$(10)这里M=y\cos x-x\sin x,N=y\sin x+x\cos x,\frac{\partial M}{\partial y}=\cos x,\frac{\partial N}{\partial x}=y\cos x+\cos x-x\sin x,方程不是恰当的$

$∵\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{-M} = 1,∴方程有积分因子μ=e^{\int 1\mathrm d y} = e^y$

$以μ=e^y乘方程两边，得到(ye^y\cos x \mathrm d x + ye^y\sin x \mathrm d y)+(e^yx\cos x \mathrm d y -e^yx\sin x \mathrm d x) = 0$

$即\mathrm d [(y-1)e^y\sin x+ e^yx\cos x] = 0$

$∴方程的通解为(y-1)\sin x + x\cos x = ce^{-y}$

$(11)∵x(4y\mathrm d x+2x\mathrm d y)=\mathrm (2x^2y)$

$对于y^3(3y\mathrm d x+5x\mathrm d y) = 0,有M=3y^4,N=5xy^3,\frac{\partial M}{\partial y}=12y^3,\frac{\partial N}{\partial x}=5y^3$

$又∵\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{5xy^3} = \frac{7}{5x},∴方程有积分因子μ=e^{\int \frac{7}{5x} \mathrm d x} = x^{\frac{7}{5}}$

$以μ=x^{\frac{7}{5}}乘以方程y^3(3y\mathrm d x+5x\mathrm d y) = 0两边，得3y^4x^\frac{7}{5}\mathrm d x + 5x^\frac{12}{5}y^3\mathrm d y = 0$

$即\mathrm d (\frac{5}{4}x^\frac{12}{5}y^4)= 0$

$∴原方程的通解为2x^2y+\frac{5}{4}x^\frac{12}{5}y^4 = c$

$4.设f(x,y)及\frac{\partial f}{\partial y}连续，试证方程\mathrm d y -f(x,y)\mathrm d x = 0为线性微分方程的充要条件是它有仅依赖于x的积分因子.$

$证明： 充分性  设\mathrm d y -f(x,y)\mathrm d x = 0有仅依赖于x的积分因子.$

$∵积分因子只依赖于x，∴存在\varphi(x)满足\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac{-\frac{\partial f}{\partial y}-0}{1}=\varphi(x)$

$即\frac{\partial f}{\partial y}=-\varphi(x)$

$积分后得f(x,y)=-\varphi(x)y+h(x),h(x)是关于x的任意可微函数$

$则原方程可以写成\frac{\mathrm d y}{\mathrm d x}= -\varphi(x)y+h(x)$

$即证明\mathrm d y -f(x,y)\mathrm d x = 0是线性的$

$必要性  设方程\mathrm d y -f(x,y)\mathrm d x = 0是线性微分方程$

$即存在g(x),h(x),使f(x,y)=yg(x)+h(x)$

$M=-f(x,y)=-yg(x)-h(x),N=1,\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac {-g(x)}{1}=-g(x)$

$∴方程具有积分因子μ=e^{\int -g(x) \mathrm d x}$

$即证明方程有仅依赖于x的积分因子$

$8.求出伯努利微分方程的积分因子$

$解：伯努利方程为\frac{\mathrm d y}{\mathrm d x}=p(x)y+q(x)y^n,(n\neq0,1)$

$dy-p(x)y\mathrm d x-q(x)y^n\mathrm d x = 0$

$y^{-n}\mathrm d y-p(x)y^{1-n}\mathrm d x-q(x)\mathrm d x = 0$

$\mathrm d y^{1-n}-(1-n)p(x)y^{1-n}\mathrm d x-(1-n)q(x)\mathrm d x = 0$

$e^{-(1-n)\int p(x) \mathrm d x}[\mathrm d y^{1-n}-(1-n)p(x)y^{1-n}\mathrm d x]-e^{-(1-n)\int p(x) \mathrm d x}(1-n)q(x)\mathrm d x = 0$

$即\mathrm d [y^{1-n}e^{-(1-n)\int p(x) \mathrm d x}]-\mathrm d [\int (1-n)q(x)e^{-(1-n)\int p(x) \mathrm d x} \mathrm d x] = 0$

$这是全微分方程，所求积分因子是μ=y^{-n}e^{-(1-n)\int p(x) \mathrm d x}$

$1.解下列方程：$

$(1)x{y^\prime}^3 = 1 + y^\prime$

$(2){y^\prime}^3 - x^3(1 - y^\prime) = 0$

$(3)y = {y^\prime}^2e^{y^\prime}$

$(4)y(1 + {y^\prime}^2) = 2a (a为常数）$

$(5)x^2 + {y^\prime}^2 = 1$

$(6)y^2 (y^\prime -1) = (2 - y^\prime)^2$

$解：(1)令\frac{\mathrm d y}{\mathrm d x} = y^\prime = \frac {1}{t},则x = t^3 + t^2$

$对y求导数，则有t =(3t^2+ 2t)\frac{\mathrm d t}{\mathrm d y}$

$积分，得y = \frac {3}{2}t^2 + 2t + c $

$∴方程的通解为\begin{cases}x=t^3 + t^2\\y = \frac{3}{2}t^2 + 2t + c \end{cases}$

$(2)令y^\prime = tx,则原方程化为x^3(t^3 -1 + tx) = 0$

$解出x得,x = \frac{1}{t} - t^2$

$两边对y求导\frac{1}{1 - t^3} = (-\frac{1}{t^2}- 2t)\frac{\mathrm d t}{\mathrm d y}$

$积分得y = -\frac{1}{2}t^2 + \frac{2}{5}t^5 + \frac{1}{t} + c$

$∴方程的通解为\begin{cases}x=\frac{1}{t} - t^2\\y = -\frac{1}{2}t^2 + \frac{2}{5}t^5 + \frac{1}{t} + c\end{cases}$

$(3)令y^\prime = p,则原方程化为y = p^2e^p$

$对x求导p = ( 2p + p^2)e^p \frac{\mathrm d p}{\mathrm d x}$

$积分得x = ( p + 1 )e^p$

$∴方程的通解为\begin{cases}x= ( p + 1 )e^p\\y = p^2e^p\end{cases}$

$(4)令y^\prime = p,则原方程化为y(1 + p^2) = 2a$

$对x求导p{(1 + p^2)}^2 + 4ap \frac{\mathrm d p}{\mathrm d x} = 0$

$积分得x = -a(2\arctan p + \frac{2p}{1 + p^2} + c)$

$∴方程的通解为\begin{cases}x = -a(2\arctan p + \frac{2p}{1 + p^2} + c)\\y = \frac{2a}{1 + p^2}\end{cases}$

$(5)令y^\prime = \cos t,则原方程化为x^2 + {\cos t}^2 = 1$

$解出x得,x = \sin t$

$对y求导\frac{1}{\cos t} = \cos t \frac{\mathrm d t}{\mathrm d y}$

$积分得y = \frac{t}{2} + \frac{1}{4}\sin 2t + c$

$∴方程的通解为\begin{cases}x = \sin t\\y = \frac{t}{2} + \frac{1}{4}\sin 2t + c\end{cases}$

$(6)令2 - y^\prime = yt,则原方程化为 y^2(1 - yt) = y^2t^2$

$ y = \frac{1}{t} - t , 且 y^\prime = 1 + t^2 ,这是原微分方程参数形式$

$∴\mathrm d x = \frac {\mathrm d y}{y^\prime} = \frac{(-\frac{1}{t^2}-1)\mathrm d t}{1 + t^2} = -\frac{1}{t^2}\mathrm d t$

$积分得 x = \frac{1}{t} + c$

$∴方程参数形式的通解为\begin{cases}x = \frac{1}{t} + c\\y = \frac{1}{t} - t\end{cases}$

$消参得 y = x - \frac{1}{x-c} - c,c为任意常数$

$当y^\prime = 0时，此方程无解$
