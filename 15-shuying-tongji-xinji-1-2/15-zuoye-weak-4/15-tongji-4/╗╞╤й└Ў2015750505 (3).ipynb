$$x\sin (x+y)=c. c为任意常数.$$

$4.设f(x,y)及\frac{\partial f}{\partial y}连续，试证方程dy-f(x,y)dx=0为线性微分方程的充要条件是$

$解：$

$充分性：$

$这里：$

$$M=-f(x,y)4;N=1.$$

$$\frac{\partial M}{\partial y}=-\frac{\partial f}{\partial y};\frac{\partial N}{\partial x}=0;$$

$因为方程的积分因子只依赖于x，所以存在φ(x)满足:$

$$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=φ(x)$$

$即$它有仅依赖于x的积分因子.

$$\frac{-\frac{\partial f}{\partial y}-0}{1}=φ(x)$$

$解出$

$$φ(x)=-\frac{\partial f}{\partial y}$$

$两边积分，有：$

$$f(x,y)=-φ(x)y+h(x)$$

$代入原方程，有：$

$$dy+φ(x)y-h(x)dx=0$$

$即$

$$\frac{dy}{dx}=-φ(x)y+h(x)$$

$其为线性微分方程$

$$即证.$$

$$ $$

$必要性$

$若方程dy-f(x,y)dx=0为线性微分方程，则存在h（x),g(x)$

$使得$

$$\frac{dy}{dx}=g(x)y+h(x)$$

$所以$

$$dy-[g(x)y+h(x)]dx=0$$

$这里$

$$M'=--[g(x)y+h(x)]);N'=1.$$

$$\frac{\partial M}{\partial y}=-g(x);\frac{\partial N}{\partial x}=0;$$

$所以$

$$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=-g(x)$$

$积分因子为$

$$μ=e^{∫-g(x)dx}.$$

$所以该方程有仅依赖于x的积分因子.$

$$即证.$$



$8.求出伯努利微分方程的积分因子.$

$解：$

$伯努利微分方程的形式为:$

$$\frac{dy}{dx}=P(x)y+Q(x)y^n.$$

$变形为$

$$dy-P(x)ydx-Q(x)y^ndx=0$$

$两边乘以y^n,得:$

$$y^{-n}dy-P(x)y^{1-n}dx-Q(x)dx=0$$

$再乘以e^{-(1-n)∫P(x)dx},得:$

$$e^{-(1-n)∫P(x)dx}[y^{-n}dy-P(x)y^{1-n}dx-Q(x)dx]=0$$

$即$

$$d[y^{1-n}e^{-(1-n)∫P(x)dx}]-d[∫(1-n)Q(x)e^{-(1-n)∫P(x)dx}]=0.$$

$这是全微分方程.$

$故所求积分因子为:$

$$y^{-n}e^{-(1-n)∫P(x)dx}.$$
