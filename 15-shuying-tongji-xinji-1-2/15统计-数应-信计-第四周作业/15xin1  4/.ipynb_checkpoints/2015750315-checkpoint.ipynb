1.（1）
令  $$ M = x + 2y $$ 
   $$ N = x $$ , $ \frac{\partial M}{\partial y} = 2 $ , $ \frac{\partial N}{\partial x} = 1 $ , 所以方程不是恰当的。
   
   因为 $ \frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N} = \frac{1}{x} $ 只与 y 有关 ， 故方程有只与 x 有关的积分因子
   
   $$
   u = e^{\int \frac{1}{x}\mathrm dx} = e^{ln|x|} = x 
   $$ 
   
   以 $ u = x $乘方程两边，得到
   
   $$
   x(x + 2y)\mathrm dx + x^2\mathrm dy = 0
   $$
   
   得到的方程为恰当微分方程。
   
   $$
   \frac{\partial h}{\partial x} = x^2+ 2xy 
   $$
   $$
   \frac{\partial h}{\partial y} = x^2 
   $$
   
   对 x 积分得到， 
   
   $$
   \frac{1}{3}x^3 + x^2y 
   $$
   
   再对 y 求导得，
   
   $$
   x^2 + \frac{\mathrm df(y)}{\mathrm dy} = x^2 
   $$
   
   所以 f(y) = 0 
   
   所以方程通解为，
   
   $$
   \frac{1}{3}x^3 + x^2y = c
   $$
   
   c为任意常数。

(2).
令  $$ M = x\cos(x + y) + \sin(x + y) $$,
   $$ N = x\cos(x + y) $$
   
   因为 $ \frac {\partial M }{\partial y} = -x\sin(x + y) + \cos(x + y) $,$\frac{\partial N}{\partial x} = \cos( x + y)  - x\sin(x + y) $
   
   所以方程为恰当微分方程。
   
   现求 u ，使它同时满足下面两个方程，
   
   $$
   \frac{\partial u}{\partial x} =  x\cos(x + y) + \sin(x + y)
   $$
   $$
   \frac{\partial u}{\partial y} = x\cos(x + y) 
   $$
   
   对 x 求积分得到，
   
   $$ 
   u = x\cos(x + y) + f(y)
   $$
   
   所以 $ f(y) = 0 $
   
   因此方程的通解为，
   
   $$ 
   x\cos(x + y) = c
   $$
   
   c为任意常数。
   

(3).
 令 $$ M = y\cos x -x\sin x $$
   $$ N = y\sin x + x\cos x $$
   
   因为 $ \frac{\partial M}{\partial y} = \cos x $ ,$ \frac{\partial N}{\partial x} = y\cos x-x\sin x + \cos x $.
   
   所以方程不是恰当方程。
   
   因为，
   
   $$
   \frac{\frac{\partial M}{\partial y} - \frac{\partial N}{\partial x}}{-M} = 1
   $$
   
   所以方程有积分因子，$ u = e^{\int \mathrm dy} = e^y $
   
   以 $ u = e^y $乘方方程两边，得到
   
   $$
   e^y(y\cos x - x\sin x)\mathrm dx + e^y(y\sin x + x\cos x)\mathrm dy = 0
   $$
   
   为恰当方程。
   
   令 $$ M_1 = e^y(y\cos x - x\sin x) $$
     $$ N_2 = e^y(y\sin x + x\cos x) $$
     
   现求 h 使它同时满足下面两个方程。
   
   $$
   \frac{\partial h}{\partial x} = e^y(y\cos x - x\sin x) 
   $$
   $$
   \frac{\partial h}{\partial y} = e^y(y\sin x + x\cos x)
   $$
   $$
   h = \int {e^y(y\cos x - x\sin x)} + f(y) 
   $$
   $$ \frac{\partial h}{\partial y} = \frac{\partial (\int {e^y(y\cos x - x\sin x)})}{\partial y} + \frac{\mathrm df(y)}{\mathrm dy} =  e^y(y\sin x + x\cos x)$$
   
   解的方程的通解为，
   
   $$ 
   e^y(y - 1)\sin x + e^y\cos x = c
   $$
   
   c为任意常数。

2.
   必要性，
   
   若方程为线性方程，则有，$ \frac{\mathrm dy}{\mathrm dx} = P(x)y + Q(x) $ , 此方程有积分因子 $ u(x) = e^{\int P(x)\mathrm dx} $ ,  u(x) 
   
   只与 x 有关。
   
   充分性，
   
   若方程有只与 x 有关的积分因子 u(x).则 $ u(x)\mathrm dy -u(x)f(x ,y)\mathrm dx = 0 $为恰当方程 ， 从而 
   
   $$
   \frac{\partial (-u(x)f(x ,y))}{\partial y} = \frac{\mathrm du(x)}{\mathrm dx}
   $$
   $$
   \frac{\partial f}{\partial y} = -\frac{u\prime(x)}{u(x)}
   $$
   $$
   f = -\int\frac{u\prime(x)}{u(x)}\mathrm dy + Q(x) = -\frac{u\prime(x)}{u(x)}y +Q(x) = P(x)y +Q(x).
   $$
   
   其中 $ P(x) = -\frac{u\prime(x)}{u(x)}.$ , 于是方程可化为 
   
   $$ 
   \mathrm dy - (P(x)y + Q(x))\mathrm dx = 0
   $$
   
   即方程为一阶线性方程。

3.
伯努利微分方程为，$ \frac{\mathrm dy}{\mathrm dx} = P(x)y + Q(x)y^n $

两边同乘以 $ y^{-n} $. 令 $ z = y^{-n} $,得到

$$
\frac{\mathrm dz}{\mathrm dx} = (1 - n)P(x)z + (1 - n)Q(x) 
$$

线性方程有积分因子 $ u = e^{-\int (1 - n)P(x)\mathrm dx} = e^{(n-1){\int P(x)\mathrm dx}} $,

所以原方程的积分因子为

$$
u = e^{-\int (1 - n)P(x)\mathrm dx} = e^{(n - 1)\int P(x)\mathrm dx},
$$

4.(1).
$ \frac{\mathrm dy}{\mathrm dx} = y\prime $ 

令 $ \frac{\mathrm dy}{\mathrm dx} = p = \frac{1}{t} $ ,

得，

$$ 
x = \frac{1 + p}{p^3} = t^3 + t^2
$$
$$
y = \int p \mathrm dx = \int \frac{1}{t} \mathrm d(t^3 + t^2) = \int 3t + 2 \mathrm dt = \frac{3}{2}t^2 + 2t +c
$$

则方程通解为,

$$
\begin{cases}
x = t^3 +t^2\\
y = \frac{3}{2}t^2 + 2t + c
\end{cases}
$$

则有

$$ 
(tx)^3 - x^3(1 - tx) = 0

$$
x = t^2 - \frac{1}{t}
$$
从而 $ y = \int p\mathrm dx = \int t(t^2 - \frac{1}{t})\mathrm d{t^2 - \frac{1}{t}} = \int (t^3 - 1)(2t + \frac{1}{t^2})\mathrm dt = \int (2t^4 - t - \frac{1}{t^2})\mathrm dt + c = \frac{2}{5}t^5 - \frac{1}{2}t^2 + \frac{1}{t} + c $

于是求得方程参数形式得通解为，

$$
\begin{cases}
x = t^2 - \frac{1}{t}\\

\end{cases}
$$

(6).
令 $ 2 - y\prime = yt $,

则 $ 1 - y\prime = yt - 1 $, 得 $ y = t + \frac{1}{t} $,

所以

$$
\mathrm dx = \frac{\mathrm dy }{y\prime} = \frac{\mathrm dy }{2 - yt } = \frac{\mathrm d(t + \frac{1}{t})}{2 - t(t + \frac{1}{t})}= \frac{(t^2 - 1)\mathrm dt}{1 - t^2} = \frac{t^2 - 1}{t^2(1 - t^2)}\mathrm dt = -\frac{1}{t^2}\mathrm dt
$$,

从而 $  x = \int -\frac{1}{t^2}\mathrm dt = \frac{1}{t} +c $

则通解为，

$$
\begin{cases}
x = \frac{1}{t} +c\\
y = t + \frac{1}{t}
\end{cases}
$$
