##$3.验证下列各函数是相应微分方程的解：$

##$(7)y=x^2 +1,y_x =y^2 −(x^2 +1)y+2x $
**解:**$$因y=x^2 +1 有y_x=2x ；$$
    $$因y_x =y^2 −(x^2 +1)y+2x 有y_x =y^2 −y^2+2x=2x ；$$
    $$得证函数是相应微分方程的解。$$

##$(8)y=-\frac{g(x)}{f(x)},y_x =\frac{f'(x)}{g(x)}*y^2-\frac{g'(x)}{f(x)}*y^2 $
**解:**$$因y=-\frac{g(x)}{f(x)} 有y'=[f'(x)*g(x)-g'(x)*f(x)]/f(x)^2 ；$$
   $$因y'=\frac{f(x)}{g(x)}*y^2-\frac{g'(x)}{f(x)}*y^2 有y'=[\frac{f'(x)}{g(x)}-\frac{g'(x)}{f(x)}]*[-\frac{g(x)}{f(x)}]^2=[\frac{f'(x)*g(x)-g'(x)*f(x)}{f(x)}]^2 ；$$
    $$得证函数是相应微分方程的解。$$

## $4.给定一阶微分方程:\frac{\mathrm {d}y}{\mathrm {d}x}=2x$

$$x^2+c=2x+3$$
$$x^2-2x-3+c=0$$
$$有(-2)^2-4(-3+c)=0$$
$$c=4 解为：y=x^2+4$$
