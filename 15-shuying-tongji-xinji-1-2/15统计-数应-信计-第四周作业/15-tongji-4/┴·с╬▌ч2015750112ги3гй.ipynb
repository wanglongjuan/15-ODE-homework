## 一.求下列方程解
## 1.$(x+2y)*d x+x*d y=0$

$M=x+2y,N=x,\frac{\partial M}{\partial y}=2,\frac{\partial N}{\partial x}=1$

$则方程不是恰当方程。$

$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac{1}{x}$

$所以，方程有积分因子\mu=e^{\int frac{1}{x}\mathrm d x}=x$

$两边同乘以\mu=x$

$有d(x^2y+\frac{1}{3}x^3)=0$

$所以x^3+3x^2y=c$

## 2.$[xcos(x+y)+sin(x+y)]*d x+xcos(x+y）*d y=0$

$M=xcos(x+y)+sin(x+y),N=xcos(x+y）,\frac{\partial M}{\partial y}=-xsin(x+y)+cos(x+y）,\frac{\partial N}{\partial x}=cos(x+y）-xsin(x+y)$

$则方程是恰当方程$

$转化为：d(xsin(x+y))=0$

$所以xsin(x+y)=c$
## 3.$(y\cos x-x\sin x)\mathrm d x+(y\sin x+x\cos x)\mathrm d y=0$

$M=y\cos x-x\sin x,N=y\sin x+x\cos x，\frac{\partial M}{\partial y}≠\frac{\partial N}{\partial x}$

$则方程不是恰当方程。$

$而\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{M}=1$

$所以方程有积分因子\mu=e^{\int1\mathrm d y}=e^y$

$方程两边同乘以\mu=e^y$

$得\mathrm d[(y-1)e^y\sin x+e^yx\cos x]=0$

$有(y-1)e^y\sin x+e^yx\cos x=c$

## 4.$x(4y\mathrm d x+2x\mathrm d y)+y^3(3y\mathrm d x+5x\mathrm d y)=0$

$因为\mathrm d(2x^2y)=x(4y\mathrm d x+2x\mathrm d y)$

$所以y^3(3y\mathrm d x+5x\mathrm d y)=0$
$则这不是恰当方程。$

$而\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac{7}{5x}$

$有积分因子\mu=e^{\int\frac{7}{5x}\mathrm d x}=x^{\frac{7}{5}}$

$方程两边同乘以\mu=x^{\frac{7}{5}}$
$3y^4x^{\frac{7}{5}}\mathrm d x+5x^{\frac{12}{5}}y^3\mathrm d y=0$

$\mathrm d(\frac{5}{4}x^{\frac{12}{5}}y^4)=0$

$得2x^2y+\frac{5}{4}x^{\frac{12}{5}}y^4=c$

## 二.设$f(x,y)$及$\frac{\partial f}{\partial y}$连续，试证方程$\mathrm d y-f(x,y)\mathrm d x=0$为线性微分方程的充要条件是它有仅依赖于x的积分因子。

先证必要性，
设方程$\mathrm d y-f(x,y)\mathrm d x=0$是线性微分方程，

存在函数$g(x),h(x)
使得 $f(x,y)=yg(x)+h(x)$

$于是，令M=-f(x,y)=-yg(x)-h(x),N=1,$

$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac{-g(x)}{1}=-g(x),$

$所以方程有积分因子\mu=e^{\int-g(x)\mathrm d x}$

所以方程有仅依赖于x的积分因子。

再证充分性，
因为积分因子依赖于x

所以存在某个$\psi(x)$满足$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac{-\frac{\partial f}{\partial y}-0}{1}=\psi(x)$

$即\frac{\partial f}{\partial y}=-\psi(x),$

$积分后得f(x,y)=-\psi（x)y+h(x)$
$h(x)是关于x的任意可微函数$

$则原方程为\frac{\mathrm y}{\mathrm x}=-\psi（x)y+h(x),$

$所以\mathrm d y-f(x,y)\mathrm d x=0$是线性的。

## 三.求出伯努利微分方程的积分因子。

$ \frac{dy}{dx}=P(x)y+Q(x)y^n$

$两边同时乘以 y^(-n)$
$\frac{d(\frac{1}{y^{n-1}})}{dx}=\frac{(1-n)P(x)}{y^n-1}+(1-n)Q(x)$

$令\frac{1}{y^{n-1}}=Z,(1-n)P(x)=P_1(x),(1-n)Q(x)=Q_1(x)$

$替代转化为[P_1(x)Z+Q_1(x)]dx-dZ=0$

设$\mu(x,Z)为方程[P_1(x)Z+Q_1(x)]dx-dZ=0的积分因子$

$则\frac{\partial \mu}{\partial Z}(P_1(x)Z+Q_1(x))+\mu(P_1(x))=-\frac{\partial \mu}{\partial x}$

$令\frac{\partial \mu}{\partial Z}=0,则-\frac{d\mu}{\mu}=P_1(x)dx$

$两边同时积分得\mu=ce^{-\int P_1(x) dx}即\mu=ce^{-(1-n)\int P(x) dx}$

# 四.求解下列方程

## 1.$xy\prime^3=1+y\prime$

$使得y\prime=p$

$则有xp^3=1+p,x=\frac{1+p}{p^3}≠0$

$dy=pdx,dy=(-\frac{2}{p^2}-\frac{3}{p^3}）dp$

$两边同时积分$
$y=\frac{3}{2p^2}+\frac{2}{p}+c，(c为任意常数）$

所以
$
\begin{cases}
x=\frac{1+p}{p^3},\\
y=\frac{3}{2p^2}+\frac{2}{p}+c,\\
\end{cases}
$
$p≠0$

## 2.${y^\prime}^3-x^3(1-y^\prime)=0$

$令y^\prime=xt$

$则有t^3x^3-x^3(1-xt)=0$

$x=\frac{1-t^3}{t},y^\prime=1-t^3$

$\mathrm d y=y^\prime\mathrm d x=(2t^4-t-\frac{1}{t^2})\mathrm d t$

$因此y=\frac{2t^5}{5}-\frac{t^2}{2}+\frac{1}{t}+c，(c为任意常数）$

所以
$
\begin{cases}
x=\frac{1-t^3}{t},y^\prime=1-t^3,\\
y=\frac{2t^5}{5}-\frac{t^2}{2}+\frac{1}{t}+c,\\
\end{cases}
$

## 3.$y={y^\prime}^2e^{y^\prime}$

$可知y=0是方程的一个解$

$y≠0$时$
$令$y^\prime=p$则原方程变为$y=p^2e^p$,

$对x求导，得p=(2p+p^2)e^p\frac{\mathrm d p}{\mathrm d x}即\mathrm d x=(p+2)e^p\mathrm d p$

$积分得x=(p+1)e^p+c，(c为任意常数）$

所以
$
\begin{cases}
x=(p+1)e^p+c,\\
y=p^2e^p,\\
\end{cases}
$

## 4.$y(1+{y^\prime}^2)=2a(a为常数)$

$可知y=2a$是方程的一个解

$y≠2a$时，令$y^\prime=\tan t，y=2a\cos^2 t$

$而\mathrm d y=y^\prime \mathrm d x$

$有\mathrm d x=-4a\cos^2 t\mathrm dt，x=-a(2t+\sin 2t)+c，(c为任意常数）$

所以
$
\begin{cases}
x=-a(2t+\sin 2t)+c\\
y=2a\cos^2 t\\
\end{cases}
$

## 5.$x^2+{y^\prime}^2=1$

$使y^\prime=\cos t,则x=\sin t$

$有\mathrm d y=\cos t\mathrm d x=(cost)^2\mathrm d t$

$两边同时积分得y=\frac{t}{2}+\frac{\sin 2t}{4}+c，(c为任意常数）$

所以
$
\begin{cases}
x=\sin t\\
y=\frac{t}{2}+\frac{\sin 2t}{4}+c\\
\end{cases}
$

## 6.y2(y′−1)=(2−y′)2

$令2-y\prime=yt，则y^2(1-yt)=y^2t^2$

$有y=\frac{1}{t}-t,y^\prime=1+t^2，dx=\frac{\mathrm d y}{y^\prime}=-\frac{1}{t^2}dt$

$两边同时积分得x=\frac{1}{t}+c，(c为任意常数）$

所以
$
\begin{cases}
x=\frac{1}{t}+c\\
y=\frac{1}{t}-t\\
\end{cases}
$

所以$y^\prime=0$时，原式为$y^2=-4$，方程无解。
