                   第三次上机作业
1.求下列方程的解。  
（8）$(x+2y)dx+xdy=0$  
解：这里$M=x+2y,N=x,\frac{\partial M}{\partial y}=2,\frac{\partial N}{\partial x}=1,$方程不是恰当的  
因为$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac{1}{N}$,所以方程有积分因子  
$$μ=e^{\int\frac{1}{x}dx}=x$$  
以$μ=x$乘方程两边得到  
$$2xydx+x^2dy+x^2dx=0$$  
即$$d(x^2y+\frac{1}{3}x^3)=0$$  
于是，方程的通解为$$x^3+3x^2y=0$$  

(9)$[xcos(x+y)+sin(x+y)]dx+xcos(x+y)dy=0$  
解：这里$M=xcos(x+y)+sin(x+y),N=xcos(x+y),$  
$\frac{\partial M}{\partial y}=-xsin(x+y)+cos(x+y),$  
$\frac{\partial N}{\partial x}=cos(x+y)-xsin(x+y)$  
方程是恰当的，并且方程可以写成  
$$d(xsin(x+y))=0$$  
于是方程的通解为$$xsin(x+y)=c$$    

(10)$(ycosx-xsinx)dx+(ysinx+xcosx)dy=0$  
解：这里$M=ycosx-xsinx,N=ysinx+xcosx,$  
$\frac{\partial M}{\partial y}=cosx,$  
$\frac{\partial N}{\partial x}=ycosx+cosx-xsinx,$  
方程不是恰当的  
因为$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{-M}=1,$  
所以方程有积分因子$$μ=e^{\int1dy}=e^y$$  
以$μ=e^y$乘方程两边得$$(ye^ycosxdx+ye^ysinxdy)+(e^yxcosdy-e^yxsinxdx)=0,$$  
即$$d[(y-1)e^ysinx+e^yxcosx]=0$$  
于是方程的通解为$$(y-1)sinx+xcosx=ce^{-y}$$  

(11) $x(4ydx+2xdy)+y^3(3ydx+5xdy)=0$  
解：显然$x(4ydx+2xdy)=d(2x^2y)$  
对于方程$y^3(3ydx+5xdy)=0$，  
有$M=3y^4,N=5xy^3,\frac{\partial M}{\partial y}=12y^3,\frac{\partial N}{\partial x}=5y^3,$  
因为$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{5xy^3}=\frac{7}{5π}$，  
所以方程有积分因子$$μ=e^{\int\frac{7}{5π}dx}=x^{\frac{7}{5}}$$  
以$μ=x^{\frac{7}{5}}$乘以方程$y^3(3ydx+5xdy)=0$两边得  
$$3y^4x^{\frac{7}{5}}dx+5x^{\frac{12}{5}}y^3dy=0$$  
即$$d(\frac{5}{4}x^{\frac{12}{5}}y^4)=0$$  
综合以上结论，得到原方程的通解为$$2x^2y+\frac{5}{4}x^{\frac{12}{5}}y^4=c$$  
或者$$8x^2y+5x^{\frac{12}{5}}y^4=c$$  

4.设$f(x,y)$及$\frac{\partial f}{\partial y}$连续，试证方程$dy-f(x,y)dx=0$为线性微分方程的充要条件是它有仅依赖于的积分因子。  
证明：充分性:设$dy-f(x,y)dx=0$有仅依赖于x的积分因子，要说明该方程是线性的。  
由于积分因子只依赖于x，所以存在某个$φ(x)$满足$$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac{-\frac{\partial f}{\partial y}-0}{1}=φ(x)$$  
即$$\frac{\partial f}{\partial y}=-φ(x)$$  
对上式积分后得$$f(x,y)=-φ(x)y+h(x)$$  
这里$h(x)$是关于x的任意可微函数。这样，原方程可以写成$$\frac{dy}{dx}=-φ(x)y+h(x)$$  
这就证明了$dy-f(x,y)dx=0$是线性的。  
必要性：设$dy-f(x,y)dx=0$是线性微分方程，即存在$g(x),h(x)$使得$$f(x,y)=yg(x)+h(x)$$  
这样$M=-f(x,y)=-yg(x)-h(x),N=1,\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac{-g(x)}{1}=-g(x)$  
所以，方程具有积分因子$$μ=e^{\int-g(x)dx}$$  
即证明了方程有仅依赖于x的积分因子。  

8，求出伯努利微分方程的积分因子。  
解：伯努利方程为$$\frac{dy}{dx}=p(x)y+q(x)y^n,(n\neq0,1)$$  
改写成$$dy-p(x)ydy-q(x)y^ndx=0$$  
乘以$y^{-n}$得$$y^{-n}dy-p(x)y^{1-n}dx-q(x)dx=0$$  
即$$d(y^{1-n})-(1-n)p(x)y^{1-n}dx-(1-n)q(x)dx=0$$  
再乘以$e^{-(1-n)\int{p(x)dx}}$得$$e^{-(1-n)\int{p(x)dx}}[d(y^{1-n})-(1-n)p(x)y^{1-n}dx]-e^{-(1-n)\int{p(x)dx}}(1-n)q(x)dx=0$$  
即$$d[y^{1-n}e^{-(1-n)\int{p(x)dx}}]-d[\int(1-n)q(x)e^{-(1-n)\int{p(x)dx}}dx]=0$$  
这是全微分方程，因此所求的积分因子是$$μ=y^{-n}e^{-(1-n)\int{p(x)dx}}$$  

1.求解下列方程。  
（1）$xy'^3=1+y'$  
解：令$\frac{dy}{dx}=y'=\frac{1}{t}$  
则$x=t^3+t^2$  
对y求导数，则有$$t=(3t^2+2t)\frac{dt}{dy}$$  
或者$$dy=(3t+2)dt$$  
积分之，即有$$y=\frac{3}{2}t^2+2t+c$$  
所以，方程的通解为$$\begin{cases}  
x=t^3+t^2,\\ 
y=\frac{3}{2}t^2+2t+c  
\end{cases}$$  

(2)$y'^3-x^3(1-y')=0$  
解：令$y'=tx$,则原方程化为  
$$x^3(t^3-1+tx)=0$$  
解出x得$$x=\frac{1}{t}-t^2$$  
两边对y求导得$$\frac{1}{1-t^3}=(-\frac{1}{t^2}-2t)\frac{dt}{dy}$$  
或者$$dy=[(1-t^3)(-\frac{1}{t^2}-2)]dt$$  
积分之，得$$y=-\frac{1}{2}t^2+\frac{2}{5}t^5+\frac{1}{t}+c$$  
所以，方程的通解为$$\begin{cases}  
x=\frac{1}{t}-t^2\\  
y=-\frac{1}{2}t^2+\frac{2}{5}t^5+\frac{1}{t}+c  
\end{cases}$$  

(3)$y=y'^2e^{y'}$  
解：令$y'=p$，则原方程化为$$y=p^2e^p$$  
对x求导数，即得$$p=(2p+p^2)e^p\frac{dp}{dx}$$  
或者$$dx=(p+2)e^pdp$$  
积分之，得$$x=(p+1)e^p$$  
所以，方程的通解为$$\begin{cases}  
x=(p+1)e^p\\  
y=p^2e^p  
\end{cases}$$  

(4)$y(1+y'^2)=2a$(a为常数)  
解：令$y'=p$，则原方程化为$$y(1+p^2)=2a$$  
对x求导数，可得$$p(1+p^2)^2+4ap\frac{dp}{dx}=0$$  
或者$$dx=\frac{-4a}{(1+p^2)^2}dp$$  
积分之，得到$$x=-a(2arctan{p}+\frac{2p}{1+p^2}+c)$$  
所以，方程的通解为$$\begin{cases}  
x=-a(2arctan{p}+\frac{2p}{1+p^2}+c)\\  
y=\frac{2a}{1+p^2}  
\end{cases}$$  

(5)$x^2+y'^2=1$  
解：令$y'=cost$，则方程化为$$x^2+{cost}^2=1$$  
解出x,得$$x=sint$$  
对y求导数，得$$\frac{1}{cost}=cost\frac{dt}{dy}$$  
或者$$dy={cost}^2dt$$  
积分之，得$$y=\frac{t}{2}+\frac{1}{4}sin{2t}+c$$  
所以，方程的通解为$$\begin{cases}  
x=sint\\
y=\frac{t}{2}+\frac{1}{4}sin{2t}+c  
\end{cases}$$  


(6)$y^2(y'-1)=(2-y')^2$  
解：令$2-y'=yt$，则原方程化为$$y^2(1-yt)=y^2t^2$$  
由此，得$$y=\frac{1}{t}-t$$  
并且$$y'=1+t^2$$  
这是原微分方程的参数形式，因此$$dx=\frac{dy}{y'}=\frac{(-\frac{1}{t^2}-1)dt}{1+t^2}=-\frac{1}{t^2}dt$$  
积分之，得到$$x=\frac{1}{t}+c$$  
于是求得方程的参数形式的通解为$$\begin{cases}  
x=\frac{1}{t}+c\\  
y=\frac{1}{t}-t  
\end{cases}$$  
或者消去参数t，得$$y=x-\frac{1}{x-c}-c$$  
其中c为任意常数.  
此外，当$y'=0$时原方程变为$-y^2=4$，此时方程无解
