$P60页2、求下列方程的解。$

$(8)(x+2y)dx+xdy=0$

$(9)[x\cos (x+y)+\sin (x+y)]dx+x\cos (x+y)dy=0$

$(10)(y\cos x-x\sin x)dx+(y\sin x+x\cos x)dy=0$

$(11)x(4ydx+2xdy)+y^3(3ydx+5xdy)=0$

$解:(8)\frac{\partial M}{\partial y}=2,\frac{\partial N}{\partial x}=1$

$\frac{\partial M}{\partial y}≠\frac{\partial N}{\partial x},所以此方程不是恰当微分方程$

$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac{1}{x}只有x有关$

$只与x有关的积分因子u=e^{\int\mathrm {\frac{1}{x}}\,\mathrm{d}x},可得u=x$

$用u=x乘以方程的两边，得(x^2+2xy)dx+x^2dy=0$

$x^2dx+2xydx+x^2dy=0$

$积分，得方程的解为\frac{1}{3}x^3+x^2y=c(c为任意常数)$

$(9)\frac{\partial M}{\partial y}=\cos (x+y)-x\sin (x+y),\frac{\partial N}{\partial x}=\cos (x+y)-x\sin (x+y)$

$所以此方程为恰当方程$

$\frac{\partial u}{\partial x}=[x\cos (x+y)+\sin (x+y)],\frac{\partial u}{\partial y}=x\cos (x+y)$

$所以u=x\sin (x+y)$

$方程的解为x\sin (x+y)=c(c为任意常数）$

$(10)\frac{\partial M}{\partial y}=\cos x,\frac{\partial N}{\partial x}=y\cos x+\cos x-x\sin x$

$\frac{\partial M}{\partial y}≠\frac{\partial N}{\partial x},所以此方程不是恰当微分方程$

$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{-M}=1$

$只与y有关的积分因子u=e^{\int 1\,\mathrm{d}y},可得u=e^y$

$用u=e^y乘以方程的两边，得e^y(y\cos x-x\sin x)dx+e^y(y\sin x+x\cos x)dy=0$

$[e^y(y\cos x-x\sin x)dx+e^yy\sin xdy]+x\cos xdy=0$

$d[e^y(y-1)\sin x]+d(xe^y\cos x)=0$

$所以方程的解为e^y(y-1)\sin x+xe^y\cos x=c(c为任意常数）$

$(11)\frac{\partial M}{\partial y}=4x+12xy^3,\frac{\partial N}{\partial x}=4x+5xy^3$

$\frac{\partial M}{\partial y}≠\frac{\partial N}{\partial x},所以此方程不是恰当微分方程$

$因为M,N为多项式，所以设积分因子的形式为u=x^my^n$

$用积分因子乘以方程的两边，得(4x^{m+1}y{n+1}+3x^my^{n+4})dx+(2x^{m+2}y^n+5x^{m+1}y^{n+3})dy=0$

$所以它是恰当微分方程的充要条件是4(n+1)x^{m+1}y^n+3(n+4)x^my^{n+3}=2(m+2)x^{m+1}y^n+5(m+1)x^my^{n+3}$

$对比系数可得，m=2,n=1$

$所以积分因子为u=x^2y$

$方程两边乘以积分因子后为4x^3y^2dx+3x^2y^5dx+2x^4ydy+5x^3y^4dy=d(x^4y^2)+d(x^3y^5)=0$

$所以方程的解为x^4y^2+x^3y^5=c,并且x=0,y=0也是方程的解。$

$P61页4、设f(x,y)及\frac{\partial f}{\partial y}连续，试证方程dy-f(x,y)dx=0为线性微分方程的充要条件是它有仅依赖于x的积分因子。$

$解：充分性：设它有仅依赖于x的积分因子$

$所以存在φ(x)满足\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial y}}{N}=\frac{-\frac{\partial f}{\partial y}-0}{1}=φ(x)$

$\frac{\partial f}{\partial y}=-φ(x)$

$积分可得f(x,y)=-φ(x)y+h(x)(h(x)是任意关于x的可微函数）$

$这样原方程为\frac{dy}{dx}=-φ(x)y+h(x)$

$即证明了dy-f(x,y)dx=0是线性可微方程$

$必要性：方程dy-f(x,y)dx=0为线性可微方程$

$即存在g(x),h(x)使得f(x,y)=yg(x)+h(x)$

$M=-yg(x)-h(x),N=1$

$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial y}}{N}=-g(x)只与x有关$

$所以方程dy-f(x,y)dx=0为线性微分方程的充要条件是它有仅依赖于x的积分因子。$

$8、求出伯努利微分方程的积分因子。$

$解：伯努利方程的一般形式为\frac{dy}{dx}=P(x)y+Q(x)y^n$

$可以改写为dy-P(x)ydx-Q(x)y^ndx=0$

$方程两边乘以y^{-n},得y^{-n}dy-P(x)y^{1-n}dx-Q(x)dx=0$

$即d(y^{1-n})-(1-n)P(x))y^{1-n}dx-(1-n)Q(x)dx=0$

$方程两边再乘以e^{-(1-n)\int\mathrm P(x)\,\mathrm{d}x},化简得d[y^{1-n}e^{-(1-n)\int\mathrm P(x)\,\mathrm{d}x}]+d[\int\mathrm(1-n)Q(x)e^{-(1-n)\int\mathrm P(x)\,\mathrm{d}x}\mathrm{d}x]=0$

$这是全微分方程，因此所求积分因子为y^{-n}e^{-(1-n)\int\mathrm P(x)\,\mathrm{d}x}$

$P69页1、求解下列方程。$

$(1)xy'^3=1+y'$

$(2)y'^3-x^3(1-y')=0$

$(3)y=y'^2e^{y'}$

$(4)y(1+y'^2)=2a(a为常数）$

$(5)x^2+y'^2=1$

$(6)y^2(y'-1)=(2-y')^2$

$解：(1)令\frac{dy}{dx}=p,则方程为y^2(p-1)=(2-p)^2$

$x=\frac{1}{p^3}+\frac{1}{p^2}$

$再对y求导，得\frac{1}{p}=(-\frac{3}{p^4}-\frac{2}{p^3})\frac{dp}{dy}$

$\frac{dy}{dp}=-\frac{3}{p^3}-\frac{2}{p^2}$

$积分，得y=\frac{3}{2p^2}+\frac{2}{p}+c$

$所以方程的解为 \begin{cases}x=\frac{1}{p^3}+\frac{1}{p^2}\\y=\frac{3}{2p^2}+\frac{2}{p}+c\end{cases}(其中p≠0为参数，c为任意常数）$

$(2)令\frac{dy}{dx}=p,则方程为p^3-x^3(1-p)=0$

$再令p=tx,得x=\frac{1}{t}-t^2,则p=1-t^3$

$dy=pdx=(-\frac{1}{t^2}-t+2t^4)dt$

$y=\frac{1}{t}-\frac{1}{2}t^2+\frac{2}{5}t^5+c$

$所以方程的解为 \begin{cases}x=\frac{1}{t}-t^2\\y=\frac{1}{t}-\frac{1}{2}t^2+\frac{2}{5}t^5+c\end{cases}(其中t为参数，c为任意常数）$

$(3)令\frac{dy}{dx}=p,则方程为y=p^2e^p$

$再对x求导,\frac{dx}{dp}=e^p(2+p)$

$积分，得x=(p+1)e^p+c$

$所以方程的解为 \begin{cases}x=(p+1)e^p+c\\y=p^2e^p\end{cases}(其中p为参数，c为任意常数）$

$另外y=0也是方程的解$

$(4)令\frac{dy}{dx}=p,则方程为y(1+p^2)=2a$

$再令p=\tan t,则y=2a\cos^2 t$

$dx=\frac{dy}{p}=-4a\cos^2 tdt$

$积分，得x=-a(2t+sin2t)+c$

$所以方程的解为 \begin{cases}x=-a(2t+sin2t)+c\\y=2a\cos^2 t\end{cases}(其中t为参数，c为任意常数）$

$另外y=2a也是方程的解$

$(5)令\frac{dy}{dx}=p,则方程为x^2+p^2=1$

$再令p=\cos t,则x=\sin t$

$dy=pdx=\cos^2 tdt$

$积分，得y=\frac{t}{2}+\frac{1}{4}\sin 2t+c$

$所以方程的解为 \begin{cases}x=\sin t\\y=\frac{t}{2}+\frac{1}{4}\sin 2t+c\end{cases}(其中t为参数，c为任意常数）$

$(6)令\frac{dy}{dx}=p,则方程为y^2(p-1)=(2-p)^2$

$再令2-p=ty,则方程为y=\frac{1}{t}-t,p=1+t^2$

$dx=\frac{dy}{p}=-\frac{1}{t^2}dt$

$积分，得x=\frac{1}{t}+c$

$所以方程的解为 \begin{cases}x=\frac{1}{t}+c\\y=\frac{1}{t}-t\end{cases}(其中t为参数，c为任意常数）$
