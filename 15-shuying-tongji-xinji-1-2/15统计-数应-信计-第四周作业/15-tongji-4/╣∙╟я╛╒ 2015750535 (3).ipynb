**1**. $求出下列方程的解。$

$(8)~(x+2y)dx+xdy=0;$

$解：~~令M=x+2y, N=x.$

$让M  ~~~~ N分别对y和x求偏导得：$

$\frac{\partial M}{\partial y}= 2,\frac{\partial N}{\partial x}=1$

$其不是恰当微分方程，则求其积分因子为：$

$u(x)=e^(\int m(x)dx)$

$m(x)=\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}$

$m(x)=\frac{1}{x},u(x)=x$

$让方程两边同时乘以x得：~~x^2dx+2xydx+x^2dy=0$

$x^2dx+(2xydx+x^2dy)=0$

$对两边分别积分得：~~\frac{1}{3}x^3+x^2y=c$

$于是得该方程的解为：~~\frac{1}{3}x^3+x^2y=c$


$(9)[xcos(x+y)+sin(x+y）]dx+xcos(x+y)dy=0;$

$解：~~令M=xcos(x+y)+sin(x+y） N=xcos(x+y)$

$让M N分别对y和x求偏导得：$

$\frac{\partial M}{\partial y}= cos(x+y)-xsin(x+y),\frac{\partial N}{\partial x}=cos(x+y)-xsin(x+y)$

$\frac{\partial M}{\partial y}=\frac{\partial N}{\partial x},则其为恰当微分方程$

$所以该方程的解就为：~~xsin(x+y)=c$



$(10)(ycosx-xsinx)dx+(ysinx+xcosx)dy=0;$

$解：~~令M=ycosx-xsinx , N=ysinx+xcosx$

$让M N分别对y和x求偏导得：$

$\frac{\partial M}{\partial y}= cosx,\frac{\partial N}{\partial x}=ycosx+cosx-xsinx$

$其不是恰当微分方程，则求其积分因子为：$

$u(x)=e^{\int m(x)dx}$

$m(y)=\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}$

$m(y)=1$

$u(y)=e^y$

$两边都乘以e^y得：~~(e^yycosxdx+e^yysinxdy+e^ysinxdy)+(e^yxcosxdy-e^yxsinxdx+e^ycosxdx)-(e^ysinxdy+e^ycosxdx)=0$

$所以得到该方程的解为：~~~e^y(y-1)sinx+e^yxcosx=c$

$(11)x(4ydx+2xdy)+y^3(3ydx+5xdy)=0$

$解：~~两边同时乘以x^2y得：$

$(4x^3y^2dx+2x^4ydy)+(3x^2y^5dx+5x^3y^4dy)=c$

$可求得该方程的解为：~~x^4y^2+x^3y^5=c$

$4.设f(x,y)及\frac{\partial f}{\partial y}连续，试证方程dy-f(x,y)dx=0为线性微分方程的充要条件是$



$它有仅依赖于x的积分因子。$


证:** 必要性 **$:若dy-f(x,y)dx=0为线性微分方程，则有 \frac{\mathrm d y}{\mathrm d x}=P（x)y+Q(x)$


$可求得其积分因子为：u(x)=e^{-\int P(x)dx},可知其仅依赖于x.$

** 充分性 **：$设dy-f(x,y)dx=0是有仅依赖于x的积分因子，下证该方程是线性的，$

$由于积分因子只依赖于x，所以存在某个m(x)满足$

$\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}=\frac{-\frac{\partial f}{\partial y}-0}{1}=m(x)$

$即\frac{\partial f}{\partial y}=-m(x)$

$积分可得f(x,y)=-m(x)y+h(x),其中h(x)是关于x的任意可微函数.$

$这样原方程就可化为\frac{\mathrm d y}{\mathrm d x}=-m(x)y+h(x),$

$即可证明~dy-f(x,y)dx=0为线性微分方程$



$8.求出伯努\frac{\frac{\partial M}{\partial y}-\frac{\partial N}{\partial x}}{N}$

利微分方程的积分因子。$

$解：伯努利微分方程为~~ \frac{\mathrm d y}{\mathrm d x}=P(x)y+Q(x)y^n.$

$对方程两边同时除以y^n得：y^{-n} \frac{\mathrm d y}{\mathrm d x}=P(x)y^{1-n}+Q(x)$

$令z=y^{1-n}. \frac{\mathrm d z}{\mathrm d x}=(1-n)y^{-n} \frac{\mathrm d y}{\mathrm d x}$

$所以y^{-n} \frac{\mathrm d y}{\mathrm d x}=\frac{1}{1-n}\frac{\mathrm d z}{\mathrm d x}$

$所以\frac{1}{1-n}\frac{\mathrm d z}{\mathrm d x}=P(x)z+Q(x)$

$所以\frac{\mathrm d z}{\mathrm d x}=(1-n)P(x)z+(1-n)Q(x)$

$即可求得其积分因子为：e^{(n-1)\int P(x)dx}$

$即有伯努利微分方程的积分因子为：e^{(n-1)\int P(x)dx}$

$1.求下列微分方程的解$

$(1)xy'^3=1+y';$

$解：其为不显含y的方程。~~$


$令~~p=y'=\frac{\mathrm d y}{\mathrm d x}=\frac{1}{t}~~$

$可解得~~x=t^3+t^2.~~$

$~~dy=pdx=\frac{1}{t}(3t^2+2t)dt~~$

$dy=(3t+2)dt，分别对两边积分得：$

$y=\frac{3}{2}t^2+2t+c$

$所以求得该方程的解为：x=t^3+t^2~~~~y=\frac{3}{2}t^2+2t+c$


$(2)y'^3-x^3(1-y')=0;$

$解：~~其为不显含y的方程。~~$

$令p=y'=\frac{\mathrm d y}{\mathrm d x}=tx$

$即可求得x=\frac{1-t^3}{t}$

$dy=pdx=(1-t^3)(\frac{-1}{t^2}-2t)dt$

$所以dy=\frac{-1}{t^2}-t+2t^4dt$

$分别对两边求积分得y=\frac{2t^5}{5}-\frac{t^2}{2}+\frac{1}{t}+c$

$所以求得该方程的解为：x=\frac{1}{t}-t^2,~y=\frac{2t^5}{5}-\frac{t^2}{2}+\frac{1}{t}+c$


$(3)y=y'^2e^{y'};$

$解：令y'=p=\frac{\mathrm d y}{\mathrm d x}=t$

$所以y=t^2e^t$

$dx=\frac{\mathrm d y}{y'}$

$dx=2e^t+te^tdt$

$分别对两边求积分得：x=(1+t)e^t+c$

$所以求得该方程的解为：y=t^2e^t，~x=(1+t)e^t+c。$


$(4)y(1+{y'^2})=2a~(a为常数)$

$解：y'=p=\frac{\mathrm d y}{\mathrm d x}=t^{\frac{1}{2}}$

$可解得y=\frac{2a}{1+t}$

$dx=\frac{\mathrm d y}{y'}$

$dx=\frac{-2a}{t^{\frac{1}{2}}(1+t^2)}dt$

$分别对两边求积分得：x=\frac{4a}{1+t^{\frac{1}{2}}}+c$

$所以求得该方程的解为:y=\frac{2a}{1+t},~x=\frac{4a}{1+t^{\frac{1}{2}}}+c$



$(5)x^2+y'^2=1;$

$解：y'=p=\frac{\mathrm d y}{\mathrm d x}=cost$

$所以求得x=sint$

$dy=pdx={cos}^2{t}dt$

$分别对两边求积分得：y=\frac{1}{2}t+\frac{sin2t}{4}+c$

$所以求得该方程的解为:y=\frac{1}{2}t+\frac{sin2t}{4}+c,~x=sint.$


$(6)y^2(y'-1)=(2-y')^2$

$解：2-y'=yt$

$y'=p=\frac{\mathrm d y}{\mathrm d x}=t^2+1$

$dx=\frac{\mathrm d y}{y'}$

$dx=-\frac{1}{t^2}dt$

$分别对两边求积分得：x=\frac{1}{t}+c$

$所以求得该方程的解为:y=-t+\frac{1}{t},~x=x=\frac{1}{t}+c$
