# 一、9月19日作业


### 1.设f(x,y)及$\frac{\partial{f}}{\partial{y}}$连续,试证方程$dy-f(x,y)dx=0$为线性微分方程的充要条件是它有仅依赖于x的积分因子。

证明：首先，充分性：设$\frac{\partial{f}}{\partial{y}}$有仅依赖于x的积分因子，则积分因子只依赖于x，
故存在某个$\varphi(x)$满足

$$\frac{\frac{\partial{M}}{\partial{y}}-\frac{\partial{N}}{\partial{x}}}{\partial{N}}=\frac{-\frac{\partial{f}}{\partial{y}}-0}{1}=\varphi(x),$$

即

$$\frac{\partial{f}}{\partial{y}}=-\varphi(x)$$

对上式进行积分，得 

$$f(x,y)=-\varphi(x)y+h(x),$$这里h(x)是关于x的任意可微函数。这样，原方程可写成

$$\frac{dy}{dx}=-\varphi(x)y+h(x),$$

这证明了原方程是线性的。

必要性：设方程$dy-f(x,y)dx=0$是线性微分方程。即存在$g(x),h(x)$使得

$$f(x,y)=yg(x)+h(x).$$  这样$M=-f(x,y)=-yg(x)-h(x),N=1,$  

$$\frac{\frac{\partial{M}}{\partial{y}}-\frac{\partial{N}}{\partial{x}}}{\partial{N}}=\frac{-g(x)}{1}=-g(x),$$

所以，方程具有积分因子$$\mu=e^{\int -g(x)dx}$$它只依赖于x。

综上，命题得证。


### 2.求出伯努利方程的积分因子。

解：对于伯努利方程

$$\frac{dy}{dx}=P(x)y+Q(x)y^n，$$

用$y^{-n}$乘以两边，得到

$$y^{-n}\frac{dy}{dx}=y^{1-n}P(x)+Q(x),$$

引入变量代换

$$z=y^{1-n},$$  

从而

$$\frac{dz}{dx}=(1-n)y^{-n}\frac{dy}{dx}.$$

以上方程联立得

$$\frac{dz}{dx}=(1-n)P(x)z+(1-n)Q(x),$$

即   

$$[(1-n)P(x)z+(1-n)Q(x)]dx-dz=0$$

这里
$M=[(1-n)P(x)z+(1-n)Q(x)],N=-1.$
   
故

$$\frac{\frac{\partial{M}}{\partial{y}}-\frac{\partial{N}}{\partial{x}}}{\partial{N}}=(1-n)P(x),$$

故方程有只关于x的积分因子

$$\mu=e^{\int (1-n)P(x)dx}$$

即伯努利方程的积分因子为

$$\mu=e^{\int (1-n)P(x)dx}$$

### 9.设$\mu(x,y)$是方程$M(x,y)dx+N(x,y)dy=0$的积分因子，从而求得可微函数$U(x,y)$，使得$dU=\mu(Mdx+Ndy)。$试证$\tilde{\mu}
(x,y)$也是方程$M(x,y)dx+N(x,y)dy=0$的积分因子的充要条件是$\tilde{\mu}(x,y)=\mu\varphi(U)$ ，其中 $\varphi(t)$是 $t$的可微函数。


证明：充分性：设$\tilde{\mu}(x,y)=\mu\varphi(U)$,因为
$$
\begin{aligned}
\frac{\partial{(\tilde{\mu}}M)}{\partial{y}}&=\frac{\partial{(\mu M\varphi(U))}}{\partial{y}}\\  
&=\frac{\partial{(\mu M)}}{\partial{y}}\varphi(U)+\mu M \frac{\partial{(\varphi(U))}}{\partial{y}}\\
&=\frac{\partial{(\mu M)}}{\partial{y}}\varphi(U)+\mu M \varphi'(U)\frac{\partial{U}}{\partial{y}}\\
&=\frac{\partial{(\mu M)}}{\partial{y}}\varphi(U)+\mu M \varphi'(U)\mu N\\
\end{aligned}
$$

$$
\begin{aligned}
\frac{\partial{(\tilde{\mu}}N)}{\partial{x}}&=\frac{\partial{(\mu N\varphi(U))}}{\partial{x}}\\
&=\frac{\partial{(\mu N)}}{\partial{x}}\varphi(U)+\mu N \frac{\partial{(\varphi(U))}}{\partial{x}}\\
&=\frac{\partial{(\mu N)}}{\partial{x}}\varphi(U)+\mu N \varphi'(U)\mu M\\
\end{aligned}
$$

又因为$\mu(x,y)$是方程$M(x,y)dx+N(x,y)dy=0$的积分因子，所以

$$\frac{\partial{(\mu M)}}{\partial{y}}=\frac{\partial{(\mu N)}}{\partial{x}}$$

比较$\frac{\partial{(\tilde{\mu}}M)}{\partial{y}}与\frac{\partial{(\tilde{\mu}}N)}{\partial{x}}$可以得到

$$\frac{\partial{(\tilde{\mu}}M)}{\partial{y}}=\frac{\partial{(\tilde{\mu}}N)}{\partial{x}},$$

因此，$\tilde{\mu}(x,y)=\mu\varphi(\tilde{U})$是方程的积分因子，充分性得证。

必要性： 设$\tilde{\mu}$是方程的积分因子，则

$$\frac{\partial{(\tilde{\mu}}M)}{\partial{y}}=\frac{\partial{(\tilde{\mu}}N)}{\partial{x}},$$

又因为
$$
\begin{aligned}
\frac{\partial{(\tilde{\mu}}M)}{\partial{y}}&=\frac{\partial{(\frac{\tilde{\mu}}{\mu}\mu M})}{\partial{y}}\\
&=\frac{\partial{(\frac{\tilde{\mu}}{\mu}})}{\partial{y}}\mu M+\frac{\tilde{\mu}}{\mu} \frac{\partial{(\mu M})} {\partial{y}}\\
\end{aligned}
$$

$$
\begin{aligned}
\frac{\partial{(\tilde{\mu}}N)}{\partial{x}}&=\frac{\partial{(\frac{\tilde{\mu}}{\mu}\mu N})}{\partial{x}}\\
&=\frac{\partial{(\frac{\tilde{\mu}}{\mu}})}{\partial{x}}\mu N+\frac{\tilde{\mu}}{\mu} \frac{\partial{(\mu N})} {\partial{x}}\\
\end{aligned}
$$

注意到$\nu$是方程的积分因子，则有

$$\frac{\partial{(\mu M)}}{\partial{y}}=\frac{\partial{(\mu N)}}{\partial{x}}$$

比较前面4个等式，则有

$$M\frac{\partial{(\frac{\tilde{\mu}}{\mu}})}{\partial{y}}=N\frac{\partial{(\frac{\tilde{\mu}}{\mu}})}{\partial{x}},$$

由$dU=\mu(Mdx+Ndy)$以及全微分与偏微分的关系，可以得出

$$\frac{\partial{U}}{\partial{x}}=\mu M,\frac{\partial{U}}{\partial{y}}=\mu N,$$
则有

$$\frac{\partial{(\frac{\tilde{\mu}}{\mu}})}{\partial{y}}\frac{\partial{u}}{\partial{x}}=\frac{\partial{(\frac{\tilde{\mu}}{\mu}})}{\partial{x}}\frac{\partial{u}}{\partial{y}},$$

即

$$\frac{\frac{\partial{(\frac{\tilde{\mu}}{\mu}})}{\partial{y}}}{\frac{\partial{(\frac{\tilde{\mu}}{\mu}})}{\partial{x}}}=\frac{\frac{\partial{u}}{\partial{y}}}{\frac{\partial{u}}{\partial{x}}}.$$


从而得到

$$\frac{\tilde{\mu}}{\mu}=\varphi{u},$$

这里$\varphi{u}$是$u$的任意可微函数，必要性得证。

# 一、9月21日作业

### 求解下列方程：
$(1)xy'^{3}=1+y';$

$(2)y'^{3}-x^{3}(1-y')=0;$

$(3)y=y'^{2}e^{y'};$

$(4)y(1+y'^(2))=2a;$

$(5)x^{2}+y'^{2}=1;$

$(6)y^{2}(y'-1)=(2-y')^{2}.$

解：(1)令$$y'=\frac{dy}{dx}=\frac{1}{t},$$

则$$x=t^{3}+t^{2}, $$

两边对y求导，得

$$\frac{dx}{dy}=t=(3t^{2}+2t)\frac{dt}{dy},$$

$$dy=(3t+2)dt,$$

积分得

$$y=\frac{3}{2}t^{2}+2t+c.$$

所以方程通解为

$$\begin{cases}x=t^{3}+t^{2}\\y=\frac{3}{2}t^{2}+2t+c\end{cases}$$


(2)令$y'=tx,$ 则$$t^{3}x^{3}-x^{3}(1-tx)=0$$

即

$$(t^{3}-1)x^{3}+tx^{4}=0$$

得

$$x=\frac{1}{t}-t^{2}$$

两边对y求导得

$$ \frac{dx}{dy}=(-\frac{1}{t^{2}}-2t)\frac{dt}{dy}$$

即

$$\frac{1}{1-t^{3}}=(-\frac{1}{t^{2}}-2t)\frac{dt}{dy}$$

即 

$$dy=(1-t^{3})(-\frac{1}{t^{2}}-2t)dt$$

积分得

$$y=\frac{1}{t}-\frac{t^{2}}{2}+\frac{2t^{5}}{5}+c$$

所以方程通解为

$$\begin{cases}x=\frac{1}{t}-t^{2}\\y=\frac{1}{t}-\frac{t^{2}}{2}+\frac{2t^{5}}{5}+c\end{cases}$$


(3)令$y'=t,$
则

$$y=t^{2}e^{t}$$

求导得

$$y'=(2te^{t}+t^{2}e^{t})\frac{dt}{dx}=t$$

即

$$dx=(2te^{t}+t^{2}e^{t}){dt}$$

积分得

$$x=te^{t}+t^{2}e^{t}$$

所以方程通解为

$$\begin{cases}x=te^{t}+t^{2}e^{t}\\y=t^{2}e^{t}\end{cases}$$

(4)令$y'=p, $则原方程化为

$$y(1+p^{2})=2a,$$

对x求导，可得

$$p(1+p^{2})+4ap\frac{dp}{dx}=0$$

即
$$dx=\frac{-4a}{(1+p^{2})}^{2}4p,$$

积分，得

$$x=-a(2arctanp+\frac{2p}{1+p^{2}}+c)$$

所以方程通解为

$$\begin{cases}x=-a(2arctanp+\frac{2p}{1+p^{2}}+c)\\y=\frac{2a}{1+p^{2}}\end{cases}$$

(5)令$y'=cost,$则

$$x^{2}+cos^{2}t=1,$$

解得

$$x=sint$$

两边求导得

$$\frac{dx}{dy}=\frac{dt}{dy}cost$$

即

$$\frac{1}{cost}=cost\frac{dt}{dy}$$

即

$$dy=cos^{2}tdt$$

得

$$y=\frac{1}{4}sin2t+\frac{1}{2}t+c$$

故所以方程通解为

$$\begin{cases}x=sint\\y=y=\frac{1}{4}sin2t+\frac{1}{2}t+c\end{cases}$$

(6)令2-y'=yt

得

$$y^{2}(1-yt)=y^{2}t^{2}$$

即

$$-yt=t^{2}$$

即

$$y=\frac{1}{t}-t,$$

而
$$1-2+y'=t^{2}$$

即
$$y'=t^{2}-1$$

故
$$dx=\frac{dy}{y'}=\frac{-\frac{1}{t^{2}}-1}{1+t^{2}}=\frac{dt}{t^{2}},$$

积分得

$$x=\frac{1}{t}+c$$

故方程的通解为

$$\begin{cases}x=\frac{1}{t}+c,\\y=\frac{1}{t}-t,\end{cases}$$
