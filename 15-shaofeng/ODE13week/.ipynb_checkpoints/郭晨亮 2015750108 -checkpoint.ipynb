#### 2.考虑方程组

$$x\prime =\mathbf A(t)\mathbf x$$

#### 其中$\mathbf A(t)$是区间$a\leq t\leq b$上的连续$n×n$矩阵，它的元为$a_{ij}(t)(i,j=1,2,\dots ,n)$

#### （1）如果$\mathbf x_1(t),\mathbf x_2(t),\cdots,\mathbf x_n(t)$是$x\prime =\mathbf A(t)\mathbf x$的任意$n$个解，那么它们的朗斯基行列式$W[x_1(t),x_2(t),\cdots,x_n(t)]\equiv W(t)$满足下面的一阶线性微分方程

$$W\prime =[a_{11}(t)+a_{22}(t)+\cdots+a_{nn}(t)]W$$

#### （2）解上面的一阶线性微分方程，证明下面的公式：

$$W(t)=W(t_0)e^{\int_{t_0}^t[a_{11}(s)+a_{22}(s)+\cdots +a_{nn}(s)]\mathrm d s},t_0,t\in [a,b]$$

解：（1）

$$
W\prime (t)=
\begin{vmatrix}
x_{11}\prime & x_{12}\prime & \cdots & x_{1n}\prime \\
x_{21}\prime & x_{22} & \cdots & x_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
x_{n1} & x_{n2} & \cdots & x_{nn}
\end{vmatrix}
+
\begin{vmatrix}
x_{11} & x_{12} & \cdots & x_{1n} \\
x_{21}\prime & x_{22}\prime & \cdots & x_{2n}\prime \\
\vdots & \vdots & \ddots & \vdots \\
x_{n1} & x_{n2} & \cdots & x_{nn}
\end{vmatrix}
+\cdots +
\begin{vmatrix}
x_{11} & x_{12} & \cdots & x_{1n} \\
x_{21} & x_{22} & \cdots & x_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
x_{n1}\prime & x_{n2}\prime & \cdots & x_{nn}\prime
\end{vmatrix}
$$

$$
=
\begin{vmatrix}
a_{11}x_{11}+a_{12}x_{21}+\cdots +a_{1n}x_{n1} & a_{11}x_{12}+a_{12}x_{22}+\cdots +a_{1n}x_{n2} & \cdots & a_{11}x_{1n}+a_{12}x_{2n}+\cdots +a_{1n}x_{nn} \\
x_{21} & x_{22} & \cdots & x_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
x_{n1} & x_{n2} & \cdots & x_{nn}
\end{vmatrix}
$$

$$
+\cdots +
\begin{vmatrix}
x_{11} & x_{12} & \cdots & x_{1n} \\
x_{21} & x_{22} & \cdots & x_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
a_{n1}x_{11}+\cdots +a_{nn}x_{n1} & a_{n1}x_{21}+\cdots +a_{nn}x_{n2} & \cdots & a_{n1}x_{1n}+\cdots +a_{nn}x_{nn}
\end{vmatrix}
$$

$$
=
\begin{vmatrix}
a_{11}x_{11} & a_{11}x_{12} & \cdots & a_{11}x_{1n} \\
x_{21} & x_{22} & \cdots & x_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
x_{n1} & x_{n2} & \cdots & x_{nn}
\end{vmatrix}
+\cdots +
\begin{vmatrix}
x_{11} & x_{12} & \cdots & x_{1n} \\
x_{21} & x_{22} & \cdots & x_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
a_{nn}x_{n1} & a_{nn}x_{n2} & \cdots & a_{nn}x_{nn}
\end{vmatrix}
$$

$$
(a_{11}+\cdots +a_{nn})
\begin{vmatrix}
x_{11} & x_{12} & \cdots & x_{1n} \\
x_{21} & x_{22} & \cdots & x_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
x_{n1} & x_{n2} & \cdots & x_{nn}
\end{vmatrix}
=(a_{11}+\cdots +a_{nn})W(t)
$$

$$=(a_{11}(t)+\cdots +a_{nn}(t))W(t)$$

（2）

$$W\prime =[a_{11}(t)+a_{22}(t)+\cdots+a_{nn}(t)]W$$

$${\mathrm d W(t) \over W(t)}=[a_{11}(t)+\cdots +a_{nn}(t)]\mathrm d t$$

$$\int_0^t{\mathrm d W(t) \over W(t)}\mathrm d s=\int_0^t\big [a_{11}(t)+\cdots +a_{nn}(t)\big ]\mathrm d s$$

$$\ln \big |W(t)\big |-\ln \big |W(t_0)\big |=\int_0^t\big [a_{11}(t)+\cdots +a_{nn}(t)\big ]\mathrm d s$$

所以
$$W(t)=W(t_0)e^{\int_{t_0}^t[a_{11}(s)+a_{22}(s)+\cdots +a_{nn}(s)]\mathrm d s},t_0,t\in [a,b]$$

#### 2.设$\mathbf A(t)$为区间$a\leq t\leq b$上的连续$n×n$实矩阵，$\mathbf \varPhi (t)$为方程$\mathbf x\prime =\mathbf A(t)\mathbf x$的基解矩阵，而$\mathbf x=\mathbf \varphi (t)$为其一解。试证：

#### （1）对于方程$\mathbf y\prime =-\mathbf A^T(t)\mathbf y$的任一解$\mathbf y\prime =\mathbf \psi(t)$必有$\mathbf \psi ^T(t)\mathbf \varphi (t)=$常数

#### （2）$\mathbf \varPsi (t)$为方程$\mathbf y\prime =-\mathbf A^T(t)\mathbf y$的基解矩阵的充要条件是存在非奇异的常数矩阵$\mathbf C$，使$\mathbf \varPsi ^T(t)\mathrm \varPhi (t)=\mathbf C$

解：（1）
$$[\mathbf \varPsi ^T(t)\mathbf \varphi (t)]\prime =(\mathbf \varPsi ^T(t))\prime \mathbf \varphi (t)+\mathbf \varPsi ^T(t)\mathbf \varphi \prime (t)$$

$$=(\mathbf \varPsi ^T(t))\prime +\mathbf \varPsi ^T(t)\mathbf A(t)\mathbf \varphi (t)$$

$$\mathbf \varPhi \prime (t)=-\mathbf A^T(t)\mathbf \varPsi(t)$$

$$(\mathbf \varPsi ^T(t))\prime =-\mathbf \varPsi ^T(t)\mathbf A(t)$$

$$[\mathbf \varPsi ^T(t)\mathbf \varphi (t)]\prime =-\mathbf \varPsi ^T(t)\mathbf A(t)\mathbf \varphi (t)+\mathbf \varPsi ^T(t)\mathbf A(t)\mathbf \varphi (t)=0$$

所以对于方程$\mathbf y\prime =-\mathbf A^T(t)\mathbf y$的任一解$\mathbf y\prime =\mathbf \psi(t)$必有

$$\mathbf \psi ^T(t)\mathbf \varphi (t)=常数$$

（2）$\mathbf \varPsi (t)$为方程$\mathbf y\prime =-\mathbf A^T(t)\mathbf y$的基解矩阵

$$[\mathbf \varPsi ^T(t)\mathbf \varPhi (t)]\prime =[\mathbf \varPsi ^T(t)]\prime \mathbf \varPhi (t)+\mathbf \varPsi ^T(t)\mathbf \varPhi \prime (t)$$

$$=[-\mathbf A^YT(t)\mathbf \varPsi (t)]^T\mathbf \varPhi(t)+\mathbf \varPsi ^T(t)\mathbf A(t)\mathbf \varPhi (t)$$

$$=-\mathbf \varPsi ^T(t)\mathbf A(t)\mathbf \varPhi (t)+\mathbf \varPsi ^T(t)\mathbf A(t)\mathbf \varPhi (t)=0$$

所以
$$\mathbf \varPsi ^T(t)\mathbf \varPhi (t)=\mathbf C$$

因为
$$\det (\mathbf \varPhi (t))\neq 0,\det (\varPsi (t))\neq 0$$

所以$\mathbf C$非奇异
