## 1.求解下列方程：

## $（2）xx^{\prime\prime}-(x^\prime)^2+(x^\prime)^3=0$

**解**
令$$x \prime=y,则x \prime\prime=y\frac{dy}{dx}$$

代入原方程得$$xy\frac{dy}{dx}-y^2+y^3=0$$

解之得$$y=0,x\frac{dy}{dx}-y+y^2=0$$

①$$y=0,可得到x=c$$

②$$y≠0,则\frac{1}{x}\mathrm d x=(\frac{1}{y}-\frac{1}{y-1})\mathrm d y$$

两边同时积分得$$y=\frac{c_1x}{c_1x-1}$$

即$$\frac{dx}{dt}=\frac{c_1x}{c_1x-1}$$

两边积分得$$x+c_1\ln |x|=t+c_2(其中c_1,c_2为任意常数）$$

综上所述，方程的解为$$x+c_1\ln |x|=t+c_2(其中c_1,c_2为任意常数）或x=c$$

## $(4)x^{\prime\prime}+\sqrt{1-(x^\prime)^2}=0$

**解**
令$x \prime=y,则x \prime\prime=\frac{dy}{dt}$

代入原方程得$$\frac{\mathrm d y}{\mathrm d t}=-\sqrt{1-y^2}$$

①$y=±1时，x=±t+c$

②$y≠±1时$

有$$\frac{1}{\sqrt{1-y^2}}\mathrm d y=-\mathrm d t$$

两边同时积分得$$\arcsin y=-t+c_1,即x\prime=sin(c_1-t)$$

解之得$$x=cos(t-c_1)+c_2$$

综上所述，通解为$$x=±t+c,或x=cos(t-c_1)+c_2$$

## $(6)x^{\prime\prime}-\frac{1}{t}x^\prime+(x^\prime)^2=0(提示：方程两边除以x^\prime)$

**解**
令$x\prime=y,x \prime\prime=y \prime=\frac{dy}{dt}$

①当$y=0,x=c$

②当$y≠0时$

代入原方程得$$\frac{dy}{dt}=\frac{1}{t}y-y^2$$

令$z=\frac{1}{y}$

得到$$\frac{dz}{dt}=-\frac{1}{t}z+1$$

由常数变易法得到$$z=\frac{1}{2}t+\frac{c_1}{t}$$

所以$$\frac{dx}{dt}=\frac{2t}{t^2+2c_1}$$

积分得$$x=ln{t^2+2c_1}+lnc_2$$

即$$e^x=c_2(t^2+2c_1)$$

综上所述方程的通解为$$e^x=c_2(t^2+2c_1)或者x=c$$

## 2.用幂级数解法求解下列方程：

## $(2)(1-t)x^{\prime\prime}+x=0$

**解**

设
$$x=a_0+a_1t+a_2t^2+…a_kt^k+…a_nt^n+…$$
为方程的解，这里$a_i(i=0,1,2,…，n,…）$是待定常数

将它对x微分两次，有
$$x^{\prime\prime}=2a_2+6a_3t+…k(k-1)a_kt^{k-2}+…n(n-1)a_nt^{n-2}+…$$

代入原方程，比较t的系数，可知
$$2a_2+a_0=0$$
$$6a_3-2a_2+a_1=0$$
$$k(k-1)a_k-(k-1)(k-2)a_{k-1}+a_{k-2}=0(k≥3）$$

求方程的通解，可设$x_(0)=c_1,x^\prime(0)=c_2$

因为$$a_k=\frac{k-2}{k}a_{k-1}-\frac{1}{k(k-1)}a_{k-2}（k>3）$$

所以有
$$a_0=c_1$$
$$a_1=c_2$$
$$a_2=-\frac{1}{2}c_1$$
$$a_3=-\frac{1}{3!}c_1-\frac{1}{3!}c_2$$

一直循环下去
因此逐次递推求得幂级数解

$$x=c_1(1-\frac{1}{2}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-\frac{2}{5!}t^5-…）+c_2(t-\frac{1}{3!}t^3-\frac{2}{4!}t^4-\frac{5}{5!}t^5-\frac{18}{6!}t^6-…)$$

## 2.将下面的初值问题化为与之等价的一阶方程组的初值问题

$（3）
\begin{cases}
x^{\prime\prime}+5y^\prime-7x+6y=e^t\\
y^{\prime\prime}-2y+13y^\prime-15x=\cos t
\end{cases}
$

$x(0)=1,x^\prime(0)=0,y(0)=0,y^\prime(0)=1.$

$(提示：令\omega_1=x,\omega_2=x^\prime,\omega_3=y,\omega_4=y^\prime.)$

**解**

令$ω1=x,ω2=x′,ω3=y,ω4=y′$
代入原方程组得
$$
\begin{cases}
\omega_2\prime+5\omega_4-7\omega_1+6\omega_3=e^t\\
\omega_4\prime-2\omega_3+13\omega_4-15\omega_1=\cos t
\end{cases}
$$
即
$$
\begin{cases}
\omega_1^\prime=x^\prime=  \omega_2\\
\omega_2^\prime=x^{\prime\prime}=7\omega_1 -6\omega_3-5\omega_4+e^t\\
\omega_3^\prime=y^\prime=\omega_4\\
\omega_4^\prime=15\omega_1+2\omega_3-13\omega_4+\cos t
\end{cases}
$$
所以
$$
\mathbf A(t)=
\begin{bmatrix}
0 & 1 & 0 & 0\\
7 & 0 & -6 & -5\\
0 & 0 & 0 & 1\\
15 & 0 & 2 & -13\\
\end{bmatrix}
$$

$$
\mathbf f(t)=
\begin{bmatrix}
0\\
e^t\\
0\\
\cos t\\
\end{bmatrix}
$$

$$
\mathbf \omega(0)=\mathbf \eta=
\begin{bmatrix}
1\\
0\\
0\\
1\\
\end{bmatrix}
$$

$\omega^\prime=A(t)\omega+f(t)$
把上式代入即可

## 3.试用逐步逼近法求方程组
$$
\mathbf x^\prime=
\begin{bmatrix}
0 &1\\
-1& 0\\
\end{bmatrix}\mathbf x
$$
## 满足初值条件
$$
\mathbf x(0)=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}
$$
## 的第三次近似解。


**解**
构造逐步逼近序列
$
\begin{cases}
\varphi_0(t)=\eta=x(0)\\
\varphi_k(t)=x(0)+\int_0^t [A(t)\varphi_{k-1}(t)+f(t)]\mathrm d t
\end {cases}
$

其中
$
\mathbf A(t)=
\begin{bmatrix}
0 &1\\
-1& 0\\
\end{bmatrix}，
\mathbf f(t)=
\begin{bmatrix}
0\\
0\\
\end{bmatrix}
$


下面进行第一次迭代
$$
\mathbf \varphi_1(t)=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+
\int_0^t 
\begin{bmatrix}
0 &1\\
-1& 0\\
\end{bmatrix}
\begin{bmatrix}
0\\
1\\
\end{bmatrix}\mathrm d t=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+\int_0^t 
\begin{bmatrix}
1\\
0\\
\end{bmatrix}\mathrm d t=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+\begin{bmatrix}
t\\
0\\
\end{bmatrix}=
\begin{bmatrix}
t\\
1\\
\end{bmatrix}
$$

同理可得
$$
\begin{aligned}
\mathbf \varphi_2(t)=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+
\int_0^t 
\begin{bmatrix}
0 &1\\
-1& 0\\
\end{bmatrix}
\begin{bmatrix}
t\\
1\\
\end{bmatrix}\mathrm d t=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+\int_0^t 
\begin{bmatrix}
1\\
-t\\
\end{bmatrix}\mathrm d t=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+\begin{bmatrix}
t\\
-\frac{1}{2}t^2\\
\end{bmatrix}=
\begin{bmatrix}
t\\
1-\frac{1}{2}t^2\\
\end{bmatrix}
\end{aligned}
$$

$$
\begin{aligned}
\mathbf \varphi_3(t)=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+
\int_0^t 
\begin{bmatrix}
0 &1\\
-1& 0\\
\end{bmatrix}
\begin{bmatrix}
t\\
1-\frac{1}{2}t^2\\
\end{bmatrix}\mathrm d t=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+\int_0^t 
\begin{bmatrix}
1-\frac{1}{2}t^2\\
-t\\
\end{bmatrix}\mathrm d t=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+\begin{bmatrix}
t-\frac{1}{6}t^3\\
-\frac{1}{2}t^2\\
\end{bmatrix}=
\begin{bmatrix}
t-\frac{1}{6}t^3\\
1-\frac{1}{2}t^2\\
\end{bmatrix}
\end{aligned}
$$

所以满足条件的第三次近似解为
$$
\mathbf \varphi_3(t)=
\begin{bmatrix}
t-\frac{1}{6}t^3\\
1-\frac{1}{2}t^2\\
\end{bmatrix}
$$
