## 习题4.3

## 1.求解下列方程：

## （2）$xx^{\prime\prime}-(x^\prime)^2+(x^\prime)^3=0$

**解**

显然，该方程不含自变量t

令$x^\prime=y$, 则 

$
x^{\prime\prime}=y^\prime=\frac{\mathrm d y}{\mathrm d t}=\frac{\mathrm d x}{\mathrm d t}\frac{\mathrm d y}{\mathrm d x}=x^\prime\frac{\mathrm d y}{\mathrm d x}=y\frac{\mathrm d y}{\mathrm d x}
$

所以原方程化为

$$
xy\frac{\mathrm d y}{\mathrm d x}-y^2+y^3=0
$$

所以

$$
xy\frac{\mathrm d y}{\mathrm d x}=y^2(1-y)
$$

1>若 $y=0$, 则 $x=c$, 满足方程。

2>若 $y≠0$, 则 $\frac{1}{x}\mathrm d x=(\frac{1}{y}-\frac{1}{y-1})\mathrm d y$

两边积分得

$$
y=\frac{c_1x}{c_1x-1}
$$

即

$$
\frac{\mathrm d x}{\mathrm d t}=x^\prime=\frac{c_1x}{c_1x-1}
$$

两边积分得

$$
x+c_1\ln |x|=t+c_2(其中c_1,c_2为任意常数）
$$

综上所述，方程的解为

$$
x+c_1\ln |x|=t+c_2(其中c_1,c_2为任意常数）或x=c
$$

## 4) $x^{\prime\prime}+\sqrt{1-(x^\prime)^2}=0$

**解**

同（2），令 $y=x^\prime$, 则 $x^{\prime\prime}=y^\prime=\frac{\mathrm d y}{\mathrm d t}$

所以原方程可化为 

$$
\frac{\mathrm d y}{\mathrm d t}=-\sqrt{1-y^2}
$$

即

$$
\frac{1}{\sqrt{1-y^2}}\mathrm d y=-\mathrm d t
$$

1>若 $y=±1$, 则 $x=±t+c$ (c为任意常数）满足方程

2>若 $y≠±1$, 两边积分得

$$
\arcsin y=-t+c_1
$$

即

$$
x^\prime=y=\sin (c_1-t)
$$

所以

$$
x=\cos(-t+c_1)+c_2=\cos(t-c_1)+c_2
$$

(其中c_1,c_2为任意常数）

综上所述，该方程的解为

$$
x=\cos(t-c_1)+c_2（其中c_1,c_2为任意常数）或者x=±t+c(c为任意常数）
$$

## (6) $x^{\prime\prime}-\frac{1}{t}x^\prime+(x^\prime)^2=0$(提示：方程两边除以 $x^\prime$)

**解**

1>当 $x^\prime=0$ 时，$x=c$，显然其满足方程

2>当 $x^\prime≠0$ 时，方程两边同除以 $x^\prime$, 则得到

$$
\frac{x^{\prime\prime}}{x^\prime}-\frac{1}{t}+x^\prime=0
$$

即

$$
\mathrm d(\ln |x^\prime|-\ln |t|+x)=0
$$

积分可得

$$
ln |\frac{x^\prime}{t}|+x=c_1
$$

所以

$$
x^\prime=t(e^{-x+c_1})
$$

两边再次积分最终化解得

$$
e^x=c_1(t^2+c_2)
$$

其中 $c_1,c_2$ 为任意常数

综上所述，方程的通解为

$$
e^x=c_1(t^2+c_2)或者x=c
$$

## 2.用幂级数解法求解下列方程：

## (2)$(1-t)x^{\prime\prime}+x=0$

**解**

设

$$
x=a_0+a_1t+a_2t^2+…a_kt^k+…a_nt^n+…
$$

为方程的解，这里 $a_i(i=0,1,2,…，n,…）$ 是待定常数

将它对 $x$ 微分两次，有

$$
x^{\prime\prime}=2a_2+6a_3t+…k(k-1)a_kt^{k-2}+…n(n-1)a_nt^{n-2}+…
$$

代入原方程，比较 $t$ 的系数，可知

$$
\begin{aligned}
2a_2+a_0&=0\\6a_3-2a_2+a_1&=0\\k(k-1)a_k-(k-1)(k-2)a_{k-1}+a_{k-2}&=0(k≥3）
\end{aligned}
$$

求方程的通解，可设 $x_(0)=c_1,x^\prime(0)=c_2$

因为

$$
a_k=\frac{k-2}{k}a_{k-1}-\frac{1}{k(k-1)}a_{k-2}（k>3）
$$

所以有

$$
\begin{aligned}
a_0&=c_1,\\a_1&=c_2,\\a_2&=-\frac{1}{2}c_1,\\a_3&=-\frac{1}{3!}c_1-\frac{1}{3!}c_2,\\\cdots
\end{aligned}
$$

因此逐次递推求得幂级数解

$$
x=c_1(1-\frac{1}{2}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-\frac{2}{5!}t^5-…）+c_2(t-\frac{1}{3!}t^3-\frac{2}{4!}t^4-\frac{5}{5!}t^5-\frac{18}{6!}t^6-…)
$$

## 2.将下面的初值问题化为与之等价的一阶方程组的初值问题

（3）
$$
\begin{cases}
x^{\prime\prime}+5y^\prime-7x+6y=e^t\\
y^{\prime\prime}-2y+13y^\prime-15x=\cos t
\end{cases}
$$

$x(0)=1,x^\prime(0)=0,y(0)=0,y^\prime(0)=1.$

(提示：令 $\omega_1=x,\omega_2=x^\prime,\omega_3=y,\omega_4=y^\prime$)

**解**

令 $\omega_1=x,\omega_2=x^\prime,\omega_3=y,\omega_4=y^\prime$

由已知条件化解可得

$$
\begin{cases}
\omega_1^\prime=x^\prime=  \omega_2\\
\omega_2^\prime=x^{\prime\prime}=7\omega_1 -6\omega_3-5\omega_4+e^t\\
\omega_3^\prime=y^\prime=\omega_4\\
\omega_4^\prime=15\omega_1+2\omega_3-13\omega_4+\cos t
\end{cases}
$$

并且

$$
\begin{cases}
\eta_1=\omega_1(0)=x(0) =1\\
\eta_2=\omega_2(0)=x^\prime(0)=0\\
\eta_3=\omega_3(0)=y(0) =0\\
\eta_4=\omega_4(0)=y^\prime(0)=1
\end{cases}
$$

所以

$$
\mathbf A(t)=
\begin{bmatrix}
0 & 1 & 0 & 0\\
7 & 0 & -6 & -5\\
0 & 0 & 0 & 1\\
15 & 0 & 2 & -13\\
\end{bmatrix}
$$

$$
\mathbf f(t)=
\begin{bmatrix}
0\\
e^t\\
0\\
\cos t\\
\end{bmatrix}
$$

$$
\mathbf \omega(0)=\mathbf \eta=
\begin{bmatrix}
1\\
0\\
0\\
1\\
\end{bmatrix}
$$


因为 $\omega^\prime=A(t)\omega+f(t)$

从而与之等价的初值问题为

$$
\begin{cases}
\mathbf \omega^\prime=
\begin{bmatrix}
0 & 1 & 0 & 0\\
7 & 0 & -6 & -5\\
0 & 0 & 0 & 1\\
15 & 0 & 2 & -13\\
\end{bmatrix}\mathbf \omega+
\begin{bmatrix}
0\\
e^t\\
0\\
\cos t\\
\end{bmatrix}\\
\mathbf \omega(0)=
\begin{bmatrix}
1\\
0\\
0\\
1\\
\end{bmatrix}
\end{cases}
$$

## 3.试用逐步逼近法求方程组

$$
\mathbf x^\prime=
\begin{bmatrix}
0 &1\\
-1& 0\\
\end{bmatrix}\mathbf x
$$

满足初值条件

$$
\mathbf x(0)=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}
$$

的第三次近似解。


**解**

依题意可构造逐步逼近序列如下

$$
\begin{cases}
\varphi_0(t)=\eta=x(0)\\
\varphi_k(t)=x(0)+\int_0^t [A(s)\varphi_{k-1}(s)+f(s)]\mathrm d s
\end {cases}
$$

其中

$$
\mathbf A(s)=
\begin{bmatrix}
0 &1\\
-1& 0\\
\end{bmatrix}，
\mathbf f(s)=
\begin{bmatrix}
0\\
0\\
\end{bmatrix}
$$

所以

$$
\begin{aligned}
\mathbf \varphi_1(t)&=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+
\int_0^t 
\begin{bmatrix}
0 &1\\
-1& 0\\
\end{bmatrix}
\begin{bmatrix}
0\\
1\\
\end{bmatrix}\mathrm d s\\&=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+\int_0^t 
\begin{bmatrix}
1\\
0\\
\end{bmatrix}\mathrm d s\\&=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+\begin{bmatrix}
t\\
0\\
\end{bmatrix}\\&=
\begin{bmatrix}
t\\
1\\
\end{bmatrix}
\end{aligned}
$$

$$
\begin{aligned}
\mathbf \varphi_2(t)&=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+
\int_0^t 
\begin{bmatrix}
0 &1\\
-1& 0\\
\end{bmatrix}
\begin{bmatrix}
s\\
1\\
\end{bmatrix}\mathrm d s\\&=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+\int_0^t 
\begin{bmatrix}
1\\
-s\\
\end{bmatrix}\mathrm d s\\&=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+\begin{bmatrix}
t\\
-\frac{1}{2}t^2\\
\end{bmatrix}\\&=
\begin{bmatrix}
t\\
1-\frac{1}{2}t^2\\
\end{bmatrix}
\end{aligned}
$$

$$
\begin{aligned}
\mathbf \varphi_3(t)&=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+
\int_0^t 
\begin{bmatrix}
0 &1\\
-1& 0\\
\end{bmatrix}
\begin{bmatrix}
s\\
1-\frac{1}{2}s^2\\
\end{bmatrix}\mathrm d s\\&=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+\int_0^t 
\begin{bmatrix}
1-\frac{1}{2}s^2\\
-s\\
\end{bmatrix}\mathrm d s\\&=
\begin{bmatrix}
0\\
1\\
\end{bmatrix}+\begin{bmatrix}
t-\frac{1}{6}t^3\\
-\frac{1}{2}t^2\\
\end{bmatrix}\\&=
\begin{bmatrix}
t-\frac{1}{6}t^3\\
1-\frac{1}{2}t^2\\
\end{bmatrix}
\end{aligned}
$$

所以满足条件的第三次近似解为

$$
\mathbf \varphi_3(t)=
\begin{bmatrix}
t-\frac{1}{6}t^3\\
1-\frac{1}{2}t^2\\
\end{bmatrix}
$$
