**1.求解下列方程。**

**$(2)\quad x x^{\prime\prime}-(x^\prime)^2+(x^\prime)^3=0$**

**解：**

令 $x^\prime=y$ ,则$x^{\prime\prime}=y\frac{\mathrm dy}{\mathrm dx}$

代入原方程得$$y\frac{\mathrm dy}{\mathrm dx}=y\frac{y-y^2}{x}$$

$y=0$ 时 $$x=c$$

$y\neq 0$ 时 $$\frac{1}{y-y^2}\mathrm dy=\frac{1}{x}\mathrm dx$$

所以$$ln\frac{y}{1-y}=cx$$

所以$$x^\prime=y=\frac{cx}{1+cx}=\frac{x}{c_1+x}$$

所以$$x+c_1ln|x|=t+c_2$$

所以方程的解为$$x=c;\quad x+c_1ln|x|=t+c_2$$

**$(4)\quad x^{\prime\prime}+\sqrt{1-(x^\prime)^2}+(x^\prime)^3=0$**

**解：**

令 $x^\prime=y$ ,则$x^{\prime\prime}=y\frac{\mathrm dy}{\mathrm dx}$

代入原方程得$$y\frac{\mathrm dy}{\mathrm dx}+\sqrt{1-y^2}=0$$

$\sqrt{1-y^2}=0$ 时 $$x=\pm t+c$$

$\sqrt{1-y^2}\neq 0$ 时 $$\frac{-y}{\sqrt{1-y^2}}\mathrm dy=\mathrm dx$$

解得$$\sqrt{1-y^2}=x+c$$

所以$$x^\prime =y=\pm\sqrt{1-(x+c)^2} $$

所以$$\frac{1}{\pm\sqrt{1-(x+c)^2}}\mathrm dx=\mathrm dt$$

所以$$arcsin(x+c)=\pm t+c_1$$

所以$$sin(\pm t+c_1)=x+c$$

所以方程的解为$$sin(\pm t+c_1)=x+c ;\quad x=\pm t+c$$

**$(6)\quad x^{\prime\prime}-\frac{1}{t}x^\prime+(x^\prime)^2=0$**

**解：**

令 $x^\prime=y$ ,则$x^{\prime\prime}=\frac{\mathrm dy}{\mathrm dt}$


代入原方程得$$\frac{\mathrm dy}{\mathrm dt}-\frac{1}{t}+y^2=0$$


两边同除以 $y^2$ 得 $$\frac{1}{y^2}\frac{\mathrm dy}{\mathrm dt}=\frac{1}{t}+1$$





令 $z=\frac{1}{y}$ 上式可化为$$\frac{\mathrm dz}{\mathrm dt}=-\frac{z}{t}+1$$

先求解方程$$\frac{\mathrm dz}{\mathrm dt}=-\frac{z}{t}$$

解得$$z=\frac{c}{t}$$

设方程$\frac{\mathrm dz}{\mathrm dt}=\frac{z}{t}+1 \quad 的解为 \quad  z=\frac{c(t)}{t}$

代入方程解得$$c(t)=\frac{1}{2}t^2+c_1$$

所以$$z=\frac{c(t)}{t}=\frac{t}{2}+\frac{c_1}{t}$$





$\therefore \quad$  $$\frac{\mathrm dx}{\mathrm dt}=\frac{t}{2}+\frac{c_1}{t}$$

所以$$x+c_2=ln|t^2+c_1| \quad $$

即 $$\quad e^x=c_3(t^2+c_1)$$

所以方程的解为$$\quad e^x=c_3(t^2+c_1) \quad x=c $$

$c ,c_1 ,c_2 ,c_3 \quad 为常数$


**2.用幂级数求解下列方程。**

**$(2)\quad (1-t)x^{\prime\prime}+x=0$**

**解：**
设解为$$y=a_0+a_1t+a_2t^2+\cdots+a_nt^n+\cdots$$

则

$$y^\prime=a_1+2a_2t+3a_3t^2+\cdots+na_nt^{n-1}+\cdots$$

$$y^{\prime\prime}=2a_2+2\times 3a_3t+\cdots+(n-1)\times na_nt^{n-2}+\cdots$$

代入原方程比较系数得

$$
\begin{aligned}
a_0+2a_2 &=0\\
a_1-2a_2+2\times 3a_3 &=0\\
a_2-2\times 3+3\times 4a_4 &=0\\
\cdots\\
a_{n-2}-3\times 4a_4+4\times 5a_5 &=0
\end{aligned}
$$

所以
$$
\begin{aligned}
a_2 &=-\frac{1}{2}a_0\\
a_3 &=-\frac{1}{3!}a_0-\frac{1}{3!}a_1\\
a_4 &=-\frac{1}{4!}a_0-\frac{2}{4!}a_1\\
a_5 &=-\frac{2}{5!}a_0-\frac{1}{5!}a_1\\
\cdots\\
a_n &=-\frac{a_0+a_1+\cdots+a_{n-2}}{(n-1)n}
\end{aligned}
$$

因此方程的解为$$x=a_0 \big( 1-\frac{1}{2}t^2-\frac{1}{3!}t^3-\frac{1}{4!}t^4-\frac{2}{5!}t^5-\cdots \big)
            +a_1 \big( t-\frac{1}{3!}t^3-\frac{2}{4!}t^4-\frac{1}{5!}t^5-\cdots \big)$$
