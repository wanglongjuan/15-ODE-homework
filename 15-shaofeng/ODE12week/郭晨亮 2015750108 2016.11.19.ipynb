#### 1.求解下列方程

（1）$xx\prime \prime -(x\prime )^2+(x\prime )^3=0$

解：设$y=x\prime $
$$xy{\mathrm d y \over \mathrm d x}-y^2+y^3=0$$

$$y=0或{\mathrm d y \over \mathrm d x}={y-y^2 \over x}$$

$${y \over 1-y}=cx$$

$$x\prime =y={cx \over 1+cx}$$

$$x+{1 \over c}\ln |x|=t+c_2$$

$$x+c_1\ln |x|=t+c_2$$

$y=0$时
$$x\prime =0,x=c$$

所以方程的解为
$$x+c_1\ln |x|=t+c_2,x=c$$

（2）$x\prime \prime +\sqrt {1-(x\prime )^2}=0$

解：设$y=x\prime $
$$y{\mathrm d y \over \mathrm d x}+\sqrt {1-y^2}=0$$

$y\neq 1,y\neq -1$时
$${y\mathrm d y \over \sqrt {1-y^2}}=-\mathrm d x$$

$$\sqrt {1-y^2}=x+c$$

$$\sqrt {1-x\prime ^2}=x+c$$

$$x\prime =-\sqrt {1-(x+c)^2}或x\prime =\sqrt {1-(x+c)^2}$$

$x\prime =-\sqrt {1-(x+c)^2}$时
$$x=\sin (t+c_1)-c_2$$

$x\prime =\sqrt {1-(x+c)^2}$时
$$x=\cos (t+c_1)-c_2$$

$y=-1或y=-1$时
$$x\prime =1或x\prime =-1$$

$$x=-t+c或x=t+c$$

所以方程的解为
$$x=\sin (t+c_1)-c_2,x=\cos (t+c_1)-c_2,x=-t+c,x=t+c$$

（3）$x\prime \prime -{1 \over t}x\prime +(x\prime )^2=0$

解：$x\prime \neq 0$时
$${x\prime \prime \over x\prime }-{1 \over t}+x\prime =0$$

$$\mathrm d (\ln |x\prime |-\ln |t|+x)=0$$

$$\ln \big |{x\prime \over t}\big |+x=c_1$$

$$x\prime =te^{-x+c_1}$$

$$e^x=e^{c_1}({t^2 \over 2})+c_2$$

$$e^x=c_1(t^2+c_2)$$

$x\prime =0$时
$$x=c$$

所以方程的解为
$$e^x=c_1(t^2+c_2),x=c$$

#### 2.用幂级数解法求解下列方程

$(1-t)x\prime \prime +x=0$

解：
$$x\prime \prime +{x \over 1-t}=0$$

$$tp(t)=0,t^2q(t)={t^2 \over 1-t}$$

设方程的解
$$x=a_0+a_1t+a_2t^2+a_3t^3+…+a_nt^n+…$$

$$x\prime =a_1+2a_2t+3a_3t^2+…+na_nt^{n-1}…$$

$$x\prime \prime =2a_2+6a_3t+…+n(n-1)a_nt^{n-2}+…$$

$$2a_2+a_0=0$$

$$6a_3-2a_2+a_1=0$$

$$k(k-1)a_k-(k-1)(k-2)a_{k-1}+a_{k-2}=0$$

设x_1(t)和x_2(t)是方程的解满足
$$x_1(0)=x_2\prime (1)=0,x_1\prime (0)=x_2(0)=1$$

对$x_1(t)$
$$a_0=0,a_1=1,…,a_n={(n-1)(n-2)a_{n-1}-a_{n-2} \over n(n-1)}$$

$$x_1(t)=t-{1 \over 3!}t^3-{2 \over 4!}t^4-{5 \over 5!}t^5-{18 \over 6!}t^6-…$$

对$x_2(t)$
$$a_0=1,a_2=0,…a_n={(n-1)(n-2)a_{n-1}-a_{n-2} \over n(n-1)}$$

$$x_2(t)=1-{1 \over 2!}t^2-{1 \over 3!}t^3-{1 \over 4!}t^4-…$$

所以方程的解为
$$x=c_1x_1(t)+c_2x_2(t)$$

$$=c_1(t-{1 \over 3!}t^3-{2 \over 4!}t^4-{5 \over 5!}t^5-{18 \over 6!}t^6-…)+c_2(1-{1 \over 2!}t^2-{1 \over 3!}t^3-{1 \over 4!}t^4-…)$$
