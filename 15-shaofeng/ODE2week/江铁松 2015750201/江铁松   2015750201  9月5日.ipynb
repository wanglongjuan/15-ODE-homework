### 4、给定一阶微分方程$$\frac{dy}{dx}=2x$$

（1）求出它的通解；

（2）求通过点（1，4）的特解；

（3）求出与直线$$y=2x+3$$相切的特解；

（4）求出满足条件$$\int_0^1 ydx=2$$的解；


（5）绘出（2），（3），（4）中的解的图形。

解：（1）对等式两边积分得$$y=x^2+c    $$

(2)将点带入得$$4=1+c$$

所以$$c=3$$

所以特解为$$y=x^2+3$$

（3）设切点为$$(x_0,2x_0+3)$$

$$因为y^\prime=2x$$

$$令2x_0=2$$
$$得x_0=1$$

所以切点$$(1,5)$$

所以$$c=4$$

所以特解$$y=x^2+4$$

(4)$$\int_0^1 ydx=\frac{1}{3}+c=2$$


所以$$c=\frac{5}{3}$$

所以解为$$y=x^2+\frac{5}{3}$$
