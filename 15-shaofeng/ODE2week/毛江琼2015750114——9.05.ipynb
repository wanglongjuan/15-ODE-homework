1.验证下列函数是相应微分方程的解

7)y=x^2+1,y{\prime}=y^2-(x^2+1)y+2x

将x^2+1=y 代入微分方程，有：$y^{\prime}=y^2-y^2+2x=2x$

由第一个等式可得:左边 = 右边，得证


<font color="red">**解**:

(1):$y^{\prime}=2x,y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)(x^2+1)+2x=2x=y^{\prime}$，验证完毕。

(2):$y^{\prime}=(-\frac{g(x)}{f(x)})^{\prime}=\frac{-g^{\prime}(x)f(x)+g(x)f^{\prime}(x)}{f^2(x)}=\frac{f^{\prime}(x)}{g(x)}\frac{g^{2}(x)}{f^{2}(x)}-\frac{g^{\prime}(x)}{f(x)}=\frac{f^{\prime}(x)}{g(x)}y^2-\frac{g^{\prime}(x)}{f(x)}$，验证完毕。<font color ="black">

8)$y=-\frac{g(x)}{f(x)},y^{\prime}=\frac{f^\prime(x)}{g(x)}y^2-\frac{g^\prime(x)}{f(x)}$

对第一个方程求导，得： $y=\frac{g(x)f^\prime(x)-f(x)g^\prime(x)}{f^2(x)}$

右边：$ \frac{f^\prime(x)}{g(x)}y^2-\frac{g^\prime(x)}{f(x)}=\frac{g(x)f^\prime(x)-f(x)g^\prime(x)}{f^2(x)}$

左边=右边，得证
