## 1.说明“有理整式”的概念。
有理整式(rational integral expression)亦称整式.一种基本的有理式.对变数字母与数仅限于有限次加、减、乘、乘方运算的代数式，即除式不含变数字母的有理式称为有理整式.整式包括单项式和多项式.可以认为有理整式即是多项式，单项式是多项式的特例.

## 2. 通解中，解对常数的独立性是指？
含有$n$个独立的任意常数$c_1,c_2,\cdots,c_n$的解$$y=\varphi(x,c_1,c_2,\cdots,c_n)$$称为$n$阶方程$F(x,y,\frac{\mathrm d y}{\mathrm d x},\cdots,\frac{\mathrm {d^{n}y}}{\mathrm {d}x^{n}})=0$的通解。关于解对常数的独立性是指，对$\varphi$及其$n-1$阶偏导数关于$n$个常数$c_1,c_2,\cdots ,c_n$的雅可比行列式不为$0$。


>其中雅可比行列式是指:

>对$n$个变元的$m$个函数$$y_{i}=f_{i}(x_1,x_2,\cdots,x_n),\,\,\,\,\,\,i=1,2,\cdots,m,$$
>定义**雅可比矩阵**为
>$$\begin{bmatrix}
\frac{\partial y_1}{\partial x_1}&\cdots&\frac{\partial y_1}{\partial x_n}\\
\vdots&&\vdots\\
\frac{\partial y_n}{\partial x_1}&\cdots&\frac{\partial y_m}{\partial x_n}\end{bmatrix}$$

>当$n=m$时，称雅可比矩阵对应的行列式为**雅可比行列式**。

 $$\frac{d\mathbf y}{dt}=\mathbf f(\mathbf y), \mathbf y(0) = \mathbf y_0$$
 
解, 则 $\mathbf y_2(t)= \mathbf y_1(t - t_0)$ 是
$$\frac{d\mathbf y}{dt}=\mathbf f(\mathbf y), \mathbf y(t_0) = \mathbf y_0$$

**解**:对$\mathbf y_2(t)$求导，$\frac{d\mathbf y_2(t)}{dt}=\frac{d\mathbf y_1(t - t_0)}{dt}=\frac{d\mathbf y_1(t - t_0)}{dt-t_0}\frac{dt-t_0}{dt}=\frac{d\mathbf y_1(t - t_0)}{dt-t_0}$

因为$\mathbf y_1(t)$ 是$\frac{d\mathbf y}{dt}=\mathbf f(\mathbf y), \mathbf y(0) = \mathbf y_0$解,所以$\frac{d\mathbf y_2(t)}{dt}=\mathbf f(\mathbf y_1(t-t_0))=\mathbf f(\mathbf y_2(t))$,并且$\mathbf y_2(t_0)=\mathbf y_1(t_0-t_0)=\mathbf y_0$。得证。

## 5.
给定一阶微分方程
$\frac{\mathrm dy}{\mathrm dx}=2x$,

(1)求出它的通解；

(2)求通过点$(1,4)$的特解；

(3)求出与直线$y=2x+3$相切的解；

(4)求出满足条件$\int_0^1y\,\mathrm{d}x=2$的解;

(5)绘出(2),(3),(4)中的解的图形。

**解**:

(1):$\frac{\mathrm dy}{\mathrm dx}=2x\iff\mathrm dy=2x\mathrm dx\iff\int\mathrm dy=\int2x\,\mathrm dx\iff y=x^2+C$

因此通解是$y=x^2+C$  (C是任意常数)。

(2):将$(1,4)$代入通解,求得$C=3$，所以通过点$(1,4)$的特解为$y=x^2+3$。

(3):$y=x^2+C$在$(x_0,y_0)=(x_0,{x_0}^2+C)$处的切线为$y=(\frac{\mathrm dy}{\mathrm dx}\mid_{x=x_0})\cdot(x-x_{0})+y_0=2x_{0}(x-x_0)+{x_0}^2+C$

联立，$\begin{cases}y=2x_{0}(x-x_0)+{x_0}^2+C\\y=2x+3\end{cases}\implies\begin{cases}x_0=1\\C=4\end{cases}$，于是与直线$y=2x+3$相切的解为$y=x^2+4$。

(4):对通解从0到1积分：$\int _0^1(x^2+C\mathrm)\,{d}x=\frac{1}{3}+C=2\implies C=\frac{5}{3}$，从而可知解为$y=x^2+\frac{5}{3}$。

(5):第(2),(3),(4)题的解的图形由以下代码显示。

```python
var('x')
p1=plot(x^2+3,(-5,5),rgbcolor=(1,0,0))
p2=plot(x^2+4,(-5,5),rgbcolor=(0,1,0))
p3=plot(x^2+5/3,(-5,5),rgbcolor=(0,0,1))
t = text('The red curve is $x^2+3$\nThe green curve is $x^2+4$ \nThe blue curve is $x^2+5/3$', (4,5), rgbcolor=(0,0,0))
show(p1+p2+p3+t)
```

## 6.求下列方程的解：
(1) $x\frac{\mathrm{d}y}{\mathrm{d}x}-y+\sqrt{x^2-y^2}=0$；

(2) $\tan{y}\,\mathrm{d}x-\cot{x}\,\mathrm{d}y=0$；

(3) $\frac{\mathrm dy}{\mathrm dx}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$；

(4) $\frac{\mathrm dy}{\mathrm dx}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$；

**解**:

(1):$x\frac{\mathrm{d}y}{\mathrm{d}x}-y+\sqrt{x^2-y^2}=0\implies(x\neq0时) \frac{\mathrm{d}y}{\mathrm{d}x}=\frac{y}{x}-sgn(x)\sqrt{1-{(\frac{y}{x}})^2}$,为一个齐次微分方程，做变量代换$u=\frac{y}{x}$得到变量分离方程$\frac{\mathrm d u}{\mathrm d x}=\frac{-sgn(x)\sqrt{1-u^2}}{x}$

在$u\neq\pm1$时，$\frac{\mathrm d u}{\mathrm d x}=\frac{-sgn(x)\sqrt{1-u^2}}{x}\iff\frac{\mathrm d u}{\sqrt{1-u^2}}=-sgn(x)\frac{\mathrm d x}{x}\iff\int\frac{\mathrm d u}{\sqrt{1-u^2}}=\int-sgn(x)\frac{\mathrm d x}{x}\iff\arcsin u=-\ln |x|+C\iff\arcsin \frac{y}{x}=-\ln |x|+C$

在$u=\pm1$时,为$y^2=x^2$。

综上解为$\arcsin \frac{y}{x}=-\ln |x|+C$或$y^2=x^2$。

(2).$\tan{y}\,\mathrm{d}x-\cot{x}\,\mathrm{d}y=0\,(y\neq\frac{\pi}{2}+k\pi\wedge x\neq k\pi,k\in \mathbb{N}时)\implies\tan x\,\mathrm d x=\cot y\,\mathrm d y(x,y\neq\frac{\pi}{2}k,k\in \mathbb{N}时)$或$(y\neq\frac{\pi}{2}+k\pi\wedge x\neq k\pi,k\in \mathbb{N})\wedge(x=\frac{\pi}{2}+k\pi\vee y= k\pi,k\in \mathbb{N}) $

$\tan x\,\mathrm d x=\cot y\,\mathrm d y(x,y\neq\frac{\pi}{2}k,k\in \mathbb{N}时)\iff\int\tan x\,\mathrm d x=\int\cot y\,\mathrm d y(x,y\neq\frac{\pi}{2}k,k\in \mathbb{N}时)\iff|\sin y\cdot\cos x|=C(x,y\neq\frac{\pi}{2}k,k\in \mathbb{N}时)$

结合$(y\neq\frac{\pi}{2}+k\pi\wedge x\neq k\pi,k\in \mathbb{N})\wedge(x=\frac{\pi}{2}+k\pi\vee y= k\pi,k\in \mathbb{N}) $有解为：

$|\sin y\cdot\cos x|=C(y\neq\frac{\pi}{2}+k\pi\wedge x\neq k\pi,k\in \mathbb{N}时)$,或$y= k\pi,k\in \mathbb{N}$

(3).$\frac{\mathrm dy}{\mathrm dx}=\frac{y^6-2x^2}{2xy^5+x^2y^2}\iff \frac{1-2(\frac{x}{y^3})^2}{2(\frac{x}{y^3})+(\frac{x}{y^3})^2}=y^2\frac{\mathrm d y}{\mathrm dx}=\frac{1}{3}\frac{\mathrm dy^3}{\mathrm dx}$，令$y^3=xu$，$\implies C|x^5|=|(3-u)^7(u+2)^3|\implies C|x^{15}|=|(3-y^3)^7(y^2+2x)^3|$

(4).$\frac{\mathrm dy}{\mathrm dx}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}\iff (3x^2y+2y^3-y)dy=(2x^3+3xy^2+x)dx\iff(3x^2+2y^2-1)\mathrm dy^2=(2x^2+3y^2+1)\mathrm dx^2\iff\frac{\mathrm d y^2}{\mathrm d x^2}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$

令$$\begin{cases}X=x^2-1\\Y=y^2+1\end{cases}$$

则有：$$\frac{\mathrm d Y}{\mathrm d X}=\frac{2X+3Y}{3X+2Y}=\frac{2+3\frac{Y}{X}}{3+2\frac{Y}{X}}$$

令$u=\frac{Y}{X}$，则有$\frac{\mathrm du}{\mathrm dX}=\frac{\frac{2+3u}{3+2u}-u}{X}=\frac{\frac{2-2u^2}{3+2u}}{X}$,

当$u^2\neq1$时，有$\frac{1}{X}\mathrm dx=\frac{3+2u}{2-2u^2}\mathrm du\iff \int\frac{1}{X}\mathrm dx=\int\frac{3+2u}{2-2u^2}\mathrm du\iff C(X-Y)^{10}=(X+Y)^2\iff (x^2-y^2-2)^5=C(x^2+y^2) $

$u^2=1$时，有$y^2=x^2$,但这并不是解，于是解为$(x^2-y^2-2)^5=C(x^2+y^2）$。
