3.验证下列各函数是微分方程的解：

(7)$y=x^2+1,y'=y^2-(x^2+1)y+2x$

(8)$y=-\frac{g(x)}{f(x)},y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$


解：(7)因为 $y=x^2+1$,$y'=2x$

故$$y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)(x^2+1)+2x=2x=y'$$

因此$y=x^2+1$是方程$y'=y^2-(x^2+1)y+2x$的解。
    
 （8）因为$y=-\frac{g(x)}{f(x)}$

所以$$y'=-\frac{g'(x)g(x)-g(x)f'(x)}{f^2(x)}=\frac{g(x)f'(x)}{f^2(x)}-\frac{(g'(x)}{f(x)}$$

而$$\frac{f'(x)y^2}{g(x)}-\frac{g'(x)}{f(x)}=\frac{f'(x)}{g(x)}[\frac{-g(x)}{f(x)}]^2-\frac{g'(x)}{f(x)}=\frac{g(x)f'(x)}{f^2(x)}-\frac{g'(x)}{f(x)}=y'$$

所以$y=-\frac{g(x)}{f(x)}$是方程 $y'=\frac{f'(x)y^2}{g(x)}-\frac{g'(x)}{f(x)}$ 的解。

（2）求通过点（1,4）的特解；

（5）绘出（2）（3）（4）中的解的图形。

解:（1）由$\frac{dy}{dx}=2x$,得$dy=2xdx$,
方程两边积分,既得$$y=x^2+c (c为任意常数)$$ 

  （2）把x=1,y=4代入$y=x^2+c$,得c=3
故通过点（1,4）的特解为$$y=x^2+3$$ 

  （3）因为与直线相切，所以抛物线与直线只有一个公共点，即方程组$$\begin{cases}y=x^2+c\\y=2x+3\end{cases}$$有且只有唯一一组解，即$x^2+c=2x+3$有唯一解，

1.求下列方程的解：

(6)$x\frac{dy}{dx}-y+\sqrt{x^2-y^2}=0;$

（7）$tanydx-cotxdy=0;$

解：（6）原方程变形为$$xdy-ydx+\sqrt{x^2-y^2}dx=0.$$

当${y^2}\neq{x^2}$，两边同除以$\sqrt{x^2-y^2}$得$$\frac{xdy-ydx}{\sqrt{x^2-y^2}}+dx=0,$$

即$$\begin{equation}
\cfrac{d(\frac{y}{x})}{\sqrt{1-(\frac{y}{x})^2}}+\frac{dx}{x}=0
\end{equation}$$

两边积分得$$arcsin\frac{y}{x}+ln{|x|}=c,(这里c为任意常数。)$$

另外，当有$y^2=x^2$时，$y^2=x^2$也是方程的解。

故原方程的解为$y^2=x^2$或$arcsin\frac{y}{x}+ln{|x|}=c$。

(7)当$tany\neq{0}$时,即$y\neq{k\pi},k=0,\pm{1},\pm{2},···$时，分离变量得$$cotydy=tanxdx.$$

两边同时积分，即得$$ln|siny|=-ln|cosx|+c_1,(这里c_1为任意常数。)$$

即$sinycosx=c,这里c=e^{c1}$为任意正常数。

另外，当$y=k\pi,k=0,\pm{1},\pm{2},···$时，容易验证也是方程的解。

所以，原方程的解为$sinycosx=c或y=k\pi,k=0,\pm{1},\pm{2},···.$

2.作适当的变量变换求解下列方程：

（6）$\frac{dy}{dx}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$

（7）$\frac{dy}{dx}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$

解：
（1）令$u=y^3$
则方程可变形为$$\frac{du}{dx}=\frac{3u^2-6x^2}{2xu+x^2}$$
再令$v=\frac{u}{x}$,即$u=vx$
则上面方程化为$$\frac{xdv}{dx}+v=\frac{3v^2-6}{2v+1}$$
整理得$$\frac{7dv}{5(v-3)}+\frac{3dv}{5(v+2)}=\frac{dx}{x}$$
两边同时积分，得$（v-3）^7(v+2)^3=cx^5$ 这里c为正常数。 
   
代入$v=\frac{u}{x}=\frac{y^3}{x}$,并且化简得原方程的解为$$（y^3-3x）^7(y^3+2x)=cx^{15}$$.

（2）原方程可变形为$$\frac{ydy}{xdx}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$$
即$$\frac{dy^2}{dx^2}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$$
令$u=x^2-1,v=y^2+1$,则上面方程化为$$\frac{dv}{du}=\frac{2u+3v}{3u+2v}$$
再令$w=\frac{v}{u}$,即$v=uw$
则上面方程化为 
$$\frac{udw}{du}+w=\frac{2+3w}{3+2w}$$ 
分离变量并且积分，得$c·(1+w)=u^4(1-w)^5$
把$w=\frac{v}{u}$,$v=y^2+1$,$u=x^2-1$,代入得$$c(x^2+y^2)=(x^2-y^2-2)^5$$ 这里c是任意常数。
