**1.求下列方程的解：**

(6)  ** $x\frac{\mathrm dy}{\mathrm dx}-y+\sqrt{\mathrm x^2-\mathrm y^2}=0$**

**解：$\frac{\mathrm dy}{\mathrm dx}=\frac{\mathrm y}{\mathrm x}-\sqrt{1-\frac{\mathrm y^2}{\mathrm x^2}}$**

令**$u=\frac{\mathrm y}{\mathrm x},则 x\frac{\mathrm du}{\mathrm dx}+u=u-\sqrt{1-u^2}$**

所以$\frac{1}{\mathrm x}dx=-\frac{1}{\sqrt{1-\mathrm u^2}}du,(\sqrt{1-u^2}\neq0)$

两边求积分得 **$\mathrm ln|x|+c=-\mathrm arcsin(u)$**

$\sqrt{1-u^2}=0时,y^2=x^2$

方程的解为$\begin{cases}\mathrm ln|x|+\mathrm arcsin(\frac{\mathrm y}{\mathrm x})+c=0\\y^2=x^2\end{cases}$

**(7)  $\mathrm tany  \mathrm dx-\mathrm cotx  \mathrm dy=0$**

**解：$\mathrm tany \neq0时，\frac{1}{\mathrm tany}\mathrm dy=\frac{1}{\mathrm cotx}\mathrm dx$**

两边积分得 **$\mathrm ln|\mathrm siny|=-\mathrm ln|\mathrm cosx|+c所以\mathrm ln|\mathrm cosx\mathrm siny|=c$**

**$\mathrm cosx\mathrm siny=\pm e^c$**

$\mathrm tany= 0时，y=k\pi，(k\in N^+)$

所以方程的解为$\begin{cases}\mathrm cosx\mathrm siny=\pm e^c\\y=k\pi，(k\in Z)\end{cases}$

**2.做适当的变量变换求解下列方程：**

**(6)$\frac{\mathrm dy}{\mathrm dx}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$;**

**解：**两边同乘$3y^2$，可得：$\frac{dy^3}{dx}=\frac{3(y^3)^2-6x^2}{2xy^3+x^2}$

令$u=y^3$则$\frac{du}{dx}=\frac{3u^2-6x^2}{2xu+x^2}=\frac{\frac{3u^2}{x^2}-6}{\frac{2u}{x}-1}$

令$z=\frac{u}{x}$，则$\frac{du}{dx}=x\frac{dz}{dx}+z$,

原方程可化为$z+x\frac{dz}{dx}=\frac{3z^2-6}{2z+1}$

移项化简得：$x\frac{dz}{dx}=\frac{z^2-z-6}{2z+1}$

1）.若$z^2-z-6=0$,则方程的解为：z=3或z=-2,即：$y^3=3x或$y^3=-2x$

2）.若$z^2-z-6$不为0，化简得，$\frac{dx}{x}=\frac{2z+1}{z^2-z-6}dz$

两边同时积分，化简得：$x^5(z-3)^7(z+3)^3=C$,

最后得：$(y^3-3x)^7(y^3+2x)^3=x^5C$

又$y^3=3x,y^3=-2x$均满足2)中等式，

综上所述，方程的解为：$(y^3-3x)^7(y^3+2x)^3=Cx^5$

**(7)$\frac{\mathrm dy}{\mathrm dx}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$**

**解:**两边同乘$\frac{y}{x}$,得：$\frac{dy^2}{dx^2}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$

令$u=y^2,v=x^2$,得$\frac{du}{dv}=\frac{2v+3u+1}{3v+2v-1}$,

解方程组：$\begin{cases}2v+3u+1=0\\3v+2u-1=0\end{cases}$得：$\begin{cases}v=1\\u=-1\end{cases}$

令：$\begin{cases}v=V+1\\u=U-1\end{cases}$得：$\frac{du}{dv}=\frac{2v+3u}{3v+2u}=\frac{2+\frac{3u}{x}}{3+\frac{2u}{x}}$

令$m=\frac{u}{v}$，所以：$u=mv,\frac{du}{dv}=v\frac{dm}{dv}+m$

得：$m+v\frac{dm}{dv}=\frac{2+3m}{3+2m},\frac{dm}{dv}=\frac{2-2m^2}{3+2m}$

1).若$2-2m^2=0,m=\pm1$为方程的解，即：$y^2=x^2-2$或$y^2=-x^2$为方程的解

2).若$2-2m^2\neq0$移项化简得：$\frac{1}{v}dv=\frac{3+2m}{2-2m^2}$

两边同时积分，化简得：$y^2+x^2=(y^2-x^2+2)^5C$

又$y^2=x^2-2，y^2=-x^2$符合上述表达式，

故方程的解为：$y^2+x^2=C(y^2-x^2+2)^5$
