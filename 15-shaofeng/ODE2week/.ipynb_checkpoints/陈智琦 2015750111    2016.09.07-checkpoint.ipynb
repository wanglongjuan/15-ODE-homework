四、验证下列各函数是相应微分方程的解。

（1）$y=x^2+1,y'=y^2-(x^2+1)y+2x$

（2）$y=-\frac{g(x)}{f(x)},y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$

解：

（1）由已知得$y'(x)=2x$,将y带入$y'=y^2-(x^2+1)y+2x=(x^2+1)^2-(x^2+1)(x^2+1)+2x=2x$,可知$y=x^2+1$是$y'=y^2-(x^2+1)y+2x$的解

（2）由$y=-\frac{g(x)}{f(x)}$可知$y'(x)=\frac{f'(x)g(x)-g'(x)f(x)}{f^2(x)}$

即$y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$

 $=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$

 $=\frac{f'(x)g(x)-g'(x)f(x)}{f^2(x)}$
 
 因此$y=-\frac{g(x)}{f(x)}$是$y'=\frac{f'(x)}{g(x)}y^2-\frac{g'(x)}{f(x)}$的解

七、作适当的变量变换求解下列方程

（1）$$\frac{dy}{dx}=\frac{y^6-2x^2}{2xy^5+x^2y^2}$$

（2）$$\frac{dy}{dx}=\frac{2x^3+3xy^2+x}{3x^2y+2y^3-y}$$

解：（1）$$\frac{dy}{dx}=\frac{(y^3)^2-2x^2}{y^2(2xy^3+x^2)}$$

既$$\frac{dy^3}{dx}=\frac{3((y^3)^2-2x^2)}{2xy^3+x^2}$$

令$$u=y^3$$

既，原方程为$$\frac{du}{dx}=\frac{3u^2-6x^2}{2xu+x^2}$$

$$\frac{du}{dx}=\frac{3(\frac{u}{x})^2-6}{2\frac{u}{x}+1}$$

令$z=\frac{u}{x}$

$$\frac{du}{dx}=z+x\frac{dz}{dx}=\frac{2z^2-6}{2z+1}$$

$$x\frac{dz}{dx}=\frac{z^2-z-6}{2z+1}$$    

(1)当$z^2-z-6=0$时，即$z=3$   or   $z=2$是上式方程的解，即$y^3=3x$     or$y^3=-2x$是原方程的解

（2）当$z^2-z-6\neq0$时，

$$\frac{dx}{x}=\frac{2z+1}{z^2-z-6}dz$$

两边同时积分得$$(z-3)^7(z+2)^3=x^15c$$

即$$(y-3x)^7(y+2x)^3=x^15c$$

又因为$y^3=3x$    or    $y=-2x$包含在通解$c=0$中

所以原方程的解为$(z-3)^7(z+2)^3=x^15c$

（2）解：$$\frac{dy}{dx}=\frac{x(2x^2+3y^2+1)}{3x^2+2y^2-1}$$

$$\frac{dy^2}{dx^2}=\frac{2x^2+3y^2+1}{3x^2+2y^2-1}$$

令 $u=y^2$          $v=x^2$

$$\frac{du}{dv}=\frac{2v+3u+1}{3u+2v+1}$$



方程组

$$
\begin{cases}
2v+3u+1=0\\
3u+2v-1=0
\end{cases}
$$的解为（1,-1）

令$Z=v-1$  $Y=u+1$

则有$$
\begin{cases}
2Z+3Y=0\\
3Z+2Y=0
\end{cases}
$$

即$$\frac{dy}{dz}=\frac{2+3\frac{y}{z}}{3+2\frac{y}{z}}$$

令$t=\frac{y}{z}$  则有$\frac{dy}{dz}=t+z\frac{dt}{dz}$

$$z\frac{dt}{dz}=\frac{2-2t^2}{3+2t}$$

当$2-2t^2=0$时，即$t=\pm1$是上式方程的解

得$y=x^2-2$  $y=x^2$是原方程的解

当$2-2t^2\neq0$时

$$\frac{3+2t}{2-2t^2}dt=\frac{dt}{z}$$

$$y^2+x^2=(y^2-x^2+2)^5c$$

且$y=x^2+2$  or  $y=-x^2$包含在其通解中，因此原方程的解为

$$y^2+x^2=(y^2-x^2+2)^5c$$
