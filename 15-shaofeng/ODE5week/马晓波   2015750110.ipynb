**3.**求初值问题

**$$
\begin{cases}
\frac{\mathrm dy}{\mathrm dx}=x^2+y^2,R:|x+1|\leq 1,|y|\leq 1,\\
y(-1)=0
\end{cases}
$$**

的解的存在区间，并求第二次近似解，给出在解的存在区间的误差估计。

**解：**

由题得 $a=1,b=1,f(x,y)=x^2-y^2$

$\therefore\quad M=\max_{(x,y)\in R}|f(x,y)| , h=\min{a,\frac{b}{M}}=\min{1,\frac{1}{4}}=\frac{1}{4}$,

在 $R$ 上函数 $f(x,y)=x^2-y^2$ 的利普希茨常数可取为 $L=2$，因为 $\frac{\partial f}{\partial y}=|-2y|\leq2=L$

$$
\begin{aligned}
\varphi_0(x)&=0\\
\varphi_1(x)&=\int_{-1}^xt^2-\varphi_0^2(t)\,\mathrm dt=\frac{x^3}{3}+\frac{1}{3}\\
\varphi_2(x)&=\int_{-1}^xt^2-\varphi_1^2(t)\,\mathrm dt=-\frac{x}{9}+\frac{x^3}{3}-\frac{x^4}{18}-\frac{x^7}{63}+\frac{11}{42} 
\end{aligned}
$$

误差估计公式 $\varphi_n(x)-\varphi(x)=\frac{ML^n}{(n+1)!}h^{n+1}$ 得$$\varphi_2(x)-\varphi(x)=\frac{4\bullet 2^2}{3!}{\frac{1}{4}}^{3}=\frac{1}{24}$$

所以，已知初值问题的解的存在区间为 $$-\frac{5}{4}^{3}\leq x \leq-\frac{3}{4}$$

第二次近似值为 $$y_2=-\frac{x}{9}+\frac{x^3}{3}-\frac{x^4}{18}-\frac{x^7}{63}+\frac{11}{42}$$

在解的存在区间的误差估计为 $$\varphi_2(x)-\varphi(x)=\frac{1}{24}$$

**8.**证明：如果函数 $f(x,y)$ 与带域 $\alpha\leq x\leq\beta$ 上连续且关于 $y$ 满足利普希茨条件，则方程 $\frac{\mathrm dy}{\mathrm dx}=f(x,y)$ 满足条件 $y(x_0)=y_0$ 的解于整个区间 $[\alpha,\beta]$ 上存在且唯一。

**证明：**

从教材中命题一可知，方程$\frac{\mathrm d y}{\mathrm d x}=f(x,y)$满足条件$y(x_0)=y_0$ 的解等价于积分方程$y=y_0+\int_{x_0}^x f(x,y)\mathrm d x ,x\in[\alpha,\beta]$ 的连续解，因此只需证明上述积分方程的解存在唯一即可。

$1)\quad$ 存在性

为了简单起见，只就区间$x_0\leq x\leq\beta$ 进行讨论，对于$\alpha\leq x\leq x_0$ 的讨论完全一样。

现取 $y_0(x)=y_0$ ，构造皮卡逼近数列如下：
$$
\begin{cases}
y_0(x)=y_0 \\
y_n(x)=y_0+\int_{x_0}^x f(\xi,y_{n-1}(x))\mathrm dx  ,  x\in[\alpha,\beta] \\  \quad  \quad  \quad (n=1,2,…) 
\end{cases}     \quad [1]       
$$

易知$y_n(x)$在$x\in[\alpha,\beta]$ 上有定义且连续。下面证明函数序列{${y_n(x)}$} 在$x\in[\alpha,\beta]$ 上一致收敛。
考虑级数$$y_0(x)+\sum_{k=1}^{+\infty}[y_k(x)-y_{k-1}(x)] , x\in[\alpha,\beta]  \quad  [2]$$

取$$M=\max_{x\in[\alpha,\beta]}\mid f(x,y_0)\mid$$ ,
$$\mid y_1(x)-y_0(x)\mid\leq \int_{x_0}^x \mid f(\xi,y_0(\xi))\mid \mathrm d\xi \leq M\mid x-x_0 \mid \quad [3]$$
$$\mid y_2(x)-y_1(x)\mid\leq \int_{x_0}^x \mid f(\xi,y_1(\xi))-f(\xi,y_0(\xi))\mid \mathrm d\xi  $$

利用利普希茨条件及[3],得到 $$\mid y_2(x)-y_1(x)\mid\leq \int_{x_0}^x \mid f(\xi,y_1(\xi))-f(\xi,y_0(\xi))\mid \mathrm d\xi \leq L\int_{x_0}^x \mid y_1(\xi)-y_0(\xi)\mid \mathrm d\xi\leq L\int_{x_0}^x M(\xi-x_0) \mathrm d\xi=\frac{ML}{2!}(x-x_0)^2$$

可用数学归纳法证明：$$\mid y_k(x)-y_{k-1}(x)\mid\leq \frac{ML^{k-1}}{k!}(x-x_0)^k\leq\frac{ML^{k-1}}{k!}(\beta-\alpha)^k$$

上式的右端是正项级数的一般项，由weierstrass判别法，级数在$x_0\leq x\leq\beta$ 上面一致收敛，因而函数序列{${y_n(x)}$} 也在 $x_0\leq x\leq\beta$上一致收敛。

现设$$\lim_{n\to\infty}y_n(x)=y(x) $$

易知y(x)在区间$x\in[\alpha,\beta]$ 上连续，且{$f(x,y_n(x))$}在区间$x\in[\alpha,\beta]$ 也一致收敛于 $f(x,y(x))$ .

因而对[1]式两端取极限，即得$$y(x)=y_0+\int_{x_0}^x f(\xi,y(\xi))\mathrm d\xi$$

即 $y(x)$是积分方程的定义于$x\in[\alpha,\beta]$ 上的连续解。

$2)\quad$唯一性。

设$\phi(x)$ 是积分方程的定义于$x_0\leq x\leq\beta$  上的另一个连续解，则$$\phi(x)=y_0+\int_{x_0}^xf\big((\xi,\phi(\xi)\big)\mathrm d\xi$$

$\because$ $$y(x)=y_0+\int_{x_0}^x f\big(\xi,y(\xi)\big)\mathrm d\xi $$

$\therefore$ 

$$\mid\phi(x)-y(x)\mid=\quad \int_{x_0}^x f\big(\xi,\phi(\xi))-f(\xi,y(\xi)\big)\mathrm d\xi$$

$$\quad\quad\quad\quad\quad\quad\quad\leq\quad \int_{x_0}^x \mid f\big(\xi,\phi(\xi)\big)-f\big(\xi,y(\xi)\big)\mid\mathrm d\xi$$

$$\quad\quad\quad\leq\quad L\int_{x_0}^x \mid\phi(\xi)-y(\xi)\mid\mathrm d\xi$$

$$\quad\quad\quad\quad\quad\quad\leq\quad \int_{x_0}^{\alpha}\mid\phi(\xi)-y(\xi)\mid\mathrm d\xi\bullet e^{\int_{\alpha}^x L\mathrm ds}$$

$$\leq\quad 0\quad(\because x_0\geq\alpha)$$

又$\because$ $$\mid\phi(x)-y(x)\mid\geq 0$$

$\therefore$ $$\mid\phi(x)-y(x)\mid=0$$

唯一性得证。

设$\phi(x)$ 是积分方程的定义于$x_0\leq x\leq\beta$  上的另一个连续解，则$$\phi(x)=y_0+\int_{x_0}^xf\big((\xi,\phi(\xi)\big)\mathrm d\xi$$

$\because$ $$y(x)=y_0+\int_{x_0}^x f\big(\xi,y(\xi)\big)\mathrm d\xi $$

$\therefore$ 

$$\mid\phi(x)-y(x)\mid=\quad \int_{x_0}^x f(\xi,\phi(\xi))-f(\xi,y(\xi))\mathrm d\xi$$

$$\quad\quad\quad\quad\quad\quad\quad\leq\quad \int_{x_0}^x \mid f\big(\xi,\phi(\xi)\big)-f\big(\xi,y(\xi)\big)\mid\mathrm d\xi$$

$$\quad\quad\quad\quad\leq\quad L\int_{x_0}^x \mid\phi(\xi)-y(\xi)\mid\mathrm d\xi$$

$$\quad\quad\quad\quad\quad\quad\leq\quad \int_{x_0}^{\alpha}\mid\phi(\xi)-y(\xi)\mid\mathrm d\xi\bullet e^{\int_{\alpha}^x L\mathrm ds}$$

$$\leq\quad 0$$

又$\because$ $$\mid\phi(x)-y(x)\mid\geq 0$$

$\therefore$ $$\mid\phi(x)-y(x)\mid=0$$

唯一性得证。

**6.证明格朗沃尔不等式：**

设 $K$ 为非负常数，$f(x)$ 和 $g(t)$ 为在区间 $\alpha\leq x\leq\beta$ 上的连续非负函数，且满足不等式

$$
f(t)\leq K+\int^t_af(s)g(s)\,\mathrm ds,\quad\alpha\leq x\leq\beta,
$$

则有$$f(t)\leq Kexp\big(\int^t_ag(s)\,\mathrm ds\big),\quad\alpha\leq x\leq\beta.$$

并由此证明定理1的命题5

**证明：**

$1)\quad K>0$ 时，

令$$\omega(t)=K+\int_\alpha^tf(s)g(s)\, \mathrm ds$$

则$$\omega‘(t)=f(t)g(t)\leq \omega(t)g(t)$$

由 $\omega(t)>0$ 可得$$\frac{\omega'(t)}{\omega(t)}\leq g(t)$$

两边从 $\alpha$ 到 $t$ 积分得$$\mathrm ln\omega(t)-ln\omega(\alpha)\leq\int_\alpha^t g(s) \mathrm ds$$

即$$\frac{\omega(t)}{\omega(\alpha)}\leq exp\big(\int_\alpha^t g(s) \mathrm ds\big)$$

$$\omega(\alpha)=K>0$$

所以$$\omega(t)\leq K\bullet exp\big(\int_\alpha^t g(s) \mathrm ds\big)$$

即有$$f(t)\leq\omega(t)\leq K\bullet exp\big(\int_\alpha^t g(s) \mathrm ds\big),\quad \alpha\leq t\leq\beta$$

2)当 $K=0$ 时，

对任意 $\varepsilon>0$,由于 $f(t)\leq\int_\alpha^tf(s)g(s)\, \mathrm ds$,

所以$$f(t)\leq\varepsilon+\int_\alpha^tf(s)g(s)\, \mathrm ds$$

由 $1)$ 有$$f(t)\leq\varepsilon exp\big(\int_\alpha^t g(s) \mathrm ds\big)$$ 

当 $\quad\varepsilon\to0^+\quad$ 时，有 $f(t)\leq 0$

又因为 $\quad f(t)\geq 0\quad$ 所以$\quad f(t)\equiv 0$

从而$$f(t)\leq exp\big(\int_\alpha^t g(s) \mathrm ds\big)\quad \alpha\leq x\leq\beta$$

由 $1),2)$ 知，格朗沃尔不等式成立。

格朗沃尔不等式（2）：设 $y(t),g(t),h(t)$ 定义于区间 $[\alpha,\beta]$ 上的非负可积函数且满足不等式$$\frac{\mathrm dy(t)}{\mathrm dt}\leq g(t)y(t)\mathrm ds+h(t)$$

下面利用上述不等式证明定理 $5$ .

证明：设$\quad\varphi(x)、\phi(x)\quad$是初值问题$$y'=f(x,y),y(x_0)=y_0$$的两个解，则
$$\varphi(x)=y_0+\int_{x_0}^xf\big(\xi,\varphi(\xi)\big)\mathrm d\xi$$

$$\phi(x)=y_0+\int_{x_0}^xf\big(\xi,\phi(\xi)\big)\mathrm d\xi$$

$\therefore$ 
$$
\begin{aligned}
\quad \mid\varphi(x)-\phi(x)\mid &\leq\int_{x_0}^x\mid f\big(\xi,\varphi(\xi)\big)-f\big(\xi,\phi(\xi)\big)\mid\mathrm d\xi\\
\quad &\leq L\int_{x_0}^x\mid\varphi(\xi)-\phi(\xi)\mid\mathrm d\xi
\end{aligned}
$$

其中 $L$ 为利普西茨常数，由格朗沃尔不等式 $(2)$ 得$$\mid\varphi(x)-\phi(x)\mid \leq\int_{x_0}^\alpha \mid\varphi(\xi)-\phi(\xi)\mid \mathrm d\xi \bullet e^{\int_\alpha^x L\mathrm ds}\leq0\quad(\because\alpha\leq x_0)$$

又$\because$ $$\mid\varphi(x)-\phi(x)\mid\geq 0$$

$\therefore$ $$\varphi(x)=\phi(x)$$

即命题 $5$ 成立。

**10.**给定积分方程

$$
\varphi(x)=f(x)+\lambda\int^b_aK(x,\xi)\varphi(\xi)\,\mathrm d\xi,\quad\quad (*)
$$

其中 $f(x)$ 是 $[\alpha,\beta]$ 上的已知连续函数，$K(x,y)$ 是 $a\leq x\leq b ,a\leq \xi\leq b$ 上的已知连续函数。证明：当 $|\lambda|$ 足够小时（$\lambda$ 是常数），（$*$）在 $[\alpha,\beta]$ 上存在唯一的连续解。

**证明：**
$1)\quad$ 存在性

作逐项逼近序列：
$$
\begin{cases}
\varphi_0(x)=f(x)\\
\varphi_{n+1}(x)=f(x)+\lambda\int_a^b{K(x,\xi)\varphi_n(\xi)}\mathrm d\xi \quad\quad (1)
\end{cases}
$$

令$\quad M=\displaystyle\max_{x\in[a,b]}\mid f(x)\mid,
\quad L=\displaystyle\max_{\substack{
x\in[a,b]\\ 
\xi\in[a,b]}}
\mid K(x,\xi)\mid$

则$$
\begin{aligned}
\quad \mid\varphi_1 (x)-\varphi_0 (x)\mid
&\leq\quad \mid\int_a^b\lambda K(x,\xi)\varphi_0 (\xi)\,\mathrm d\xi\mid \\
&\leq\quad \mid \lambda \mid \int_a^b\mid K(x,\xi)\mid\mid f(\xi)\mid,\mathrm d\xi\\
&\leq\quad \mid \lambda\mid ML(b-a)
\end{aligned}
$$

假设$\quad\mid\varphi_n (x)-\varphi_{n-1} (x)\mid \leq\mid\lambda\mid^n L^n M(b-a)^n\quad$成立

则有
$$
\begin{aligned}
\mid\varphi_{n+1} (x)-\varphi_n (x)\mid
&=\quad \mid\int_a^b\lambda K(x,\xi)\big(\varphi_n (\xi)-\varphi_{n-1} (\xi)\big)\,\mathrm d\xi\mid \\ 
&\leq\quad  \mid \lambda \mid \int_a^b\mid K(x,\xi)\mid\mid\varphi_n (\xi)-\varphi_{n-1} (\xi)\big)\,\mathrm d\xi\\ 
&\leq\quad  \mid \lambda \mid \int_a^b\mid\lambda\mid^n L^{n+1}M(b-a)^n\mathrm d\xi\\
&=\quad     \mid\lambda\mid^{n+1} L^{n+1}M(b-a)^{n+1}
\end{aligned}
$$

由归纳假设可知，对任意 $ n\in N^+$ 

有$$\mid\varphi_{n+1} (x)-\varphi_n (x)\mid\leq\mid\lambda\mid^{n+1} L^{n+1}M(b-a)^{n+1}$$

证下列序列 $\{\varphi_n(x)\}$ 在 $[a,b]$ 上一致收敛

考虑级数 $$\quad\varphi_0(x)+\displaystyle\sum_{k=1}^\infty \big(\varphi_k(x)-\varphi_{k-1}(x)\big)$$

则$$\varphi_n(x)=\varphi_0(x)+\displaystyle\sum_{k=1}^n \big(\varphi_k(x)-\varphi_{k-1}(x)\big)$$

所以要证序列 $\{\varphi_n(x)\}$ 在 $[a,b]$ 上一致收敛，只需证级数 $(2)$ 在 $[a,b]$ 上一致收敛。

$\because$ $$\mid\varphi_{n+1} (x)-\varphi_n (x)\mid\leq\mid\lambda\mid^{n+1} L^{n+1}M(b-a)^{n+1}$$

$\therefore$ 

当 $\lambda<\frac{1}{(b-a)L}$ 时，级数 $\displaystyle\sum_{k=1}^\infty\mid\lambda\mid^{n+1} L^{n+1}M(b-a)^{n+1}$ 收敛

当 $\lambda>\frac{1}{(b-a)L}$ 时，级数 $\displaystyle\sum_{k=1}^\infty\mid\lambda\mid^{n+1} L^{n+1}M(b-a)^{n+1}$ 收敛


设极限为 $\varphi(x)$, 即 $任意的x\in[a,b],n\to\infty$ 时, $\varphi_n(x) 一致收敛于 \varphi(x) $

由 $\varphi_n(x) 的连续性可知 \varphi(x) 在 [a,b] 连续  $    

由 $(1)$ 式得，当 $\quad n\to\infty\quad $ 时，有    
$$\varphi(x)=f(x)+\lambda\int_a^bK(x,\xi)\varphi(x)\mathrm d\xi\quad \big(\mid\lambda\mid<\frac{1}{(b-a)L}\big)$$

这证明了当 $\quad\mid\lambda\mid < \frac{1}{(b-a)L}\quad$ 时，积分方程存在连续解。


$2)\quad$唯一性

 设有另一解$\quad\phi(x)\quad$ 且 $\quad\phi(x)\neq\phi(x)$
     
即 $$\phi(x)=f(x)+\lambda\int_a^bK(x,\xi)\phi(x)\mathrm d\xi$$

 记$\quad A=\max_{x\in[a,b]}\mid\varphi(x)-\phi(x)\mid>0\quad$ 则
 $$
 \begin{aligned}
 \mid\varphi(x)-\phi(x)\mid 
 &=\quad \mid\lambda\int_a^b K(x,\xi) \big(\varphi(\xi)-\phi(\xi)\big)\mathrm d\xi\mid\\
 &\leq\quad \mid\lambda\mid\int_a^b L\mid\varphi(\xi)-\phi(\xi)\mid\mathrm d\xi\\
 &\leq\quad \mid\lambda\mid A(b-a)
 \end{aligned}
 $$
 
 即 $\quad A\leq\mid\lambda\mid AL(b-a)$
   
 $又\because\quad A=\max_{x\in[a,b]}\mid\varphi(x)-\phi(x)\mid>0\quad\therefore\quad\lambda\geq\frac{1}{(b-a)L} 与  \lambda<\frac{1}{(b-a)L}\quad矛盾$ 
 
  故唯一性得证 
  
  因此当 $|\lambda|$ 足够小时（$\lambda$ 是常数），（$*$）在 $[\alpha,\beta]$ 上存在唯一的连续解。
